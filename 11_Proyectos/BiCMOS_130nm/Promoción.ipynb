# Proyecto para promoción de Electrónica Aplicada III

Para la promoción del proyecto sera condición necesaria analizar un trabajo tecnico referido a los tema de la materia. Cada grupo recibirá un documento de IEEE que cubre varios temas sobre RFIC. El examen será oral e indivudual.

El proyecto reviste carácter individual. El proyecto consiste de analizar y verificar una publicación técnica o científica.  La evaluación del proyecto se realizará mediante un informe y una presentación oral. 

El objetivo del proyecto es que los alumnos de la materia Electrónica Aplicada III apliquen los temas estudiados durante el curso en un proyecto de análisis técnico. Para obtener la promoción de la materia, deberán cumplir con los requistos que se desarrollan a continuación


Los alumnos deberán realizar un análisis riguroso de un trabajo técnico relacionado con los temas abordados en la materia. Además, deberán llevar a cabo simulaciones para verificar los conceptos presentados en el documento de IEEE que se proporcionará. Es importante que las simulaciones demuestren una comprensión completa de los conceptos técnicos.

Los resultados de las simulaciones deberán ser presentados en un informe técnico detallado que demuestre la comprensión y habilidades técnicas de los alumnos. Asimismo, deberán preparar una presentación de 10 a 15 minutos que vincule los detalles técnicos del documento de IEEE con los contenidos de la materia. La presentación debe ser clara, concisa y demostrar una comprensión completa de los conceptos técnicos.

La evaluación del proyecto se realizará de manera oral y será un requisito obligatorio para obtener la promoción de la materia. Por lo tanto, es fundamental que los alumnos trabajen con anticipación y estén preparados para demostrar sus habilidades técnicas y su comprensión de los conceptos presentados. Si tienen alguna duda o consulta, los alumnos pueden comunicarse con los responsables de la materia.


## Informe

La presentación del informe tiene como objetivo dar a conocer, en base a fundamentos teóricos, el funcionamiento del circuito. En el informe debe explicar el funcionamiento del circuito y cuales pueden ser sus aplicaciones. Además, se deben justificar la topología y componentes del circuito en base a cálculos y simulaciones. 


Se debe detallar que mediciones o pruebas deben ser realizadas para constatar el funcionamiento del circuito. 
La fecha límite de entrega será antes del examen de recuperación del curso. 

El informe debe estar estructurado con el siguiente formato:

    . Hoja presentación: Titulo, autores.

    . Índice

    . Resumen

    . Introducción

    . Diseño y consideraciones prácticas

    . Mediciones y verificaciones.

    . Conclusiones

    . Bibliografía



## Presentación oral

El informe debe ser sustentado por una presentación oral, donde se de una explicación sobre el proyecto.

Al final de la presentación, tendremos una sección de preguntas y respuestas.

# Nodo de proceso SG13G2

SG13G2 es una tecnología BiCMOS de alto rendimiento con un proceso CMOS de 0,13 μm. Contiene dispositivos bipolares basados en SiGe:C npn-HBT con una frecuencia transitoria de hasta 300 GHz y una frecuencia de oscilación de 500 GHz de oscilación. Este proceso proporciona 2 óxidos de puerta: Un óxido de puerta fino para la lógica digital de 1,2 V y un óxido grueso para una tensión de alimentación de 3,3 V. Para ambos módulos se ofrecen transistores NMOS, PMOS y NMOS, también están disponibles otros componentes pasivos como resistencias de polisilicio y condensadores MIM. 
La opción ofrece 5 capas metálicas finas, dos capas metálicas gruesas (de 2 y 3 μm de grosor) y una capa MIM.
