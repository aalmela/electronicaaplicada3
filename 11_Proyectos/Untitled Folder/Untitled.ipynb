```python
import matplotlib
import matplotlib.pyplot as plt

import numpy as np
```

```python
N  = 16384
fs = 125e6

fm =2*fs/N
fc =500*fs/N
m  = 0.6
zo = 50 # ohms

print(fm, fc)

ap = -10 #dBm


av = ((10**(ap/10)*1e-3)*(2*zo))**0.5
print(av)
```

```python
ns = np.arange(0, N) # generates x axis in range 0 to 6 with 20000 points


s1 = 1*np.sin(2*np.pi*fm/fs*ns) # the first sinus signal with the amplitude 0.8
s2 = av*(1+m*s1)*np.sin(2*np.pi*fc/fs*ns) # the second sinus signal with a frequency 20 times higher than the first one and the amplitude of 0.2

# plot both signals
plt.plot(ns, s2, 'r')
plt.plot(ns, s1, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('base signals')
plt.show()
```
