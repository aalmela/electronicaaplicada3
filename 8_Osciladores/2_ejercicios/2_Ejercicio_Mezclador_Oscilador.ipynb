### Diseño el oscilador local empleando una configuración Collpits.  

1. Obtener del simulador los parámetros del transistor a la frecuencia de trabajo.  

Emplear el archivo "MRF9411_AC.asc". 

Recordando para la medición de los parametros:

$$i_i = v_i y_{11} + v_o y_{12}$$

$$i_o = v_i y_{21} + v_o y_{22}$$

Entonces haciendo $v_o = 0$ 

$$y_{11} = \frac{i_i}{v_i}$$

$$y_{21} = \frac{i_o}{v_i}$$

Y haciendo $v_i = 0$ 

$$y_{12} = \frac{i_i}{v_o}$$

$$y_{22} = \frac{i_o}{v_o}$$

Nota: medir los parametros a la frecuencia de trabajo y verificar los sentidos de las corrientes (entrantes al cuadripolo)
