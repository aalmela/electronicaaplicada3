# Applying the oscillator of the NE602 in low-power mixer applications


## INTRODUCCIÓN
Para el diseñador de sistemas de RF de baja potencia, el mezclador/oscilador NE602 de Philips Semiconductors proporciona un funcionamiento de mezclador más allá de $500MHz$, un oscilador versátil capaz de funcionar hasta $200MHz$, y una ganancia de conversión, con un consumo total de corriente de sólo $2,5mA$. 

Con una comprensión adecuada de las consideraciones de diseño del oscilador, el NE602 puede ponerse a trabajar rápidamente en muchas aplicaciones. 






## DESCRIPCIÓN

![](fig/SA602_fig1.png)

La figura 1 muestra el circuito equivalente del dispositivo. El chip es en realidad tres subsistemas: Un mezclador de células Gilbert (que proporciona una ganancia de entrada diferencial), un oscilador seguidor del emisor con búfer y la regulación de la corriente y el voltaje de RF. 
La integración completa del sesgo de CC permite una aplicación simple y compacta.
La simplicidad del oscilador permite muchas configuraciones. 
Mientras que el oscilador es simple, el diseño del oscilador no lo es. Este artículo no abordará el equipo...
o de diseño del oscilador, pero algunas pautas prácticas permitirán al diseñador lograr un buen rendimiento con la mínima dificultad.

Los circuitos de los tanques de cristal o de LC pueden ser empleados de manera efectiva. La figura 2 muestra las cuatro configuraciones más comúnmente usadas en su forma más básica.

En cada caso, el Q del tanque afectará los límites de frecuencia superior de la oscilación: el
Cuanto más alto sea el Q, más alta será la frecuencia. 

El NE602 se fabrica con un proceso de 6GHz, pero la resistencia del emisor de la clavija 7 a tierra es nominalmente de 20k. Con una corriente de polarización típica de 0,25mA, se puede lograr una oscilación de 200MHz con un Q alto y una retroalimentación adecuada.

La retroalimentación, por supuesto, depende del Q del tanque. 

En general se acepta que se debe utilizar una cantidad mínima de retroalimentación, de modo que aunque la elección sea totalmente empírica, se puede determinar rápidamente un buen equilibrio entre las características de arranque, la distorsión y la estabilidad de la frecuencia.

## Circuitos de tanques LC


Los tanques LC presentan un desafío un poco mayor para el diseñador. 
Si el Q es demasiado bajo, el oscilador no va a empezar. 

Un truco que ayudará si todo lo demás falla es desviar el Pin 7 al suelo con un 22k
resistencia. En aplicaciones reales esto ha sido efectivo hasta 200MHz con cerámica de alta Q
condensadores y un inductor de tanque de 0.08mH y un Q de 90. Un valor de resistencia más pequeño alterará Sesgo de DC debido a un sesgo de base inadecuado en la entrada del oscilador. Un sesgo externo podría añadirse una resistencia de VCC al Pin 6, pero esto introducirá el ruido de la fuente de alimentación a el espectro de frecuencias. 

La configuración de Hartley (Figura 2D) ofrece simplicidad. Con un condensador variable sintonizando el el Hartley sintonizará un rango muy grande ya que toda la capacitancia es variable.
Tenga en cuenta que el inductor debe estar acoplado al Pin 7 con una baja impedancia
condensador. 

El oscilador Colpitts mostrará un rango de sintonía más pequeño ya que los condensadores de retroalimentación fijos limitan el rango de capacitancia variable; sin embargo, el Colpitts tiene una buena estabilidad de frecuencia con componentes adecuados.
Control de frecuencia sintetizada 

El NE602 puede ser muy eficaz con un sintetizador si se toman las precauciones adecuadas para minimizar la carga del tanque y la introducción de transitorios de conmutación digital en el espectro. La Figura 4 muestra un circuito adecuado para las frecuencias de navegación de los aviones (108-118MHz) con 10.7MHz de FI.

El MOSFET de doble puerta proporciona un alto grado de aislamiento de la conmutación del preescalador
picos. Como se muestra en la Figura 4, el consumo total de corriente del NE602 y
3SK126 es típicamente 3mA. La entrada del MOSFET es del emisor del transistor del oscilador
para evitar cargar el tanque. 

La capacitancia de la puerta 1 del MOSFET en serie con el condensador de acoplamiento de 2pF se añade ligeramente a la relación de capacitancia de retroalimentación. 

El uso de la resistencia de 22k en el Pin 7 ayuda a asegurar la oscilación sin alterar el sesgo de DC.
Para aplicaciones donde el amortiguamiento óptimo del tanque, o la corriente mínima no son obligatorios, o donde la complejidad del circuito debe ser minimizada, los amortiguadores mostrados en la Figura 5 puede ser considerado.

La eficacia del MRF931 (u otros transistores bipolares VHF) dependerá de la frecuencia y el nivel de entrada requerido para el preescalador. 

Un transistor bipolar generalmente proporcionará el menor aislamiento. 

A bajas frecuencias el transistor puede ser usado como un seguidor del emisor, pero por VHF la base de la unión del emisor
comenzará a convertirse en un condensador bidireccional y el buffer se pierde.
El 2N5484 tiene un IDSS de 5mA máximo y el 2SK126 tiene un IDSS de 6mA máximo, lo que los hace adecuados para un bajo número de piezas, modestos buffers de corriente. El aislamiento es bueno.

## Resumen 

El NE602 puede ser un efectivo mezclador de baja potencia en frecuencias de hasta 500MHz con funcionamiento de oscilador hasta 200MHz. 
Todo el sesgo de CC se proporciona internamente al dispositivo, por lo que son posibles diseños muy compactos. 
El sesgo interno establece la corriente continua del oscilador a un nivel relativamente bajo, por lo que el diseñador debe elegir componentes de frecuencia selectiva que no carguen el transistor.

## Analisis oscilador Colpitts

![](fig/collptis_1.png)


![](fig/collpits_2.png)

El Colpitts emplea dos capacitores y un inductor en la red de realimentación.

![ex 2.13](fig/colpittBJT.png)

En la figura se representa el esquema del oscilador Colpitts, redibujado para poner en evidencia la red de realimentación. También en esta figura se indica el punto M, elegido para abrir el lazo de realimentación. 

![ex 2.13](fig/colpittRealimenta.png)


En la siguiente figura se muestra el circuito que resulta después de abrir el lazo y de sustituir el BJT por su circuito equivalente en pequeña señal. Notar que la impedancia de entrada en el punto de inicio es $Z_{in} = r_{\pi}$. 

De circuito se obtienen los valores de la impedancia.

$$ Z_{1} = \frac{r_{\pi}}{1 + j \omega C_{1} r_{\pi} } $$
 
$$ Z_{2} = \frac{1}{j \omega C_{2}} $$

$$ Z     =  j \omega L $$


$$A = -gm \cdot  \frac{Z_{2} \cdot (Z + Z_{1}) }{Z_{1}+Z+Z_{2}}$$

$$ \beta =  \frac{Z_{1}}{Z + Z_{1}}$$

Entonces, para la condición de Barkhausen.

$$ A \beta =   -gm \cdot  \frac{Z_{1} \cdot Z_{2}}{Z_{1}+Z+Z_{2}} = 1$$

Sustituyendo en la ecuación anterior las expresiones correspondientes.


$$ A \beta =   -gm \cdot (\frac{r_{\pi}}{1+j\omega C_{1} r_{\pi}}) \cdot (\frac{1}{j \omega C_{2}}) \cdot  \frac{1}{(\frac{r_{\pi}}{1 + j \omega C_{1} r_{\pi} } )+ (j \omega L) +(\frac{1}{j \omega C_{2}})} = 1$$


$$   \frac{ -gm \cdot r_{\pi} }{(1 + j \omega C_{1} r_{\pi}  -  \omega^2 C_{2} L (1 + j \omega C_{1} r_{\pi}) + j \omega C_{2}) } = 1$$

$$   \frac{ -gm \cdot r_{\pi} }{(1 - \omega^2 C_{2} L) + j \omega r_{\pi} ( C_{2} + C_{1} - \omega^2 C_{2} C_{1} L  ) } = 1$$

Aplicando el criterio de Barkhausen para la fase, $\phi( A\beta)= 0$, la frecuencia de oscilación: 

$$\omega = \sqrt{ \frac{C_{2} + C_{1}}{ C_{2} C_{1} L} }$$


Sustituyendo este resultado en la expresión de $A \beta $ y aplicando el criterio de Barkhausen para el módulo, $|A \beta | > 1$,  obtenemos la condición de arranque 

$$\frac{ -gm \cdot r_{\pi} }{ 1 - \omega^2 C_{2} L }  > 1$$

$$ gm \cdot r_{\pi}  > \frac{C_{1}}{C_{2}} $$
