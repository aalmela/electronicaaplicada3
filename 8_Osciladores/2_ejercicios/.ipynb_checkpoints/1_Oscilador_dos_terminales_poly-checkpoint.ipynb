```python
import warnings

x = np.array([0.0,  0.05, 0.222,  0.637])

y = np.array([0.0, 6.035e-3, 13.84e-3,  1.896e-3])

z = np.polyfit(x, y, 3)
p = np.poly1d(z)


xp = np.linspace(0, 1, 100)
yp = p(xp)

print(p)

```

```python
plt.plot(xp, yp)
plt.plot(x,y,'ro')

plt.ylabel('$i_{o}$ [A]')
plt.xlabel('$v$[V]')
plt.grid()
plt.savefig('fig/dos_term_tras.png')
plt.show()
```

```python
def Id_tunel(V):
    return 0.1414*V-0.4302*V**2+0.3343*V**3

xp = np.linspace(-0.1, 1, 100)
yp = Id_tunel(xp)
plt.plot(xp, yp)


plt.ylabel('$i_{o}$ [A]')
plt.xlabel('$v$[V]')
plt.grid()

plt.show()
```

```python
def IDT(v, diode):
    ID = np.sum([d*v**(n) for n,d in enumerate(diode)])
    return ID



MBD1057 = [0.0 ,0.007358,-0.08607,0.291100, 0.039220,-1.69300,2.09900] 
MBD2057 = [0.0 ,0.011540,-0.12690,0.450900,-0.319000,-1.12400,1.63800]
MBD3057 = [0.0 ,0.013738,-0.15709,0.546242, 0.076587,-3.55975,4.75774]
MBD4057 = [0.0 ,0.017188,-0.19620,0.571800, 1.067800,-7.34250,8.82072] 
MBD5057 = [0.0 ,0.019600,-0.20173,0.483730, 1.599600,-8.45630,9.63780]


vp = np.linspace(0, 0.46, 500)
idt = np.array([IDT(v, MBD2057 ) for v in vp])  
plt.plot(vp, idt )


plt.ylabel('$i_{o}$ [A]')
plt.xlabel('$v$[V]')
plt.grid()

plt.show()
```

```python
g = np.diff(idt)/np.diff(vp)

print(len(g),len(vp))
```

```python
vpp = ((0.24+0.07)/2)
#vs = ((0.24-0.07)/2)
vs = 0.1
plt.plot(vp[1:],g )
print(vp[np.argmin(g)])


plt.plot(vpp,0,'b*')
plt.plot(vs+vpp,0,'*')
plt.plot(vpp-vs,0,'*')

plt.ylabel('$g_{o}$ [mho]')
plt.xlabel('$v$[V]')
plt.grid()

plt.show()
```

```python
ph = np.pi*np.linspace(0,2, 100)

vd = vpp + vs*np.sin(ph)

idt = np.array([IDT(v, MBD2057 ) for v in vd])  

ge = np.diff(idt)/np.diff(vd)


plt.plot(ph,idt )

plt.ylabel('$g_{ph}$ [mho]')
plt.xlabel('$ph$[V]')
plt.grid()

plt.show()


plt.plot(ph[1:],ge )
plt.ylabel('$g_{ph}$ [mho]')
plt.xlabel('$ph$[V]')
plt.grid()

plt.show()
```
