De circuito se obtienen los valores de la impedancia.

$$Z_{1} = \frac{r_{\pi}}{1 + j \omega C_{1} r_{\pi} }$$
 
$$Z_{2} = \frac{1}{j \omega C_{2}}$$

$$Z    =  j \omega L$$


$$A = -gm \cdot  \frac{Z_{2} \cdot (Z + Z_{1}) }{Z_{1}+Z+Z_{2}}$$

$$\beta =  \frac{Z_{1}}{Z + Z_{1}}$$

Entonces, para la condición de Barkhausen.

$$A \beta =   -gm \cdot  \frac{Z_{1} \cdot Z_{2}}{Z_{1}+Z+Z_{2}} = 1$$

Donde

$Z_1 = \frac{r_{11}}{1+j\omega C_{1} r_{11}}$

$Z_2 = \frac{R_{1}}{1+j\omega C_{2} R_{1}}$

$Z   = j \omega L$

Sustituyendo en la ecuación anterior las expresiones correspondientes.


$$A \beta =   -gm \cdot (\frac{r_{11}}{1+j\omega C_{1} r_{11}}) \cdot (\frac{R_1}{1+ j \omega C_{2} R_1}) \cdot  \frac{1}{(\frac{r_{11}}{1 + j \omega C_{1} r_{11} } )+ (j \omega L) +(\frac{R_1}{1+ j \omega C_{2} R_1})} = 1$$





![ex 2.13](fig/oscilador_c.png)

De circuito se obtienen los valores de la impedancia.

$$Z_{A} = \frac{R_A}{1 + j \omega C_{A} R_{A}}$$

$$Z     =  j \omega L $$

$$Z_{B} = \frac{R_B}{1 + j \omega C_{B} R_{B}}$$
 


$$A = -gm \cdot  \frac{Z_{A} \cdot (Z + Z_{B}) }{Z_{A}+Z+Z_{B}}$$

$$\beta =  \frac{Z_{B}}{Z + Z_{B}}$$

Entonces, para la condición de Barkhausen.

$$A \beta =   -gm \cdot  \frac{Z_{A} \cdot Z_{B}}{Z_{A}+Z+Z_{B}} = 1$$

Sustituyendo en la ecuación anterior las expresiones correspondientes.

$$-gm \cdot  \frac{Z_{A} \cdot Z_{B}}{Z_{A}+Z+Z_{B}} = 1$$

$$Z_{A} \cdot Z_{B} = \frac{R_A}{1 + j \omega C_{A} R_{A}} \frac{R_B}{1 + j \omega C_{B} R_{B}}$$

$$Z_{A}+Z+Z_{B} = \frac{R_A}{1 + j \omega C_{A} R_{A}}+ j \omega L + \frac{R_B}{1 + j \omega C_{B} R_{B}}$$


$$\frac{-gm R_{A} R_{B}}{( R_A + j \omega C_{B} R_{B} R_A + j \omega L (1 - \omega^2 C_{A} R_{A} C_{B} R_{B} + j (\omega C_{A} R_{A}+ \omega C_{B} R_{B})) + R_B + j \omega C_{A} R_{A} R_B)} = 1$$

$$\frac{-gm R_{A} R_{B}}{(R_A + R_B -  \omega^2 L ( C_{A} R_{A}+ C_{B} R_{B})) + j( \omega L (1-\omega^2 C_{A} R_{A} C_{B} R_{B}) + \omega (C_{A}+C_{B})R_{B} R_A) }=1$$


Haciendo la parte imaginaria cero para $\varphi = 0$

$$\omega (C_{A}+C_{B}) R_{B} R_A +   \omega L -  \omega^3 L C_{A} R_{A} C_{B} R_{B} = 0$$

$$(C_{A}+C_{B}) R_{B} R_A +   L =  \omega^2 L C_{A} R_{A} C_{B} R_{B}$$

$$\frac{C_{A}+C_{B}}{L C_{A} C_{B}}  +  \frac{1}{R_B R_A C_{A} C_{B}} =  \omega^2$$

Despresiando el segundo termino

$$\omega^2 = \frac{C_{A}+C_{B}}{L C_{A} C_{B}}$$
 
Volviendo a la ecuación

$$\frac{-gm R_{A} R_{B}}{(R_A + R_B -  \omega^2 L ( C_{A} R_{A}+ C_{B} R_{B})} > 1$$

$$gm R_{A} R_{B} > (R_A + R_B -  \frac{C_{A}+C_{B}}{C_{A} C_{B}}  ( C_{A} R_{A}+ C_{B} R_{B})$$

$$gm  > \frac{1}{R_A} + \frac{1}{R_B} -  (C_{A}+C_{B})  ( \frac{1}{C_{A} R_{A}} + \frac{1}{C_{B} R_{B}})$$


$$C_s = \frac{C_{A} C_{B}}{C_{A}+C_{B}}$$

$$gm  > \frac{1}{R_A} (1 - \frac{C_A}{C_s}) + \frac{1}{R_B} (1 - \frac{C_B}{C_s})$$


Haciendo $L$ el inductor efectivo se puede calcular como 

$L = L_8 - \frac{1}{wo^2 C_d}$

$L$ debe resonar con $C_s$

$L = \frac{1}{wo^2 C_s}$

Entonces, para este valor de $L$

$C_d(26MHz) = \frac{1}{(L - L_8) wo^2}$

$$L_e = 421.0488 nHy$$

$$C_d = 261.1061 pF$$

3. <span style="color:red">**[1 punto]**</span> Exprese la transconductancia de conversión $gm_c$.

$$i_{D} = i_{DQ} + gm(t) \Delta v_{GS}$$


La transconductancia $gm$ se calcula como 

$$gm(t) = \frac{d i_{DQ}}{d v_{GS}}$$

Donde:
$i_{DQ} = I_{DSS} (1 - \frac{v_{GS}}{V_c})^2$


$$gm(t) = \frac{d i_{DQ}}{d v_{GS}} =  -2 \cdot \frac{I_{DSS}}{V_c} \cdot (1 - \frac{v_{GS}}{V_c})$$

Los siguientes terminos de la serie dependeran de la variación de $\Delta v_{GS}$ y el termino de la serie de Taylor.

La transconductancia instantanea entonces

$$gm =  -2 \frac{I_{DSS}}{V_c} \cdot ( 1 + \frac{( V_{SS}  + V_{OL} \cdot cos(2 \pi f_{OL} t))}{V_c})$$


$$gm =   - \frac{2 I_{DSS}}{V_c} \cdot  ( 1 + \frac{V_{SS}}{V_c} ) -  \frac{2 \cdot I_{DSS}  V_{OL}}{V_c^2}  \cdot cos(2 \pi f_{OL} t)$$

Dado el divisor de entrada de OL 

$V_{OL} = 0.5\,V$

$V_{GS} = 1.5\,V$ 

$$gm = gm_0 + gm_1$$


$$gm = -12.00 mS -1.33 mS cos(2 \pi f_{OL})$$

$gm_c = \frac{gm_1}{2}$

$$gm_c = -0.67 mS$$
