## 1. Obtener el valor de C que maximiza la sensibilidad del modulador $\frac{\Delta \varphi}{\Delta g_m}$
Justifique su respuesta. 


$$I_{Dq} = I_{DSS}(1-\frac {V_{GSq}} {V_{c}} )^2$$  

$$I_{Dq} = 333.33 \mu A$$

$$g_{mq} = \frac{-2 I_{DSS}}{Vc} (1-\frac {V_{GSq}} {V_{c}} )$$

$$g_{mq} = 4 mS$$


$$\Delta \varphi = \frac{\omega_c C}{g_{mq}^2 + (\omega_c C)^2} \Delta g_{m}$$


$$\frac{d \varphi}{d C} =   \frac{g_{mq}^2-(\omega_c C)^2}{(g_{mo}^2 + (\omega_c C)^2)^2} = 0$$

$$g_{mq}^2-(\omega_c C)^2 = 0$$

$$C = \frac{g_{mq}}{ \omega_c}$$

$$C = 3.18 nF$$





## 2. Determinar el valor de $R_1$ que produce un $\Delta f = 50Hz$ a la salida del Tx.

## $\Delta V_{GS}$

$$Z_1 = R_1$$

$$Z_2 = \frac{R_2}{1 + S C_2 R_2}$$

$$A_v = -\frac{Z_2}{Z_1} = -\frac{R_2}{R_1} \frac{1}{1+S C_2 R_2}$$ 

$$A_v = -\frac{R_2}{R_1} \frac{1}{1+j \omega C_2 R_2}$$ 

$$\Delta V_{GS} = -\frac{1}{R_1} \frac{R_2}{1+j \omega_m C_2 R_2} V_m$$

$$\Delta V_{GS} = -\frac{1571.76}{R_1} V_m$$


## $\Delta g_{m}$

$$\Delta g_{m} = -\frac{2 I_{DSS}}{V_c^2} \Delta V_{GS}$$

$$\Delta g_{m} = 2.66\frac{mS}{V} \Delta V_{GS}$$


## $\Delta \varphi$ 

$$\Delta \varphi = \frac{\omega_c C}{g_{mq}^2 + (\omega_c C)^2} \Delta g_{m}$$

$$\Delta \varphi = \frac{1}{2 g_{mq}} \Delta g_{m}$$

$$\Delta \varphi  = 125 \frac{rad}{S} \Delta g_{m}$$


## $\Delta f_m$
$$\varphi = \int \omega_i(t) dt$$ 

$$\varphi = \omega_c t +  \frac{\Delta \omega_m}{\omega_m} \cdot \sin(\omega_m t)$$ 

$$\Delta \varphi = \frac{\Delta \omega_m}{\omega_m} = \frac{\Delta f_m}{f_m}$$

$$\Delta \varphi =  \frac{\Delta 50 Hz}{1 KHz}$$




## Calculando

$$\Delta \varphi =  \frac{\Delta 50 Hz}{1 KHz}$$

$$\Delta \varphi = 50 \times 10^{-3} rad$$


$$\Delta g_{m}  =  \frac{S}{125 rad} \Delta \varphi$$

$$\Delta g_{m}  = 400 \mu S$$

$$\Delta V_{GS} = \frac{V}{2.66 mS} \Delta g_{m}$$

$$\Delta V_{GS} = \frac{V}{2.66 mS} 400 \mu S$$

$$\Delta V_{GS} = 150 mV$$

$$V_{m(p)} = 100 mV$$

$$\frac{\Delta V_{GS}}{V_{m(p)}} = 1.5$$

$$R_1 = -\frac{R_2}{1.5} \frac{1}{1+j \omega_m C_2 R_2}$$

$$R_1 = 1047 \Omega$$

## 4. <span style="color:red">**[1 punto]**</span> Obtener la potencia de salida a la frecuencia de $27 MHz$. Exprese el $gm_c$ (gm de conversión)

El $gm$ se puede expresar como

$$gm = \frac{-2 I_{DSS}}{Vc} (1-\frac{V_{GS}}{Vc})$$

Dado que $V_{GS}$ varia con el tiempo:

$$gm(t) = \frac{-2 I_{DSS}}{Vc} (1-\frac{V_{GS}(t)}{Vc})$$

Expresando $V_{GS}(t)$

$$V_{GS}(t) = 0.878 V + V_{OLG}(t)$$

Donde (se calculan las variaciones, no las fases de la señal)

$$V_{OLG}(t) = V_{OL} \frac{|Z_{in}|}{|Z_{in}|+r_{gOL}}$$

$$|Z_{in}|= \frac{r_{rRF} r_{gOL}}{(r_{rRF}+ r_{gOL})}$$

$$|Z_{in}|=  5000\Omega$$


$$V_{OLG} = V_{OL} \frac{|Z_{in}|}{|Z_{in}|+r_{gOL}}$$ 

$$V_{GS} = V_{OL} + 0.50V cos(\omega_{OL} t)$$

$$V_{GS} = -0.878 V - 0.5V cos(\omega_{OL} t)$$

Dado que la transferencia es proporcional, se puede obtener:

$$gm_{MIN} =  \frac{-2 I_{DSS}}{Vc} (1-\frac{V_{GS_{MAX}}}{Vc})$$

$$V_{GS_{MAX}} = -1.135 V$$

$$gm_{MIN} = 4.32 mS$$



$$gm_{q} =  \frac{-2 I_{DSS}}{Vc} (1-\frac{V_{GS_{q}}}{Vc})$$

$$gm_{q} =  5.65 mS$$


$$gm_{MAX} =  \frac{-2 I_{DSS}}{Vc} (1-\frac{V_{GS_{MIN}}}{Vc})$$

$$V_{GS_{MIN}} = -0.621 V$$

$$gm_{MAX} = 6.992 mS$$


De la excursión entonces:

$$gm_1 =  5.65 mS + 1.432 mS cos(\omega_{OL} t)$$


El $gm_c$ se calcula como

$$gm_c = \frac{gm_1}{2}$$

$$gm_c = 716 \mu S$$

La tensión de entrada dada la fuente $V_{RF}$


$$|Z_{in}|=  5000\Omega$$


$$V_{gRF} = V_{RF} \frac{|Z_{in}|}{|Z_{in}|+r_{RF}}$$ 

$$V_{gRF} = 20mV$$

En la salida del mezclador

La resistencia de carga total a la fuente de corriente $gm$

$r_t = \frac{100K\Omega  100K\Omega}{100K\Omega + 100K\Omega}$

$r_t = 50 K\Omega$

$$v_o = gm_c v_{RF} 50K \Omega$$

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

#FET
Idss = 12e-3
Vc   =  -3
VGS  = -0.878
Vs   = -0.5

def Idq(Vgs, Idss, vc):
    return Idss *(1-(Vgs/Vc))**2

# gm
def gmVgs(Vgs, Idss, vc):
    return -2 * Idss/Vc *(1-(Vgs/Vc))

# Qc
def SS(f, fo, Qc):
    return 1/(1+(Qc*(f/fo-fo/f) )**2 )**0.5
    
Vgso = -2.5   
print('V_GSq= {:1.3f} V'.format(Vgso))

gmq   = gmVgs(Vgso,Idss,Vc)
print('$g_mq = {:1.3f} mS$'.format(gmq/1e-3))


thi = np.arange(-np.pi,np.pi,2*np.pi/8)

vgs = VGS + Vs*np.cos(thi)

idd = np.array([Idq(v, Idss, Vc) for v in vgs])



plt.plot(thi, idd)
plt.plot(thi, idd, 'ro')
plt.ylabel('$i_{o}$ [A]')
plt.xlabel('${\ theta}$')

plt.grid()
plt.show()

for i,p in enumerate(thi):
    if p  >= 0 :
        print('|{:1.2f}| {:1.2f}| {:1.2e}| {:1.2f}| {:1.2f} | {:1.2f} |'.format( 
            p*180/np.pi, vgs[i], idd[i], np.cos(p) , np.cos(2*p), np.cos(3*p)   ) ) 
        
        
ns = len(thi)        
# Continua
idco = 1/ns * np.sum(idd) 


# primer arminico
idc1 = 2/ns * np.sum([idd[i]*np.cos(1*thi[i]) for i in range(ns)] ) 

# segundo arminico
idc2 = 2/ns * np.sum([idd[i]*np.cos(2*thi[i]) for i in range(ns)] ) 

# tercero arminico
idc3 = 2/ns * np.sum([idd[i]*np.cos(3*thi[i]) for i in range(ns)] ) 

print('idco= {:.2e} A'.format(idco))
print('idc1= {:.2e} A'.format(idc1))  

print('idc2= {:.2e} A'.format(idc2))  


print('idc3= {:.2e} A'.format(idc3))  


vdc2 =  idc2 * 50e3*SS(28.4e6,27e6,150)

print('vdc2= {:.2e} V'.format(vdc2))  
```
