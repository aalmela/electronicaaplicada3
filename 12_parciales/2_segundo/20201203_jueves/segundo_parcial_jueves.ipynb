La corriente $i_i$ entonces

$$i_i = i_1 + i_2 = v_i \cdot \frac{1}{(R+\frac{1}{S C})} + g_m \cdot v_i \cdot \frac{R}{(R+\frac{1}{S C})}$$


La admintacia de entrada $Y_i$ es 

$$Y_i = \frac{i_i}{v_i} =  \frac{1}{(R+\frac{1}{S C})} + g_m \cdot \frac{R}{(R+\frac{1}{S C})}$$

$$Y_i =   \frac{j\omega C \cdot (1 + g_m R) }{( j\omega C R+1)}$$

$$C_e =  C \cdot (1 + g_m R)$$

De circuito se obtienen los valores de la impedancia.

$$Z_{A} = \frac{R_A}{1 + j \omega C_{A} R_{A}}$$

$$Z     =  j \omega L $$

$$Z_{B} = \frac{R_B}{1 + j \omega C_{B} R_{B}}$$
 


$$A = -gm \cdot  \frac{Z_{A} \cdot (Z + Z_{B}) }{Z_{A}+Z+Z_{B}}$$

$$\beta =  \frac{Z_{B}}{Z + Z_{B}}$$

Entonces, para la condición de Barkhausen.

$$A \beta =   -gm \cdot  \frac{Z_{A} \cdot Z_{B}}{Z_{A}+Z+Z_{B}} = 1$$

Sustituyendo en la ecuación anterior las expresiones correspondientes.

$$-gm \cdot  \frac{Z_{A} \cdot Z_{B}}{Z_{A}+Z+Z_{B}} = 1$$

$$Z_{A} \cdot Z_{B} = \frac{R_A}{1 + j \omega C_{A} R_{A}} \frac{R_B}{1 + j \omega C_{B} R_{B}}$$

$$Z_{A}+Z+Z_{B} = \frac{R_A}{1 + j \omega C_{A} R_{A}}+ j \omega L + \frac{R_B}{1 + j \omega C_{B} R_{B}}$$


$$\frac{-gm R_{A} R_{B}}{( R_A + j \omega C_{B} R_{B} R_A + j \omega L (1 - \omega^2 C_{A} R_{A} C_{B} R_{B} + j (\omega C_{A} R_{A}+ \omega C_{B} R_{B})) + R_B + j \omega C_{A} R_{A} R_B)} = 1$$

$$\frac{-gm R_{A} R_{B}}{(R_A + R_B -  \omega^2 L ( C_{A} R_{A}+ C_{B} R_{B})) + j( \omega L (1-\omega^2 C_{A} R_{A} C_{B} R_{B}) + \omega (C_{A}+C_{B})R_{B} R_A) }=1$$


Haciendo la parte imaginaria cero para $\varphi = 0$

$$\omega (C_{A}+C_{B}) R_{B} R_A +   \omega L -  \omega^3 L C_{A} R_{A} C_{B} R_{B} = 0$$

$$(C_{A}+C_{B}) R_{B} R_A +   L =  \omega^2 L C_{A} R_{A} C_{B} R_{B}$$

$$\frac{C_{A}+C_{B}}{L C_{A} C_{B}}  +  \frac{1}{R_B R_A C_{A} C_{B}} =  \omega^2$$

Despresiando el segundo termino

$$\omega^2 = \frac{C_{A}+C_{B}}{L C_{A} C_{B}}$$
 
Volviendo a la ecuación

$$\frac{-gm R_{A} R_{B}}{(R_A + R_B -  \omega^2 L ( C_{A} R_{A}+ C_{B} R_{B})} > 1$$

$$gm R_{A} R_{B} > (R_A + R_B -  \frac{C_{A}+C_{B}}{C_{A} C_{B}}  ( C_{A} R_{A}+ C_{B} R_{B})$$

$$gm  > \frac{1}{R_A} + \frac{1}{R_B} -  (C_{A}+C_{B})  ( \frac{1}{C_{A} R_{A}} + \frac{1}{C_{B} R_{B}})$$


$$C_s = \frac{C_{A} C_{B}}{C_{A}+C_{B}}$$

$b_n = \frac{2}{2 \pi} \int_{-\pi}^{\pi}  V_{GS}(\theta) sin(\theta n) d\theta$

Donde encontramos que para una señal cuadrada (Square) con amplitud $A$ pico a pico:

$a_n = \frac{2 V_{GSpp}}{\pi n} sin( \frac{\pi}{2} n)$

```python
def IcqFET(vgs,IDSS,VC):
    return IDSS*(1-vgs/VC)**2

def gmFET(vgs,IDSS,VC):
    return -(2*IDSS/VC)*(1-vgs/VC)



## oscilador
VCC = 9

frf = 72e6
wrf = frf*2*np.pi

C1 = 100e-12
C2 = 100e-12
Re = 1200
RB1 = 5e3
RB2 = 5e3
VBE = 0.7

Icq = ((VCC * RB2)/(RB1+RB2)-VBE)/Re 

# Transistor
y11 = (1e-3/10) + 1j* 4.5e-3
y22 = (1e-3/100) + 1j* 4.5e-3
VT  = 26e-3
gm  = Icq/VT

r11 = 1/(y11.real)
c11 = y11.imag/(2*np.pi*frf)
r22 = 1/(y11.real)
c22 = y22.imag/(2*np.pi*frf)



## Fet reactancia
R   = 22
C   = 1e-12 
# Transistor
IDSS = 12e-3
VC   = -3
ID   = 6e-3


```

```python
## Calculo de Rs
VG   = 0
VGSq = ((ID/IDSS)**0.5-1)*VC

VS   = VG - VGSq  
print('VGSq ={:1.2f} V'.format(VGSq) )
print('VS ={:1.2f} V'.format(VS) )

Rg = (-VS)/ID

print('Rg ={:1.2f} ohm'.format(Rg) )
```

```python
### 
gmq = gmFET(VGSq,IDSS,VC) 

Ce  = C*(1+gmq*R)

print('Ce ={:1.3f} pHy'.format(Ce/1e-12) )
```

```python
### OSCILADOR
RA = r11
CA = (C1+c22)

RB = r22*Re/(r22+Re)  
CB = (C2+c11)+Ce

Cs = CA*CB/(CA+CB)

L = (1)/(wrf**2*Cs)

print('L ={:1.2f} nHy'.format(L/1e-9) )

```

```python
DgmDVgs = -2*IDSS/VC**2
DceDgm  = -R*C
Dc2Dce  = 1
DcDc2   = CA**2/(CA+CB)**2
DfDc    = 1/(4*np.pi*L**0.5) * Cs**(-3/2)

print('Dgm/DVgs = {:1.2e}'.format(DgmDVgs))
print('DCe/Dgm = {:1.2e}'.format(DceDgm))
print('DC2/DCe = {:1.2e}'.format(Dc2Dce))
print('DC/DC2 = {:1.2e}'.format(DcDc2))
print('Df/DC = {:1.2e}'.format(DfDc))


DfDVgs = DfDc* DcDc2*Dc2Dce*DceDgm*DgmDVgs

print('Df/DVgs = {:1.2e}'.format(DfDVgs))

```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy                  # http://scipy.org/
from scipy import signal

from IPython.display import Image  
%matplotlib inline
params = {'legend.fontsize': 24,
          'figure.figsize': (8, 8),
         'axes.labelsize': 24,
         'axes.titlesize': 24,
         'xtick.labelsize':24,
         'ytick.labelsize':24}
plt.rcParams.update(params) 

#Calculos 
#parametros de la senal 
fc     =  72e6 #Hz
th_c   =  np.pi


#parametros del filtro de entrada y de salida 
Qo = 50
Qc = 20

#parametros del transistor
Vds_min = 1  #V
Vgs_max = 8 #V
Vth     = 3 #V

# Carga
Zl      = 50 #ohms

VDD     = 9 #V

### Para los calculos!
#tiempo de un periodo 
Tc  = 1/fc

# muestreo (empleando 24 ya que corresponde a el muestreo del empleado comunmente en el metodo de los trapecios)
ns  =  24       
fs  =  fc*ns
ds  =  1/(fs)



Vs  = (Vgs_max-Vth)/(1-np.cos(th_c/2))
VGG = Vgs_max - Vs
print('$Vs  = {:.2f}$'.format(Vs))
print('$VGG = {:.2f}$'.format(VGG))

def vgsi(VGG,Vs,th):
    return VGG+Vs*np.cos(th)


def idi(v):
    i =  0.             if (v < 3) else 0
    i = (1.6/5)*(v-3) if (v >= 3  and v < 8) else i
    i = (1.6) if (v >= 8  ) else i
    return i


thi = np.arange(-np.pi,np.pi,2*np.pi/24)

vgs = vgsi(VGG,Vs,thi)

idd = np.array([idi(v) for v in vgs])



plt.plot(thi, idd)
plt.plot(thi, idd, 'ro')
plt.ylabel('$i_{o}$ [A]')
plt.xlabel('${\ theta}$')

plt.grid()
plt.savefig('fig/id_trans.png')
plt.show()


```
