# Densidad Espectral de Potencia

La densidad espectral de potencia, $S_x(\omega)$ describe la distribución de la potencia respecto a la frecuencia, y esta dada por

$$\bar{x(t)^2} = lim_{T -> \inf} \frac{1}{T} \int_{-T/2}^{T/2} |x(t)|^2dt$$

Es el promedio o la media del cuadrado de la señal $x(t)$. 

Esta cantidad se llama valor cuadratico medio de la señal. 

$S_x(\omega)$ tiene unidad de $[\frac{W}{Hz}]$

La integral de densidad espectral de potencia, nos da la potencia de la señal.

$$P = \frac{1}{2 \pi} \int_{-\inf}^{\inf} S_x(w) dw$$



## Señales definidas en potencia 


Una señal $x(t)$ es definida en potencia si su potencia media es finita, i.e, $0 < P_x < \infty$ y por tanto, su energía media es infinita, $E_x=\infty$.

La Densidad Espectral de Potencia se calcula usando el Teorema de Wiener-Khinchin el cual relaciona la Densidad Espectral de Potencia con la transformada de Fourier de la función de autocorrelación

$$S_{xx}(f)=\mbox{TF}\left\{R_{xx}(\tau)\right\}=\int_{-\infty}^{\infty}\,R_{xx}(\tau)\,e^{-2\pi\,if\tau}\,d \tau$$

expresado en $[W/Hz]$

donde $\mbox{TF}$ significa Transformada de Fourier y $R_{xx}$ es la función de autocorrelación de $x(t)$. 

El valor $S_{xx}(0)$ es la potencia de la componente continua (DC) de la señal. La integral de esta función en todo el eje $f$ es el valor de la potencia total de la señal $x(t)$

$$P=\int_{-\infty}^{+\infty} S_{xx}(f) df$$

Usando el concepto de correlación cruzada es posible definir también la '''densidad espectral cruzada'''.

$$S_{xy}(f)=\mbox{TF}\left\{R_{xy}(\tau)\right\}$$





## Autocorrelación

En el campo de procesamiento de señales, dada una señal temporal $x(t)$, la autocorrelación continua $R_x(\tau)$ es la  correlación cruzada  de $x(t)$ consigo mismo tras un desfase $\tau$, y se define como:

$$ R_x(\tau)= x^*(-\tau) \circ x(\tau) = \int_{-\infty}^{\infty} x(t+\tau)x^*(t)\, dt = \int_{-\infty}^{\infty} x(t)x^*(t-\tau)\, dt$$

donde $x^*$ representa el  números complejos  y el círculo representa una  convolución . 
Para una función real, $x^* = x$.
