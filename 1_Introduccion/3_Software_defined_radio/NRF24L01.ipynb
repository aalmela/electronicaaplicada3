# NRF24L01 + RFX2401

## Introducción 

El módulo transceptor nRF24L01 + (plus) es muy económico en términos de módulos RF, lo que lo convierte en una de las opciones de comunicación de datos más económicas que puede obtener.

Este documento es una introducción del kit  nrf24L01 + pa + lna;  que viene con un chip amplificador de seal, para aumentar el alcance de la señal. 
Con este módulo podemos lograr alcances de hasta 300m en exteriores (sin obstáculos) transmitiendo a 2mbps contra 10 a 30M que logra el modulo sin amplificador. 
En interiores podemos transmitir a la misma velocidad a distancias de hasta 20m y si reducimos la velocidad de transmisión logramos mayor alcance todavía.


El nRF24L01+ es un transceptor de 2.4GHz de protocolo de banda base integrado (Enhanced ShockBurst ™), adecuado para aplicaciones inalámbricas de ultra baja potencia. 
El nRF24L01+ está diseñado para funcionar en la banda de frecuencia ISM mundial a 2.400 - 2.4835 GHz. 

Para diseñar un sistema de radio con el nRF24L01+, simplemente necesita un MCU (microcontrolador) y algunos componentes pasivos externos. Puede operar y configurar el nRF24L01 + a través de una interfaz periférica en serie (SPI). 

Los registros, al que se puede acceder a través del SPI, permiten la configuración en el nRF24L01+ y está disponible en todos los modos de operación del chip. El protocolo de banda base integrado (Enhanced ShockBurst ™) se basa en la comunicación de paquetes y admite varios modos, desde operación manual hasta operación avanzada de protocolo autónomo. 
Los FIFO internos aseguran un flujo de datos fluido entre la interfaz de radio y la MCU del sistema. 
El Shock-Burst mejorado reduce el costo del sistema al manejar todas las operaciones de capa de enlace de alta velocidad. La interfaz de radio utiliza la modulación GFSK. 
Tiene parámetros configurables por el usuario como el canal de frecuencia, la potencia de salida y la velocidad de datos del aire. 
El nRF24L01+ admite una velocidad de transmisión de datos de 250 kbps, 1 Mbps y 2Mbps. La alta velocidad de datos de aire combinada con dos modos de ahorro de energía hacen que el nRF24L01+ sea muy adecuado para diseños de ultra bajo consumo. nRF24L01 + es compatible con nRF24L01 y compatible con nRF2401A, nRF2402, nRF24E1 y nRF24E2. 
La intermodulación y los valores de bloqueo de banda ancha en nRF24L01 + se han mejorado mucho en comparación con el nRF24L01 y la adición de filtrado interno a nRF24L01 + ha mejorado los márgenes para cumplir con los estándares regulatorios de RF. Los reguladores de voltaje internos aseguran una alta relación de rechazo de suministro de energía (PSRR) rango de suministro

## Radio Frecuencia

El módulo transceptor nRF24L01 + está diseñado para operar en la banda de frecuencia ISM mundial de 2.4 GHz y utiliza la modulación GFSK para la transmisión de datos. La velocidad de transferencia de datos puede ser de 250 kbps, 1 Mbps y 2 Mbps.

### ¿Qué es la banda ISM de 2.4 GHz?
La banda de 2,4 GHz es una de las bandas industriales, científicas y médicas (ISM) reservadas internacionalmente para el uso de dispositivos de baja potencia sin licencia. Algunos ejemplos son los teléfonos inalámbricos, los dispositivos Bluetooth, los dispositivos de comunicación de campo cercano (NFC) y las redes de computadoras inalámbricas (WiFi) que usan las frecuencias ISM.

## Antena

Salida de antena Los pines de salida ANT1 y ANT2 proporcionan una salida RF balanceada a la antena. 
Los pines deben tener una ruta de CC a VDD_PA, ya sea a través de un estrangulador de RF o a través del punto central en una antena dipolo equilibrada. 
Se recomienda una carga de 15Ω + j88Ω para la potencia de salida máxima (0dBm). 
Se puede obtener una impedancia de carga más baja (por ejemplo, 50Ω) instalando una red de coincidencia simple entre la carga y ANT1 y ANT2.

### Power Amplifier y Low Noise Amplifier

El PA significa amplificador de potencia. Simplemente aumenta la potencia de la señal que se transmite desde el chip nRF24L01 +. Mientras que, LNA significa amplificador de bajo ruido. La función de la LNA es tomar la señal extremadamente débil y pequeña de la antena (generalmente del orden de microvoltios o por debajo de -100 dBm) y amplificarla a un nivel más útil (generalmente de 0.5 a 1V).

<img src="fig/RFX2401C_PA_LNA.png">

El amplificador de bajo ruido (LNA) de la ruta de recepción y el amplificador de potencia (PA) de la ruta de transmisión se conectan a la antena a través de un duplexor, que separa las dos señales y evita que la salida de PA relativamente potente sobrecargue la entrada de LNA sensible.

# RFX2401C

El RFX2401C es un RFeIC (Circuito Integrado FrontEnd de RF) completamente integrado, de un solo chip que incorpora toda la funcionalidad de RF necesaria para IEEE802.15.4/ZigBee, red inalámbrica de sensores y cualquier otro sistema inalámbrico en 2.4GHz Banda ISM. 

La arquitectura RFX2401C integra los circuitos de conmutación PA, LNA, transmisión y recepción, la red de correspondencia asociada y el filtro de armónicos, todo en un dispositivo de chip único CMOS. 
Las aplicaciones típicas de alta potencia incluyen la automatización doméstica e industrial, la energía inteligente y RF4CE, entre otras. 
Combinando un rendimiento superior, alta sensibilidad y eficiencia, bajo ruido, factor de forma pequeño y bajo costo, RFX2401C es la solución perfecta para aplicaciones que requieren un rango extendido y ancho de banda. RFX2401C tiene una lógica de control CMOS simple y de bajo voltaje, y requiere componentes externos mínimos para la implementación del sistema.
