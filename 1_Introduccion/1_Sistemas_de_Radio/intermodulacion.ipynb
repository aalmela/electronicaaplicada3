```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq
from time import time

%matplotlib inline

params = {'legend.fontsize': 24,
          'figure.figsize': (15, 5),
         'axes.labelsize': 24,
         'axes.titlesize': 24,
         'xtick.labelsize':24,
         'ytick.labelsize':24}
plt.rcParams.update(params)



def Salida(x):
    return -0.0001811* x**6 + 0.008696* x**5 - 0.1528* x**4 + 1.152* x**3 - 3.165* x**2 + 3.003* x**1 - 0.432




volt  = np.arange(2, 10, 0.01)


plt.plot(volt, (Salida(volt) ))

plt.ylabel('salida [V]')
plt.xlabel('entrada[us]')
plt.show()
```

```python
f1   = 100 
A1   = 1   

f2   = 120
A2   = 1


# defino el sampleo
Fs = 10*(f1+f2) #Sps 
dt = 1/Fs
# tiempo maximo del analisis 
T = 10/f1 
# vector de tiempo 
time  = np.arange(0, T, dt)


# seales
S1   = A1*(np.cos(2*np.pi*f1*time) )
S2   = A2*(np.cos(2*np.pi*f2*time) )
Si    = (S1 + S2) +5

plt.plot(time, Si)
plt.ylabel('Si [V]')
plt.xlabel('time[us]')
plt.show()

# Salida    
So = Salida(Si)

plt.plot(time, So)
plt.ylabel('So [V]')
plt.xlabel('time[us]')
plt.show()
```

```python
# FFT

Fam = fft(Si) / len(Si)  # Normalizada

frq = fftfreq(len(Si), dt)  # Recuperamos las frecuencias

plt.stem(frq, 20*np.log(np.abs(Fam)/1e-3 ) )  

plt.ylim(-1, 200)
plt.xlim(-2,500)

plt.xlabel('Frecuencia (MHz)')
plt.ylabel('Abs($Y$)')
plt.grid()
plt.show()
```

```python
# FFT

Fam = fft(So) / len(So)  # Normalizada

frq = fftfreq(len(So), dt)  # Recuperamos las frecuencias

plt.stem(frq, 20*np.log(np.abs(Fam)/1e-3 ) )  

plt.ylim(-1, 200)
plt.xlim(-2,500)

plt.xlabel('Frecuencia (MHz)')
plt.ylabel('Abs($Y$)')
plt.grid()
plt.show()
```
