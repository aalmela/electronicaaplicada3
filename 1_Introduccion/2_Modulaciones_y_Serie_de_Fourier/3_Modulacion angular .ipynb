# Modulacion angular.


En las modulaciones angulares, es la fase  o la frecuencia de la portadora la que es modulada.

$$F_{c}(t) = A_c \cos(\omega_c t + \theta(t))$$ 

$\theta_{i}( t ) = \omega_c t + \theta(t)$ : Fase instantánea total de la señal

Los tipos de modulaciones angulares son:

FM (Frequency Modulation)

PM (Phase Modulation)

¿Como se relacionan la frecuencia y la fase de una señal?
La frecuencia es la derivada de la fase!

$\omega_{i}(t) = \frac{\partial \theta _{i}(t)}{\partial t}$

si $\theta = sin(\omega_m t)$

$\omega_{i}(t) = \omega_{m} \frac{cos(\omega_{m}t)$


$f_{i}(t)=\frac{\omega_{m}}{2\pi} \frac{cos(\omega_{m}t)$ 

Igualmente, podemos decir que la fase, es la integral de la frecuencia

$\begin{align}
  & \omega _{i}\left( t \right)=\frac{\partial \theta _{i}\left( t \right)}{\partial t}\Leftrightarrow \theta _{i}\left( t \right)=\int_{-\infty }^{t}{\omega _{i}\left( \lambda  \right)\partial \lambda } \\ 
 & f_{i}\left( t \right)=\frac{1}{2\pi }\frac{\partial \theta _{i}\left( t \right)}{\partial t}\Leftrightarrow \theta _{i}\left( t \right)=2\pi \int_{-\infty }^{t}{f_{i}\left( \lambda  \right)\partial \lambda } \\ 
\end{align}$


## Moduladores y demoduladores de fase y de frecuencia

Un modulador de fase es un circuito en el cual la portadora varía de tal manera que su fase instantánea es proporcional a la señal modulante. 
La portadora no modulada es una sinusoide de frecuencia única y se llama comúnmente la frecuencia de reposo. 
Un modulador de frecuencia (frecuentemente llamado un desviador de frecuencia), es un circuito en el cual la portadora varía, de tal manera, que su fase instantánea es proporcional a la integral de la señal modulante. 

Por lo tanto, con un modulador de frecuencia, si la señal modulante $v(t)$ es diferenciada, antes de ser aplicada al modulador, la desviación de fase instantánea es proporcional a la integral de  $v(t)$ o, en otras palabras, proporcional a $v'(t)$ porque $v'(t) = v(t)$ 

De manera semejante, un diferenciador que precede a un modulador de FM produce una onda de salida en la cual la desviación de fase es proporcional a la señal modulante y es, por lo tanto, equivalente a un modulador de fase. 

Son posibles varias equivalencias interesantes. 

Por ejemplo, un demodulador de frecuencia, seguido por un integrador es equivalente a un demodulador de fase. 

Cuatro equivalencias comúnmente usadas son mencionadas a continuación e ilustradas.

1. Modulador de PM = diferenciador seguido por un modulador FM

2. Demodulador de PM = un demodulador de FM seguido por un integrador

3. Modulador de FM = integrador seguido por un modulador de PM

4. Demodulador de FM = demodulador de PM seguido por un diferenciador


## FM (Frequency Modulation) 

La modulación de frecuencia utiliza la señal de información, $V_m(t)$, para variar la frecuencia portadora dentro de un pequeño rango sobre su valor original. Aquí están las tres señales en forma matemática:

$$F_{FM}(t) = A_c\cos( (\omega_c t + k_{FM} \cdot V_m(t) ))$$

Siendo $V_m(t) = A_m cos(\omega_m t)$

$$F_{FM}(t) = A_c\cos( (\omega_c + k_{FM} \cdot A_m cos(\omega_m t) ) )$$

donde $k_{FM} \cdot A_m$ se denomina desviación instantánea, $\theta(t)$

La máxima desviación de fase se llama índice de modulación. 
Una diferencia importante, entre la modulación en frecuencia (FM )y fase (PM), es la manera en que se define el índice de modulación. 

Para PM, el índice de modulación es proporcional a la amplitud de la señal modulante, independientemente de su frecuencia. 
El índice de modulación para una portadora de fase modulada se muestra matemáticamente como

$$m_{PM} = K_{PM} A_m$$

Para FM, el índice de modulación es directamente proporcional a la amplitud de la señal modulante e inversamente proporcional a su frecuencia y se muestra matemáticamente
como

$$m_{FM} = \frac{K_{FM} A_m}{\omega_m}$$




## Desviación en frecuencia $Df$

Para las modulaciones en frecuencia FM se varia la frecuencia angular $\theta_i(t)$

$$\theta _{i}(t) = (\omega_c + k_{FM} V_m(t) ) \cdot t $$

$$k_{FM} = \frac{2 \pi Df}{V_{mo}}$$ 

Se ha sustituido el término de frecuencia portadora, por una frecuencia variable en el tiempo. 
También hemos introducido un nuevo término: $Df$, la desviación de frecuencia máxima. 
De esta forma, deberías ser capaz de ver que el término de frecuencia portadora: 

$$f_c + \frac{2 \pi Df}{V_{mo}} V_m(t)$$ 

ahora varía entre los extremos de $f_c - Df$ y $f_c + Df$. 

La interpretación de Df se hace clara: es la más alejada de la frecuencia original que puede estar la señal de FM. 
A veces se denomina "oscilación" de la frecuencia.

## Indice de modulación $\beta$

También podemos definir un índice de modulación para FM, análogo al de AM:

$\beta = \frac{Df}{fm}$ , donde $fm$ es la máxima frecuencia de modulación utilizada.

La interpretación más simple del índice de modulación, $\beta$, es como una medida de la desviación de la frecuencia máxima, $Df$. 

En otras palabras, $\beta$ representa una forma de expresar la frecuencia de desviación de pico como un múltiplo de la máxima frecuencia de modulación, $fm$, es decir, $Df = \beta fm$.

$$F_{fm}= A_c \cos[w_c t + k_f \int f{(t)} dt]$$

si $f_m(t) = \Delta v_m \cos (\omega_m \cdot t)$

$$F_{fm}= \cos[w_c t + k_f \int \Delta v_m \cos (\omega_m \cdot t) dt]$$

$$F_{fm}= \cos[w_c t + \frac{k_f  \Delta v_m}{\omega_m} \sin (\omega_m \cdot t)]$$




## Ejemplo: Radio FM

La radio FM utiliza la modulación de frecuencia. 

La banda de frecuencia de la radio FM es de unos $88$ a $108 MHz$. 
La señal de información es música y voz que cae en el espectro de audio. 
La gama completa del espectro de audio va de $20$ a $20.000 Hz$, pero la radio FM limita la frecuencia de modulación superior a $15 kHz$ (véase la radio AM que limita la frecuencia superior a $5 kHz$). 

Supongamos en la radio FM que la señal de audio que se va a transmitir se encuentra entre $20$ y $15.000 Hz$. 
Si el sistema de FM utilizara un índice de modulación máximo, $\beta = 5.0$, entonces la frecuencia "oscilaría" un máximo de $5 \times 15 kHz = 75 kHz$ por encima y por debajo de la frecuencia portadora.

Aunque, parte de la señal puede perderse por encima de los $15 kHz$, la mayoría de las personas no pueden oírla de todas formas, por lo que hay poca pérdida de fidelidad. La radio FM puede ser apropiadamente llamada "alta fidelidad".

Si los transmisores de FM utilizan un índice de modulación máximo de alrededor de $5.0$, entonces el ancho de banda resultante es de $180 kHz$ (aproximadamente $0.2 MHz$). 
La FCC asigna las estaciones ) separadas por $0,2 MHz$ para evitar que las señales se superpongan. 
Si se llenara la banda de FM con estaciones, se podría obtener $108 - 88 / .2 = 100$ estaciones, aproximadamente el mismo número que la radio AM (107). 

La radio FM se emite en estéreo, lo que significa dos canales de información. 
En la práctica, generan tres señales antes de aplicar la modulación:

    la señal L + R (izquierda + derecha) en el rango de $50$ a $15.000 Hz$.
    
    una portadora piloto de $19 kHz$.
    
    la señal L-R centrada en una portadora piloto de $38 kHz$ (que se suprime) que se extiende de $23 a 53$ kHz. 

Así pues, la señal de información tiene en realidad una frecuencia de modulación máxima de $53 kHz$, lo que requiere una reducción del índice de modulación a cerca de $1,0$ para mantener el ancho de banda total de la señal en torno a los $200 kHz$.


## Porcentaje de modulación. 

El porcentaje de modulación para una onda de modulación angular se determina de diferente manera que con una onda modulada en amplitud. 

Con la modulación angular, el porcentaje de modulación simplemente es la relación de la desviación de frecuencia realmente producida a la máxima desviación de frecuencia permitida por la ley establecida en forma porcentual. 

Matemáticamente, el porcentaje de modulación es 

$$\% modulación = \frac{\Delta f}{\Delta f_{max}} \times 100 $$

Por ejemplo, en Argentina, la Enacom limita la desviación de frecuencia para transmisores de la banda de radiodifusión comercial de FM a $\pm 75 KHz$.
Si una señal modulante produce $\pm 50 KHz$ de desviación de frecuencia, el porcentaje de modulación es 

$$\% modulación = \frac{50 kHz}{75 kHz} \times 100 = 67 \%$$


## Ancho de banda

### Aproximación de Carson 

Como ya hemos demostrado, el ancho de banda de una señal de FM se puede predecir usando:

$BW = 2 (\beta + 1 ) fm$

donde $\beta$ es el índice de modulación y $fm$ es la máxima frecuencia de modulación utilizada.

La radio FM tiene un ancho de banda significativamente mayor que la radio AM, pero la banda de radio FM también es mayor. La combinación mantiene el número de canales disponibles más o menos igual.

El ancho de banda de una señal FM tiene una dependencia más complicada que en el caso de la AM (recuerde que el ancho de banda de las señales AM depende sólo de la máxima frecuencia de modulación). 
En FM, tanto el índice de modulación como la frecuencia de modulación afectan el ancho de banda. 
A medida que la información se hace más fuerte, el ancho de banda también crece.



```python
import numpy as np
import matplotlib.pyplot as plt

import scipy                  # http://scipy.org/
from scipy.fftpack import fft, fftfreq
from time import time



%matplotlib inline

params = {'legend.fontsize': 24,
          'figure.figsize': (15, 5),
         'axes.labelsize': 24,
         'axes.titlesize': 24,
         'xtick.labelsize':24,
         'ytick.labelsize':24}
plt.rcParams.update(params)


# Ejemplo FM en el tiempo
# modulante
fm   = 10 #Hz
Am   = 1 

# carrier
fc   = 100
beta = 2
Ac   = 1

# defino el sampleo
Fs = 20*fc #Sps 
dt = 1/Fs
# tiempo maximo del analisis 
T = 2/fm 
# vector de tiempo 
time  = np.arange(0, T, dt)


# modulante
Sm   = Am*np.cos(2*np.pi*fm*time) 
# modulada
Sfm   = Ac*np.cos(2*np.pi*fc*time + beta * Sm  ) 


plt.plot(time, Sm)
plt.plot(time, Sfm)

plt.ylabel('Sm [V]')
plt.xlabel('time[us]')
plt.show()
```

```python
# FFT

Ffm = fft(Sfm) / len(Sfm)  # Normalizada

frq = fftfreq(len(Sfm), dt)  # Recuperamos las frecuencias

plt.stem(frq, np.abs(Ffm) )  

#plt.xlim(-1.20, 1.20)
plt.xlabel('Frecuencia (MHz)')
plt.ylabel('Abs($Y$)')
plt.grid()
plt.show()
```

## Banda Ancha - Modulación por una sinusoide de frecuencia única.

El análisis de frecuencia de una onda modulada angular, por una sinusoide de frecuencia única, produce una desviación pico de fase de m radianes, en donde m es el índice de modulación. 
Nuevamente, para una modulación en frecuencia igual a $\omega_m$, $F(t)$ se escribe como


$$F_{FM}(t) = A_c \cos( (\omega_c  \cdot t + k_{FM} \cdot V_m(t)))$$

Siendo $V_m(t) = sin(\omega_m t)$

$$F_{FM}(t) = A_c \cos( (\omega_c \cdot t+   m sin(2 \pi f_m t) ))$$


$$F_{FM}(t) = A_c [ cos(\omega_c t) cos(m sin(\omega_m t)) - sin(\omega_c t) sin(m sin(\omega_c t))]$$

Empleando la identidad de Bessel 

$$cos(\alpha + m cos(\beta)) = \sum_{n=-\infty}^{\infty} J_n(m) cos(\alpha+ n \beta +\frac{n\pi}{2})$$

$J_n(m)$ es la función Bessel de primera clase de enésimo orden con argumento $m$. 

Si la identidad se aplica, podemos escribir 

$$F_{FM}(t) = A_c \sum J_n(m) cos( \omega_c \cdot t +n \omega_m t+ \frac{n \pi}{2})$$




```python
import matplotlib.pyplot as plt
from scipy.fftpack import fft
import numpy as np
from math import pi
import pylab as py
import scipy.special as sp
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.jv.html#scipy.special.jv

from ipywidgets import *
%matplotlib inline
plt.close('all')
plt.rcParams['figure.figsize'] = [20, 6]


Fs=1000
t=np.arange(0,1,1/Fs)
fc=100;
fm=15;
beta=15;
#________________Funcion de Bessel_____________________________
x = np.linspace(0, 2, 100)

for v in range(0, 8):
    plt.plot(x, sp.jv(v, x), label = "$J_{}(x)$".format(v))
    
plt.xlim((1, 2))
plt.ylim((-0.5, 1.1))

plt.legend(loc='lower right')

plt.xlabel('$β$ (indice de modulación)')
plt.ylabel('$\mathcal{J}_n(x)$')
plt.title('Funcion de Bessel de primera clase')                                
plt.grid(True)
#py.savefig('besseln0to6.eps', format = 'eps')                                      
plt.show()

```

```python
# Ejemplo FM en el tiempo
# modulante
fm   = 10 #Hz
Am   = 1   
# carrier
fc   = 100
kf   = 5 # Hz/volt
Ac   = 1

# defino el sampleo
Fs = 50*fc #Sps 
# tiempo maximo del analisis 
# Periodo de la frecuencia minima
fmmin = 10 
fmmax = 20
T = 2/fmmin

# vector de tiempo 
t  = np.arange(0, T, 1/Fs)

def modulante(Am, fm, t):
    # modulante
    Sm   = Am*np.cos(2*np.pi*fm*t) 
    return Sm



def modFM(Am,fm):
    Sm  = modulante(Am,fm,t)
    # modulada
    ISm  = np.poly1d(Sm)
    wi   = kf*ISm
    Sfm  = Ac*np.cos(2*np.pi*fc*t + wi)  
    plt.plot(t, ISm)
    plt.plot(t, Sfm)
    plt.ylabel('Sm [V]')
    plt.xlabel('freq[Hz]')
    plt.show()
```

```python
# Ejemplo FM en el tiempo
# modulante
fm    = 5 #Hz
# Periodo de la frecuencia minima
fmmin = 4 
fmmax = 10
T     = 2/fmmin
Am    = 0.5   

# carrier
fc    = 40

kf    = (2*np.pi*fc)/10  # rad/volt
kp    = (2*np.pi)/4     # fase/volt
Ac    = 1

# defino el sampleo
Fs = 20*fc #Sps 
dt = 1/Fs
# tiempo maximo del analisis 

# vector de tiempo 
t  = np.arange(0., T, dt)


def mod(Am,fm,time,stype='square', modulation = 'FM',fc = fc, Ac = Ac ,show = 'time'):

    if stype == 'square':  Sm = Am*signal.square(2*np.pi*fm*time)
    else                :  Sm = Am*np.cos(2*np.pi*fm*time) 

    # modulada
    if modulation == 'FM': 
        Smo  = Ac*np.cos( (2*np.pi*fc+ kf*Sm)*time )  
    else    : 
        Smo  = Ac*np.cos(2*np.pi*fc*time + kp*Sm) 
    

    if show == 'time':
        plt.plot(t, Sm, 'b-')
        plt.plot(t, Smo, 'r-')
        plt.ylabel('PM(t)/V')
        plt.xlabel('Time/s')
        plt.show()    
    else :
        Y = fft(Smo) / len(Smo)  # Normalizada
        frq = fftfreq(len(Smo), dt)  # Recuperamos las frecuencias

        plt.plot(frq, np.abs(Y), 'ro-' )  # Representamos la parte imaginaria
        plt.ylim(-0.1, 0.8)
        plt.xlabel('$Frecuencia/Hz$')
        plt.ylabel('$Abs(Y)$')
        plt.xlim([-200,200])
        plt.grid()
        plt.show()
```
