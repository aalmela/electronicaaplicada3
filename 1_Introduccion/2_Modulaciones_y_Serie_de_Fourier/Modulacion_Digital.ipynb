4. Modulación digital

El objetivo principal de la modulación es comprimir la mayor cantidad de datos en la menor cantidad de espectro posible. 
Técnicamente hablando, queremos maximizar la "eficiencia espectral" en unidades bits / seg / Hz. Transmitir unos y ceros más rápido aumentará el ancho de banda de nuestra señal (recuerde las propiedades de Fourier), lo que significa que se utiliza más espectro. También examinaremos otras técnicas además de transmitir más rápido. Habrá muchas compensaciones al decidir cómo modular, pero también habrá espacio para la creatividad.

Amplitude Shift Keying (ASK) es el primer esquema de modulación digital que discutiremos porque la modulación de amplitud es la más simple de visualizar de las tres propiedades sinusoidales. Observe cómo el valor promedio es cero; siempre preferimos esto siempre que sea posible.

Podemos usar más de dos niveles, lo que permite más bits por símbolo. A continuación se muestra un ejemplo de 4-ASK. En este caso, cada símbolo lleva 2 bits de información.

¿Cómo creamos realmente esta señal digitalmente, a través del código? Todo lo que tenemos que hacer es crear un vector con N muestras por símbolo, luego multiplicar ese vector por una sinusoide. Esto modula la señal en una portadora (la sinusoide actúa como esa portadora). El siguiente ejemplo muestra 2-ASK con 10 muestras por símbolo.
../_images/ask3.svg

El gráfico superior muestra las muestras discretas representadas por puntos rojos, es decir, nuestra señal digital. En sistemas reales, la frecuencia de la portadora suele ser mucho más alta que la velocidad a la que cambian los símbolos. En este ejemplo, solo hay tres ciclos de la sinusoide en cada símbolo, pero en la práctica puede haber miles, dependiendo de qué tan alto en el espectro se transmita la señal.

Modificación por desplazamiento de frecuencia (FSK)

El último en la lista es la manipulación por desplazamiento de frecuencia (FSK). FSK es bastante simple de entender: simplemente cambiamos entre N frecuencias donde cada frecuencia es un símbolo posible. Sin embargo, debido a que estamos modulando una portadora, en realidad es nuestra frecuencia portadora +/- estas N frecuencias. P.ej. podríamos estar en una portadora de 1,2 GHz y cambiar entre estas cuatro frecuencias:

    1.2005 GHz
    1,2010 GHz
    1.1995 GHz
    1.1990 GHz

El ejemplo anterior sería 4-FSK y habría dos bits por símbolo. Una señal 4-FSK en el dominio de la frecuencia podría verse así:
../_images/fsk.svg

Si usa FSK, debe hacer una pregunta crítica: ¿Cuál debería ser el espacio entre frecuencias? A menudo denotamos este espaciado como \ Delta fen Hz. Queremos evitar la superposición en el dominio de la frecuencia, por lo que \ Delta fdebe ser lo suficientemente grande. El ancho de cada portadora en frecuencia es una función de nuestra tasa de símbolos. Más símbolos por segundo significan símbolos más cortos, lo que significa un ancho de banda más amplio (recuerde la relación inversa entre el tiempo y la escala de frecuencia). Cuanto más rápido transmitamos los símbolos, más ancha se volverá cada portadora y, en consecuencia, más grande tendremos que hacer \ Delta fpara evitar la superposición de portadoras. No entraremos en más detalles sobre el diseño de FSK en este libro de texto.

Los gráficos de IQ no se pueden usar para mostrar diferentes frecuencias. Muestran magnitud y fase. Si bien es posible mostrar FSK en el dominio del tiempo, más de 2 frecuencias dificulta la distinción entre símbolos:

Ejemplo de Python

Como un breve ejemplo de Python, generemos QPSK en la banda base y tracemos la constelación.

Aunque podríamos generar los símbolos complejos directamente, comencemos con el conocimiento de que QPSK tiene cuatro símbolos a intervalos de 90 grados alrededor del círculo unitario. Usaremos 45, 135, 225 y 315 grados para nuestros puntos. Primero generaremos números aleatorios entre 0 y 3 y realizaremos matemáticas para obtener los grados que queremos antes de convertir a radianes.

```python
import numpy as np
import matplotlib.pyplot as plt

num_symbols = 1000

x_int = np.random.randint(0, 4, num_symbols) # 0 to 3
x_degrees = x_int*360/4.0 + 45 # 45, 135, 225, 315 degrees
x_radians = x_degrees*np.pi/180.0 # sin() and cos() takes in radians
x_symbols = np.cos(x_radians) + 1j*np.sin(x_radians) # this produces our QPSK complex symbols
plt.plot(np.real(x_symbols), np.imag(x_symbols), '.')
plt.grid(True)
plt.show()
```

Observa cómo se superponen todos los símbolos que generamos. No hay ruido, por lo que todos los símbolos tienen el mismo valor. Agreguemos algo de ruido:

```python
n = (np.random.randn(num_symbols) + 1j*np.random.randn(num_symbols))/np.sqrt(2) # AWGN with unity power
noise_power = 0.01
r = x_symbols + n * np.sqrt(noise_power)
plt.plot(np.real(r), np.imag(r), '.')
plt.grid(True)
plt.show()
```

Considere cómo el ruido gaussiano blanco aditivo (AGWN) produce una dispersión uniforme alrededor de cada punto de la constelación. Si hay demasiado ruido, los símbolos comienzan a pasar el límite (los cuatro cuadrantes) y el receptor los interpretará como un símbolo incorrecto. Intente aumentar noise_powerhasta que eso suceda.

Para aquellos interesados en simular ruido de fase, que podría resultar de la fluctuación de fase dentro del oscilador local (LO), reemplace rcon:

```python
phase_noise = np.random.randn(len(x_symbols)) * 0.1 # adjust multiplier for "strength" of phase noise
r = x_symbols * np.exp(1j*phase_noise)

plt.plot(np.real(r), np.imag(r), '.')
plt.grid(True)
plt.show()
```

Incluso puede combinar el ruido de fase con AWGN para obtener la experiencia completa:

```python
n = (np.random.randn(num_symbols) + 1j*np.random.randn(num_symbols))/np.sqrt(2) # AWGN with unity power
noise_power = 0.01
phase_noise = np.random.randn(len(x_symbols)) * 0.1 # adjust multiplier for "strength" of phase noise



r = x_symbols * np.exp(1j*phase_noise) + n * np.sqrt(noise_power)
plt.plot(np.real(r), np.imag(r), '.')
plt.grid(True)
plt.show()
```
