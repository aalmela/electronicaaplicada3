```python
import myplot as myp
import numpy as np
import matplotlib.pyplot as plt

def AvF(f,fo,Qc,Avo = 1):
    return -Avo/(1+1j*Qc*(f/fo-fo/f))

def AvA(f,fo,Qc,Avo = 1):
    X = 2*Qc*(f-fo)/fo
    return -Avo/(1+1j*X)


# Datos para los calculos 
fo    = 1e6
wo    = 2*np.pi*fo
R     = 100
Qc    = 1 
gm    = 100e-3

# Ahora calculando en la hoja de calculo 
f  =  np.arange(100e3,10e6,1e3)                   # Defino el array de frecuencia
Av   = AvF(f,fo,Qc,Avo = gm*R)
AvAp = AvA(f,fo,Qc,Avo = gm*R)

plt.semilogx(f , np.abs(Av)  , 'r-', label='$|Av|$' )
plt.semilogx(f , np.abs(AvAp), 'b--', label='$ABA(|Av|)$' )
plt.grid(True)
plt.legend()
plt.ylabel('$|A_{v}|$')
plt.xlabel('Frecuencia/Hz')

#plt.savefig('figures/ejemplo2ABA.png')
plt.show()


plt.plot(f , 20*np.log10(Av)  , 'r-', label='$|Av|$' )
plt.plot(f , 20*np.log10(AvAp), 'b--', label='$ABA(|Av|)$' )
plt.xlim([0,0.4e7])
plt.grid(True)
plt.legend()
plt.ylabel('$|A_{v}|$')
plt.xlabel('Frecuencia/Hz')

#plt.savefig('figures/ejemplo2ABAdB.png')
plt.show()
```
