```python
import myplot as myp
import numpy as np
import matplotlib.pyplot as plt

def AvF(f,fo,Qc,Avo = 1):
    return -Avo/(1+1j*Qc*(f/fo-fo/f))

def AvA(f,fo,Qc,Avo = 1):
    X = 2*Qc*(f-fo)/fo
    return -Avo/(1+1j*X)


# Datos para los calculos 
fo    = 1e6
wo    = 2*np.pi*fo
R     = 10
BW    = 100e3 
gm    = 100e-3
n     = 10

Qc = fo/BW * (2**(1/n)-1)**0.5
print(Qc)
# Ahora calculando en la hoja de calculo 
f    =  np.arange(100e3,10e6,1e3)                   # Defino el array de frecuencia
Av   = AvF(f,fo,Qc,Avo = gm*R)


plt.semilogx(f , np.abs(Av**1)  , 'r--', label='$|Av|^1$' )
plt.semilogx(f , np.abs(Av**2)  , 'g-.', label='$|Av|^2$' )
plt.semilogx(f , np.abs(Av**n)  , 'b-',  label='$|Av|^n$' )
plt.grid(True)
plt.legend()
plt.ylabel('$|A_{v}|$')
plt.xlabel('Frecuencia/Hz')

#plt.savefig('figures/ejemplo3Av3.png')
plt.show()


plt.semilogx(f , 20*np.log10(Av**1)  , 'r--', label='$dB(|Av|^1)$' )
plt.semilogx(f , 20*np.log10(Av**2)  , 'g-.', label='$dB(|Av|^2)$' )
plt.semilogx(f , 20*np.log10(Av**n)  , 'b-',  label='$dB(|Av|^n)$' )
plt.grid(True)
plt.legend()
plt.ylabel('$|A_{v}|$')
plt.xlabel('Frecuencia/Hz')

#plt.savefig('figures/ejemplo3Av3dB.png')
plt.show()
```
