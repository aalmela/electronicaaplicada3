```python
import myplot as myp
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

def AvF(f,fo,Qc,Avo = 1):
    return -Avo/(1+1j*Qc*(f/fo-fo/f))

# Dado valores a wo y el Q del circuito
fo    = 1e6
wo    = 2*np.pi*fo
R     = 100
Qc    = 0.5 
gm    = 100e-3

# Transferencia en wo
gmR= gm*R # gmR

# Valores del denominador y numerador de la ganancia de tensión
#H($) =  $ (Gvo*wo/Q) / ($**2+$wo/Q+1) 
num=[  gmR*fo/Qc, 0    ]   
den=[ 1, fo/Qc  , fo**2]

# Transferencia
sys1 = signal.lti(num,den) 

plt.figure(figsize=(2,5))
plt.plot(sys1.zeros.real, sys1.zeros.imag, 'o')
plt.plot(sys1.poles.real, sys1.poles.imag, 'rx')
plt.ylabel('Im()')
plt.xlabel('Re()')
plt.grid()
#plt.xlim([-60,1])
plt.xlim([-1e6,1e6])
#plt.savefig('figures/ejemplo1PZ.png')
plt.show()
```

```python
w, mag, phase = signal.bode(sys1)

plt.figure()
plt.semilogx(w, mag)    # Bode magnitude plot
plt.xlim([1e4,1e7])
plt.figure()
plt.semilogx(w, phase)  # Bode phase plot
plt.show()
```
