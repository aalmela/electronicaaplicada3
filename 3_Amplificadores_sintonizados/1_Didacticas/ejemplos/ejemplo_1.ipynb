```python
import myplot as myp
import numpy as np
import matplotlib.pyplot as plt

def AvF(f,fo,Qc,Avo = 1):
    return -Avo/(1+1j*Qc*(f/fo-fo/f))

# Datos para los calculos 
fo    = 1e6
wo    = 2*np.pi*fo
R     = 100
Qc    = 1 
gm    = 100e-3

# Ahora calculando en la hoja de calculo 
f  =  np.arange(100e3,10e6,1e3)                   # Defino el array de frecuencia
Av = AvF(f,fo,Qc,Avo = gm*R)

# Plot transferencia tensi√≥n
plt.plot(f,Av )

#plt.xlim([2.99e6, 3.01e6])
#plt.savefig('figures/ejemplo1AvABS.png')
plt.show()  

plt.plot(f,20*np.log10(Av))
plt.grid()
#plt.savefig('figures/ejemplo1AvdB.png')
plt.show()
```
