1. $Q_c$

Para el calculo de $Q_c$ empleamos la expresión de la transferencia  de tensión del simple sintonizado.

$$|\bar{A_v}| =   \frac{1}{ \sqrt{1 + Q_c^2(\frac{f}{f_o}-\frac{ f_o}{f})^2} }$$

Buscamos una atenuación de $40 dB$ a la frecuencia de canal adyacente.

La respuesta en frecuencia del fitro muestra mayor atenuación a bajas frecuencias, por lo que para garantizar una atenuanción de almenos $40dB$, debemos realizar el calculo en el canal superior. 
Dado que la separación entre canales es de $df= 50 KHz$, la frecuencia a la que debe garantizar la atenuación es $f_{ca} = 505 KHz$.

$$|\bar{A_v}|(505KHz) = \frac{1}{ 10^{\frac{40}{20} } }$$

$$\frac{1}{100} =   \frac{1}{ (\sqrt{1 + Q_c^2(\frac{120MHz}{100MHz}-\frac{ 100MHz}{120MHz})^2})^2 }$$

$$100 =    \sqrt{1 + Q_c^2(\frac{505KHz}{455KHz}-\frac{ 455KHz}{505KHz})^2}$$

$$Q_c = \frac{\sqrt{100 - 1}}{\frac{505KHz}{455KHz}-\frac{ 455KHz}{505KHz}}$$


$$Q_c = 47.62$$
