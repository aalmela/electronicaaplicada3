# Ejercicio 1


Un Rx superheterodino de simple conversión, con FI en $455 kHz$, es sintonizado para recibir un canal de $f_c=1 MHz$. 
Los canales adyacentes se encuentran a $\pm 50 kHz$. 
Se especifica un rechazo mínimo a los mismos de $44 dB$.

A la salida del Rx (sobre el parlante) hay una señal alterna de $1V$ y $SNR = 10 dB$.

El circuito de la figura corresponde al amplificador de RF, el cual debe presentar máxima planicidad en la banda de paso, con un ancho de banda de $-3dB$ de $100 KHz$. 
Todas las interfaces adaptadas a MTE-Qcte

$$y_{22} =         0.1mS +j\cdot \omega 10\times10^{-12}S$$
