### $L_1$

Para el calculo del inductor $L_1$, es importante notar la condicion $L_1 = L_2 = L_3$. 

El inductor $L_2$ es el unico que tiene impuesta la carga, por lo tanto, es necesario realizar el calculo de este componente primero. 

### Calculo del inductor $L_2$


![ex 2.13](fig/20200803_circuito_out.png)

Ese inductor corresponde a uno de los sintonizado del doble sintonizado y tiene como carga la impedancia de salida del transistor.

$$\frac{1}{Q_c} = \frac{1}{Q_o} + \frac{\omega L_2}{r_{ext}}$$

Al ser un doble sintonizado, la resistencia de carga del sintonizado corresponde al parte resistiva de la admtancia de salida del transistor.

$$r_{ext} = r_{22}$$

El calculo de $Q_c$ para el sintonizado conformado por $L_2$

$$\frac{1}{Q_c} = \frac{1}{Q_o} + \frac{\omega L_2}{r_{22}}$$

Despejando $L_2$

$$L_2 = (\frac{1}{Q_c} - \frac{1}{Q_o}) \cdot \frac{r_{22}}{\omega}$$

$$L_2 = (\frac{1}{20} - \frac{1}{\inf}) \cdot \frac{10k\Omega}{2 \pi 1e6MHz}$$

$$L_2 = 79.577e-06 Hy$$

$$L_1 = L_2 = 7.9577e-05 Hy$$

$$\frac{1}{Q_c} = \frac{1}{Q_o} + \frac{\omega L_1}{r_{ext}}$$

## 4. Determinar el rechazo de Frecuencia Imagen del Rx

## Butterwoth de tercer orden

La transferencia normalizada de un Butterwoth de orden $n$ puede expresarse como: 

$$\bar{|H(f_{c})|}=\frac{1}{\sqrt{1+(\frac{\chi(f)}{\chi(f_c)})^{2\cdot n}}}$$ 

donde:

$$\chi(f)   = 2 \cdot Q \cdot \frac{(f-f_o)}{f_o}$$

$$\chi(f_c) = 2 \cdot Q \cdot \frac{(f_c-f_o)}{f_o}$$

$$n = 3$$

$$\bar{|H(f_{c})|}=\frac{1}{\sqrt{1+(\frac{(f-f_o)}{(f_c-f_o)})^{6}}}$$ 

$$|\bar{A_v(img)} | = 0.00016$$

$$|\bar{A_v(img)} | = -75.60dB$$



## 5. Determinar la ganancia de potencia (G=Pout/Pin) del amplificador de RF en frecuencia central


### Calculo de $g_m$

Antes de calcular la ganancia de potencia, debemos conocer $g_m$. 

$$g_m = 40 \cdot I_{cq}$$

Para calcular $I_{cq}$, debemos calcular la tensión $V_{BB}$. 

Suponiendo que la corriente $I_B$ es despresiable a la que circula por el divisor resistivos (luego debemos demostrarlo).

$$V_{BB} = V_{CC} \frac{R_2}{R_1+R_2}$$



$$V_BB=4.00e+00 V$$

$$V_E  = V_{BB}-V_{BE}$$

$$V_E=3.40e+00 V$$

$$I_E  = \frac{V_E}{R_E}$$

$$I_E=2.83e-03A$$


$$I_C  = \frac{I_E}{(1+\frac{1}{H_{FE}})}$$

$$I_C=2.82e-03A$$

$$I_B  = \frac{I_C}{HFE}$$

$$I_B=9.41e-06A$$

Verificando entonces.

$$I_{RB} = \frac{V_{BB}}{R_2}$$

$$I_{RB} = 222e-06A$$

Que es mucho mayor a la corriente de la base. 

$$g_m = 40 \cdot I_C$$

$$gm = 113e-03S$$

### Ganancia de potencia
La ganacia de potencia se calcula como:
$$G_{1} = (\frac{v_o}{v_i})^2 \frac{r_i}{r_o}$$

Por lo que es necesario calcular la ganacia de tensión:

$$| \bar{Av} | = \frac{v_o}{v_i}$$

La ganancia de tensión se se puede calcular mediante las ganancias parciales. 

$$| \bar{Av} | = \frac{v_o}{v_c} \cdot \frac{v_c}{v_b} \cdot \frac{v_b}{v_a}  \cdot \frac{v_a}{v_i}$$



$$\frac{v_o}{v_c} = \frac{1}{N_{c2}}$$

$$|\frac{v_o}{v_c}| = 0.50$$



$$|\frac{v_c}{v_b}| =  gm L_2 w_o Q_2\frac{h}{(1+h^2) }$$ 


$$|\frac{v_c}{v_b}| = 489.12$$




$$\frac{v_b}{v_a} =\frac{N_2}{N_1}$$

$$|\frac{v_b}{v_a}| = 0.35$$

$$\frac{v_a}{v_i} = N_{c1}$$

$$|\frac{v_a}{v_i}| = 2.83$$

$$| \bar{Av} | = \frac{v_o}{v_c} \cdot \frac{v_c}{v_b} \cdot \frac{v_b}{v_a}  \cdot \frac{v_a}{v_i}$$

$$|Avo| = 244.56$$
$$|Avo| = 47.77dB$$


$$G_{1} = (\frac{v_o}{v_i})^2 \frac{r_i}{r_o}$$

$$|G_1| = 29904.47$$
$$|G_1| =44.76dB$$


## 8. Determinar el ancho de banda del Rx sabiendo que la etapa de FI está conformada por 3 Simple Sintonizados sincrónicos

Los canales adyacentes se encuentran a $+/- 50 kHz$. 
Se especifica un rechazo mínimo a los mismos de $44 dB$

$$|\bar{A_{3ss}}| = \frac{1}{(1+Q_{if}^2 (\frac{f_{ca}}{f_{if}} - \frac{f_{if}}{f_{ca}})^2 )^{\frac{3}{2}}} = \frac{1}{10^{\frac{Attca}{10}}}$$

$$Q_{if} = \frac{\sqrt{(10^{\frac{Attca}{10}})^{\frac{2}{3}} - 1}}{(\frac{f_{ca}}{f_{if}} - \frac{f_{if}}{f_{ca}})}$$

$$Q_if = 25.460$$

$$BW_{1ss} = \frac{f_o}{Q_if}$$

$$BW_{3ss} = BW_{1ss} \sqrt{2^{\frac{1}{3}}-1}$$

$$BW_{3ss}  = 9111.322 Hz$$

```python
import numpy as np

# Parametros del transistor 
r11   =   1/0.8e-3 
c11   =   10e-12
r22   =   1/0.1e-3 
c22   =   10e-12

rg    = 1250
rl    =   2.5e3
N1    =   100
# Calculos auxiliares
####### 1
BW = 100e3
fo = 1e6
wo = (2*np.pi*fo)
# Simple sintofizado
Q1 = fo/BW

# Doble 

Q2 = fo/(BW*np.sin(30*np.pi/180))

k  = (BW*np.cos(30*np.pi/180))/fo

print('$$Q_1= {:1.2f}$$'.format(Q1))

print('$$Q_2= {:1.2f}$$'.format(Q2))

print('$$k = {:1.2e}$$'.format(k))
```

```python
# DOBLE SINTONIZADO
#  calculo a la salida sobre COLECTOR

Qo = np.inf

L2 = (1/Q2-1/Qo)*r22/wo

L1 = L2

print('$$L2= {:1.2e} Hy$$'.format(L2))

print('$$L1= {:1.2e} F$$'.format(L1))
```

```python
# ENTRADA!
L1 = L2
rext = wo*L1/(1/Q1-1/Qo)
print('$$rext = {:1.3f} \Omega$$'.format(rext))
rt   = rext*2

print('$$rt = {:1.3f} \Omega$$'.format(rt))

# (N1/N2)**2 = rt/r11
N2 = (r11/rt)**0.5*N1

print('$$N2= {:1.2f}$$'.format(N2))

L12 = L1*(N2/N1)**2
```

```python
CS = 1/(L1*wo**2) 
print('$$C_S= {:1.2e} F$$'.format(CS))

c11t = c11/((N1/N2)**2)
print('$$C_11t= {:1.2e} F$$'.format(c11t))

CSi = CS - c11t
print('$$C_Si= {:1.2e} F$$'.format(CSi))


# Divisor Capacitivo transistor
# R//C == --C1-C2//Ro
R = rt
C = CSi 
print(C)
Ro= rg

# (ro/ri) = (nc)^2
nc1 =  (R/Ro)**(1/2) 
print('$$Nc1 = {:1.2f} $$'.format(nc1))

Qm2 = wo*C*R 
print('$$Qm2 = {:1.2f} $$'.format(Qm2))

Qm1 = ( ((1+Qm2**2)/nc1**2) - 1)**0.5
print('$$Qm1 = {:1.2f} $$'.format(Qm1))

# para el calculo de los capacitorees
C1  = Qm1/(Ro*wo)

C1p = C1*(1+1/Qm1**2)

Cp  = C /(1+1/Qm2**2)

C2  = C1p*Cp/(C1p-Cp)

print('$$C_1= {:1.3e} F$$'.format(C1))
print('$$C_2= {:1.3e} F$$'.format(C2))

C2a = nc1/(nc1-1)*C
C1a = nc1*C
print('$$C_1a= {:1.3e} F$$'.format(C1a))
print('$$C_2a= {:1.3e} F$$'.format(C2a))
```

```python
#  calculo a la salida sobre CARGA
cant = 150e-12
L3 = L2

C3 = CS - c22
print('$$C_3= {:1.2e} F$$'.format(C3))

C4 = k*CS
print('$$C_4= {:1.2e} F$$'.format(C4))


# Divisor Capacitivo transistor
R = r22
C = CS 
Ro= rl
# (ro/ri) = (nc)^2
nc2 =  (R/Ro)**(1/2) 
print('$$Nc2 = {:1.2f}$$'.format(nc2))

Qm2 = wo*C*R 
print('$$Qm2 = {:1.2f}$$'.format(Qm2))

Qm1 = ((1+Qm2**2)/nc2**2-1)**0.5
print('$$Qm1 = {:1.2f}$$'.format(Qm1))


# Dado que Qm2 y Qm1 son mayores a 10, puedo emplear las ecuaciones aproximadas \
# para el calculo de los capacitorees
#C5 = nc/(nc-1)*C
#C6 = nc*C
#print(nc, C5, C6)

# para el calculo de los capacitorees
Ct  = Qm1/(Ro*wo)
C6  = Ct-cant
Ctp = Ct*(1+1/Qm1**2)
Cp  = C /(1+1/Qm2**2)
C5  = Ctp*Cp/(Ctp-Cp)

print('$$C_5= {:1.2e} F$$'.format(C5))
print('$$C_6= {:1.2e} F$$'.format(C6))

C5a = nc2/(nc2-1)*C
Cta = nc2*C
C6a = Cta - cant
print('$$C_5a= {:1.3e} F$$'.format(C5a))
print('$$C_ta= {:1.3e} F$$'.format(Cta))
print('$$C_6a= {:1.3e} F$$'.format(C6a))
```
