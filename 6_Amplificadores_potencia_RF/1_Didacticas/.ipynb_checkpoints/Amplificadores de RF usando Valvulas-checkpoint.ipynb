# Introducción 


Actualmente las válvulas o tubos electrónicos son consideradas como piezas de museo y ya no se utilizan. Efectivamente, los tipos de válvulas para manejo de señales a potencias pequeñas que se utilizaron extensamente fueron remplazadas por transistores y otros dispositivos de estado sólido, y han dejado de usarse y, aunque vuelven a utilizarse en algunos casos aislados como ciertos amplificadores  de  sonido,  o algunas aplicaciones.

En sistemas de comunicaciones, cierto tipo de válvulas continúan empleándose ampliamente, sin que se vislumbre un futuro cercano en que puedan ser substituidas por componentes de  estado  sólido.  
Entre  ellas,  los  tubos  de  onda  progresiva  (TWT)  en  los  amplificadores  de los transpondedores de la mayoría de los satélites, los magnetrones utilizados extensamente en  radares  y  los  diversos  tipos  de  tubos  amplificadores  de  potencia  en  los  transmisores  de radio y televisión. 
En los terrenos científico, industrial y médico las válvulas electrónicas se emplean, por ejemplo, en aceleradores de partículas, sistemas de calentamiento industrial y aparatos de rayos X. 
Por ello, es conveniente dedicar, aunque sólo sea de forma descriptiva y  relativamente  superficial,  un  breve  tiempo  al  estudio  de  las  válvulas  amplificadoras  de potencia.




## Las valvulas 

Las válvulas son dispositivos electrostáticos que utilizan el flujo de electrones, la base de todo el flujo de corriente eléctrica. 
En un material, la cantidad de electrones depende del elemento atómico: el hidrógeno tiene un solo electrón, mientras que los metales tienen muchos más electrones en órbita. 
Los electrones son móviles y pueden intercambiar libremente órbitas con sus átomos vecinos. 
Cuando un electrón cambia de órbita, deja una deficiencia en ese átomo en particular. 
Durante un breve período, el átomo está cargado positivamente hasta que un electrón equilibre la carga. 

Los materiales metálicos tienen electrones flotando muy cerca de la superficie, como una fina niebla. Pueden ser atraídos fácilmente por un objeto suficientemente cargado positivamente. 
A temperatura ambiente, los electrones son un poco lentos y no tienen la capacidad de saltar lejos del metal principal. 
Sin embargo, a medida que el metal se calienta se vuelve más activo. 
Cuando la temperatura es lo suficientemente alta, los electrones tienen mucha energía y pueden imaginarse como una espesa niebla flotando sobre el metal principal. 
Algunos materiales muy especiales emiten una gran cantidad de electrones incluso a temperatura ambiente. Estos se utilizaron para hacer válvulas especiales, llamadas válvulas de cátodo frío, pero hoy estos son raros.

Esto en términos simples es lo que constituye un cátodo, es una fuente de "electrones libres".
Si un objeto cargado positivamente se acerca ahora al metal, los electrones libres fluiran al objeto. 
Los metales puros no son la mejor fuente de electrones libres, a menos que la temperatura sea muy alta. 
Sin embargo, hay algunos otros materiales que liberan electrones con mucha más facilidad, como materiales de tierras raras, como bario, torio y estroncio. 
Si el metal está recubierto con estos óxidos, la generación de electrones libres aumenta dramáticamente.

El objeto cargado positivamente se llama ánodo y también es un material metálico. 
Si se aplica un voltaje externo entre el ánodo y el cátodo, podemos hacer que los electrones fluyan entre ellos, incluso en el vacío. 


Si la distancia entre el electrón y el cuerpo con carga positiva es grande, la fuerza de atracción es pequeña. A medida que el electrón comienza a moverse hacia el cuerpo cargado positivamente, inicialmente viaja lentamente, luego, a medida que se reduce la distancia, la fuerza de atracción aumenta. 
En poco tiempo viajan a una velocidad cercana a la de la luz, típicamente el $70\%$. 

Cuando el electrón llega al ánodo, dada la aceleración del campo eléctrico, tiene energia cinética. 
Aunque el electrón tiene una masa muy baja, gracias a esta energia, cuando golpea la superficie del ánodo, cede la energía cinética que ha acumulado al acelerar.
Esto desaloja otros electrones, que ya están en la superficie. Los electrones desplazados, aunque ahora son catapultados al espacio sobre el ánodo, son fuertemente atraídos porque están muy cerca del ánodo y regresan al ánodo en un período muy corto. Este desplazamiento de electrones debido a una colisión se llama emisión secundaria.

La temperatura del cátodo se mide en la escala de Kelvin. 
El tungsteno tiene la emisión más baja a baja temperatura. 
Para emitir suficientes electrones, debe calentarse a una temperatura muy alta. 
El cátodo recubierto de óxido necesita la temperatura más baja y el cátodo de tungsteno toriado se encuentra entre estos otros dos tipos. 
El torio es un metal de tierras raras que se usa comúnmente como recubrimiento de alambre de tungsteno.

Para una emisión máxima de un cátodo recubierto de óxido, la temperatura es de aproximadamente $1350 K$; el cátodo de tungsteno toriado es de aproximadamente $1900 K$. 
Los cátodos de tungsteno son cables desnudos y deben estar a 2700 K para obtener la mejor emisión. 
El blanco brillante es de unos 3300 K. 
El tungsteno es el material del que están hechas las lámparas eléctricas y un cátodo de tungsteno funciona a una temperatura similar. 
En terminología de válvulas, se clasifican como "emisores brillantes".


Podemos ver en la tabla que un cátodo recubierto de óxido puede suministrar $\sim 2A/cm^2$.
Se trata del límite superior con la tecnología moderna.

La nube de electrones en la superficie del cátodo da lugar a otra terminología que se utiliza. 
La nube de electrones libres se conoce como "nube espacial". 
A medida que el ánodo atrae algunos de los electrones lejos de la nube, surgen reemplazos del cátodo de modo que en cualquier momento hay un número relativamente constante esperando ser transportado al ánodo. 
Si la tasa de producción de electrones libres no es suficiente, no se puede alcanzar la corriente máxima del ánodo. 
La corriente de electrones que se mueve entre el cátodo y el ánodo se denomina "carga espacial".

## Tipos de valvulas

### Diodo


Hoy en día, los diodos semiconductores utilizan la misma terminología de denominación, ánodo y cátodo; es un retroceso a la era de las válvulas.
Las válvulas se podían quitar fácilmente ya que los pines de la base se enchufaron en el equipo. 
Por lo tanto, es rápido y fácil reemplazar una válvula defectuosa. 
Los pines están polarizados para que la válvula solo pueda encajar si los pines están alineados correctamente con los enchufes. 

### Válvula de triodo

En el triodo, el ánodo y el cátodo son similares al tipo de diodo simple. 
El electrodo adicional se coloca cerca del cátodo, mientras que el ánodo está a una distancia similar a la del tipo de diodo. 
Este nuevo electrodo se llama rejilla porque usa una rejilla de alambres finos para formar una pantalla entre el cátodo y el ánodo. 
Aunque la rejilla tiene aberturas muy pequeñas, en comparación con un electrón, son agujeros abiertos. 
Por tanto, los electrones tienen poca dificultad para encontrar su camino entre el cátodo y el ánodo.

La forma común de construcción es utilizar tubos cilíndricos concéntricos. 
El cátodo es el tubo más pequeño. 
Alrededor de esta se encuentra la rejilla de control, enrollada como una bobina helicoidal o una serie de rejillas de malla enrolladas en un tubo hueco. 
El ánodo es el tubo exterior, que forma el soporte, mediante pasadores rígidos y aisladores de mica, para mantener los diferentes cilindros en la alineación correcta.

La rejilla de control permite controlar el flujo de electrones desde el cátodo al ánodo. 
En un triodo permite controlar el flujo de electrones variando el voltaje que se le aplica. 
Los electrones, al ser partículas con carga negativa, repelen cualquier otro objeto con carga negativa. 
Al aplicar un voltaje negativo suficientemente alto a la rejilla, los electrones en la nube espacial son repelidos. 
Cuando la red no tiene una carga tan negativa, algunos de los electrones pueden pasar a través de la red porque el alto voltaje positivo en el ánodo los atrae. Si el voltaje del ánodo aumenta y se vuelve más positivo, la atracción también aumenta y más electrones pasarán a través de la red. 
Por lo tanto, aumentar el voltaje del ánodo también crea un mayor flujo de corriente.

Si la tensión de la red se eleva para igualar la del cátodo, la diferencia de tensión cae a cero. Se comporta como si la cuadrícula se hubiera eliminado por completo. 
En este estado fluirá la máxima corriente de ánodo. 
Si la tensión de la red aumenta de modo que se vuelva positiva con respecto al cátodo, se aplica otro estado.

Los electrones buscan un lugar con carga positiva adonde ir. Importa poco si es el ánodo o algún otro objeto cargado positivamente. 
Con la rejilla ahora muy positiva, y estando mucho más cerca del cátodo que del ánodo, algunos de los electrones aterrizarán en la rejilla de la misma manera que aterrizarán en el ánodo. 
Esto significa que la amplitud del flujo del ánodo comienza a disminuir a medida que pasa una gran cantidad de electrones a la estructura de rejilla positiva. 
Si la rejilla es suficientemente positiva, reduce el flujo de corriente del ánodo a un nivel muy bajo. La corriente que ahora fluye en la red hace que se caliente ya que los cables muy finos tienen una alta resistencia. 
Esto puede provocar daños si se permite que sea demasiado alto.

Entre la condición de potencial de cuadrícula positiva muy alta y solo una condición débilmente positiva, entra en juego otro factor. 
Los electrones que salen del cátodo son atraídos inicialmente por el potencial anódico muy alto. 
Mientras viajan hacia él, también experimentan la atracción de la rejilla de control débilmente positiva. 
Esta atracción adicional acelera a los electrones. 
Cuando los electrones llegan a la red, ya están viajando casi a su máxima velocidad. 
Vuelan y, a medida que se alejan de la red, ahora viajando hacia el ánodo, experimentan cada vez menos atracción por el potencial positivo débil de la red y más y más por el ánodo. 
La pequeña carga positiva, si no es demasiado alta, actúa como un acelerador inicial de los electrones. 
Por encima de esta condición débil,la rejilla atraerá más y más electrones hasta que el ánodo se quede sin electrones. 
Por lo tanto, cuando la red está unos pocos voltios por encima de cero, la corriente del ánodo aumenta antes de caer.

El esquema en el diseño de válvulas es maximizar la aceleración de los electrones y minimizar el número que son detenidos por la red. 
Por supuesto, algunos electrones siempre no pasarán a través de la estructura de rejilla fina, simplemente porque chocan con ella. 
Siempre que el número de impactos en la red sea pequeño, no se produce ningún daño. 
Al igual que con el ánodo, cuando un electrón choca contra un objeto estático, provoca calor al ceder la energía cinética que contiene. 
Por lo tanto, siempre existirá una corriente de red muy pequeña, aunque la red no esté cargada positivamente.
La amplitud de esta corriente es tan baja que apenas se puede medir. 
En el caso de que el voltaje de la red sea negativo, pero no lo suficiente para cortar completamente la corriente del ánodo, los electrones se doblan alrededor de los cables cargados y los evitan. 

A la derecha se muestra una estructura de cuadrícula típica. 
Esto es de un pequeño triodo de microondas (2C39A, https://frank.pocnet.net/sheets/140/2/2C39A.pdf), la rejilla es del tipo clásico de cable cruzado y el espacio entre el cátodo es de solo $\sim 0,25 mm$. 
El espaciado del ánodo está a $\sim 1,2 mm$ de la rejilla, por lo que la trayectoria de vuelo de los electrones es muy corta. 
La porción del ánodo de la válvula se ha roto para ver el interior de la válvula. 
El sobre de vidrio es un simple tubo redondo.

Los cables de la rejilla tienen un diámetro de aproximadamente $0,01 mm$ y la abertura entre los cables es de aproximadamente $0,05 mm$. 
El objeto oscuro en el centro es la superficie del cátodo. 
Los cables de la rejilla se tensan y se sueldan por puntos al exterior en forma de arandela que se conecta al contacto de la rejilla.

### Válvula Tetrode

La siguiente válvula comúnmente utilizada incluye otro electrodo adicional y, dado que ahora tiene cuatro electrodos, se llama tetrodo.

La válvula de triodo se puede mejorar y el tetrode fue el siguiente paso lógico. 
En el triodo, cuando los electrones chocan contra el ánodo, tienden a desprender otros electrones y provocar una emisión secundaria. 
Estos electrones desplazados causan una "confusión" del flujo total de electrones entre el cátodo y el ánodo y pueden ser perjudiciales para la operación. 
Si se coloca otra estructura de rejilla cerca del ánodo y se carga positivamente a un voltaje muy por debajo del potencial del ánodo, detendrá los electrones de emisión secundaria y ayudará a limpiar el flujo total. Esta nueva estructura de cuadrícula se llama cuadrícula de pantalla o Grid-2. Barre los electrones de emisión secundaria errantes y los desvía del ánodo de vuelta a la fuente de alimentación.

Al hacerlo, existe una pequeña corriente adicional en la nueva estructura de rejilla. En funcionamiento normal, esta corriente es bastante pequeña.
Sin embargo, si el voltaje del ánodo se reduce hacia el potencial de la rejilla de la pantalla, este último comienza a actuar como otro ánodo. 
Si el voltaje del ánodo cae para ser igual al voltaje de la rejilla de la pantalla, entonces ahora tenemos dos 'ánodos' polarizados de manera similar que luchan por los electrones en tránsito. 
Por lo tanto, el voltaje de la rejilla de la pantalla nunca debe ser tan alto como el ánodo o, por el contrario, el voltaje del ánodo nunca debe caer por debajo del voltaje de la rejilla de la pantalla.

Sin embargo, existen excepciones válidas a esta regla. Si la rejilla de la pantalla y el ánodo están conectados, la válvula tetrode se convierte en un triodo simple. Esto a veces se hace por varias razones que se tratarán más adelante.

Una de las acciones de la cuadrícula de la pantalla no es solo detener los electrones errantes, sino también comportarse de la misma manera que una cuadrícula de control ligeramente positiva. 
Provoca un empuje adicional a los electrones que viajan entre el cátodo y el ánodo para dar luego una pequeña aceleración adicional en la parte final de la trayectoria de vuelo. 
El resultado de esto es que, no solo la rejilla de control tiene un mayor control sobre la cantidad de electrones liberados para volar hacia el ánodo, sino que ahora también la rejilla de la pantalla tiene el mismo efecto. 
Actúa como una red de control subsidiaria. 
Cuando los dos actúan al unísono, el resultado es que aumenta el flujo de electrones. Por lo tanto, la válvula de tetrodo de rejilla de pantalla tiene un mayor flujo de corriente que un triodo de tamaño similar. 
Esto significa que para una cierta variación del voltaje de la red de control hay un cambio mayor en el flujo de corriente del ánodo.
Es aproximadamente de tres a cinco veces mayor que una válvula de triodo de tamaño similar.

Sin embargo, el tetrodo también tiene cierto inconveniente. Al ser una estructura adicional colocada entre la rejilla de control y el ánodo, la rejilla de la pantalla también puede experimentar efectos de emisión secundaria, porque algunos de los electrones de alta velocidad pueden impactar y hacer que haya más electrones de emisión secundaria presentes. 
Para corregir este problema se utiliza otra estructura de cuadrícula. 
Por lo tanto, ahora tenemos una válvula con tres estructuras de rejilla, el pentodo.

### Válvula de pentodo

El pentodo introduce la cuadrícula final y esta se llama cuadrícula supresora o cuadrícula-3. 
Esta rejilla se coloca entre la rejilla de la pantalla (Rejilla-2) y el ánodo, pero está conectada al cátodo. Esta es una conexión interna o una que el usuario tiene que conectar en los pines de la base de la válvula.

Al estar en el potencial del cátodo, los electrones lo ven como otra estructura con el mismo potencial que ellos.
Por lo tanto, tratan de evitarlo y prefieren refugiarse en un lugar más positivo. 
Este refugio seguro los está esperando; es la cuadrícula de la pantalla. 
Entonces, la rejilla supresora suprime los electrones secundarios errantes del ánodo y los obliga a refugiarse en la rejilla de la pantalla.

De los dos tipos de válvulas de rejilla de pantalla, el tetrodo y el pentodo son ambos muy similares en características. 
Muchas y variadas son las diferentes técnicas de construcción interna utilizadas, pero todas funcionan de la misma manera. 

Tanto los pentodos como los tetrodos tienen mayores ganancias de corriente que los triodos. 
En determinadas aplicaciones es preferible utilizar un pentodo y en otras un tetrodo. 
Las características de tetrodo y pentodo también se pueden adaptar para adaptarse a diferentes aplicaciones cambiando el tamaño, el espaciado o la forma de las estructuras de rejilla para satisfacer requisitos especiales.

### Construcción de filamento de cátodo

Se utilizan varias estructuras de cátodo diferentes. 
Algunos son alambres de tungsteno lisos, a menudo recubiertos de óxido, y sirven como calentador y cátodo combinados. 
Estos tipos se utilizan comúnmente en válvulas rectificadoras grandes y algunos triodos de RF y se conocen como cátodos calentados directamente. 
Otros tipos usan un filamento calefactor dentro de un tubo, que está recubierto por fuera con un óxido. 
Estos tipos se conocen como cátodos calentados indirectamente ( Fig. 2.5 ). 
Muchas válvulas pequeñas de recepción y transmisión utilizan el cátodo calentado indirectamente: simplifican el diseño pero tienen una ruptura de voltaje finita en voltajes de calentador a cátodo altos. 
El cátodo calentado directamente no tiene este problema, pero dificulta el diseño del equipo.


![5-cuatrocientos](fig/valvula_25.png)
Fig 2.5: Construcción de cátodo calentado indirectamente.

Generalmente, todos los calentadores de cátodo requieren un suministro de filamento de CA o CC. 
En válvulas de transmisión grandes, la energía consumida por el filamento puede ser bastante alta. 
El triodo Eimac 3-500Z, por ejemplo, requiere un suministro de filamento de 5 V de 14,5 A para cada válvula. Si el amplificador tiene dos válvulas, el suministro de filamento debe suministrar casi 30A.

El método más común para suministrar el filamento es un devanado de baja tensión en el transformador de red. Si el equipo utiliza diferentes voltajes de filamento, se requiere un devanado para cada tipo diferente. 
Los receptores pequeños y las válvulas de transmisión de baja potencia a menudo utilizan 6,3 V o 12,6 V, por lo que se adaptan al funcionamiento de la radio móvil con una batería de coche de 12 V. Otros tipos tienen filamentos de doble voltaje. 
Suelen ser dos filamentos conectados en serie con una clavija central. 
Estos pueden utilizar 6,3 V o 12,6 V cableando los pines de la base en consecuencia.

En la numeración de válvulas estadounidense, a menudo, el primer número indica el voltaje del filamento, por ejemplo, una 6SA7 y una 12SA7 son la misma válvula, excepto que una tiene un filamento de 6,3 V y la otra un filamento de 12,6 V. 
Para los receptores de televisión y radio, a menudo, todas las válvulas tienen la misma corriente de filamento, por ejemplo 300 mA, y pueden conectarse en serie y alimentarse desde la red de CA con una resistencia cuentagotas adecuada en serie.

### Disipación y enfriamiento del ánodo

Para cada tipo de válvula existe una temperatura de funcionamiento segura y una disipación del ánodo. 
La disipación se calcula como vatios de potencia disipados en el ánodo. 
La válvula debe mantenerse por debajo de una cierta temperatura máxima para una vida útil prolongada y un funcionamiento confiable. 
Hay varias formas diferentes en que el ánodo se deshace del calor disipado y varía de un tipo a otro.

En las válvulas de envoltura de vidrio pequeñas, normalmente es una mezcla de radiación y conducción. 
El calor infrarrojo pasará a través de un vacío y llegará al sobre de vidrio exterior, que luego emite el calor al aire circundante. 
En válvulas de baja potencia, a menudo esto es suficiente para mantener la válvula fría. 
Algunos ánodos de válvulas más grandes brillan en rojo apagado a máxima potencia, como una forma de aumentar el efecto de enfriamiento infrarrojo. 
En algunos otros tipos, el calor sube por los cables de soporte del ánodo y sale a las clavijas, donde se puede disipar, ya sea a través del cableado del enchufe de la base o, en el caso de los ánodos de la tapa superior, con accesorios especiales de enfriador con aletas. Otros tipos requieren enfriamiento por aire forzado soplado en el sobre o mediante un enfriador de aletas. Algunas válvulas de transmisión muy grandes necesitan refrigeración líquida para disipar el calor.

Los fabricantes normalmente establecen la disipación segura máxima y la temperatura máxima recomendada de envoltura y sellado para garantizar un funcionamiento confiable. 
Es importante prestar atención a estas recomendaciones para lograr la máxima vida útil de la válvula.

Para las válvulas de tetrodo transmisor de gran tamaño, no solo el ánodo disipa una cantidad apreciable de energía, sino también la rejilla de la pantalla. 
Si se opera a una temperatura insegura, a menudo conducirá a fallas catastróficas. 
Si la temperatura alcanza la temperatura de ablandamiento de la envoltura de vidrio, se sabe que el alto vacío hace que la envoltura implosione al succionar el vidrio blando.

Por lo general, no hay peligro de sobreenfriar una válvula y puede mejorar significativamente la vida útil.
Eimac proporciona algunos datos sobre la vida útil de la válvula con valores de temperatura para válvulas de encapsulado de vidrio. 
Por ejemplo, una reducción de la temperatura de la envoltura de vidrio tan solo 25ºC por debajo del máximo puede aumentar la vida útil en un 25%. Una reducción de 50ºC casi duplica la vida.


### Eficiencia

Cualquier dispositivo, al convertir energía de una forma a otra, siempre sufre alguna pérdida o ineficiencia. Las válvulas no son una excepción a esta regla. Por cada vatio de potencia de entrada de CC, no es razonable esperar la misma potencia de salida. La cantidad de potencia de entrada de CC que se convertirá en salida se define por la eficiencia.

Generalmente, usamos la fórmula simple:

Eficiencia 

$\eta = \frac{P_L}{P_{dc}$


De los distintos modos de funcionamiento, algunos son más eficientes que otros. Para los amplificadores de RF, los mas eficientes son los Clase C, los siguientes  son los Clase B y la menor es la Clase A. 
Se puede calcular exactamente qué tan bien cada modo de funcionamiento puede convertir la potencia de entrada de CC en la potencia de salida a partir de los primeros principios. 
Y aquí es donde el tema se vuelve interesante.

Durante muchos años, el tema estuvo empañado por una tabla que apareció originalmente en un libro de texto que mostraba el valor teórico de la eficiencia que sería posible con una válvula ideal. 

![5-cuatrocientos](fig/valvula_27.png)
Fig 2.7: Eficiencia teórica para varios ángulos de conducción. (Tomado de Thomas L Martin Jr, 'Circuitos electrónicos', Prentice-Hall Inc, 1955.)


Sin embargo, este gráfico y los cálculos hicieron una suposición ideales. 
Se suponía que la válvula podía oscilar el voltaje del ánodo en un rango igual al voltaje de suministro de CC. 
Esto significa que el voltaje del ánodo bajaría a cero voltios y hasta el doble del voltaje de suministro. 
No se conoce ninguna válvula que pueda funcionar de esta manera. 
Es prácticamente imposible hacer descender el voltaje del ánodo a cero voltios, incluso si se usa un tipo de triodo grande.

En consecuencia, la oscilación hacia arriba no puede ser mayor que la oscilación hacia abajo para un circuito de ánodo resonante que está estrechamente acoplado a la carga para transferir la potencia máxima. 
En un amplificador de Clase C, el ángulo de conducción normalmente no es inferior a unos 60 grados y, muy a menudo, es necesario utilizar más, hasta unos 150 grados, y esto requiere una gran cantidad de excitación para alcanzarlo. 
Si se utilizara este método, la potencia de accionamiento requerida sería excesiva y la disipación de la red sería destructiva. 
En la práctica, el ángulo de conducción mínimo es de aproximadamente 90 a 120 grados. 
Entonces, aunque el gráfico muestra que un amplificador de Clase C puede alcanzar el $100\%$ de eficiencia, en la práctica no es así. 
Si se considera el punto teórico del $100\%$ en la curva, entonces el ángulo de conducción es 0 grados, lo cual es imposible de alcanzar. 
Aunque, con amplificadores pulsados de muy alta potencia, el ángulo de conducción puede acercarse a $45$ grados, el espacio entre pulsos es largo y la corriente media de la red es bastante baja.

Terman [1] , sin embargo, sí calculó los límites teóricos con una válvula comportándose de manera correcta, bajo lo que se consideraría funcionamiento normal. Los valores que calculó han resistido la prueba del tiempo y se ha demostrado que son correctos para el caso de la válvula ideal. Dio el valor para un amplificador Clase C como máximo 84%, Clase B como máximo 78.5% y Clase A como máximo 25%. Las clases intermedias de AB 1 y AB 2 son 50% y 55% respectivamente. A partir de muchas pruebas, estos valores están bien probados cuando se examinan los amplificadores prácticos. Sin embargo, en un amplificador práctico nunca alcanzamos las condiciones ideales, por lo que los números siempre son un poco más bajos.


[1] Frederick Emmons 'Fred' Terman (1900 - 1982), fue profesor de Ingeniería Eléctrica, Decano de la Escuela de Ingeniería de la Universidad de Stanford, Estados Unidos. Escribió dos libros de texto estándar, Radio Engineering , publicado por primera vez en 1932, y Electronic and Radio Engineering en 1955.



## Trabajo con curvas características de válvulas

Al diseñar el equipo de válvulas, necesitamos encontrar los datos correctos para la
válvula en particular que se está utilizando. 
Esto se puede obtener de la hoja de datos del fabricante, que hoy en día a menudo se puede encontrar en Internet sin problemas. 
La hoja de datos contendrá mucha información, algunas en formato corto y otras en una serie de curvas. 
A menudo, los datos contenidos en las curvas parecen ser solo muchas líneas y curvas y, a primera vista, no ayudan mucho. 
Sin embargo, si estas curvas se interpretan correctamente, se dispone de una gran cantidad de datos de diseño válidos.

Los datos de forma abreviada se darán a menudo en tablas y normalmente contendrán datos sobre las clasificaciones máximas, voltaje y corriente del filamento, capacitancia de los diversos electrodos y, por lo general, alguna indicación de la frecuencia de funcionamiento máxima y cualquier reducción necesaria. 
Echemos un vistazo a estos datos de formato corto para ver qué nos dicen.

### Máximos ratings

Muy a menudo, estos se dividen en áreas críticas y otras que no lo son tanto. 
Los elementos críticos serán cosas como el voltaje máximo del ánodo, la corriente del ánodo, la disipación del ánodo y la temperatura. 
Exceder cualquiera de estos en una gran cantidad es probable que acorte la vida útil de la válvula o cause una falla catastrófica. 
De manera similar, algunas válvulas necesitan una buena regulación del suministro de filamento para una larga vida. 
Mullard afirma que, en general, se necesita una tolerancia de ± 2,5% o más para una vida útil más larga, aunque algunas válvulas diseñadas para aplicaciones de radio móvil pueden tolerar hasta ± 10% durante períodos cortos.

Otros elementos de clasificación máxima serán la frecuencia de funcionamiento y la potencia o corriente de entrada de CC. 
Obviamente, si la válvula está diseñada para amplificación de RF, no tiene sentido seleccionar una válvula que no funcione correctamente con la frecuencia elegida. 
Normalmente, las válvulas no dejan de funcionar cuando se excede la frecuencia máxima, sino que disminuyen gradualmente en ganancia y eficiencia a medida que aumenta la frecuencia.

### Recomendaciones operativas típicas

Normalmente, el fabricante dará algunos ejemplos de los valores óptimos de voltaje de ánodo y potencia de salida que se esperan cuando se opera con los voltajes y corrientes de suministro recomendados. Por lo general, son una muy buena indicación de si la válvula cumplirá con sus requisitos. 
Es prudente seleccionar una válvula que tenga un poco de capacidad en la mano, para que no se sienta decepcionado cuando finalmente se construya el equipo. Por ejemplo, si la frecuencia de funcionamiento es de 145 MHz, se requiere una válvula con al menos esta clasificación y, a menudo, los fabricantes dan un ejemplo de funcionamiento a 175 MHz. En este caso 145MHz es bastante más bajo y esta válvula sería suficiente. Sin embargo, si se necesitan 432MHz, esta válvula no será de mucha utilidad.

Los valores recomendados para el voltaje y la corriente del ánodo darán una buena vida útil a la válvula cuando se cumpla con la clasificación CCS, Servicio comercial continuo . Estos asumen que la operación es el 100% de las calificaciones las 24 horas del día. Durante períodos cortos (clasificación intermitente), algunos de los parámetros pueden superarse en aproximadamente un 25% sin degradar gravemente la vida útil de la válvula. Estas clasificaciones son las clasificaciones ICAS - Servicio comercial y aficionado intermitentecalificaciones. Estos deben considerarse con cierta cautela. Normalmente se permite aumentar el voltaje del ánodo hasta en un 25% de la clasificación CCS máxima, pero se asume que la corriente del ánodo es la misma que la clasificación CCS. Es decir, puede aumentar uno de ellos pero no ambos simultáneamente. Mullard afirma que donde la eficiencia de la etapa es crítica (menor calor desarrollado) es preferible mantener el voltaje del ánodo tan alto como sea práctico dentro de los valores nominales normales y reducir la corriente del ánodo. Reducir el voltaje del ánodo y aumentar la corriente del ánodo para lograr la misma potencia de entrada no es tan eficiente y, en algunos casos, puede reducir radicalmente la vida útil del cátodo.

Hay tantas válvulas diferentes disponibles que tratar de cubrir todos los tipos sería imposible. 
Por tanto, este capítulo se centrará en una serie de válvulas en particular. 
Este es el popular tetrodo dual, el Mullard QQV06-40, que es parte de una familia de válvulas VHF / UHF. Esto también va bajo el número estadounidense de 5894 y QQE06-40 es la designación posterior. Otros de la serie son las series QQV03-20, QQV07-50, YL-1060 y YL-1070. También cubre la versión de filamento de calentamiento rápido QQZ06-40 /
YL-1030.

Para dar sentido a los datos de la válvula, necesitamos saber cómo interpretar las curvas características y lo que cada curva nos puede decir.

    La conductancia es la inversa de la resistencia: su unidad abreviada se muestra como G o g. Es una medida de qué tan bien un conductor, como un cable de cobre, transporta una corriente. Si el cable tiene un diámetro grande, tiene poca resistencia y puede soportar una gran corriente. El valor de conductancia es el opuesto, si la conductancia es alta, la resistencia es baja.

### Transconductancia S ($g_m$)

Esta es una terminología utilizada tanto para válvulas como para dispositivos semiconductores y es la abreviatura de conductancia de transferencia . En días anteriores se conocía como conductancia mutua y es por eso que la notación abreviada es g m . Si miramos una hoja de datos de válvula o transistor típica, normalmente se indica de una de varias maneras diferentes. La razón es que a lo largo de los años, la industria cambió del antiguo sistema de medición al moderno sistema ISO.

### Unidades de conductancia

En el antiguo sistema de medición, la unidad de conductancia era el mho . Esta es la palabra Ohm escrita al revés y tenía sentido usarla ya que conductancia es el recíproco de resistencia. Una resistencia de 1 Ω es una conductancia de 1 mho. Sin embargo, una resistencia de $0,1 \Omega$ equivale a una conductancia de 10 mho. (El Ohm lleva el nombre de un científico alemán Dr. GS Ohm, por lo tanto, cuando se usa, usa la letra mayúscula. El mho no lleva el nombre de una persona, por lo que usa la letra minúscula).

En las hojas de datos de los fabricantes estadounidenses vemos en los dispositivos más antiguos la transconductancia expresada en micro-mhos , que se abrevia como µ mho . Hoy hemos avanzado y, después de pasar por un período de transición, llegamos al método moderno de establecer la conductancia. En válvulas y semiconductores, la conductancia de transferencia es una medida de cuánto variará la corriente cuando se varíe la señal de entrada. Tradicionalmente, la unidad de medición de transconductancia es el mili-Amp (mA) para un cambio de 1V en la señal de entrada. El símbolo abreviado de transconductancia se conoce como g m . Por tanto, el g m de un dispositivo se da en mA / V.

    Las hojas de datos de válvulas de los fabricantes estadounidenses más antiguos, cuando se caracterizaron los valores de capacitancia de las diversas partes de la válvula, también usaban el símbolo obsoleto micro-micro Farad ( $\mu \mu F$). Hoy se lo conoce como pico-Farad ($pF$).


Cuanto mayor sea el valor de transconductancia, menor debe ser la señal de activación de la red para proporcionar una cantidad adecuada de corriente de ánodo para que fluya. 

Un triodo pequeño puede tener una transconductancia tan baja como $2.5mA/V$ mientras que un tetrodo de transmisión grande puede ser tan alto como $50mA/V$. 
El valor de transconductancia es, por tanto, una indicación de la ganancia de potencia de una válvula, una alta transconductancia equivale a una alta ganancia de potencia.


### Gráficos y curvas de válvulas

Cuando se examinan las curvas de corriente del ánodo de una válvula en una hoja de datos, es normal encontrar que las curvas en el gráfico son de hecho curvas y no líneas rectas. 
Esto puede ser engañoso porque los gráficos de los fabricantes se obtienen de una manera diferente a como se operará normalmente la válvula.

Las curvas se conocen normalmente como curvas de corriente constante, ya que la válvula recibe un voltaje de suministro de ánodo fijo, quizás $100 V$, y la corriente del ánodo varía variando el voltaje de la red de control. 
La válvula se comporta como una resistencia variable a medida que varía la tensión de la red. 
El voltaje de la red se cambia en pasos lógicos y en cada voltaje de la red se registra la corriente del ánodo. 
Cuando se ha trazado un conjunto de curvas, el suministro del ánodo se cambia a otro voltaje de CC, quizás $150 V$, y se mide un nuevo conjunto de puntos. 
Esto se repite tantas veces como sea necesario para cubrir el rango de voltaje del ánodo y luego todos los puntos trazados se dibujan en un solo gráfico. 
Todo esto requiere mucho esfuerzo para obtener resultados precisos y es un proceso largo.

### Funcionamiento push-pull y paralelo

Esta familia de válvulas, aunque originalmente diseñada para operación push-pull, también se puede usar con las dos mitades atadas en paralelo. 
La 6146 tiene una disipación de ánodo de $20W$ (también fue fabricado por Mullard como QV06-20) y cada mitad del QQV06-40 tiene la misma disipación de ánodo.

Hay que aclarar que actualmente solo estamos tratando con la mitad de la válvula, por lo que las corrientes que determinamos son la mitad del total que la válvula consumirá en funcionamiento, pero solo si las dos mitades están atadas en paralelo. 
Las corrientes de red y ánodo serán el doble de los valores que hemos determinado hasta ahora.

Sin embargo, en un amplificador push-pull, las dos válvulas son impulsadas por una tensión de red, que está desfasada 180. 
Cuando la mitad se acciona con una corriente de ánodo creciente, debido a que el voltaje de la red aumenta hacia la condición positiva, la mitad opuesta tiene un voltaje de red negativo en aumento y su corriente de ánodo cae hacia cero. 
El resultado neto es que la corriente de ánodo medida es aproximadamente la corriente de ánodo verdadera de la mitad. 
Al igual que con el amplificador paralelo, cada mitad de la válvula proporciona la mitad de la potencia de salida total, pero en el amplificador push-pull se turnan para suministrar la potencia. 
En un amplificador paralelo, ambas válvulas funcionan juntas.

La otra cosa a apreciar es que las corrientes son el valor pico al que oscilarán la red y el ánodo en la mitad positiva de la señal de entrada. 
El valor medio de una forma de onda sinusoidal es 0,637 del pico. 
Cualquier medidor de corriente de ánodo o rejilla es un tipo de lectura promedio, por lo que las corrientes mostradas no son los valores máximos reales.

### Resistencia del ánodo

La resistencia que exhibe el ánodo está determinada por el voltaje rejilla-cátodo. 
Cuando el voltaje de la red es menor que el voltaje de corte, la válvula parece comportarse como un ánodo a una resistencia catódica muy alta. 
A medida que la tensión de la red aumenta de forma positiva, la resistencia del ánodo desciende a un valor más bajo. 
La figura muestra cómo se comporta una válvula típica cuando se varía el voltaje de la red.

![5-cuatrocientos](fig/valvula_35.png)
Fig: Variación típica de la resistencia del ánodo frente al voltaje de la red de control.

A una tensión de red de $-20 V$, la resistencia del ánodo es alta. 
A medida que se reduce la polarización de la red, la resistencia cae a un valor bajo. La resistencia también está determinada por el voltaje de la rejilla de la pantalla ($V_{g2}$) y los valores más altos de $V_{g2}$ hacen que fluya más corriente ya que la resistencia puede reducirse. 
Cuando el voltaje de la red supera los 0 V, la red ahora es positiva con respecto al cátodo y fluye una mayor corriente de ánodo para el mismo voltaje de ánodo y voltaje de la red de pantalla.

Como era de esperar, cuando el voltaje de la red de la pantalla es mayor, la válvula puede consumir más corriente de ánodo para un voltaje de red particular. 
Tiene más ganancia de corriente y, por lo tanto, una menor resistencia del ánodo. De esta tabla podemos determinar la carga del ánodo requerida para extraer la potencia máxima de la válvula, el criterio de coincidencia conjugado . Se puede determinar que cada mitad de la válvula exhibe una resistencia de aproximadamente $2k\Omega$ cuando el voltaje de la red oscila hasta 0V, que es el máximo impulso posible en la Clase AB 1 . Por lo tanto, en un amplificador push-pull, el circuito del tanque del ánodo debe ser de $\sim 4k\Omega$ ánodo-ánodo para una transferencia de potencia máxima en la salida máxima. 
En funcionamiento en paralelo, la resistencia del ánodo o la carga se reduce a la mitad.

Cuando la válvula no está completamente accionada, la resistencia del ánodo es mayor y ya no cumple con los criterios de coincidencia conjugada, ya que el circuito del tanque del ánodo está fijado a una resistencia de carga particular. Entonces, la eficiencia será mucho menor que en la producción máxima. Teniendo en cuenta la eficiencia típica que se puede lograr en la Clase AB1 , en el mejor de los casos, es solo del $55\%$ aproximadamente. La eficiencia cuando la válvula solo se impulsa a la mitad de la potencia será del $27,5\%$ y si solo se impulsa a un cuarto de potencia será del $12,75\%$ en el mejor de los casos. En el funcionamiento en SSB, la potencia máxima sólo se produce raras veces, ya que la potencia media suele ser mucho menor.

Para la Clase AB2, la válvula se conecta a la corriente de la red. La cantidad que se puede tolerar depende en gran medida de cuán lineal sea la válvula. 
El QQV06-40 no fue diseñado específicamente para funcionar en modos lineales y normalmente no puede tolerar más de aproximadamente 1 mA de corriente de red antes de que se vuelva no lineal. 
Esto limita la potencia adicional que la válvula podría proporcionar en funcionamiento AB2. 
Para la Clase C, donde la linealidad no es un problema, la válvula puede proporcionar una salida de potencia considerablemente mayor con una mayor eficiencia.

### Comprobación de vida util

Si su válvula tiene una vida anterior dudosa, una simple prueba determinará si es adecuada para un uso posterior. 
Todas las válvulas se degradan lentamente a lo largo de su vida, ya que el cátodo puede haber perdido parte de su emisión y ya no se ajusta a los datos del fabricante. Comprobar la emisión es sencillo. Encienda el filamento y aplique un ánodo conocido, una rejilla de pantalla y un suministro de polarización variable a la rejilla de control. Intente elegir un voltaje de ánodo y un voltaje de red de pantalla cercano a los valores recomendados. Un método más seguro y simple es atar la rejilla de la pantalla y los ánodos juntos y las dos rejillas de control juntas para operar las dos mitades en paralelo como un triodo doble. Use un voltaje de suministro bastante bajo, +200V dará un resultado válido ya que esto se traza en las curvas de voltaje del ánodo para el gráfico de 200V V g2.

Al medir la corriente de ánodo a medida que varía el voltaje de la red, podemos ver qué corriente de ánodo es posible. Esto debe hacerse con una corriente de ánodo baja porque si la corriente es demasiado alta, el peligro de dañar la válvula es mayor: no mantenga la corriente de ánodo encendida durante más tiempo del necesario para realizar una medición.

Aplique un voltaje de red negativo alto y disminuya gradualmente hasta que vea que comienza a fluir una pequeña corriente de ánodo. Ajuste el voltaje de la red para obtener una corriente de ánodo lógica, digamos 200mA, esta es la curva marcada como 100mA ya que las dos mitades están ahora conectadas en paralelo y lea el voltaje de la red para esta corriente. Esta corriente de ánodo es la disipación de ánodo máxima permitida de las dos válvulas, siendo $(200V \cdot 0.2A) = 40W$. 
Siempre que la medición se realice rápidamente, no dañará la válvula. 

La QQV06-40 puede tolerar esta disipación durante varios minutos si tiene un enfriamiento adecuado. La válvula se calentará, pero no tanto como para dañarla. Ahora, al encontrar la curva de corriente de ánodo aplicable, puede determinar qué voltaje de red debería ser necesario para esta corriente de ánodo. (En la figura 3.3 el voltaje de la red de corte de 0 mA es de aproximadamente $-26 V$ y la curva de $100 mA$ requiere aproximadamente $-7,5 V$). 
Puede que tenga que estimar a partir de las curvas el voltaje de red aproximado. Si el voltaje de la red es mucho más bajo (menos negativo) que el valor del fabricante, la emisión es baja y puede que no valga la pena continuar.

Las válvulas utilizadas en receptores y válvulas de transmisión de baja potencia son dispositivos de vida útil asombrosamente larga. Es común encontrar que una válvula de 50 años todavía está dentro del 10% de los nuevos valores. Este tipo de válvulas envejecen con gracia y se desvanecen lentamente. Si la emisión es un poco baja, pero superior al 70% de la nueva válvula, a menudo es posible reformar el cátodo en algunos casos. Para hacer esto, se eliminan todos los demás voltajes de suministro y el voltaje del filamento aumenta en aproximadamente un 20% y la válvula se 'hornea' durante aproximadamente una hora. Esto hierve cualquier impureza en el cátodo y la próxima vez que se revisa, la emisión es normalmente mucho mayor. ¡Pero no es un método garantizado!

### Datos de potencia de salida del fabricante

Algunos fabricantes proporcionan datos completos y otros simplemente lo mínimo. Mullard tiende a proporcionar datos completos que contienen algunos datos adicionales útiles. En un amplificador de RF de válvula práctico, hay dos valores de potencia de salida cotizados. La primera es la potencia que la válvula puede suministrar al circuito del tanque resonante: Mullard y la mayoría de los demás fabricantes la denominan "Potencia de salida disponible". Es la potencia que la válvula ha transferido desde la potencia de entrada de CC para aparecer en el ánodo y está disponible para transformarse en la impedancia de salida final de la carga.

El circuito del tanque de ánodo tiene algunas pérdidas asociadas dentro de él, por lo que parte de la potencia del ánodo se pierde en esta red. Lo que queda es la 'Potencia de carga' disponible para excitar la antena. Otros fabricantes utilizan una terminología ligeramente diferente, pero el significado es el mismo. La eficiencia ($\eta$) se determinó construyendo una etapa de salida de muy alta calidad utilizando los mejores componentes disponibles (bajas pérdidas) y luego probando la válvula en esta configuración. 
La eficiencia determinada encaja bien con el valor teórico para una etapa de Clase C. 
Como podemos ver, los ánodos proporcionaron un total de $90W$ al circuito del tanque. 
Después de que la impedancia se transformó a la carga, la potencia medida fue de $78W$.

La potencia de entrada de continua en los ánodos es de  $P= (600 V \cdot 0,2 A) = 120 W$. 
La potencia en los ánodos es de $90 W$ y la entrada de continua es de $120 W$ da una eficiencia del $\eta = 75\%$. 
Sin embargo, la potencia de salida real de la carga es de solo $78 W$ y si se usa para calcular la eficiencia, da un valor de $\eta = \frac{78}{120} = 65\%$. 
Los $12W$ perdidos en el circuito del tanque del ánodo es la razón por la que la eficiencia no es el valor que cita Mullard: dan la eficiencia del ánodo calculada para la potencia de entrada del ánodo medida.

Esto se debe a que después de que se utilizó el amplificador de prueba y se tomaron las lecturas, se apagó el amplificador y se quitó la válvula. Luego, se midió la red de salida del amplificador para determinar la pérdida de inserción que proporcionó cuando estaba bajo energía. Por lo tanto, el valor de 90W indicado no es una medida real sino inferida. La potencia perdida en el circuito del tanque es de 12 W de 90 W, que como porcentaje es 13,33%. La pérdida de 1 dB en números redondos es del 20%, por lo que la pérdida del circuito del tanque asciende a aproximadamente 0,66 dB.

Mullard no contó la entrada de CC de la rejilla de la pantalla ni la entrada de potencia del filamento como parte de la entrada de CC suministrada a la válvula. 
El suministro de la rejilla de la pantalla proporcionó una entrada adicional $P = (250 \cdot 0.018) = 4.5W$. Del mismo modo, la potencia del filamento fue $(6,3 V \cdot 1,8 A) = 11,33 W$.

Si se incluyen estas entradas de energía adicionales, el valor de la eficiencia cae bastante. La potencia total de entrada a la válvula es $(120 + 4,5 + 11,33) = 135,83 W$ y la potencia de salida a la antena es $78 W$. Por tanto, la eficacia real es del 57,4%. Si tenemos en cuenta la fuente de alimentación y otras partes del amplificador que consumen energía (ventiladores, resistencias cuentagotas, etc.), empeora aún más. Si tuviéramos que hacer una medición de entrada de potencia total de la red de CA y compararla con la potencia de salida de RF, es posible terminar con una eficiencia general por debajo del 30%.

Entonces, conocer estos hechos explica por qué cuando medimos un amplificador obtenemos resultados diferentes y no tan buenos como esperábamos. También aclara por qué, al construir un amplificador de RF, la parte crítica es el circuito del tanque del ánodo. Si utiliza componentes inadecuados o con pérdidas potenciales, se puede perder una gran cantidad de energía.

### Trazado de corriente de ánodo cuando se polariza al corte

Cuando se opera una válvula en Clase C, la polarización de la red es un voltaje negativo alto y más de lo necesario para cortar la corriente del ánodo. Cuando la señal de entrada se aplica a baja amplitud, no se desarrolla voltaje suficiente para hacer que la válvula comience a conducir. Cuando se eleva la señal de entrada, la corriente del ánodo comienza a fluir en pulsos de corta duración: este es un tipo de operación no lineal. Cuando la señal de excitación está un poco por encima del voltaje de corte, los pulsos de corriente alcanzan una mayor amplitud y la válvula se puede operar con una alta eficiencia. En la Fig. 3.6 se muestra un gráfico de ejemplo para una señal de entrada que está un poco por debajo de la amplitud máxima.

![5-cuatrocientos](fig/valvula_36.png)

La polarización de la red se establece en –45 V para que no fluya corriente sin accionamiento. A medida que la amplitud aumenta gradualmente, la válvula comienza a consumir corriente a –42 V ya que el pico de la señal de entrada está superando la polarización de la red fija. A un nivel de entrada mayor, la corriente del ánodo aumenta. El eje X muestra el voltaje efectivo de cátodo de rejilla. Cuando el voltaje de la red es positivo, la corriente se eleva a una gran amplitud. Donde la linealidad no es una preocupación, pero se requiere la mayor potencia de salida con buena eficiencia, permitir que la red se vuelva positiva en los picos de la forma de onda sinusoidal de entrada conduce a una mayor potencia y una mejor eficiencia. Sin embargo, hay que pagar una multa por esta opción, que se tratará más adelante.

    Cuando vemos la 'calificación CW' en una hoja de datos, no es lo mismo que entienden los radioaficionados. 'CW' es la abreviatura de Onda continua (o portadora) y significa portadora completa por un tiempo indefinido, a menos que el fabricante establezca un límite de tiempo. Esto es radicalmente diferente al significado de radioaficionado del código Morse, que es una forma de transmisión intermitente.

    La curva típica de corriente de ánodo en la Fig. 3.6 destaca uno de los problemas potenciales de una válvula de alta transconductancia. ¡Si falla la polarización de la red, la corriente del ánodo puede alcanzar un nivel que provoque una fusión en la válvula! Si la polarización de la red desaparece debido a una falla, el voltaje de la red aumentará a 0V. Para una tensión de ánodo de 800 V y una Vg2 de 250 V, intentará extraer unos 350 mA por sección. Esta es una corriente de ánodo total de 700 mA y la disipación del ánodo será (0,7 x 800) = 560 vatios, asumiendo que el suministro del ánodo no colapsa bajo la tensión. La disipación máxima del ánodo del YL-1060 es 60W. (El 1060 en el número de pieza denota un voltaje de ánodo máximo de 1 kV y una disipación de ánodo de 60 W). Esto es 500 W más de lo que la válvula está clasificada y caducará muy rápidamente, probablemente con una envoltura implosionada a medida que el vacío succiona el vidrio ablandado. !

### Corriente máxima del ánodo

La corriente máxima que experimentará la válvula depende de la condición de polarización seleccionada y de la amplitud de la señal de activación. Para el amplificador lineal AF y RF hay tres modos de funcionamiento posibles. Estos son Clase B, Clase AB 2 y Clase AB 1 . Para la Clase B, la válvula está polarizada solo para cortar y el voltaje de la red hace oscilar la corriente del ánodo de manera sinusoidal. Para la Clase AB 2, el ánodo consume una pequeña cantidad de corriente inactiva y para la Clase AB 1, la corriente inactiva es mayor. La figura 3.7 ilustra el pulso de corriente de ánodo diferente para un nivel de excitación de entrada similar.

![5-cuatrocientos](fig/valvula_37.png)

### Efectos del voltaje de suministro

Normalmente, el suministro de alto voltaje utilizado para el ánodo es un tipo simple no regulado, que utiliza un rectificador y un condensador de suavizado. Para muchas aplicaciones esto es aceptable. Si volvemos a las figuras 3.2 y 3.3podemos ver que aunque el voltaje del ánodo puede variar en un amplio rango de voltaje, las curvas son sustancialmente rectas, casi horizontales. La mayoría de los suministros de alto voltaje de ánodo se hunden un poco en el voltaje a medida que aumenta el drenaje de corriente, debido al factor de regulación. Para una válvula de triodo, es el único voltaje de suministro requerido, además del suministro de filamento / calentador. Sin embargo, para el tetrodo y el pentodo se requiere un suministro de voltaje inferior separado y este suministro debe estar bien regulado. Lo mismo se aplica al suministro de la red de control, si no se genera por 'autopolarización' utilizando la corriente de la red.

Muy a menudo, el suministro de la red de la pantalla se deriva del mismo transformador que el suministro del ánodo, utilizando otro devanado secundario y componentes rectificadores. Para suministros de polarización fija, también es común usar otro devanado secundario para derivar este suministro. Como la corriente del ánodo suele ser la más grande, ya que varía, cambia los otros voltajes secundarios. Para los amplificadores lineales que utilizan válvulas de tetrodo / pentodo, esta variación en el voltaje de la rejilla de la pantalla provoca una no linealidad adicional, así como cualquier no linealidad inherente debido a las curvas de transconductancia no perfectas. Si la linealidad es crítica en estos amplificadores, se requieren estabilizadores de voltaje tanto para la red de pantalla como para los suministros de la red de control para mantenerlos dentro de límites estrechos. Este tema se tratará con más detalle más adelante.

    La rejilla de la pantalla en una válvula de tetrodo / pentodo tiene casi tanto control sobre la corriente del ánodo como la rejilla de control. Por lo tanto, si el voltaje de la rejilla de la pantalla se tambalea a medida que varía la corriente del ánodo, alterará la operación.





### Modos de funcionamiento

El amplificador de clase B se utiliza cuando se requiere una buena eficiencia con una linealidad razonable. 
Se usa más comúnmente para amplificadores de audio que necesitan entregar alta potencia pero donde la distorsión no es tan importante. 
En la mayor parte del rango de potencia de salida, la distorsión es bastante baja y es aceptable para el propósito previsto. 
Un amplificador de alta fidelidad no lo es, pero cuando la distorsión no tiene que ser muy baja es suficiente.

La clase B casi siempre se usa en amplificadores de audio push-pull, ya que al operar dos válvulas en anti-fase, los productos de distorsión generados en una mitad se cancelan por una distorsión igual y opuesta en la otra mitad, al menos en teoría. 
Cuando se utilizan amplificadores de RF, no se necesita la operación push-pull: un circuito de un solo extremo hace todo lo que necesitamos. Esto se debe a que el circuito del tanque resonante llena el medio ciclo que falta debido al "efecto volante" de la red. Por tanto, se elimina parte de la distorsión, pero no toda.

Para obtener una verdadera operación de Clase B, la corriente inactiva de la válvula simplemente se lleva a cero ajustando la polarización de la red. En esta condición, el semiciclo positivo de la forma de onda de entrada conducirá la corriente del ánodo sobre la pendiente completa de la curva característica. Esto significa que la corriente de la red fluye sobre la parte superior del pico de la pendiente de la corriente del ánodo de la forma de onda de entrada del variador cuando la red es positiva. Esto se puede ver en la figura 5.1 .

![5-cuatrocientos](fig/valvula_51.png)
Fig 5.1: Curva de funcionamiento de Clase B.

Aunque el caso ideal es que la corriente del ánodo comience en cero, muy a menudo esto introduce una gran distorsión para señales de entrada muy pequeñas y, por lo tanto, normalmente se eleva a aproximadamente 2 a 5% de la oscilación máxima de la corriente del ánodo para evitar este problema. La corriente del ánodo fluye por los 180º completos de la señal de entrada en el semiciclo positivo. Como puede verse, la parte inferior de la curva de transferencia está doblada y esta curvatura causa una distorsión de segundo orden, comúnmente llamada distorsión armónica. Siempre que la desviación de la línea perfectamente recta no sea demasiado pronunciada, la distorsión no será un problema grave. A partir de esta estimación inicial se realizan pequeñas variaciones para optimizar la linealidad. Se observará en el segundo gráfico que cuando la señal de entrada lleva la red a 0 V, interseca la pendiente de la corriente del ánodo en la región del 50% de la corriente máxima del ánodo utilizable. Esto indica que se ha alcanzado el punto de polarización óptimo y, por lo tanto, se puede alcanzar la máxima potencia sin distorsión. Otro método para encontrar el punto de polarización óptimo es ajustar el voltaje de la red hasta que sea cero. Si la pendiente es como se muestra arriba, la corriente del ánodo estará exactamente al 50% del valor pico.

La potencia de excitación requerida para un amplificador de Clase B es bastante alta ya que la válvula consume la corriente de la red durante una gran parte del ciclo de entrada y, por lo tanto, la fuente de excitación debe poder suministrar suficiente oscilación de voltaje de la red sin una caída en la amplitud de la señal. Esto implica una fuente con una impedancia de salida muy baja. Para garantizar que no se produzca una distorsión de la forma de onda de la señal de entrada, esto a menudo significa que la fuente de activación debe poder suministrar una potencia mucho mayor de la que realmente se necesita para hacer oscilar el voltaje de la red en el rango requerido.

### Derivación de la eficiencia de clase B

Terman derivó la eficiencia de una válvula ideal que opera en Clase B al amplificar una forma de onda sinusoidal. La máxima eficiencia posible es $\frac{\pi}{4}$ (78,5%). 
Esto supone que la válvula puede hacer oscilar el voltaje del ánodo desde el voltaje de suministro del ánodo hasta cero voltios. 
Sin embargo, ninguna válvula fabricada puede hacer esto. 
Por lo tanto, la ecuación completa se modifica con el término ($V_{a(min)}/V_{fuente}$). 
La fórmula completa es, por tanto:

$\eta = \frac{\pi}{4} (1 - \frac{V_{a(min)}}{V_{fuente}})$

Para el caso de una válvula de rejilla de pantalla (tetrodo, haz-tetrodo, pentodo, etc.), la limitación de cuán baja Va min puede ser definida por el voltaje en la rejilla de pantalla ($V_{g2}$). 
En caso de que el voltaje del ánodo baje o disminuya, se producirá una distorsión $V_{g2}$.
Por lo tanto, si se considerara una válvula que opera con un suministro de ánodo de $750 V$ y un suministro de pantalla de $250 V$, la eficiencia máxima no puede ser mayor que:

$\eta = \frac{\pi}{4}  (1 - \frac{V_{a(min)}}{V_{fuente}})$

$0,785  (1 - 0,333) = 0,523 (52,3\%)$

Por lo tanto, para una máxima eficiencia, el suministro de V debe ser mucho mayor o $V_{g2}$ debe ser menor.

En la práctica, las cifras de eficiencia típicas son más como el $60\%$ para la caja de válvula ideal a bajas frecuencias, como el audio, y cuando se agregan pérdidas de circuito, se logrará algo menos en la práctica. 
Si la forma de onda se distorsiona, ya no se comporta como una verdadera onda sinusoidal y es necesario modificar la ecuación. 
Esto significa que las mediciones de eficiencia aparente mejor se deben normalmente a alguna distorsión que ocurre en la forma de onda de salida.

Cabe señalar que si el amplificador solo se lleva a la mitad de su potencia potencial, la eficiencia será la mitad de la eficiencia máxima calculada, por lo que es de esperar alrededor del $30\%$. 
Esto está directamente relacionado con el término $V_{a(min)}/V_{fuente}$. 
Si $V_{a(min)}$ solo se reduce a la mitad del valor total ($500 V$ en lugar de $250 V$), el término ( suministro de 1 - Va min / V ) se reduce. En el ejemplo anterior, la eficiencia será $26\%$.

Otro factor es cuánto cae el voltaje de suministro del ánodo por debajo de la corriente máxima del ánodo, si esto es más que un pequeño porcentaje, mejora ligeramente la eficiencia, ya que el valor del suministro V se toma en el caso de la corriente pico del ánodo ($I_{max}$) cuando se alcanza $V_{a(min)}$. 
Por lo tanto, no se le da tanta importancia a la regulación de la fuente de alimentación para amplificadores de Clase B.

La potencia de salida posible para válvulas ideales en un amplificador de clase B push-pull viene dada por:

$P_o = (\frac{I_{max} V_{fuente}}{2})  (1 - \frac{V_{a(min)}}{V_{fuente}})$

Donde $I_{max}$ es la corriente máxima del ánodo en cada válvula.

Nuevamente, no se han considerado pérdidas en el circuito, lo que disipará parte de la potencia de salida potencial.

### Clase AB

Para mejorar la linealidad y disminuir la potencia de accionamiento requerida, a menudo se usa un cambio en el método de polarización. 
Esto ahora se convierte en una mezcla de la Clase B y la Clase A más lineal y, por lo tanto, se llama Clase AB. 
En este método de polarización, la válvula se comporta como un tipo de Clase A a niveles de control de entrada pequeños y se mueve hacia la Clase B más eficiente a medida que aumenta la señal de entrada. 
Hay dos tipos distintos de Clase AB, uno en el que no fluye corriente de red y el otro donde hay algo de corriente de red.

El primer tipo se conoce como Clase AB  y se comporta más como un verdadero amplificador de Clase A en un rango más amplio de corriente de ánodo. 
La penalidad que pagamos es que la disipación del ánodo es mayor que la de la Clase AB 2 y la corriente inactiva es mucho mayor. Por tanto, la eficiencia es menor que AB 2 .

En la Clase AB 2 se utiliza la misma pendiente del ánodo pero la corriente inactiva es más alta que la Clase B normal. Esto significa que la disipación total del ánodo es mayor y, como tal, no podemos obtener la misma potencia pico sin exceder la clasificación de disipación del ánodo. Sin embargo, para un tipo de señal de entrada intermitente como SSB, esto no es una limitación seria.

El factor decisivo de si es AB 1 o AB 2 es si fluye alguna corriente de red. Si la corriente de la red no fluye sobre ninguna parte de la forma de onda de entrada, es Clase AB 1 . Si alguna corriente de la red fluye sobre una parte sustancial de la forma de onda de entrada, es AB 2 . Desafortunadamente, hay muchas permutaciones en este método y si ocurriera una pequeña cantidad de corriente de red justo en el pico de la forma de onda de entrada estrictamente hablando, tiene que clasificarse como AB 2 , cuando en realidad para el 99,9% del ciclo de entrada no flujos de corriente de red. Esto llevó a algunos a llamarlo Clase AB 1.5 , lo cual es una tontería.

Muchas de las válvulas de tetrodo populares para amplificadores de RF muestran una mejora pequeña pero significativa en los productos IMD si fluye un poco de corriente de red, a pesar de que están polarizadas en Clase AB 1 . El gráfico de las dos clases se muestra en la figura 5.3 . Como se puede ver, el modo AB 2 necesita más señal de excitación, pero hace que la corriente del ánodo ascienda por la pendiente y, por lo tanto, permite más potencia de salida. Aunque la Clase AB 2 hace correr la corriente del ánodo más arriba de la pendiente de la corriente del ánodo, luego comienza a acercarse a la región de curvatura superior. Cuanto más empujamos la corriente del ánodo hacia arriba de la pendiente, mayores son los productos de distorsión generados debido a la inclinación de la pendiente en la parte superior.

### Limitaciones de la válvula Tetrode

El tipo de rejilla de la pantalla de válvulas como tetrodes, pentodes y tetrodes de haz tienen un problema inherente particular cuando se polarizan para Clase B, AB 1 y AB 2. Si se examina el gráfico de la figura 5.4, se muestra una versión ampliada de la parte inferior de la curva de transferencia de corriente del ánodo.

![5-cuatrocientos](fig/valvula_54.png)
Fig 5.4: Diagrama de curvatura de la curva de transferencia inferior.

El punto P en el eje de voltaje de la red denota el punto de polarización ideal. En este punto de polarización, la corriente del ánodo inactivo o en reposo es IaS en el eje Y de la corriente del ánodo. Esto se corresponde con el punto Q en el medio de la curvatura en el extremo inferior de la curva de transferencia de corriente del ánodo. La corriente máxima del ánodo lineal se indica con el punto C. La curvatura superior no se muestra en el diagrama.

La verdadera parte en línea recta de la pendiente de transferencia del ánodo comienza en el punto B. Aquí es aproximadamente donde un amplificador Clase AB 1 estaría polarizado; sin embargo, la polarización elegida es para la Clase AB 2 ya que proporciona una mejor eficiencia y potencia de salida. Entre los puntos A y B, la curva es predominantemente de segundo orden e introducirá la generación de distorsión armónica. Sobre la mayor parte de la pendiente del ánodo, desde B es una línea recta, por lo que es sustancialmente lineal. Para una verdadera operación de Clase B, el punto de polarización correcto sería el punto A. Pero esto introduce más distorsión y, a menudo, necesita estar más alineado hacia el punto Q.Un punto de polarización a medio camino entre A y Q generalmente se elige como un compromiso cuando las válvulas de rejilla de pantalla son usado.

Las válvulas de rejilla de pantalla, como un tetrodo de haz, son sensibles al voltaje de la rejilla de pantalla. Si las curvas de corriente constante del ánodo se examinan en detalle para diferentes voltajes de la rejilla de la pantalla, la relación entre la corriente del ánodo y el voltaje de la rejilla de la pantalla se puede representar en un gráfico separado. Muestra que la corriente del ánodo depende en gran medida del voltaje de la rejilla de la pantalla y tiene una ley de 3/2. Esto significa que si el voltaje de la rejilla de la pantalla se eleva de, digamos, 300 V a 500 V, la corriente del ánodo se duplica para el mismo voltaje de la rejilla de control y voltaje del ánodo. La razón por la que esto ocurre es que la rejilla de la pantalla actúa como un mejor acelerador de los electrones del cátodo que fluyen hacia el ánodo.

Para corregir esto, la corriente del ánodo debe reducirse para mantener la corriente y la disipación del ánodo dentro de los límites seguros para la válvula. El aumento del voltaje de polarización de la red puede hacer esto, de modo que sea más negativo. Esto restaura la corriente inactiva de nuevo al valor requerido y la coloca firmemente en la parte inferior de la curva de transferencia del ánodo inferior en el punto Q.Como el voltaje de la red ahora es mucho más alto, se requiere un voltaje de conducción de la red mayor y, a medida que la corriente de la red fluye por debajo Operación AB 2 , esto significa que se requiere un nivel de accionamiento más alto.

Esto va en contra de todo el razonamiento aparente de aumentar el voltaje de la red de la pantalla para obtener más ganancia y potencia de salida. La potencia de salida de primer orden es la oscilación del voltaje del ánodo y la capacidad de la válvula para extraer suficiente corriente. Si el voltaje del ánodo es bajo, la oscilación del voltaje del ánodo y la corriente también serán bajos para la misma señal de excitación. Para una válvula típica, si el voltaje del ánodo se eleva de 1 kV a 2 kV, la potencia de salida potencial aumenta en un factor de cuatro.

El otro gran problema de aumentar el voltaje de la rejilla de la pantalla es que el voltaje del ánodo no puede oscilar mucho por debajo del voltaje de la rejilla de la pantalla. Si el voltaje del ánodo oscila por debajo del voltaje de la rejilla de la pantalla, existe un cátodo virtual, la rejilla de la pantalla intenta asumir el papel del ánodo y consume una corriente de pantalla excesiva. Esta corriente de ánodo desviada hace que la forma de onda de voltaje del ánodo se aplana en la parte superior, ya que la corriente de ánodo no puede oscilar lo suficiente para seguir la forma sinusoidal correcta. El resultado es que se generan productos de distorsión cuando esto ocurre.

Al aumentar el voltaje de la rejilla de la pantalla, llega el punto en el que puede hacer oscilar el voltaje del ánodo hacia abajo a un voltaje más alto y, por lo tanto, limita la corriente del ánodo que es posible extraer sin que fluya una corriente excesiva de la rejilla de la pantalla. Reducir el voltaje de la rejilla de la pantalla tiene el efecto contrario. El ánodo ahora puede bajar el voltaje antes de alcanzar el voltaje de la rejilla de la pantalla. 
Este efecto fue aprovechado en el amplificador lineal G2DAF, que cubriremos más adelante.

### Clase C

Aunque el amplificador de Clase C no es adecuado para modos que requieren una curva de transferencia lineal, tiene sus usos para transmisores modulados por ánodo AM, así como para modos FM y CW. La base de funcionamiento es que la corriente del ánodo se corta por un gran voltaje de rejilla negativo y la corriente del ánodo solo fluye en pulsos estrechos para una parte de la forma de onda de la señal de rejilla positiva. La curva de transferencia de corriente del ánodo se muestra en la Fig . 5.5 .

![5-cuatrocientos](fig/valvula_55.png)
Fig 5.5: Curva de corriente del ánodo de clase C.

Normalmente, la polarización de la red se establece en aproximadamente el doble del valor requerido para simplemente cortar la corriente del ánodo. Esto significa que en la mitad positiva de la señal de activación de entrada no fluye corriente de ánodo hasta que el voltaje de la red se acerca al punto de cátodo de la red de 0V. Una vez que se ha superado el punto de 0 V, la corriente del ánodo se conduce directamente a la parte superior de la pendiente y por encima de la parte superior en una pequeña cantidad. Esto coloca la corriente del ánodo en saturación y cualquier pequeña variación en el nivel de la señal de entrada tiene poco efecto sobre la corriente del ánodo y, por lo tanto, la potencia de salida.

La potencia de conducción de la red es el más alto de todos los modos de funcionamiento, pero como la resistencia de la red es bastante alta, la potencia requerida no es excesiva. La ganancia que puede producir un amplificador de potencia de Clase C típico es de alrededor de 20 veces (13 dB) a alrededor de 40 veces (16 dB), dependiendo de la frecuencia y el nivel de salida de potencia. La corriente de la red es bastante alta y, por lo tanto, es esencial una válvula diseñada para atender esta alta corriente. Afortunadamente, a los niveles de potencia que utilizan los aficionados, esto significa que muchas válvulas normales pueden funcionar en Clase C sin problemas graves. Sin embargo, hay algunas válvulas que no deben usarse con una corriente de red alta, ya que la estructura de la red puede dañarse.

De todas las clases operativas, la Clase C tiene la mayor eficiencia y es común una cifra del 75% o un poco más alta. En un circuito de ánodo realmente eficiente es posible ver cifras un poco por encima del 80% a bajas frecuencias.

Aunque los pulsos de corriente del ánodo son mucho menores que los 180º de la señal de entrada y, por lo tanto, son ricos en armónicos, el circuito del tanque del ánodo resonante restaura la forma de onda de la señal de entrada a una forma de onda sinusoidal casi perfecta debido al efecto de volante. Solo existen armónicos de la frecuencia fundamental en el ánodo. Estos son suprimidos en gran medida por el circuito del tanque cargado Q, y cualquier supresión adicional se puede realizar con un filtro de paso bajo conectado en serie con la salida si se considera que son excesivos. En un circuito de tanque correctamente proporcionado, cuando el circuito del ánodo entra en resonancia, los armónicos normalmente disminuyen cuando el circuito del tanque se ajusta a la potencia de salida máxima.

El tema de los circuitos del tanque de ánodo y la necesidad de seleccionar el valor correcto de Q cargado se tratan más adelante en otro capítulo.

### Controlar la potencia de salida de RF

En el amplificador de Clase C, el método normal para impulsar la rejilla de un tetrodo de haz es suministrar un nivel de portadora fijo desde la etapa de activación. La polarización de la red puede ser mediante un suministro de polarización fijo, que tiene alguna variación para establecer el voltaje de corte correcto, o para utilizar la corriente de la red que fluye en una resistencia conectada desde la red a tierra. De los dos, este es el método más simple. Dado que la corriente del ánodo depende del voltaje de polarización de la red, si se elimina la señal de excitación, el voltaje de la red aumenta a 0 V y la corriente del ánodo aumentará al máximo, ya que el voltaje de polarización de la red desarrollado a través de la resistencia de la red ha desaparecido.

Para evitar esta condición, un método común con los transmisores de CW es disponer algún método de modo que si la polarización de la red falla, el voltaje de la pantalla se lleva a un voltaje bajo, reduciendo así la corriente del ánodo a un nivel bajo. Esta técnica se llama etapa de sujeción . En la figura 5.6 se muestra una etapa de sujeción simple que usa un triodo para controlar el voltaje de la rejilla de la pantalla del amplificador de RF .

### Capacidad de salida de potencia de una válvula

Una pregunta que se hace a menudo es: "¿Cuánta potencia de RF puedo obtener de una válvula en particular?" La respuesta simple es: "Depende del tipo de amplificador que desee utilizar y de la frecuencia de funcionamiento". 
Es importante comprender cómo se clasifican las válvulas en cuanto a potencia. 
En elementos tales como motores de combustión interna y motores eléctricos, la forma general de calificarlos es cotizar la potencia de salida en caballos de fuerza o kilovatios. 
Las válvulas generalmente no se clasifican de esta manera, aunque el fabricante a menudo dará un ejemplo para mostrar su capacidad. 
Las válvulas se clasifican en lo que no pueden entregar como potencia. 
Aunque esta parece una forma extraña de calificarlos, es por una razón muy lógica. 
La razón se debe a la eficiencia con la que la válvula se puede operar de manera segura en un circuito práctico.

### Eficiencia

La eficiencia es la medida de cuánta potencia de salida podemos lograr para una determinada cantidad de potencia de entrada. 
Se da en términos porcentuales. Si toda la potencia de CC alimentada a una válvula estuviera disponible en alguna otra forma, por ejemplo, audio o potencia de RF, entonces tendría una eficiencia del 100%.

La Ley de Conservación de Energía establece que: “La energía no se puede crear ni destruir. Solo se puede convertir de una forma a otra ". El quid de esta ley es que, aunque la potencia de salida es diferente a la potencia de entrada, siempre se puede explicar la diferencia entre las dos.

En una válvula, o cualquier otro dispositivo que convierte la energía en que la energía hacia fuera , la diferencia se encuentra por lo general para ser calor generado dentro del dispositivo durante el proceso de conversión. En las válvulas, esto se denomina disipación del ánodo y se expresa en vatios. La potencia de CC alimentada al dispositivo que no se pudo convertir en potencia de salida del tipo requerido aparece como calor. Solo se 'pierde' en la medida en que no tiene la forma que nos gustaría: se contabiliza completamente como energía térmica. La eficiencia ($\eta$) generalmente se considera que es:


$\eta = \frac{P_o}{P_{DC}}$


Si se alimenta una potencia de entrada de $P_{CC} = 150 W$ a un amplificador y aparece una potencia de $P_o =100 W$ en la salida, la eficiencia es del $\eta = 66,66\%$.

#### Disipación de ánodo

La gran mayoría de la energía perdida se atribuye a ser consumida por el ánodo. Esto no es estrictamente cierto: algunos pueden usarse para calentar otras partes de la válvula, pero el concepto es una forma viable de expresar el mecanismo. Si no se perdiera energía en el ánodo, entonces el proceso de conversión sería perfecto y la eficiencia sería de 1l$ 100%$.

La cantidad de energía que una válvula en particular puede convertir de la potencia de entrada de CC a la potencia de salida útil está determinada únicamente por la disipación segura del ánodo de la válvula. Si la potencia que se está disipando es muy superior a la clasificación máxima del fabricante, se sobrecalentará y dañará la válvula. La válvula solo puede deshacerse de la energía térmica hasta una cierta velocidad, transfiriendo la energía térmica a otro objeto. El enfriamiento del ánodo y otras partes de la válvula es, por lo tanto, un factor importante para determinar cuánta disipación del ánodo es posible.

Calcular la potencia de salida de la disipación del ánodo

La fórmula para convertir la clasificación de disipación del ánodo en vatios en potencia de entrada utiliza la fórmula:

$P_{in} = P_{diss} (1 - \eta )$

Dónde:

$P_{dc}$ es la entrada de energía de CC en vatios

$P_{diss}$ es la disipación nominal del ánodo

$\eta$ es la eficiencia numérica (por ejemplo, $75\% = 0,75$).

Para una válvula como la RCA-811, la disipación máxima del ánodo es de $45 W$. 
Suponiendo una eficiencia de ánodo del $75\%$ para un amplificador de audio de Clase B, puede manejar con seguridad una entrada de CC de 180 W.

La potencia de salida es simplemente la potencia de entrada multiplicada por la eficiencia:

$P = 180W \cdot 0,75 = 135 W$

La disipación del ánodo es (entrada de potencia - salida de potencia) es 

$P_{diss} = (180 - 135) = 45W$.

Esto supone que la salida de potencia es continua. 
Cuando la salida no está al 100% del valor nominal, se puede aplicar una reducción inteligente para obtener una condición de trabajo segura. A esto lo llamamos ciclo de trabajo .

### Ciclo de trabajo

En el caso del código Morse, la relación entre puntos y espacio es 1:1, por lo que si se transmite una cadena de puntos, el transmisor está encendido la mitad del tiempo y apagado durante el resto. 
Por tanto, el ciclo de trabajo es del $50%$. 
Bajo esta condición, usando el ejemplo anterior, se podría duplicar de manera segura para la entrada de CC y la potencia de salida de RF sin exceder la disipación del ánodo de 45 W. La disipación citada es el valor promedio , por lo que aunque durante la transmisión del punto se disipa el doble de su disipación nominal, durante el período espacial la disipación es cero. Tiene un período de reposo donde el calor puede ser transportado.

En el radar y transmisores pulsados similares, el nivel de potencia es muy alto durante el pulso de transmisión, pero el tiempo entre pulsos es largo. 
Por lo tanto, la válvula tiene tiempo suficiente para alejar el calor y la disipación promedio del ánodo es baja.

Al evaluar hasta dónde se puede empujar la envolvente antes de que la válvula se someta a una tensión indebida, necesitamos conocer el ciclo de trabajo de los diversos modos de uso común. Para el código Morse, si se transmite una cadena de guiones, el tiempo de activación es el 75% del tiempo total, porque la relación de guiones a espacio es 3:1. 
Por lo tanto, en este modo solo podríamos aumentar la potencia de entrada de CC en 1,33. 

Para FM, donde la portadora está al 100% durante las transmisiones, claramente la disipación del ánodo de 45 W es el factor determinante y funcionar a más de la potencia de entrada de CC segura sería inaceptable.

Para la operación de banda lateral única, la voz promedio tiene un ciclo de trabajo de entre el 30 y el 40% de los picos, por lo que la clasificación podría aumentarse bastante antes de que se exceda la disipación promedio del ánodo. Sin embargo, si se usa un procesador de voz, esto aumenta la amplitud de voz promedio a aproximadamente un 60% y dicta la cantidad de energía adicional que podemos ejecutar con seguridad.

Sin embargo, los transmisores de aficionados normalmente no funcionan de forma continua: tienen períodos cortos de transmisión y aproximadamente el mismo período de recepción, por lo que la potencia máxima podría aumentarse un poco. Si el equipo se usa en un concurso, las cosas son muy diferentes, especialmente si se usa un procesamiento de voz pesado. Estos factores determinan cuánto 'abuso inteligente' es capaz de tolerar la válvula.

### Eficiencia potencial del ánodo

Varias fuentes han calculado a partir de los primeros principios cuál puede ser la eficiencia máxima del ánodo para los diversos modos de conducción de uso común. Los modos de funcionamiento más populares son Clase A, Clase B y Clase C. A estos se añaden otras dos variaciones de Clase AB 1 y Clase AB 2 , que son combinaciones de Clase A y B. El modo de funcionamiento que se utilice dependerá de la aplicación y otros factores.

Terman primero calculó estos valores máximos y varios otros han confirmado que los valores son el límite superior para una válvula ideal. Sin embargo, nadie hace una válvula ideal, solo están disponibles válvulas prácticas. Por lo tanto, aunque útil, es simplemente una guía sobre la máxima eficiencia absoluta que sería posible lograr si estuviera disponible la válvula ideal.

Generalmente, la eficiencia del ánodo disminuye a medida que aumenta la frecuencia, por lo que un valor para un amplificador de audio siempre será mejor que un amplificador de HF y un amplificador de VHF / UHF. Por lo general, no es posible citar las hojas de datos de las válvulas de los fabricantes para los tres tipos de operación, ya que una válvula en particular no está hecha para atender a todos los tipos. Sin embargo, existen datos suficientes para operar en áreas de frecuencia, por ejemplo, audio y RF baja o HF y UHF.

Una válvula típica de alta potencia a frecuencia de audio se acercará al valor teórico pero se quedará corta por una pequeña cantidad. De manera similar, una válvula de microondas a baja frecuencia de RF es notablemente mejor que en el rango superior en las bandas de microondas. 
El triodo 2C39A a 3GHz tiene una eficiencia de aproximadamente 25 a 30% en el mejor de los casos para operación de Clase C, a 144MHz su eficiencia es de alrededor del 70% y a 432MHz es típicamente del 60%. Esta tendencia es común para todas las válvulas. Pueden indicarse valores teóricos típicos para las distintas condiciones de funcionamiento:

La clase C tiene la más alta al $\eta = 84\%$;

La clase B es la siguiente mejor con $\eta = 78,5\%$;

La clase A es la mas baja en alrededor del $\eta = 50\%$.

Los modos intermedios de la Clase AB 1 y AB 2 son similares, la Clase AB2 es ligeramente mejor al 65% y la Clase AB1 al 55%. Sin embargo, debe repetirse que estos son los valores teóricos máximos cuando se opera a bajas frecuencias . A mayor frecuencia, estos valores son más bajos, ¡a veces mucho más bajos! Generalmente, cuanto mayor es la linealidad del amplificador, menor es la eficiencia potencial.

### Eficiencia del circuito

La eficiencia general de un amplificador de potencia no está completamente determinada por la eficiencia del ánodo de la válvula. Hay otros factores que agregan una pérdida adicional al proceso de transformación de la entrada de CC a la salida de RF. El factor principal es la pérdida inherente en la red de ánodos. A frecuencias de audio y RF bajas, estas pérdidas son bastante pequeñas y, a menudo, pueden ignorarse. Este no es el caso cuando la frecuencia es muy alta. En UHF y por encima de la red de ánodo, las pérdidas pueden ser significativas y pueden consumir la energía disponible para alimentar la antena.

Como ya se ha mencionado, los fabricantes de válvulas suelen proporcionar dos medidas de salida de potencia de RF diferentes en sus hojas de datos. Estos son la potencia de ánodo disponible y la potencia de salida útil . El primero de ellos es la potencia calculada que sale del ánodo, el otro es la potencia real medida en el conector de salida del amplificador. La diferencia entre los dos es la potencia perdida en la red de transformación del ánodo, que se puede medir sin la válvula en el circuito.

### Enfriamiento

El enfriamiento es una de las áreas más críticas en cualquier diseño de amplificador de alta potencia. Si el calor generado en la válvula no se elimina con la suficiente rapidez, superará la temperatura de funcionamiento segura.

Cuando se diseña una válvula, el fabricante decide el método de enfriamiento óptimo para adaptarse a la aplicación que probablemente se utilizará. Para el tipo de válvulas de envoltura de vidrio más pequeñas, a menudo el enfriamiento por convección simple es suficiente para mantener la válvula por debajo de cierta temperatura máxima. Para válvulas grandes y de mayor potencia, la envolvente tiene un área de superficie mayor y, a veces, el enfriamiento por convección es adecuado. Las válvulas de envoltura de vidrio disiparán el calor más fácilmente si se les aplica una suave corriente de aire frío. A menudo, el volumen de aire requerido no es grande y esto es suficiente para enfriar la válvula. En otros, se fabrican chimeneas de vidrio especiales que encajan sobre la válvula y se sopla un mayor volumen de aire desde debajo de la válvula para formar un método de transferencia de calor eficiente.

Para las válvulas físicamente pequeñas, donde el área disponible es pequeña, a menudo se instala un radiador de enfriamiento con aletas en el ánodo y se fuerza aire a través de él para mantener el ánodo a una temperatura segura. En otras válvulas de alta potencia, la refrigeración por aire simplemente no es suficiente y se instalan camisas de refrigeración líquida para transportar el calor.

Si la velocidad a la que el calor puede ser transportado y disipado es mucho mayor de lo que genera la válvula, entonces es factible aumentar la disipación del ánodo a un nivel mucho más alto. La serie 2C39 de triodos de microondas está equipada con un enfriador de ánodo con aletas que permite una disipación máxima de ánodo de 100W. Sin embargo, es sencillo reemplazar este enfriador de aletas con una camisa de enfriamiento líquido. Con un sistema de refrigeración líquida, la disipación del ánodo se puede aumentar hasta cuatro veces más que la refrigeración por aire normal.

 
#### Eficiencia real

Aunque el cálculo de la eficiencia es un factor fácil de determinar, nos puede sorprender que lo que logremos en la práctica sea mucho menor de lo esperado. Es fácil medir la potencia de entrada de CC: simplemente mida la corriente del ánodo y el voltaje y multiplique los dos. También es sencillo medir la salida de potencia de RF utilizando un vatímetro calibrado. Pero hay casos en los que obtenemos resultados inesperados.

Suponga que el amplificador en uso tiene una eficiencia conocida del 60% cuando se conduce a su salida nominal completa. El modo de funcionamiento es el código Morse. Cuando el amplificador se cambia al modo de transmisión, pero antes de presionar la tecla, muestra algo de corriente inactiva en el medidor de corriente del ánodo y, por supuesto, el vatímetro de RF muestra una salida cero. ¿Cuál es la eficiencia en esta condición? La respuesta es 0%, porque aunque el amplificador está consumiendo energía, no hay energía de salida.

Ahora supongamos que enviamos una larga cadena de puntos que están correctamente espaciados con la longitud del punto siendo el mismo período de tiempo que el período de espacio. ¿Cuál es la eficiencia ahora? Cuando la tecla está presionada, el amplificador suministra toda la potencia a la eficiencia medida, por lo que está funcionando al 60% de eficiencia. Pero en el tiempo de funcionamiento clave tiene una eficiencia del 0%. Por tanto, la eficiencia media es el máximo dividido entre dos o 30%. Lo mismo se aplica a la potencia de salida media. Cuando la tecla está presionada, transmite, digamos, 100W, pero con la tecla hacia arriba es, por supuesto, cero potencia. Por tanto, la potencia media es de 50 W.

Si ahora enviamos una larga serie de guiones, y nuevamente asumimos que se observa el período correcto del guión al espacio, ¿cuál es la cifra de eficiencia en esta condición? El período del guión de tecla hacia abajo es tres veces más largo que el período de espacio, por lo que el ciclo de trabajo es del 75%. Por tanto, la eficiencia media es del 60% x 0,75 = 45%. La potencia media es de 75 W.

Cuando se utiliza SSB como entrada, se aplica el mismo argumento. La SSB con voz humana tiene un ciclo de trabajo de aproximadamente un 30% de relación promedio a pico. Entonces, la eficiencia promedio que logramos es 60% x 0.3 = 18%. Aunque esto suene terrible, debe verse en el contexto correcto. Bajo esta condición de operación, la disipación del ánodo también promedia solo el 30% de la condición de llave completa. Para un modo como FM, donde la portadora está al 100% de la potencia, la disipación del ánodo es la máxima.

#### Máxima eficiencia

En un amplificador de RF, la red de adaptación del ánodo normalmente se ajusta para entregar la máxima potencia cuando la válvula está completamente accionada. Esto significa que cuando la válvula no se acciona hasta su potencia máxima de entrada, la eficiencia se ve afectada. Esto se debe a que la combinación se ha ajustado para adaptarse a la potencia máxima, en salidas de potencia más bajas, la válvula ahora no está correctamente emparejada para entregar la potencia potencial total. Esto escala casi linealmente en amplificadores prácticos. Al igual que con el ejemplo de puntos de código Morse anterior, se aplica el mismo efecto.

Si la válvula solo se acciona para dar la mitad de la potencia máxima, la eficiencia cae al 50% de lo que sería a plena potencia máxima. Si la eficiencia máxima es del 60%, la válvula solo ofrece un 30% de eficiencia para esa entrada de alimentación de CC en particular. Por lo general, esto no es un problema, ya que la clasificación de disipación segura del ánodo es mucho más alta a máxima potencia. Podemos trazar la curva de eficiencia contra la entrada de energía y es esencialmente una línea recta que comienza en 0% y alcanza el punto de eficiencia máxima en la parte superior de la línea.

#### Disipación de ánodo en condiciones inactivas

La disipación permitida del ánodo cuando una válvula está polarizada en un modo lineal es motivo de preocupación. Normalmente, la corriente inactiva se establece en un valor bastante alto, de modo que la válvula en condiciones de baja transmisión funciona en Clase A. A medida que la transmisión aumenta gradualmente, se mueve a la región de operación de Clase B. El modo de funcionamiento con la corriente inactiva más alta es la Clase AB 1 . Encontrar la corriente inactiva óptima es a menudo una cuestión de experimentación. Por un lado, deseamos tener una buena linealidad a niveles de excitación pequeños, pero la disipación del ánodo es una preocupación. Si reducimos la corriente inactiva, reducimos la disipación del ánodo, pero podemos reducir la linealidad para niveles de excitación pequeños. Afortunadamente, Eimac encontró la respuesta definitiva a este problema.

Eimac recomienda que cuando una válvula está polarizada para la Clase AB 1, la corriente inactiva debe establecerse de modo que el nivel de disipación del ánodo sea solo del 80% de la disipación máxima del ánodo de la válvula. Esta es la mejor configuración de compromiso. Esto significa que para una válvula con una disipación máxima de ánodo de 250W, como la serie 4CX250, la disipación de ánodo será de 200W. Para el voltaje de suministro de ánodo de CC recomendado de 2 kV, la válvula 4CX250 debe consumir 100 mA de corriente inactiva.

Si examinamos la hoja de datos de Eimac para el triodo 3-500Z, vemos su recomendación para el funcionamiento en Clase AB 2 . El 3-500Z es más lineal con un suministro de ánodo de alrededor de 3 kV y cuando está inactivo a este voltaje consume 62 mA. Esta es una disipación de ánodo de 186W. Sin embargo, para proporcionar este estado, el voltaje del cátodo debe elevarse a + 10 V en lugar de la condición normal de polarización cero. Si el suministro del ánodo se reduce a 2,5 kV, se puede prescindir de la polarización adicional de + 10 V. Ahora consume una corriente inactiva de 130 mA cuando se ejecuta en una verdadera condición de polarización cero. En esta condición, la disipación del ánodo inactivo es de 325 W.

Si el suministro del ánodo se eleva al máximo de 3,5 kV, la polarización del cátodo debe elevarse a + 15 V para mantener la corriente del ánodo bajo control. Esto está inactivo a 53 mA, una disipación de 185 W.

Con la unidad completa en el suministro de 3 kV, consume una potencia de entrada de CC de un solo tono de 1200 W y normalmente produce 740 W. Por tanto, la disipación del ánodo es (1200 - 740) = 460W. En el voltaje de ánodo más bajo de 2.5kV, consume una entrada de CC de 1kW y produce 600W, una disipación de 400W.

Si elegimos el suministro máximo de ánodo de 3.5kV, la entrada de CC es 1400W, la salida es 890W y la disipación del ánodo es 510W. Se establece que la disipación máxima del ánodo es de 500 W.

En todos esos modos, necesita aproximadamente 45 W de potencia de accionamiento cuando está en una configuración de red con conexión a tierra.

#### Amplificadores de red con conexión a tierra

Estos se tratan con más detalle más adelante, pero un factor interesante acerca de ellos es la aparente eficiencia posible. Al medir la potencia de salida y la potencia de entrada, se obtiene una cifra de eficiencia mayor de lo que parece realista. Esto finalmente se atribuyó al hecho de que parte de la potencia de accionamiento de entrada se alimenta a través de la válvula desde el cátodo al ánodo y aumenta la potencia del ánodo. Si se utiliza una cantidad excesiva de unidad, es posible un aumento de la potencia de salida de hasta un 11%.

Como resultado de esto, la FCC en Estados Unidos cambió la redacción de la licencia de aficionado para atender este efecto. Afirmaron que si el porcentaje de potencia de alimentación estaba por encima de un cierto valor, la entrada de CC al ánodo debía reducirse para mantenerse dentro del nivel de entrada de CC de 1 kW permitido.

En un amplificador de rejilla con conexión a tierra, el cátodo y la señal del ánodo están en fase y se suman. En el amplificador de cátodo conectado a tierra, están desfasados 180º y, por lo tanto, la potencia de alimentación se resta de la potencia del ánodo.

    Esto fue en los días antes de que la salida de RF pudiera medirse con precisión y los transmisores eran de los tipos AM y CW, es decir, antes de SSB, donde solo se medía la potencia de salida. Hoy en día, la potencia de entrada de CC no se usa excepto, en algunos países, para los modos FM y CW. Si tiene que usar una potencia de entrada de CC de 1200W para alcanzar 400W PEP, eso no es motivo de preocupación, solo afecta al operador que tiene que pagar la factura de la luz. Muchos de los primeros equipos comerciales de aficionados se especificaron como entrada PEP y no como salida. Esto hizo que el equipo se viera mejor, pero cuando se consideró el factor de eficiencia, a menudo dieron menos de la mitad de la potencia de la cifra de entrada citada. El primer Yaesu FTdx500 era un transceptor de HF con una potencia de entrada PEP de 500W: normalmente daba una salida PEP de unos 180W.





## Diseño de amplificadores Clase C empleando valvulas

Los amplificadores de potencia clase C se caracterizan por proveer amplificación con alta eficiencia y son utilizados principalmente en aplicaciones de RF, donde se requiere un incremento en el nivel de potencia y no se requiere linealidad entre la tensión de entrada y tensión de salida. 

Los amplificadores Clase C pueden ser modulados en amplitud por inyección de la tensión modulante en la placa, reja ó grilla pantalla, reja ó grilla control ó cátodo. También son utilizados para amplificar una portadora modulada en frecuencia. 

En el circuito de placa la eficiencia puede ser tan alta como el $85\%$, pero la mayoría de los amplificadores se diseñan para operar con eficiencia del orden de $75\%$, debido a que incrementar el rendimiento del  $75\%$ al $85\%$ significa un aumento importante de la potencia de reja, además grandes eficiencias son obtenidas con elevadas tensiones de placa. 

En un amplificador Clase C, la reja se polariza por  debajo del corte y la corriente de placa circula durante menos de $180$ por cada ciclo de la tensión de excitación de reja, por esto los amplificadores Clase C, no son usados en aplicaciones de audio debido a que la tensión de salida no es proporcional a la tensión de entrada y los pulsos de placa tendrían una excesiva distorsión. 

Sin embargo un amplificador Clase C puede ser usado en aplicaciones de RF, si el circuito de placa es resonante a la componente fundamental o algunas de las armónicas de la frecuencia de excitación de reja, de esta forma el pulso de la corriente de placa es filtrado por el circuito resonante de placa para proveer una salida muy cercana a una sinusoide. 
Cuando el amplificador es usado como multiplicador de frecuencia, el circuito de placa deberá resonar a alguna de las armónicas de la tensión de excitación, la potencia de salida se reduce considerablemente, siendo la eficiencia del orden del $40$ a $50\%$ cuando se los utiliza como dobladores y del $15$ a $20\%$ como triplicadores.

### DISEÑO DE UN AMPLIFICADOR CLASE C CON VALVULA DE VACIO 

El análisis de un amplificador clase C, se lleva a cabo en un circuito básico con una sola válvula. 
Si se usa operación en paralelo o push pull, la potencia de salida, entrada a grilla y requerimientos de la fuente de potencia se multiplican por el número de tubos utilizados. 

![5-cuatrocientos](fig/MALLO_1.png)



#### 1. Determinación de la potencia de salida ($P_o$): 
Además de la potencia real sobre la carga que necesitamos, se deben incluir las pérdidas de potencia en el circuito tanque de placa y en el circuito de acoplamiento. 
Si la frecuencia de operación está por encima de los $50 MHz$, se debe considerar la pérdida en los dieléctricos de los materiales aislantes de la envolvente de la válvula. 

#### 2. Cálculo de la disipación de placa por tubo aproximada ($P_{bs}$): 

$P_{bs} = k P_o$           
     
     
|k    |$\theta_p$ (Recomendado)|Armónicas   | 
|-----|--------------|------------| 
|0,33 |de 120º a 150º| Fundamental| 
|1,0  |de 90º a 120º | 2° armónico| 
|2,33 |de 80º a 110º | 3° armónico| 
|3,44 |de 70º a 90º  | 4° armónica| 
|4,35 |de 60º a 75º  | 5° armónica|

Existen diversos ángulos de circulación de la corriente de placa recomendados, de acuerdo con el uso. 

#### 3. Selección de la válvula que satisfaga las condiciones requeridas: 

1. Disipación de placa debe ser mayor al encontrado en el punto anterior. 
2. Frecuencia de operación. 
3. Tensión de placa, grilla pantalla y filamento requerida. 
4. Sistema de enfriamiento (aire, agua, aire forzado). 
5. Características físicas: dimensión y forma, resistencia mecánica a golpes y vibraciones, etc. 
6. Potencia de excitación de grilla. 
7. Tipo de servicio (CW, AM, etc.) 

#### 4. Determinación de la tensión de polarización $E_{bb}$:

En la mayoría de los manuales, la información que se da es la máxima tensión de placa admisible. 

En este caso debemos tomar un margen de seguridad o tomar los valores recomendados de acuerdo al tipo de servicio que se va a utilizar. 

#### 5. Cálculo de la corriente promedio de placa $I_{ba}'$: 
    
$I_{ba}' = (P_o + P_{bs})/ E_{bb}$

Este valor no debe exceder la máxima corriente admisible de la válvula. 

#### 6. Cálculo del valor pico de la corriente de placa por tubo: 

Se estima un valor:               
$i_{b(max)} = 4 I_{ba}'$

Esta es la relación aproximada entre el valor pico y el valor medio de la corriente de placa. 
Es necesario hacer una estimación de $i_{b(max)}$ para realizar el análisis gráfico. 

#### 7. Localización del punto A sobre las características de corriente constante: 


![5-cuatrocientos](fig/MALLO_5.png)


El punto A determina: $e_{c(max)}$, $e_{b(min)}$, $i_{b(max)}$, $i_{c(max)}$, $i_{c2(max)}$, este punto A se ubica  sobre la curva de corriente constante cuyo valor se obtuvo en el pto. 6. 

En caso de utilizar un triodo con una gran excitación de grilla, el punto puede localizarse muy cerca de la recta diodo, la que se encuentra definida por la relación $E_c = E_b$, pero si la potencia de excitación de grilla es limitada, el potencial mínimo instantáneo de placa ($e_{b(min)}$) debe ser algo mayor que la tensión pico positiva de reja ($e_{c(max)}$), por esto se hace necesario desplazar el punto A de la recta diodo. 

En tetrodos el punto A debe localizarse cerca del punto donde las curvas de corriente de placa giran abruptamente hacia arriba. 
Se debe tener en cuenta que si $e_{b(min)}$ es grande, disminuye la potencia de excitación requerida, pero se reduce la máxima tensión de placa provocando esto una disminución de la potencia de salida, lo que hace aumentar la potencia de disipación, obteniéndose un menor rendimiento. 


![5-cuatrocientos](fig/valvula_103.png)


Si el punto A se ubica de forma que $e_{b(min)}$ sea pequeño, la corriente de reja pantalla y control podrían ser excesivas y el incremento en la potencia de salida seria relativamente pequeño. 

Nota: puede ampliarse en rango de tensión y Corriente del gráfico multiplicando la tensión por un factor de escala y la corriente por el mismo factor de escala elevado a la potencia $3/2$ (ley de Child), cuando se toma otra tensión de reja pantalla distinta a la de la gráfica. Es preferible a veces sacrificar linealidad para que ebmin no sea grande. 

#### 8. Determinación de $E_{cc}$ y localización del punto G: 

El punto G está determinado por $E_{cc}$ y $E_{bb}$. 

Sin embargo el valor de la polarización de grilla está determinado por el ángulo de conducción de placa p deseado. 
Luego es necesario adoptar un valor de $\theta_p$ y determinar gráficamente el valor de $E_{CC}$, que produce ese $\theta_p$ deseado.



La ecuación siguiente permite determinar en forma aproximada el valor de $e_{cc}$, esta permite minimizar el número de pasos prueba-error en la determinación gráfica de Ecc. 

$E_{cc} = \frac{|E_{co}|+e_{c(max)} cos(\frac{\theta_p}{2})}{1-cos(\frac{\theta_p}{2})}$

Donde $E_{co}$ es la tensión de polarización de reja de corte (se corta la corriente de circulación de placa), ésta se obtiene a partir de la tensión de polarización de placa $E_{bb}$, subiendo hasta cortar la curva de corriente de placa igual a cero. 
El valor de $\theta_p$ se elige de acuerdo con la tabla indicada en el punto 2, por ejemplo del orden de 140º.
Con estos valores se puede determinar el valor de $E_{cc}$. 
El punto G se localiza en las coordenadas ($E_{cc}$, $E_{bb}$). 
La unión de A y G es la trayectoria de operación. 
Con los puntos A y G se puede calcular el valor de $\theta_p$ con:  

$\theta_p = 2 arccos(HG / AG)$

La determinación de HG y AG se realiza midiendo las distancias sobre la línea de operación. 
Calculado $\theta_p$ se compara con el valor elegido, si el calculado es menor que el elegido, se deberá  hacer la polarización menos negativa, esto incrementa el ángulo de circulación de placa, haciendo que el punto G se desplace hacia arriba. 

Con este nuevo par de valores HG y AG se calcula nuevamente el ángulo $\theta_p$ y se vuelve a comparar, esto se repite hasta que coincidan los valores.

#### 9. Determinación del valor pico de la tensión de excitación: 

Esta es igual al valor absoluto de la tensión de polarización de reja más el pico positivo de la tensión reja cátodo, esto es: 
     
$e_{g(max)} = |E_{cc}| + e_{c(max)}$

$e_{p(max)} = E_{bb} – e_{b(min)}$


#### 10. Determinación de las componentes de continua, fundamental y armónicos de la corriente de  placa (análisis armónico de Chaffee): 


![5-cuatrocientos](fig/MALLO_4.png)

Los amplificadores de potencia clase C se caracterizan por proveer amplificación con alta eficiencia y son utilizados principalmente en aplicaciones de RF, donde se requiere un incremento en el nivel de potencia y no se requiere linealidad entre la tensión de entrada y tensión de salida. 

Los amplificadores Clase C pueden ser modulados en amplitud por inyección de la tensión modulante en la placa, reja ó grilla pantalla, reja ó grilla control ó cátodo. También son utilizados para amplificar una portadora modulada en frecuencia. 

En el circuito de placa la eficiencia puede ser tan alta como el 85%, pero la mayoría de los amplificadores se diseñan para operar con eficiencia del orden de 75%, debido a que incrementar el rendimiento del  75% al 85% significa un aumento importante de la potencia de reja, además grandes eficiencias son obtenidas con elevadas tensiones de placa. 

En un amplificador Clase C, la reja se polariza por  debajo del corte y la corriente de placa circula durante menos de 180º por cada ciclo de la tensión de excitación de reja, por esto los amplificadores Clase C, no son usados en aplicaciones de audio debido a que la tensión de salida no es proporcional a la tensión de entrada y los pulsos de placa tendrían una excesiva distorsión. 

Sin embargo un amplificador Clase C puede ser usado en aplicaciones de RF, si el circuito de placa es resonante a la componente fundamental o algunas de las armónicas de la frecuencia de excitación de reja, de esta forma el pulso de la corriente de placa es filtrado por el circuito resonante de placa para proveer una salida muy cercana a una sinusoide. Cuando el amplificador es usado como multiplicador de frecuencia, el circuito de placa deberá resonar a alguna de las armónicas de la tensión de excitación, la potencia de salida se reduce considerablemente, siendo la eficiencia del orden del 40 a 50% cuando se los utiliza como dobladores y del 15 a 20% como triplicadores.

4: El amplificador de clase A

El tipo de amplificador más simple es el que opera en Clase A. No solo es
simple de diseñar, sino que también tiene la mejor linealidad de todos los diversos tipos. Sin embargo, no es el tipo más eficiente y está limitado a aplicaciones de baja potencia, ya que la disipación de potencia es alta. Se usa comúnmente en etapas de señales pequeñas de audio, como amplificadores de micrófono, ya que también es un tipo de bajo ruido cuando se usan válvulas de triodo. En aplicaciones de RF, se utiliza donde la linealidad es crítica en etapas de baja potencia de transmisores y receptores.

La ganancia que puede lograr un amplificador de Clase A es sorprendentemente alta para un circuito tan simple. Las ganancias de voltaje de 50 a 100 son relativamente fáciles de lograr y es fácil conectar en cascada dos o más etapas para obtener una ganancia aún mayor. Aunque la ganancia de voltaje es alta, la ganancia de potencia es algo limitada.

Para calcular la resistencia de carga del ánodo RA usamos la ley de Ohm: R = V / I

Necesitamos una resistencia que baje el riel de suministro de 150 V a 75 V cuando fluya una corriente de 2,5 mA. Por lo tanto, RA = (75V / 2.5mA) da una respuesta de 30k Ω . Este no es un valor de resistencia E12 estándar, los valores más cercanos son 27k Ω y 33k Ω e inicialmente seleccionaremos la opción 33k Ω . Ahora verificamos cuál es el voltaje del ánodo para esta resistencia.

Usando la ley de Ohm (V = I x R), para una corriente de 2.5mA y una resistencia de 33k Ω , V = 82.5V de caída. Esto coloca el voltaje del ánodo inactivo en (150 - 82,5) = 67,5 V, que normalmente está lo suficientemente cerca. Si quisiéramos exactamente el voltaje de carril medio de 75 V, podríamos usar 2 x 15k Ω en serie.

Si seleccionamos la resistencia de 27k Ω, la caída de voltaje sería de 67,5 V y el voltaje del ánodo de reposo sería de + 82,5 V. Esto también está cerca del valor requerido. Un método simple para hacer que el voltaje sea exactamente medio riel es cambiar un poco la corriente del ánodo inactivo. Para el resistor de 33k Ω necesitaríamos una corriente de ánodo inactivo de 2.27mA y para el resistor de 27k Ω es 2.77mA. Dado que ambas opciones son un pequeño cambio, ignoraremos el error en este momento.

Al estudiar la hoja de datos del fabricante, ahora encontramos el voltaje de polarización correcto en la rejilla de control leyendo las curvas de corriente constante. Encontramos que para la corriente y el voltaje de ánodo inactivo requeridos, necesitamos un voltaje de cátodo de red de -2V. Dado que la corriente del ánodo también fluye a tierra a través de la resistencia del cátodo RK, necesitamos aumentar el voltaje del cátodo + 2V por encima del suelo cuando fluye la corriente de 2.5mA. Usando la ley de Ohm encontramos: RK = (2V / 2.5mA) nos da un valor de 800 Ω y el valor E12 más cercano es 820 Ω .

Para determinar el valor de CK, necesitamos calcular el valor del condensador que es al menos diez veces menor en reactancia en comparación con RK. Debe ser inferior a 82 Ω a la frecuencia más baja encontrada. Si se trata de un amplificador de micrófono para un transmisor, normalmente no necesitamos bajar de unos 300Hz. Un condensador de 10 µ F a 300 Hz tiene una reactancia de 53 Ω y esto sería adecuado. Si el amplificador fuera para un amplificador de alta fidelidad, entonces se necesitaría un valor mayor para cubrir hasta aproximadamente 20Hz, un 100 µF sería adecuado.

Los condensadores de acoplamiento de entrada y salida se eligen para que tengan una reactancia baja a la frecuencia de funcionamiento. Para los amplificadores de audio, un valor de 100nF suele ser adecuado ya que la impedancia de entrada y salida es alta. El condensador de acoplamiento conectado al ánodo debe tener una tensión nominal adecuada, un condensador de 250 V sería adecuado. El condensador de entrada necesita muy poca tensión nominal y sería adecuado un tipo de 50 V.

Se han determinado todos los valores de los componentes, excepto la resistencia de red de entrada RG. Si el micrófono fuera del tipo de alta impedancia, una resistencia adecuada sería mayor que el doble de la impedancia del micrófono. Si el micrófono es un 47k w escriba una mayor resistencia de 100K Ω sería adecuado. Elegiremos 1M Ω . (Algunos micrófonos son un poco exigentes si no se usan con la carga de impedancia correcta: puede ser mejor elegir una resistencia de 47k Ω ).

Ahora que todos los valores están determinados, ¿cuánta ganancia de voltaje podemos esperar?

    Si cuando comprobamos las condiciones de CC en el ánodo a tierra encontramos que el voltaje no está en el punto medio, un pequeño ajuste en la resistencia del cátodo RK traerá el voltaje al valor requerido. Esto se puede hacer con una resistencia variable en el retorno del cátodo. Si la resistencia del cátodo está formada por una resistencia fija de 560 Ω y una resistencia variable de 500 Ω , esto nos da la capacidad de recortar el voltaje del ánodo al valor exacto requerido. Sin embargo, en amplificadores simples normalmente no es necesario llegar a tales longitudes. Un amplificador de instrumentación de CC probablemente requeriría este ajuste fino.

Ganancia de voltaje

En una primera aproximación, la ganancia de voltaje es (RA xg m ) esto da un valor de (33 x 10 3 ) x (2.5 x 10 –3 ). Dado que los términos 10 3 y 10 –3 se cancelan, la respuesta es (33 x 2.5) = 82.5.

Suponga que la señal de entrada máxima es 100 mV RMS. El voltaje de entrada pico a pico será de 282,8 mV (0,2828 V). Con una ganancia de voltaje de 82,5, la oscilación del voltaje del ánodo será (0,2828 x 82,5). Este es un voltaje pico a pico de 23,33 V, en valores RMS es de 8,25 V.

El voltaje del ánodo se moverá hacia arriba y hacia abajo, centrado en el punto de reposo nominal de + 75 V, y alcanzará un máximo de (23,33 / 2) + 75 = 86,66 V y un mínimo de 75 - (23,33 / 2) = 63,33 V.

Dado que la oscilación total del ánodo es mucho menor que el punto de polarización nominal de + 75V, la oscilación es pequeña en términos porcentuales y la linealidad será muy buena.

Ahora hacemos la pregunta, ¿cuánta señal de entrada puede tolerar el amplificador antes de que se distorsione?

Suponga que la señal de entrada se eleva ahora a 0,2 V RMS. Esta es una oscilación de entrada pico a pico de 0,565 V. La corriente del ánodo cambiará en ± (0.565 x 2.5) = 1.414mA a cada lado de la corriente de reposo de 2.5mA. Esto da las corrientes de ánodo mínima y máxima de 1.08mA y 3.53mA. La oscilación del voltaje del ánodo será de ~ 47 V a cada lado del punto de reposo de 75 V.

Bajará a + 28V y hasta + 122V. Dado que el riel de suministro es de + 150 V, se está acercando a este voltaje. De manera similar, una válvula necesita un cierto voltaje mínimo ánodo-cátodo para funcionar correctamente. El swing desarrollado es aproximadamente el máximo que de manera realista podemos esperar que sea aceptable.

    Sabiendo que la ganancia de voltaje es 82.5, también podríamos encontrar la oscilación aproximada del voltaje de salida simplemente multiplicando el valor pico-pico del voltaje de entrada por 82.5. Por lo tanto (82,5 x 0,565) = 46,66V.

Para una mayor oscilación del voltaje de salida, es mejor usar un voltaje de suministro más alto y una resistencia de carga de ánodo más alta para lograr el mismo punto de polarización inactivo de 2.5mA. Si estuviera disponible una fuente de 250 V, la oscilación de salida puede ser mayor para la misma señal de entrada. Por lo tanto, la oscilación potencial del voltaje de salida y, por lo tanto, la ganancia de voltaje, está directamente relacionada con el valor de RA.

Si calculamos el nuevo valor de RA para una tensión de alimentación de 250V, con la nueva tensión de reposo del ánodo de 125V se requiere una resistencia de 50k Ω .

Etapas en cascada

Cuando requerimos más ganancia de la que puede lograr una sola etapa, podemos lograr valores de ganancia considerables conectando dos o más etapas en serie. Si cada etapa tiene una ganancia nominal de 30, dos etapas darán una ganancia de voltaje total de 900. Tres etapas lograrían una ganancia de voltaje total de 27,000. En la figura 4.4 se muestra un ejemplo de un amplificador en cascada simple .

Fig 4.4: Dos etapas amplificadoras de triodo en cascada.

Tenga en cuenta que el valor de la resistencia de red en la segunda etapa influye en la oscilación del voltaje del ánodo de la primera etapa. Si el valor de esta resistencia de red es demasiado bajo, la red actúa como divisor de potencial. Si la resistencia de la red de la segunda etapa tiene el mismo valor que RA, el voltaje en la red de la segunda etapa es el 50% del voltaje de salida de V1. Generalmente, es necesario usar un valor mucho mayor para RG1 para permitir que ocurra el cambio de voltaje completo.

Un problema potencial con este tipo de amplificador 'acoplado por resistencia' es la corriente de fuga en el condensador de acoplamiento entre el ánodo de la primera etapa y la rejilla de la segunda etapa. Si este condensador tiene una corriente de fuga alta, obligará a la red de V2 a elevarse por encima del punto de polarización requerido. Esto suele ser un problema en equipos antiguos. Si la corriente de fuga es de solo 1 µ A, el voltaje desarrollado a través de RG1 es + 1V. Si la red necesita polarizarse a –2V, esto altera el punto de polarización correcto. Por lo tanto, estos condensadores deben elegirse con cuidado.

Amplificadores de pentodo

El amplificador de triodo simple está un poco limitado en los valores de transconductancia que se pueden lograr. Cuando se requieren mayores ganancias de etapa, el uso de una válvula de pentodo suele ser una buena opción. La válvula de pentodo es similar a un tetrodo y tiene una estructura de rejilla adicional, la rejilla de pantalla o Grid-2. El aumento de la transconductancia en comparación con un triodo equivalente puede llegar a cinco veces. La variación de corriente de ánodo que puede generar es mucho mayor para la misma señal de entrada. En términos simples, tiene más ganancia de voltaje potencial.

Muchas de las válvulas de amplificador de audio comunes se fabrican como del tipo triodo-pentodo o triodo-tetrodo, con las dos válvulas en una sola envoltura. (Uno de esos tipos es el ECL-82, que es un dispositivo popular para controlar altavoces de baja potencia. Este es un triodo y un tetrodo de haz en una envolvente común). No solo tiene el pentodo más ganancia de voltaje, sino que tiene una mayor corriente ganancia, por lo que es capaz de suministrar una potencia razonable. Los pentodos y tetrodos son válvulas más ruidosas que los triodos, por lo que la primera etapa suele ser la sección de triodo y el pentodo se utiliza para impulsar el altavoz a través de un transformador de adaptación de impedancia.

Algunas de las válvulas de pentodo de amplificador de audio clásico son populares como etapas de entrada de micrófono, guitarra o fono, ya que la ganancia es considerable y pueden oscilar el voltaje del ánodo en un rango bastante amplio. Normalmente es posible una ganancia de voltaje de más de 100. Una válvula popular es la EF86 / 6F22, que es un pentodo de baja corriente que puede proporcionar una ganancia de etapa de 150 con una fuente de 250 V ( Fig 4.5 ).

La EF86 es una válvula de baja corriente y normalmente funcionará con un suministro de 300 V a 450 V y consumirá una corriente de reposo de ~ 1 mA. Los valores de resistencia a la carga del ánodo se encuentran de la misma manera que una válvula triodo, consultando la hoja de datos del fabricante para obtener el valor g m .

La rejilla de la pantalla (g 2 ) requerirá una resistencia cuentagotas del riel de suministro de alto voltaje y un condensador de desacoplamiento adecuado para evitar variaciones de voltaje. A menudo, este condensador puede tener un valor bajo ya que la resistencia del cuentagotas es normalmente un valor alto. Este condensador y CD deben tener una clasificación adecuada para el voltaje.

La tercera rejilla (rejilla supresora) debe estar conectada directamente a tierra o al cátodo. En algunas válvulas, esta es una conexión interna. El EF86 grid-3 se lleva a los pines de la base para que el usuario lo conecte. Además, el EF86 también tiene pantallas o pantallas electrostáticas internas para reducir la captación de zumbidos y estos dos pines también deben estar conectados a tierra. El blindaje interno a menudo significa que no es necesaria una lata de protección sobre la válvula. Prevención del zumbido

Al igual que todas las etapas de alta ganancia, la prevención de zumbidos y otras señales extrañas depende en gran medida del diseño y la conexión a tierra. El cableado del calentador es una fuente potencial de zumbido, por lo que el enrutamiento de estos cables es importante. A menudo, el mejor método es retorcer los dos cables firmemente y no conectar a tierra el devanado secundario del calentador.

Otra fuente de zumbido es la ondulación en el riel de la fuente de alimentación. Para reducir esto, es común suministrarlo desde un voltaje más alto a través de una resistencia cuentagotas adecuada y un condensador de desacoplamiento de gran valor para formar un filtro de paso bajo RC. En la figura 4.5, el condensador de desacoplamiento CD realiza esta función. Para el funcionamiento con una fuente de 350 V, se puede usar una resistencia cuentagotas de aproximadamente 10 a 33 k Ω .

Amplificadores RF Clase A

Para etapas de baja potencia en un transmisor donde la linealidad es crítica, la etapa de Clase A es una buena opción. La resistencia de carga del ánodo se reemplaza con un 'circuito de tanque' sintonizado en paralelo resonante sintonizado a la frecuencia requerida.

La resistencia dinámica de la carga del ánodo se elige para que sea alta seleccionando el valor del inductor y el condensador con reconocimiento del inductor Q descargado. Muy a menudo, el Q cargado es demasiado alto y esto causa una banda estrecha sobre la cual el amplificador entregará potencia. Para corregir esto, se puede derivar una resistencia de amortiguación a través del circuito sintonizado para ampliar la respuesta, a expensas de una pérdida de potencia de salida. Cuando se requiere un filtrado máximo, para rechazar señales no deseadas, a menudo se usa un capacitor variable para alcanzar el pico del circuito del tanque y la respuesta es bastante aguda. Normalmente no se utiliza ninguna resistencia de amortiguación. Este tipo de amplificador sería necesario como amplificador de búfer después de un mezclador de frecuencia en un transmisor SSB donde la frecuencia de generación SSB original se mezcla con un VFO para llegar a la frecuencia de salida final.

El circuito sintonizado, aunque se comporta como una resistencia de alto valor en resonancia, prácticamente no tiene resistencia en la condición de CC. Para desviar el voltaje y la corriente del ánodo al punto 50-50 se requiere una resistencia de alimentación en serie para llevar el voltaje del ánodo hacia la mitad de la pendiente de la corriente del ánodo. Si no se utiliza, sin señal de entrada, el voltaje del ánodo estará en reposo al voltaje de suministro completo. Por lo tanto, el uso de una resistencia en serie de ánodo significa que sacrificamos una parte de la capacidad de oscilación de voltaje. Para contrarrestar esto, la tensión de alimentación del ánodo debe ser mayor. La resistencia de la serie de rejilla de pantalla debe tomarse del nuevo riel de ánodo para que los dos mantengan la relación correcta.

Una válvula típica sería un tetrodo o pentodo de haz con polarización de resistencia de cátodo para amplificadores de baja potencia o una polarización de rejilla fija para amplificadores de potencia superior. Sin embargo, debido a la poca eficiencia del amplificador de Clase A, existe un límite en la cantidad de potencia que podemos lograr con este tipo de amplificador. El voltaje de la rejilla de la pantalla debe estar bien regulado para que no varíe mucho cuando esté completamente accionado. Como normalmente no fluye una corriente apreciable de la rejilla de la pantalla, a menudo una resistencia en serie simple es suficiente para los amplificadores de potencia más bajos. Cuando el amplificador es de mayor potencia, un divisor de potencial del suministro del ánodo a la rejilla de la pantalla que extrae aproximadamente tres veces la corriente de la pantalla prevista es normalmente suficiente para estabilizar la tensión de alimentación de la rejilla de la pantalla. Una alternativa es utilizar un diodo Zener de alto voltaje, que proporciona una regulación mucho más estricta.El pin de la rejilla de la pantalla debe estar bien desacoplado a la frecuencia de funcionamiento mediante condensadores de baja inductancia con cables cortos. En la mayoría de los casos, se puede prescindir de la neutralización, ya que la mayoría de los tetrodos y pentodos de haz pequeño tienen una baja capacitancia de paso de la red del ánodo. Sin embargo, el diseño y la ubicación de los componentes tienen un efecto considerable en la estabilidad del circuito y se debe tener cuidado de no introducir inadvertidamente un mecanismo de retroalimentación no deseado. A menudo, se requiere una pantalla, bien conectada a tierra, en la base de la válvula para evitar que los pines de entrada y salida causen retroalimentación. El pin del ánodo, si es capaz de "ver" el pin de la rejilla, puede causar inestabilidad, ya que tiene un alto voltaje de RF cuando está completamente activado. Es posible generar un voltaje de RF de 100 V pico a pico o más y la red es un pin sensible a pequeñas señales. Similar,El circuito sintonizado con ánodo debe estar bien protegido del circuito de la red.

Un tetrodo o pentodo de haz, como el tipo 5763 o 12BY7, puede proporcionar una ganancia de potencia de aproximadamente 13 a 20 dB si se diseña correctamente. Esto significa que la fuente de alimentación necesita suministrar muy poca energía, simplemente suficiente oscilación de voltaje en la red de control. Este tipo de amplificador se usaba comúnmente en los transceptores híbridos de HF en días anteriores, donde las etapas de baja potencia eran de estado sólido y el controlador y el PA eran válvulas. La salida se puede acoplar a través de un condensador de bajo valor directamente a las rejillas de las válvulas PA.

Sesgo

La polarización de la red necesaria se puede obtener convenientemente con una resistencia de cátodo bien desacoplada a RF con cables cortos para que la inductancia en serie sea mínima. En algunos casos, es mejor conectar a tierra directamente el cátodo y suministrar la red a través de una fuente de polarización negativa ajustada para adaptarse a la corriente de ánodo inactiva requerida. El punto de polarización se puede determinar a partir de la hoja de datos del fabricante, si no se cita en un circuito de ejemplo, examinando las curvas de corriente constante para encontrar el valor óptimo.

6: Causas de la distorsión

El término distorsión se usa normalmente para indicar que la señal de salida difiere de
alguna manera de la señal de entrada. En el amplificador ideal, la señal de salida debe ser una réplica exacta de la señal de entrada, pero de mayor amplitud.

La distorsión puede ser de diferentes formas, como distorsión armónica o distorsión de intermodulación. En los amplificadores de audio, la distorsión armónica suele ser el tipo que se encuentra. Por el contrario, en los amplificadores lineales de RF la distorsión de intermodulación es normalmente el tipo que más nos preocupa. Los armónicos desarrollados en un amplificador de RF se pueden filtrar, pero en un amplificador de audio no pueden y son audibles para el oyente.

Distorsión de emisión

Este tipo de distorsión no se encuentra comúnmente en los amplificadores, pero se nota en otros circuitos. Como se ha explicado, los electrones generados por el cátodo están disponibles para proporcionar un flujo de corriente al ánodo, al ser atraídos por el alto potencial positivo de esa parte de la válvula. Si la capacidad del cátodo para evaporar electrones es ilimitada, se podría lograr cualquier valor deseado de corriente de ánodo. En la práctica, las válvulas se diseñan y fabrican normalmente para proporcionar solo los electrones suficientes, con la debida tolerancia para las condiciones de sobrecarga, de modo que a lo largo de la vida útil de la válvula nunca debe quedarse sin electrones si se opera en las condiciones recomendadas.

Sin embargo, hay ciertas válvulas en las que es prudente limitar la capacidad de generación de electrones para proporcionar una forma de dispositivo de seguridad. Si se empuja mucho más allá de su capacidad de cátodo, cualquier válvula eventualmente dejará de proporcionar suficientes electrones. Cuando esto ocurre, la corriente del ánodo deja de aumentar para aumentar el nivel de activación.

Esta característica se utiliza deliberadamente en válvulas rectificadoras ( Fig. 6.1 ). Al privar al ánodo de electrones, proporciona una función de limitación de corriente simple y eficaz. Si la salida de CC de un rectificador de válvula está en cortocircuito, lo que puede proporcionar un condensador de suavizado de cortocircuito o algún otro circuito en una condición de falla, el voltaje de salida del rectificador cae significativamente. Los ánodos de la válvula rectificadora normalmente se pondrán al rojo vivo para alertar al usuario de la condición de falla. Si el equipo está bien diseñado, normalmente un fusible, un circuito de disparo por sobrecorriente o sobrecalentamiento desconectarán la alimentación de entrada de CA.

Fig 6.1: Saturación de corriente de ánodo de un diodo rectificador.

Por el contrario, un rectificador semiconductor moderno normalmente fallará bajo esta condición estresante. La válvula rectificadora generalmente puede tolerar cortocircuitos a corto plazo sin efectos nocivos. Parte de la razón por la que esto es así es que la válvula rectificadora tiene un voltaje directo bastante alto en comparación con un diodo semiconductor. En un diodo de silicio, el voltaje directo es de aproximadamente 0,6 V a corrientes pequeñas y de aproximadamente 1 V a máxima potencia. Una válvula rectificadora tiene un voltaje directo medido en varios voltios. La caída de voltaje mínima es de aproximadamente 10 V y la máxima es de aproximadamente 70 V a plena corriente nominal. Aunque estas cifras parecen altas, cuando la tensión de salida de CC es de ~ 500 V y superior, son un pequeño porcentaje de la tensión de alimentación y no son significativas. También son un mejor método para suministrar alta tensión CC a otras etapas de la válvula. Esto se debe a que, al encender, el filamento del rectificador está frío y, aunque el transformador está suministrando al rectificador el voltaje de CA requerido, la válvula no comenzará a conducir hasta que el cátodo se eleve por encima de una cierta temperatura. En consecuencia, la tensión de CC de salida aumenta gradualmente durante un período de varios segundos. Esto proporciona un buen comienzo suave ya que las otras válvulas también se están calentando. En los rectificadores de semiconductores que alimentan circuitos de alto voltaje, a menudo es necesario proporcionar algún tipo de mecanismo de arranque suave para ralentizar el rápido aumento en el voltaje de salida cuando se enciende por primera vez. Los condensadores electrolíticos de alto voltaje y los diodos pueden resultar dañados por el interruptor de sobretensión.

Esto también es un problema con algunos equipos donde la válvula rectificadora normal se reemplaza con diodos de silicio modernos. No solo se pierde la capacidad de arranque suave, sino que el voltaje de salida sin carga puede ser demasiado alto para los condensadores electrolíticos. Si se instalan diodos de silicio, es imperativo instalar una resistencia de valor adecuado en serie para limitar la corriente de entrada y proporcionar cierta limitación de corriente en condiciones de cortocircuito.

Al igual que todas las válvulas, la válvula rectificadora necesita un cierto voltaje ánodo-cátodo antes de comenzar a conducir la corriente adecuada. En las válvulas rectificadoras pequeñas diseñadas para rectificación de alto voltaje, puede ser tan alto como 50 V, por lo que la parte inferior de la forma de onda sinusoidal se corta y solo se observa la parte superior a aproximadamente 50 V en la salida del cátodo. Esto es motivo de preocupación, ya que el voltaje máximo es menor de lo teóricamente posible. (Para compensar la pérdida de voltaje se requieren algunas vueltas más en el secundario, llamadas vueltas de compensación. Esto a menudo puede significar que si se utilizan diodos semiconductores, la tensión directa más baja es demasiado alta para el condensador de suavizado instalado). Distorsión armónica

Esto se acepta normalmente como una distorsión causada por la generación de frecuencias que son múltiplos exactos de la señal de entrada. En un amplificador de audio alimentado con una forma de onda de entrada sinusoidal pura, el segundo armónico es el doble de la frecuencia de entrada y el tercer armónico es tres veces la señal de entrada. Generalmente, el oído humano es más sensible a los armónicos de orden impar y tienden a "colorear" la señal de salida, haciéndola menos atractiva para escuchar. En los amplificadores de audio, el término distorsión armónica total(THD) se utiliza comúnmente. Esta es una medida porcentual de todas las señales de salida armónicas sumadas, en comparación con la señal de entrada original deseada. Para realizar una medición de THD, la frecuencia fundamental de la señal de salida se anula y todo lo que quede se considera como la cifra de THD. En la práctica, también se produce una cierta cantidad de ruido generado internamente dentro del amplificador que también es detectado por el instrumento de medición. Por lo tanto, la THD a veces se da como 'THD + ruido', como una relación de distorsión + ruido a la señal deseada, generalmente expresada en decibelios.

En los amplificadores de RF, la distorsión armónica para tipos de señales como CW y FM no es de gran preocupación, ya que se pueden filtrar utilizando un filtro de paso bajo después del amplificador. Sin embargo, la distorsión armónica tiene relación con otro tipo de distorsión llamada distorsión de intermodulación , o IMD para abreviar. El mecanismo que genera la distorsión armónica también es responsable de la generación de IMD.

En un amplificador ideal no tendría no linealidad en el dispositivo, pero en dispositivos prácticos como válvulas y semiconductores siempre habrá alguna. Es la magnitud de la no linealidad lo que determina si el dispositivo es aceptable o no. Cuanto más lineal sea el amplificador, menor será la amplitud de los productos no deseados.

Distorsión de intermodulación

En un amplificador lineal, además del mecanismo fundamental de generación de armónicos, también hay un segundo mecanismo que también es causado por la no linealidad del dispositivo. Estos productos adicionales surgen porque la señal de entrada no es un solo tono, sino una señal compleja que contiene dos o más portadores o tonos discretos. El proceso de mezcla o modulación que surge cuando dos o más señales pasan a través de un amplificador genera estas señales adicionales llamadas productos de distorsión de intermodulación.

Si las dos portadoras están poco espaciadas en frecuencia dentro de la banda de paso del amplificador, cada señal se mezclará con la otra señal. Si el amplificador fuera perfectamente lineal, no se generarían armónicos y no se produciría ninguna mezcla, por lo que no se podrían generar IMD. ¡Pero tal amplificador no existe en el mundo real!

Distorsión de recorte

Este tipo de distorsión ocurre cuando el nivel de la señal de entrada es excesivo. El recorte es una forma útil de limitar la amplitud máxima de ciertas señales, por ejemplo, como limitador de ruido, pero para la amplificación lineal debe evitarse. En el amplificador de Clase A, si el voltaje de la red oscila demasiado alto en el semiciclo positivo, puede impulsar la red en positivo con respecto al cátodo. Esto provoca varios efectos diferentes.

En el caso de recorte o cobertura plana , la corriente del ánodo se satura en el pico positivo de la señal de entrada. Como el voltaje de salida del amplificador es el inverso de la corriente del ánodo, la forma de onda del voltaje de salida muestra una porción positiva superior casi ideal de la onda sinusoidal, pero el semiciclo inferior negativo de la forma de onda sinusoidal está muy distorsionado.

La mitad de entrada positiva alta de la señal de entrada también puede hacer que fluya la corriente de la red. Este efecto se conoce en la industria de las válvulas como una condición de ánodo virtual . Si la rejilla es positiva con respecto al cátodo, actúa como un ánodo. Dado que la estructura de la rejilla está normalmente muy cerca del cátodo, solo unos pocos voltios positivos tienen un efecto similar al alto voltaje positivo en el ánodo que, sin embargo, está mucho más separado. Un espaciamiento pequeño necesita menos diferencia de voltaje para producir un gran cambio en el flujo de la corriente de electrones, por lo tanto, un potencial de cátodo de rejilla pequeño tiene tanto efecto como un potencial de cátodo de ánodo grande con un espaciado mucho mayor.

El espacio entre la rejilla y el cátodo suele ser muy pequeño, para garantizar un control estricto del flujo de electrones en la carga espacial. La distancia ánodo-cátodo es normalmente mucho mayor. Entonces, una pequeña diferencia de potencial entre la rejilla y el cátodo tiene un efecto significativo, aunque el voltaje ánodo-cátodo es mucho mayor. Todos los electrones que salen del cátodo finalmente tienen que ir a algún otro electrodo que tenga un potencial positivo. La mayoría terminará en el ánodo, pero si la rejilla es ligeramente positiva, interceptará algunos de los electrones. Si la cuadrícula es muy positiva interceptará un número significativo, agotando así el total que está disponible para viajar al ánodo.

La relación entre la rejilla y el cátodo y, por extensión, las distancias entre la rejilla y el ánodo con los potenciales dicta cuántos de los electrones disponibles se compartirán entre los dos electrodos. Por ejemplo, si el voltaje del ánodo es de 100 V y la distancia que recorren los electrones es de 10 mm, esto se puede expresar como una relación entre el voltaje y la distancia. (Para distancias más grandes, la fuerza de atracción electrostática del ánodo a los electrones emitidos por el cátodo es más débil). En este ejemplo es un gradiente de 10V / mm. En consecuencia, si el espacio entre la rejilla y el cátodo es pequeño, digamos 1 mm, solo se requieren 10 V positivos en la rejilla para interceptar todos los electrones disponibles emitidos por el cátodo. Cuando la tensión de la red es menor que la relación crítica, se desvían menos, pero sigue siendo un porcentaje significativo.El resultado de esto es que la rejilla desvía una cantidad significativa de electrones de la corriente del ánodo.

Estos electrones normalmente servirían para producir un flujo de corriente de ánodo más alto. Esto significa que la resistencia del ánodo de la válvula no alcanza el valor más bajo posible, ya que se dispone de electrones insuficientes. Se satura a una resistencia fija durante el período del ciclo en el que la red es suficientemente positiva. En realidad, esto es lo que provoca el recorte / cobertura plana.

Un efecto similar ocurre en las válvulas de rejilla de pantalla. En este mecanismo, es la oscilación excesiva hacia abajo del voltaje del ánodo la responsable. Si el voltaje del ánodo se conduce por debajo del voltaje de la rejilla de la pantalla, la rejilla de la pantalla asume el papel del ánodo. Esto es similar al efecto de ánodo virtual que utiliza la rejilla de control para desviar los electrones del ánodo. Se conoce como efecto de cátodo virtual o simplemente efecto de fondo . Si el voltaje del ánodo oscila por debajo del voltaje de la rejilla de la pantalla, la corriente de la rejilla de la pantalla aumenta de un valor bajo a ser mucho mayor. Cuando el voltaje del ánodo ha caído unos voltios por debajo del voltaje de la rejilla de la pantalla, comienza una reacción en cadena. En primer lugar, la corriente de la rejilla de la pantalla aumenta rápidamente, casi de manera exponencial, y puede asumir niveles peligrosamente altos. La rejilla de la pantalla de un tetrodo y pentodo no es una estructura de alta disipación, a diferencia del ánodo que puede disipar una potencia considerable, y puede calentarse debido a la corriente que fluye y alcanzar una temperatura en la que el metal comienza a fundirse. Si esto ocurre, el vapor metálico liberado hace que se forme una nube de ionización cerca del ánodo. Esto es como un rectificador de arco de mercurio y normalmente se establece una corriente de arco muy grande entre el ánodo y la rejilla de la pantalla. Si el arco se puede establecer,la válvula se destruye rápidamente.

Sin embargo, incluso si se evita que la corriente de la rejilla de la pantalla alcance niveles peligrosos mediante resistencias limitadoras de corriente, o algún otro método, el efecto de sujeción del voltaje del ánodo todavía hace que aparezca un fondo plano en la mitad inferior de la forma de onda del voltaje de salida. Esto es idéntico al aplanamiento inducido por la red de control de la parte inferior del ciclo de voltaje de salida. Puede ser difícil saber cuál de las dos rejillas es el culpable sin realizar algunos ajustes experimentales en el voltaje de la rejilla de la pantalla. Si al reducir el voltaje de la rejilla de la pantalla, la distorsión es menor, el voltaje de la rejilla de la pantalla era demasiado alto. Muy a menudo, al reducir el voltaje de la rejilla de la pantalla, encontramos que podemos impulsar la válvula un poco más fuerte y lograr una mayor potencia de salida. Esto va en contra de la sabiduría común, pero se puede explicar por qué es así.

La potencia de salida está directamente relacionada con la corriente de ánodo disponible y la oscilación de voltaje. Si al reducir el voltaje de la rejilla de la pantalla en, digamos un 20%, podemos lograr un 20% adicional de oscilación del voltaje del ánodo, ya que la potencia es una función de V 2 / R, si la carga del ánodo permanece sustancialmente constante, el aumento en la potencia de salida potencial es ~ 20%. Para alcanzar la corriente de pico de corriente adicional del ánodo hay un pequeño aumento en el voltaje de conducción. Al reducir el voltaje de la rejilla de la pantalla, la ganancia cae ligeramente, pero la mayor potencia de salida disponible compensa con creces esta pequeña disminución.

Recorte inducido por corriente de red

El otro efecto, que no suele apreciarse, es que la generación de corriente de red provoca una reducción radical de la impedancia de la red. La estructura de cátodo de rejilla de una válvula a bajas frecuencias es esencialmente una resistencia muy alta. En las frecuencias de audio, es de varios millones de ohmios y está derivado por una pequeña capacitancia paralela. Por lo general, se trata de unos pocos pico-faradios y no tiene ningún efecto mientras la frecuencia sea baja. Sin embargo, cuando la rejilla es positiva con respecto al cátodo, la resistencia muy alta cae significativamente. Esto se debe a que un diodo fantasmaregión ahora existe. A medida que la red se impulsa de manera más positiva, la corriente de la red aumenta y, por lo tanto, la resistencia efectiva cae, desde esencialmente un circuito abierto, a un valor más bajo. Si 3 V de voltaje de red positivo hacen que fluyan 5 mA de corriente de red, entonces la resistencia de red efectiva es ahora de solo 600 Ω .

Hay una terminología especial para este efecto: perverancia . Se dice que las válvulas que requieren muy poco voltaje de red positivo para inducir un gran flujo de corriente de red tienen una alta pervencia, mientras que las que tienen un flujo de corriente de red bajo para la misma condición son de baja pervencia.

Algunas válvulas tetrode como la Eimac 4CX150, 250 y 350 son particularmente susceptibles a este tipo de distorsión. De hecho, el Eimac 4CX350 no puede tolerar ningún flujo de corriente de la red antes de que se produzca la distorsión. La razón por la que una válvula tiene una alta pervencia es porque el espacio entre el cátodo y la rejilla de control es muy pequeño. En la serie 4CX150 / 250/350, el espacio entre el cátodo y la rejilla es de tan solo 0,5 mm. Esto permite un control estricto de la corriente del ánodo, un valor de g m alto , pero una perveancia más alta.

Si la fuente de excitación o el circuito de la red tienen una impedancia alta, la reducción de la resistencia de la red reduce la oscilación del voltaje de la red a menos de lo que sería si la fuente de excitación o el circuito de la red tuvieran impedancia cero. Por esta razón, la fuente de alimentación debe poder generar suficiente corriente con una caída de voltaje insignificante. La capacitancia de la red sirve como un capacitor de suavizado para mantener la carga.

Por lo tanto, es imperativo utilizar una fuente de polarización de baja impedancia y minimizar el valor de cualquier resistencia que la alimente al circuito de la red. A menudo, un RFC es una mejor opción que una resistencia.

Efecto de resistencia de la serie del controlador

De manera similar, si la fuente de excitación tiene una impedancia de salida alta, el aumento de carga a medida que el voltaje de la red oscila en positivo provoca una disminución del voltaje de excitación debido a la resistencia en serie en la red de salida de la etapa de excitación.

Si la caída repentina en la resistencia de la red hace que la tensión de entrada también caiga plana, se producirá un tope. La válvula no tiene la culpa: es la etapa de conducción la que necesita mejorar para corregir el efecto.

    El escenario es el mismo que el de un amplificador de audio que activa altavoces de baja impedancia. Aunque el altavoz puede ser de 4 Ω , la impedancia de salida de un amplificador típico es una fracción de ohmio. Por lo tanto, puede desarrollar una alta corriente en la carga con una caída de voltaje cercana a cero.

Este efecto no suele ser un problema con los amplificadores de tipo Clase A de señal pequeña, porque la señal de entrada no suele ser lo suficientemente grande como para acercarse al punto de corriente de la red. Sin embargo, en cualquier tipo de amplificador que utilice un tipo de método de Clase B, que incluye los tipos de Clase AB 2 , la fuente de activación debe poder suministrar la potencia de activación adecuada para obligar a la red a seguir fielmente la forma de onda.

Un método simple es utilizar un controlador de baja impedancia, es decir, uno que pueda suministrar un exceso de potencia del controlador, y cargar la red con una resistencia de amortiguación adecuada para terminar la etapa del controlador. Esto es especialmente cierto para los amplificadores de clase B donde debe fluir una corriente de red apreciable.

    Si se examina el generador de señales de RF, ocurre una situación similar. En este escenario, el voltaje de circuito abierto es el doble del voltaje cargado, la situación de PD y EMF. Cuando la carga está conectada, la mitad de la energía disponible se disipa en la carga y la mitad dentro del generador.

Distorsión inducida por el controlador

Si un amplificador tiene una gran variación en la impedancia de entrada cuando se varía la excitación, es posible que esta gran oscilación de impedancia tenga un efecto en la etapa de excitación. Esto se observa con mayor frecuencia en amplificadores lineales de RF que utilizan triodos de rejilla conectados a tierra. El terminal de entrada del triodo de rejilla con conexión a tierra es el cátodo, mientras que la rejilla de control está conectada a tierra a RF, ya sea directamente o mediante un condensador de bloqueo de CC. Los tipos de válvulas que se emplean comúnmente para este tipo de amplificador se conocen generalmente como tipos de polarización cero de Clase B. Ejemplos de estas válvulas son la serie 811, 572B y 3-500Z.

Todos estos triodos de rejilla conectados a tierra muestran valores altos de corriente de rejilla cuando están completamente activados. Normalmente, la corriente de la red alcanzará un pico de aproximadamente el 30 al 40% de la corriente del ánodo. Esto significa que para un triodo que indica una corriente de ánodo de 300 mA cuando está completamente accionado, tiene una corriente de red de al menos 100 mA. Si la impedancia del cátodo se mide sin condiciones de excitación, el valor que medimos es de varios cientos de ohmios o más. Sin embargo, tan pronto como la corriente de la red comienza a fluir, la impedancia de entrada del cátodo cae a un valor bajo y, a medida que la unidad aumenta al máximo, disminuye progresivamente.

Esto significa que la etapa de conducción se presenta con una impedancia de carga muy variable. Si la etapa del controlador es un tipo de válvula, esto generalmente no es un problema serio, ya que la red del tanque pi se puede ajustar para proporcionar una cantidad adecuada de energía. Sin embargo, el transceptor de HF moderno ya no usa válvulas, sino amplificadores de transistores de banda ancha. Estos no tienen ajustes del operador para adaptar el transmisor a la carga y dependen en gran medida de que la carga sea una carga de 50 Ω relativamente constante con VSWR bajo.

Aunque muchos tipos tienen una ATU incluida que puede coincidir en un rango bastante amplio de impedancia de antena, este no es el caso con un amplificador de rejilla con conexión a tierra. En una condición de desajuste de antena, no cambia con el nivel de potencia, por lo que la ATU puede proporcionar una coincidencia de VSWR baja con el transmisor en cualquier nivel de salida de potencia. En el caso del amplificador lineal de rejilla con conexión a tierra, la impedancia cambia con el nivel de potencia, generalmente en un factor de 5: 1 o más, dependiendo del nivel de excitación. Aunque es posible encontrar una buena combinación a plena potencia de transmisión utilizando la ATU, cuando se reduce el nivel de transmisión, aumenta el VSWR.

Los tipos de amplificadores de banda ancha push-pull de HF que utilizan dispositivos de estado sólido con transformadores de ferrita son particularmente sensibles al desajuste reactivo. Cuando funcionan con una carga ficticia de 50 Ω o una antena bien adaptada, se comportan correctamente. Pero en una carga muy desigual, pueden exhibir productos de alta IMD. Estos se presentan a la entrada del amplificador lineal externo y simplemente se amplifican. Muy a menudo se afirma que el culpable es el amplificador lineal. Cuando el amplificador se desconecta del circuito y el transmisor ahora impulsa la antena bien adaptada, los productos IMD caen a niveles bajos. Pero en realidad no es el amplificador lineal el culpable sino el transmisor de conducción.

Esto puede demostrarse simplemente conectando un acoplador direccional entre el transmisor y el amplificador lineal y alimentando la señal muestreada a un analizador de espectro. Cuando el lineal se cambia al circuito, el transmisor genera productos de alta IMD porque no le gusta trabajar con una impedancia de carga tan terrible.

El escritor ha pasado muchas horas investigando este tipo de problemas. Algunos de los amplificadores lineales HF comerciales que utilizan triodos de rejilla conectados a tierra no son buenos, incluso cuando son impulsados ​​por una fuente inmune a la autogeneración de IMD. En muchos casos, las redes de adaptación de entrada al cátodo están lejos de ser óptimas para la adaptación invariable de fuentes. Los peores tipos no tienen redes coincidentes en absoluto, simplemente se acoplan a los cátodos con un condensador de bloqueo de CC. Otros, que al menos intentan utilizar una red coincidente, la diseñan para una Q cargada que es demasiado baja para ser efectiva.

De los diversos tipos de válvulas de uso común, la serie 811 es la peor. Esta válvula originalmente entró en producción alrededor de 1935. La última 811A tuvo algunos ajustes en el diseño para permitir el uso de hasta 30MHz en amplificadores de RF de Clase C, noamplificadores lineales - ¡nota! Estas son válvulas pequeñas de disipación de baja potencia y un operador no calificado las sobreactiva fácilmente. El 811 nunca fue diseñado como un amplificador de RF lineal, originalmente fue diseñado para trabajos de amplificador de audio de Clase B donde se necesitaba alta potencia. RCA seleccionó arbitrariamente un valor de distorsión armónica del 8% para categorizar la válvula. Una distorsión del 8% podría ser aceptable para el servicio de amplificador de megafonía, pero está lejos de ser ideal para la amplificación SSB. Suponiendo que cuando se utiliza como amplificador lineal de RF todavía alcanza un 8% de distorsión, esto equivale a un IMD de tercer orden de aproximadamente
–22dBc. Lograr una medición más aceptable de –30dBc significa que la salida de potencia no puede superar el 50% de la capacidad nominal total.

Distorsión de suministro de cuadrícula de pantalla

Como ya se ha mencionado, el tipo de válvula de rejilla de pantalla tiene una rejilla adicional que se comporta de manera similar a la rejilla de control. Al variar el voltaje positivo de la rejilla de la pantalla, también se puede variar la corriente del ánodo. En los transmisores modulados con rejilla de pantalla y ánodo AM, la inclusión de la rejilla de pantalla permite lograr una mayor profundidad de modulación. Por tanto, cualquier variación del voltaje de la rejilla de la pantalla provoca una modulación adicional. En el servicio de amplificador lineal SSB, esto significa que para garantizar la curva de transferencia más lineal, el voltaje de la rejilla de la pantalla debe estabilizarse a una cifra muy ajustada. Si el voltaje de la rejilla de la pantalla cambia bajo el accionamiento, se producen mecanismos no lineales adicionales.

Algunos de los transceptores de HF clásicos de la época anterior utilizaban un enfoque híbrido. Estos eran en gran parte de estado sólido para el receptor y el transmisor hasta la etapa del controlador. Posteriormente, las etapas de amplificación de potencia utilizaron válvulas. A menudo, se trataba de un controlador 12BY7 y un par de 6JS6C o un par de válvulas 6146 para obtener una salida PEP de aproximadamente 100W. Lamentablemente, muchos de estos transceptores de HF omitieron cualquier tipo de estabilizador para el suministro de rejilla de pantalla de las válvulas de PA. En consecuencia, el voltaje de la pantalla oscilaba por todo el lugar mientras se accionaba el transmisor. No solo eso, sino que se permitió que las válvulas se accionaran en una pequeña cantidad de corriente de la red antes de que el circuito ALC retrocediera el accionamiento.

El suministro de polarización de la red también era de tipo no estabilizado, a menudo una red de divisor de potencial crudo que utiliza resistencias de alto valor. Esto agravó el cambio de voltaje de la red bajo picos de señal de excitación. A medida que varía el voltaje de la red, también lo hace el ánodo, la rejilla de la pantalla y la rejilla de control. Si la corriente extraída de la red de CA es alta en los picos de voz, también sufre un pequeño cambio y esto se refleja en los voltajes del ánodo, la rejilla de la pantalla y la rejilla de control. Una variación del voltaje del ánodo de aproximadamente el 10% como máximo normalmente no es una preocupación, pero los otros voltajes que se tambalean ciertamente son una preocupación.

Caso de estudio

El autor modificó un transceptor Kenwood TS-820 para ver qué mejora se podía obtener colocando en primer lugar un estabilizador de derivación de rejilla de pantalla y, más tarde, un estabilizador de polarización de rejilla. Esto surgió después de que se hubieran realizado considerables experimentos en los amplificadores VHF 4CX250 y QQV06-40, donde se lograron mejoras significativas tanto en el IMD como en la salida de potencia.

La rejilla de la pantalla del 6146 se redujo del diseño original de Kenwood de aproximadamente + 300 V a un valor más adecuado de + 175 V. (RCA recomienda que el voltaje de la pantalla 6146 no supere los + 200 V para obtener la mejor linealidad. Parece que los diseñadores no conocían este factor o decidieron ignorarlo). La instalación del estabilizador de derivación de la rejilla de la pantalla redujo en ~ 10dB a máxima potencia. Aunque las figuras Kenwood TS-820 y TS-830 eran bastante buenas antes de la modificación, entonces eran excepcionalmente buenas. El peor de los casos fue -37dBc IMD 3con una potencia de salida de 120W PEP. Reducir el voltaje de la red de la pantalla permitió aproximadamente 15 W PEP adicionales simplemente ajustando los valores del circuito ALC para permitir que fluya un poco más de corriente de la red. La oscilación del voltaje del ánodo se mejoró al reducir el voltaje de la pantalla, antes de disminuir el voltaje , se notó mucho un efecto de cátodo virtual .

A continuación, se abordó el sesgo de la rejilla y se diseñó y colocó un estabilizador de derivación. Después de que esto se ajustó, no hubo un cambio medible en el voltaje de la red de CC cuando se accionó. Antes de la modificación, el desplazamiento de CC medido en el impulso máximo era de casi 20 V, lo que movía la condición de funcionamiento casi a la Clase C. Esto ahora eliminó el efecto de ánodo virtual y el voltaje de RF de la red se parecía más a una verdadera forma de onda sinusoidal. Antes, la cobertura plana era muy evidente. Esto permitió que fluyera más corriente de la red en los picos y, como consecuencia, una potencia de salida adicional.

La válvula impulsora 12BY7 también se alimentó desde el estabilizador de derivación de la pantalla 6146 y esto mejoró aún más las cosas. El conductor estaba contribuyendo con algunos de los productos IMD que eran visibles. El resultado general fue de 100 W PEP, la cifra de IMD3 cayó por debajo de –40 dBc. Los productos IMD 5 casi desaparecieron en el piso de ruido del analizador de espectro alrededor de –55dBc.

amplificadores de válvula y de estado sólido comparados

Con la tendencia a utilizar amplificadores de estado sólido, se debe considerar la cantidad de IMD generada. Aunque el enfoque de estado sólido elimina la necesidad de altos voltajes desagradables y la necesidad de que el operador sepa cómo ajustar el amplificador, el amplificador de válvula sigue siendo superior. Como ya se mencionó, los amplificadores de estado sólido no pueden tolerar mucho la sobreexcitación y hacerlo genera productos IMD excesivos. Por debajo del punto de compresión de 1dB, ambos tipos son bastante similares y tienen un rendimiento IMD comparable para los productos IMD de orden inferior. Pero los productos IMD de los órdenes superiores (7º, 9º, 11º, etc.) no desaparecen rápidamente como los IMD del amplificador de válvula.

Si bien las IMD cercanas de los dos tipos son muy parecidas, en el séptimo orden las IMD del transmisor de la válvula están desapareciendo en el piso de ruido del analizador de espectro y no se puede medir nada más bajo que esto. Los IMD del transmisor de estado sólido desaparecen de una manera más gradual y los productos hasta el orden 15 son visibles por encima del piso de ruido del analizador de espectro. Se puede ver una ilustración de las diferencias en la figura 6.7 , que es para dos amplificadores de potencia de salida similares accionados con la misma señal de prueba limpia de dos tonos.

Si el espaciado de dos tonos es ~ 1 kHz, la salpicadura que genera el transmisor de estado sólido se extiende sobre ~ 7 kHz a cada lado de la frecuencia central; es decir, ocupa al menos 14 kHz de espectro. En comparación, el transmisor de la válvula ocupa solo 3 kHz a cada lado de la señal principal y desaparece rápidamente.

Sin embargo, algunos amplificadores lineales que utilizan triodos de rejilla conectados a tierra son similares al tipo de estado sólido debido a la alta corriente de rejilla que fluye en los triodos. En comparación, un amplificador tetrodo de cátodo conectado a tierra sin corriente de red es unos 20 dB mejor. Esto se debe a que el tetrodo tiene inherentemente una mejor característica de transferencia y es más lineal. Si fluye algo de corriente de la red, siempre que no sea excesiva, el rendimiento del IMD no se degrada significativamente. De hecho, con una pequeña cantidad de corriente de red permitida en el pico del ciclo de conducción, los IMD pueden descender ligeramente. Pero si el impulso aumenta más allá de este punto, comienzan a aumentar drásticamente. Por lo tanto, es imperativo que cualquier amplificador lineal externo debe estar dentro del rango operativo lineal. Una forma sencilla de garantizarlo es utilizar el control automático de nivel (ALC).

Control de nivel automático

ALC es un circuito de retroalimentación entre el amplificador externo y el transmisor de conducción. Esencialmente, funciona de la misma manera que el control automático de ganancia (AGC) del receptor. En un receptor, las etapas de ganancia de RF e IF se controlan variando la ganancia de modo que se presente una señal de amplitud constante a las etapas del detector. Esto significa que una señal que se está desvaneciendo (es decir, que varía en amplitud) se mantiene en un valor estable y el resultado neto es que la recepción mejora.

En un bucle ALC, el transmisor tiene etapas similares de ganancia ajustable en las secciones de baja potencia. Sin un amplificador externo, la salida de la etapa de salida del transmisor se detecta en nivel y las etapas de ganancia se ajustan para evitar que el variador exceda un cierto nivel. Esto asegura que la señal de salida tenga productos con IMD baja. El problema es que la mayoría de los amplificadores externos necesitan menos excitación de la que el transmisor es capaz de generar, por lo que pueden sobreexigirse fácilmente y esto genera productos de alta IMD.

En un bucle ALC externo correctamente ajustado, se detecta una pequeña muestra de la salida del amplificador y genera una tensión CC proporcional a la potencia de salida. Por razones históricas, el voltaje de salida de CC es un voltaje negativo: en los primeros días todos los transmisores usaban válvulas y para reducir la ganancia en un amplificador de válvula, el voltaje de la red de las etapas de baja potencia se cambió a un voltaje más negativo. Por lo tanto, el voltaje de salida del ALC del amplificador lineal externo es un voltaje negativo de entre 0 V y aproximadamente –9 V. Esto es suficiente para desviar completamente las etapas de ganancia, por lo que el voltaje ALC promedio es generalmente un poco menor que esto, quizás entre –4V y
–6V.

Cuando esta señal de retroalimentación está conectada al transmisor de excitación, la potencia de salida del amplificador controla el nivel de excitación posible. Si el ALC está correctamente ajustado, es imposible sobrecargar el amplificador externo.

Para ajustar el ALC, primero se aplica una pequeña cantidad de excitación y se carga el amplificador hasta la salida nominal completa. Luego, se enchufa el cable ALC y se vuelve a aplicar la unidad mientras se monitorea la potencia de salida. La unidad está configurada para producir la potencia de salida deseada. El ajuste ALC del amplificador se utiliza para hacer que la potencia de salida comience a reducir la potencia de salida. A partir de entonces, la unidad se puede aumentar al máximo y cualquier nivel en el que se haya establecido el ALC mantendrá el amplificador en esta potencia de salida máxima.

Como la mayoría de los amplificadores tienen una pendiente de entrada-salida que tiende a curvarse cerca de la parte superior del rango de potencia, es prudente configurar el ALC a un nivel aproximadamente un 20% más bajo que la potencia nominal real. Esto asegura que el amplificador no pueda exceder esta potencia y esté por debajo del punto de compresión de 1dB. 1dB es 20% en números redondos. Por ejemplo, si el amplificador está clasificado por el fabricante en 600 W PEP, no debe funcionar a más de 480 W PEP para garantizar un buen rendimiento de IMD.

Si se ha probado el amplificador y el rendimiento del IMD se conoce con precisión a diferentes niveles de potencia, se puede determinar con mayor precisión el punto de potencia superior exacto.




7: medición de la linealidad del amplificador

T aquí varios métodos diferentes de uso común para evaluar la linealidad de RF
amplificadores. Aunque los radioaficionados rara vez consideran otra cosa que no sea la prueba de dos tonos, existen otros medios para evaluar la linealidad. Lo que hay que tener muy presente es lo que estamos tratando de lograr. En la mayoría de los casos, el énfasis está en prevenir, o al menos minimizar, el nivel de productos de intermodulación de orden impar (IMD). Si se permite que se salga de control, hacen que la señal transmitida se extienda por una gran parte de la banda, comúnmente conocida como 'salpicadura', y hace que las estaciones que escuchan en una frecuencia adyacente escuchen estos productos no deseados.

En días pasados, antes de que la banda lateral única se hiciera popular, los transmisores generalmente eran tipos de portadora completa modulada en amplitud. Estos transmitían una portadora así como las dos bandas laterales y, como tales, desperdiciaban la energía que se usaba. Al eliminar la portadora y una banda lateral, se hizo un mejor uso de la potencia disponible ya que toda la potencia estaba contenida en una única banda lateral.

Si se mira en retrospectiva cómo eran las cosas en los días anteriores a la SSB, la prueba de un transmisor o amplificador era mucho más simple y requería muy poco equipo de prueba. Aunque los amplificadores utilizados para la etapa final no se clasificaron como amplificadores lineales , de hecho eran bastante buenos para producir una señal limpia con muy poca distorsión. Esto se debió a que la modulación se aplicó directamente a la etapa final del amplificador.

Modulación de ánodo

El transmisor de AM más simple usaría una válvula de triodo en la etapa de salida. Esto se muestra en la Fig . 7.1 .

Fig 7.1: Etapa de salida del transmisor de amplitud modulada.

En este método, el alto voltaje del ánodo se alimenta al secundario de un transformador de audio al que el devanado primario tiene conectado un amplificador de potencia de audio. Por tanto, la señal moduladora hace que varíe el voltaje del ánodo. Sin modulación aplicada, la portadora no está modulada y la potencia de salida es constante. Cuando se alimenta una señal de audio sinusoidal a la entrada del micrófono, el amplificador de audio genera un voltaje adicional que aumenta y disminuye el voltaje del ánodo aplicado. En el semiciclo positivo de la señal de audio, el voltaje del ánodo aumenta hasta un máximo y en el otro medio ciclo disminuye hasta un mínimo. La profundidad de modulación se mide en relación con la alta tensión normal alimentada al ánodo. Si el voltaje del ánodo sin modulación es de 750 V CC,cuando la señal de audio hace oscilar el voltaje del ánodo hacia arriba hasta la modulación completa, el voltaje del ánodo aumenta al doble del nivel normal. Por lo tanto, alcanza un pico en la parte superior del semiciclo a 1500V. De manera similar, en la otra mitad del ciclo desciende a cero voltios. En la práctica, es muy difícil hacer una válvula que pueda bajar a cero voltios, por lo que la modulación máxima posible nunca puede alcanzar exactamente el 100%, pero el 90% es generalmente posible.

Si la señal de salida de RF se ve en un osciloscopio, la portadora se verá como el gráfico que se muestra en la figura 7.2 .

Fig 7.2: Una portadora de RF modulada al 100%.

Notará que la portadora sube a un pico pero también cae a amplitud cero antes de subir al siguiente pico. Una estimación del porcentaje de modulación es que si la envolvente de la portadora no cae a cero, la modulación es inferior al 100%. Para la condición en la que el voltaje del ánodo aumenta al doble de la condición estática, generalmente se asumen ciertas cosas.

En primer lugar, se supone que si se duplica el voltaje del ánodo, la corriente del ánodo también debería duplicarse. La entrada de energía es una función del producto del voltaje del ánodo multiplicado por la corriente del ánodo, asumiendo que la etapa funciona de manera lineal. Para el caso en el que el voltaje del ánodo y la corriente del ánodo aumentan en un factor de dos, la potencia de entrada instantánea aumenta a un valor que es cuatro veces el de la condición no modulada.

La segunda suposición que se hace es que la ganancia y la eficiencia del amplificador permanecen constantes para diferentes niveles de potencia de entrada. Eso es, después de todo, lo que define un amplificador lineal: la potencia de salida frente a la potencia de entrada es una línea recta inclinada en algún ángulo en el gráfico. Por lo tanto, si la potencia de entrada aumenta en un factor de cuatro veces, la potencia de salida también debe aumentar en la misma cantidad. Suponiendo que la potencia de entrada de CC bajo una portadora no modulada es de 150 W y la eficiencia de la etapa es del 66,6%, la potencia de la portadora es de 100 W. Por debajo de los picos de modulación del 100%, la potencia de salida debe ser 400W PEP. Aquí es de donde se derivan las condiciones de la licencia de aficionados del Reino Unido.

    Antes de los días de la banda lateral única, los transmisores eran del tipo de amplitud modulada y la potencia de entrada de CC se limitaba a 150 W de CC en la etapa final del amplificador. Se asumió que la eficiencia general de la etapa de salida de Clase C sería un promedio del 66,6%, lo que era una suposición razonable. Sin embargo, con una muy buena etapa de Clase C, fue posible una eficiencia del 75% a bajas frecuencias. La eficiencia tiende a disminuir a medida que aumenta la frecuencia, pero el 66,6% cubría el 99% de las estaciones de aficionados. Cuando SSB se hizo popular, los reguladores simplemente aplicaron los límites de AM para un transmisor portador de 100 W, por lo que la licencia SSB es la misma a 400 W PEP, que hoy se da como + 26 dBW, es decir, 26 dB por encima de la salida de 1 W.

Linealidad de un amplificador de clase C

Una etapa de Clase C modulada en amplitud es de hecho notablemente lineal cuando se ajusta correctamente. Ver la envolvente de RF en un osciloscopio no muestra prácticamente ninguna distorsión por debajo de aproximadamente el 90% de profundidad de modulación. Para medir con precisión la distorsión, se requiere un equipo de prueba especializado además de un osciloscopio.

La prueba estándar de la industria para la distorsión AM no se realiza al 100% de modulación sino al 60% o al 80%, ya que la mayoría de los amplificadores tendrán cierta no linealidad con un porcentaje de modulación muy alto. En transmisores comerciales, la profundidad de modulación está limitada por audio AGC o algún otro método, por lo que el transmisor no puede sobremodularse.

El método de prueba consiste en alimentar al conector del micrófono un tono de audio sinusoidal puro, ajustando el nivel de entrada para alcanzar el 60% de modulación y medir la distorsión generada en la portadora modulada. El tono de prueba estándar es de 1 kHz, pero también se aplican 400 Hz y, a veces, 3 kHz para ver si la distorsión está relacionada con la frecuencia.

La señal de RF se alimenta a través de un atenuador adecuado a un medidor de modulación . Este es un detector de RF lineal para demodular la señal. El medidor de modulación puede determinar el porcentaje exacto de modulación. El audio recuperado luego se envía a un analizador de distorsión de audio , como el Hewlett Packard HP-331A, que puede medir la distorsión armónica presente. El uso de un instrumento como este puede medir la distorsión armónica total a niveles muy bajos. Es común encontrar que se pueden lograr niveles de distorsión tan bajos como 0.1% con un transmisor AM correctamente ajustado. Esto es a pesar del hecho de que un amplificador de Clase C es muy no lineal para señales SSB.

Se entiende que si la señal de entrada de audio contiene productos armónicos, la señal de salida siempre mostrará algunos productos de distorsión. La señal de audio debe ser lo más perfecta posible y, en general, es necesario tener una distorsión armónica total de ~ 0,1% para realizar una medición válida del rendimiento del amplificador. El HP-331A puede medir hasta un 0,01% de distorsión.

La diferencia entre transmisores AM y SSB

La razón por la que un amplificador de Clase C puede amplificar una señal de AM de manera lineal mientras que el mismo amplificador, cuando se usa SSB, genera un alto nivel de distorsión, se puede explicar simplemente.

En el amplificador AM, la señal de activación de RF en la válvula es una señal de nivel constante; la modulación se aplica al ánodo. Por lo tanto, la señal de excitación suministra el voltaje de RF necesario para mantener la válvula en la región de conducción lineal. Por el contrario, el mismo amplificador cuando se alimenta con una señal SSB no se comporta de la misma manera. Aquí la entrada de la portadora es muy baja para algunas partes del ciclo de entrada y en otras ocasiones es suficiente para desviar la válvula correctamente. Este efecto también se nota con amplificadores de estado sólido polarizados para la Clase C. Si la señal de excitación está por encima de un cierto nivel, el amplificador funciona en un modo cuasi-lineal. Cuando la señal de entrada cae por debajo de cierto nivel, el transistor se vuelve no lineal, ya que no hay suficiente impulso para encenderlo.Para reducir la no linealidad, tiene que haber una corriente de reposo mínima que fluya en todo momento y la polarización de CC alimentada al dispositivo hace que se mueva del modo no lineal al modo semilineal.

Algunas de las técnicas para mejorar la linealidad en un amplificador de válvula también se pueden aplicar a amplificadores de estado sólido, y era común en los días de los transmisores AM que usaban transistores. A medida que aumenta la profundidad de modulación, la corriente del ánodo no siempre sigue la curva deseada de la característica de la válvula. Esto se debe a que a una alta profundidad de modulación, el nivel de excitación es insuficiente.

Para solucionar este problema, se adoptó el uso de válvulas de rejilla de pantalla, como pentodos y tetrodos. En estos, no solo se modula el voltaje del ánodo, sino que también se alimenta la rejilla de la pantalla con una porción más pequeña del alto voltaje modulado. Este aumento del voltaje de la rejilla de la pantalla aumenta la ganancia de la válvula y compensa la falta de impulso. Por lo general, esto se obtenía del suministro modulado del ánodo a través de una resistencia de caída de modo que el voltaje del ánodo y la rejilla de la pantalla subía y bajaba en simpatía con la relación correcta de voltaje. Generalmente ocurre que cuando se ha seleccionado el valor correcto de resistencia para la condición no modulada, también se ajusta bien a la condición completamente modulada. A veces, la modulación de la rejilla de la pantalla se toma de una toma en el secundario del transformador de modulación para adaptarse mejor a la válvula que se está utilizando.La inclusión de la cuadrícula de pantalla en el sistema de modulación permite lograr una modulación más completa y con una etapa de salida de tetrodo bien ajustada se puede lograr una profundidad del 100% con baja distorsión.

Otra solución es modular parcialmente hacia arriba la etapa del controlador de modo que en la oscilación ascendente de la corriente del ánodo, el voltaje del ánodo del controlador también se impulse hacia arriba. Esto aumenta el nivel de transmisión de RF durante los picos y reduce considerablemente la distorsión. Este método casi siempre se usó con transmisores de AM de estado sólido para garantizar un impulso adecuado en los picos de voz. Es fácil de arreglar con un diodo para pasar el voltaje ascendente requerido y bloquear el voltaje descendente. El diodo alimentó el pulso de CC extra de media onda sinusoidal al ánodo del controlador en paralelo con el suministro normal del ánodo.

Potencia portadora vs potencia pico de modulación al 100%

Esto ya se ha discutido, pero es útil recordar al lector que si se utiliza un amplificador lineal para amplificar una transmisión de AM, el ajuste correcto es fundamental para garantizar una baja distorsión.

En condiciones no moduladas, la potencia de salida no debe ser superior al 25% de la potencia nominal máxima del amplificador. Suponiendo que el amplificador tiene una potencia nominal de 100 W PEP, la potencia de la portadora no debe ser superior a 25 W. Si el amplificador no es perfectamente lineal, y muy pocos amplificadores lineales son realmente lineales, un ajuste más seguro sería no más de 15 a 20 W de portadora no modulada. Esto significa que el PEP máximo posible es solo de 60 a 80 W y dependerá de la linealidad de potencia superior del amplificador.

En la figura 7.3 se muestra una indicación de un amplificador típico . El amplificador está clasificado oficialmente como un tipo PEP de 100 W, pero las afirmaciones están un poco fuera del caso real. Para evaluar la linealidad de un amplificador es necesario trazar la curva salida-entrada. En un amplificador perfecto, la ganancia sería constante, sin importar cuál sea la potencia de salida. Habrá un límite superior finito donde la línea recta ideal se desvía.

Fig 7.3: Curva de entrada vs salida para un amplificador típico.

Aunque se dice que es un amplificador PEP de 100 W, no lo es. De hecho, el amplificador entrega un poco más de 100W, se traza hasta una salida de aproximadamente 105W. Pero la porción lineal real se detiene en aproximadamente 80W de salida. Si el amplificador siguiera el caso ideal, estaría disponible una salida de aproximadamente 130W. A niveles bajos, la ganancia es de 10 dB, pero cuando la salida supera los 80 W, comienza a caer. Con una entrada de 9W, solo suministra 85W y no 90W. En niveles de conducción más altos, se vuelve cada vez más no lineal. Con una entrada de 13W, apenas supera los 100W. Sería arriesgado afirmar que esto es más que un amplificador PEP de 70 W, ya que la linealidad se desvía de una condición ideal. Para FM o CW esto no importa, pero para AM y SSB definitivamente afecta la pureza de la señal transmitida. Si la unidad de entrada aumenta más, la potencia de salida eventualmentela línea plana y la adición de más impulso no darán ningún aumento adicional. El amplificador estaría totalmente saturado. Para garantizar productos con IMD baja, es necesario operar el amplificador muy por debajo del punto de compresión de 1dB.

La generación de productos IMD

Los productos de distorsión de intermodulación son el resultado de un proceso de mezcla que ocurre en un amplificador no lineal. Si se examinara un amplificador ideal, la señal de salida sería una réplica exacta de la señal de entrada, solo que más grande. En un amplificador típico se produce cierta no linealidad. Esto da lugar a la producción de señales de salida adicionales además de la deseada. El tipo predominante de señales adicionales son los armónicos de la entrada y estos pueden ser considerables en un amplificador de Clase C. Si el amplificador solo se usa para CW o FM, esto no es una preocupación real, ya que la potencia armónica se puede reducir colocando un filtro de paso bajo adecuado después del amplificador para limpiar la salida.

En el caso de un amplificador de Clase C de amplitud modulada, es la mezcla de la portadora de RF con la señal de modulación de audio lo que produce las dos bandas laterales a cada lado de la portadora. Este es un caso en el que necesitamos una etapa muy no lineal para producir la señal de salida AM.

    Si el amplificador fuera ideal, sin no linealidad, sería imposible generar una señal AM, ya que no se puede producir ninguna mezcla. Tratar de generar una señal AM usando un amplificador lineal tiene malos resultados ya que no hay suficiente no linealidad para realizar el proceso de mezcla.

En el caso de que la señal de entrada no sea una sola frecuencia portadora, sino dos portadoras, los productos de mezcla tienen dos posibles señales extra generadas. En el transmisor de AM sabemos que si la señal de audio es de 1 kHz, las dos bandas laterales se producirán –1 kHz y + 1 kHz a cada lado de la portadora. En IMD de orden impar, el segundo armónico de una de las portadoras se mezcla con la frecuencia fundamental de la entrada de la otra portadora para generar el producto extra. Estos se repliegan para ocupar un lugar adyacente a las portadoras, arriba y abajo, definido por la diferencia de frecuencia de las dos señales de entrada. Este primer producto IMD se conoce como producto de tercer orden, porque es el resultado de la mezcla del segundo armónico (una señal de segundo orden) más el primer armónico, conocido como la frecuencia fundamental, por lo tanto 2 + 1 = 3.Para el caso del producto de quinto orden, es el resultado del tercer armónico y el segundo armónico de los dos portadores que también se pliegan hacia atrás para situarse encima y debajo del producto de tercer orden. Si el amplificador es muy no lineal, estos productos IMD se extienden hasta el infinito, disminuyendo gradualmente en amplitud. Los mejores productos son normalmente de tercer orden, luego el quinto orden es menor y así sucesivamente hasta la serie IMD.

En un transmisor de AM típico, estos productos adicionales también son visibles como bandas laterales adicionales por encima y por debajo de las normales, pero generalmente son tan bajas que no representan un problema en la práctica. Estos se deben a los productos de distorsión (armónicos) contenidos en la señal de entrada de audio, a menudo generados en la etapa de salida del amplificador de potencia AF, y son exactamente el mismo mecanismo que la generación SSB IMD. Por lo tanto, con una señal de audio de 1 kHz, el segundo armónico del audio ocurre a 2 kHz y da lugar a las bandas laterales adicionales observadas. Si la señal de audio es pura, no se pueden generar.

Generando una señal de prueba

En los laboratorios de diseño de RF, el método clásico de generar una señal de prueba es utilizar dos generadores de señales de RF de muy alta calidad y combinar las dos señales en un combinador de potencia de 0º. Las amplitudes se ajustan para que todas sean idénticas y las frecuencias se establecen para producir una pequeña diferencia de frecuencia, digamos 2 kHz. Esta señal luego se aplica al amplificador bajo prueba y la salida se visualiza en un analizador de espectro. Cualquier no linealidad será evidente por los productos IMD vistos. En un amplificador perfectamente lineal, solo las dos portadoras discretas deberían ser visibles. En un amplificador práctico, se generará cierta distorsión en forma de productos de distorsión de intermodulación.

Para los casos en los que la potencia de entrada al amplificador es una señal grande, la salida de los generadores de señales combinados debe amplificarse. Los generadores de señales de RF generalmente solo pueden desarrollar una salida de aproximadamente 100 mW de portadora no modulada. Estos cuando se combinan producen una señal compuesta de 400mW PEP, asumiendo que no hay pérdida en el combinador. La linealidad del amplificador que sigue a los generadores de señales tiene que ser bastante excepcional si la señal de prueba va a tener productos de IMD muy bajos. Esto a menudo significa amplificadores costosos con filtrado adicional y obtener una señal lo suficientemente limpia no es un asunto trivial, especialmente si el nivel de excitación requerido es alto.

Para los aficionados, este no es un sistema viable, por lo que una alternativa es generar la señal de prueba a frecuencias de audio y alimentar esta señal compuesta de dos tonos en el conector del micrófono del transmisor. Los tonos de audio seleccionados no deben estar relacionados armónicamente, por lo que no se pueden utilizar tonos de 1 kHz y 2 kHz. Generalmente, los tonos de audio deben poder pasar a través del filtro SSB, por lo que deben estar por debajo de aproximadamente 2,5 kHz. Una buena elección sería 1kHz y 1.8kHz, ya que no están relacionados y ambos están en la banda de paso completo del filtro.

    De los dos métodos diferentes, la coherencia de fase de la señal compuesta no es importante. En el caso del método de generador de dos señales, la fase no es un valor fijo, varía ya que las dos frecuencias son diferentes. En el método alternativo de audio de dos tonos, la señal de salida compuesta del modulador doble balanceado es coherente en fase, ya que se generan a partir de una frecuencia portadora común. En el modulador doble balanceado, la portadora se anula con los ajustes de balanceo de modo que solo las dos bandas laterales son visibles. En la práctica, es posible un nulo de portadora de –50 dB o menos. Si la portadora no se puede anular lo suficientemente baja, esto equivale a una tercera señal portadora, similar a un verdadero transmisor de AM.En un transmisor SSB, el filtro de cristal de FI ayuda a eliminar una de las bandas laterales y la portadora residual para que la señal portadora de dos tonos sea generalmente lo suficientemente limpia, si el transmisor está correctamente ajustado.

Cualquiera de estos dos métodos generará una señal compuesta que, cuando se ve en un osciloscopio, se verá idéntica a la de la figura 7.2.: ambos son iguales a una señal AM modulada al 100%. Sin embargo, cuando se ven en un analizador de espectro, no son lo mismo. La señal de AM tiene tres señales principales, la portadora central grande y las dos bandas laterales de menor amplitud, espaciadas a cada lado de la portadora por una distancia correspondiente a la frecuencia de modulación. Si la frecuencia portadora es de 1 MHz y la señal de modulación es de 1 kHz, las bandas laterales se producirán a 999 kHz y 1001 kHz, por lo que están separadas por 2 kHz. En una señal AM modulada al 100%, las bandas laterales están exactamente 6 dB por debajo del nivel de la portadora. En un transmisor SSB, la señal de audio de dos tonos mostrará solo las dos bandas laterales, asumiendo que el balance de la portadora está correctamente ajustado para anularlo en el ruido.

De las dos señales diferentes, la portadora modulada AM es la más buscadora, ya que es un estímulo de tres tonos. La medición de los productos IMD con solo el estímulo de dos tonos a menudo muestra un buen rendimiento de IMD, pero la señal de tres tonos muestra que los IMD no están presentes en la prueba de dos tonos.

La modulación de AM está más cerca de lo que sucede cuando se aplica una señal de voz compleja al transmisor. La voz humana contiene no solo dos señales discretas de frecuencia fija, sino un espectro completo de amplitudes variables y separación de frecuencias. Si el analizador de espectro tiene una función de retención de picos y una persona habla por el micrófono durante unos segundos, la imagen que se ve muestra la respuesta de banda de paso del filtro SSB a medida que se llena gradualmente.

    El método de prueba mencionado anteriormente se usa ampliamente en la televisión por cable y otras industrias donde se usa una señal de estímulo de tres tonos para explorar mejor la linealidad de los amplificadores usados. En los sistemas de televisión por cable (CATV) pasan hasta 10 canales a través de los amplificadores y cada uno tiene un nivel de señal variable. Los IMD pueden resultar interesantes en este tipo de sistemas. Esta es la razón por la que los amplificadores utilizados son capaces de una potencia muy alta pero funcionan a una fracción de su capacidad máxima.

Pureza del requisito de señal de dos tonos

La pureza de la señal de audio de dos tonos tiene que ser muy buena. Si se considera la conversión de THD en términos de porcentaje a dB, una distorsión de audio del 1% equivale a
–40dB. Si es probable que el IMD del amplificador sea menor que éste, la distorsión de audio debe ser considerablemente menor para que este método funcione. En la figura 7.4 se muestra un diagrama de bloques de un oscilador de dos tonos .

Fig 7.4: Diagrama de bloques del oscilador de audio de dos tonos.

La señal de audio de dos tonos se usa para generar la señal SSB y luego pasa a través de varios mezcladores y etapas de ganancia en el transmisor para generar la frecuencia final en el nivel de salida requerido para impulsar el amplificador bajo prueba. Si la THD de los tonos combinados es solo del 1% en el micrófono, es probable que a la salida del transmisor sea considerablemente peor. Es común encontrar que hay una degradación de 10 a 15 dB en el transmisor de excitación. Por lo tanto, la señal de entrada de –40 dB se degrada a tan solo –25 dB, que es demasiado alta para usarse en pruebas precisas.

Es la norma en las pruebas de IMD requerir una señal de excitación que sea al menos 6dB mejor que la que se está midiendo. Por lo tanto, la señal compuesta de dos tonos debe ser de al menos –60 dB cuando se aplica a la entrada del micrófono y esto generará una señal de salida de aproximadamente –45 dB IMD en la salida del transmisor. Esto significa que lo más bajo que se puede medir es de aproximadamente –39dB IMD. Una señal de entrada de –60dB equivale a una distorsión de audio de 0.1% y eso se está volviendo difícil de generar con osciladores de audio simples.

Conversión de THD a decibelios

La cifra de THD en porcentaje es una medida de voltaje. Para convertir a la medida de potencia correspondiente en decibelios se utiliza la siguiente fórmula:

Distorsión en dB = 20 log (% / 100)

El 10% de distorsión es igual a 1/10 = 0,1 y equivale a –20dB

1% de distorsión es igual a 1/100 = 0.01 y equivale a –40dB.

0.1% de distorsión es igual a 1/1000 = 0.001 y equivale a –60dB.

Generación de dos tonos y precauciones necesarias

Los osciladores de audio para generar los dos tonos discretos pueden ser osciladores de retroalimentación simples. Uno de los tipos más simples es el oscilador de cambio de fase que usa componentes comunes y es simple de hacer. En la figura 7.5 se muestra un esquema de un oscilador de cambio de fase . Cada oscilador tiene un interruptor, por lo que cualquiera o ambos pueden operarse en cualquier momento.

Fig 7.5: Oscilador de cambio de fase de audio.

La frecuencia de oscilación está definida por los valores de C2 y las resistencias R6, R7 y R8. Para aumentar la frecuencia, estas resistencias se reducen de valor. Con los valores mostrados, la frecuencia es de aproximadamente 1100Hz. Para modificar para un tono más alto, R6, R7 y R8 se cambian a 8.2k Ω , lo que aumenta la frecuencia a ~ 1700Hz. Las frecuencias exactas utilizadas no son críticas siempre que no estén relacionadas armónicamente. La señal compuesta cuando se suman las dos salidas del oscilador crea una nota de tiempo de aproximadamente 600 Hz.

El oscilador consta de TR1, que es un amplificador inversor. La salida en el colector impulsa TR2 que funciona como otro amplificador inversor para la señal en el colector. El emisor actúa como seguidor de voltaje y tiene un cambio de fase cero entre la base y el emisor. Esta señal de emisor tiene una impedancia baja para impulsar la red RC que consta de R6, R7 y R8. Los condensadores, que son todos del mismo valor, y las resistencias forman una red de desplazamiento de fase de tres etapas de 180º, con 60º por sección. Esta señal luego se aplica a la base de TR1. El desplazamiento de fase total alrededor de la red es (180 + 180 = 360º) o (180 - 180 = 0º), ambos equivalen a 0º. Por lo tanto, el oscilador funcionará a la frecuencia donde el cambio de fase en la red RC es igual a 180º.La red RC se comporta como un filtro de paso alto, por lo que los armónicos de los transistores del amplificador están contenidos en la señal de salida.

Para garantizar que la salida del oscilador esté limpia, se requiere un filtro de paso bajo entre cada oscilador y la red del combinador. Cualquier oscilador contiene armónicos y si se les permite pasar a las etapas de audio del micrófono transmisor, generarán sus propios productos IMD. En consecuencia, se inserta un filtro de paso bajo de corte agudo entre cada oscilador y el combinador. El combinador es una red de resistencias simple que permite el ajuste de los niveles de los osciladores individuales y la salida combinada al transmisor. Esto se muestra en la Fig . 7.6 .

Fig 7.6: Red de combinación de osciladores.

Los resistores RS establecen el voltaje de salida máximo en el combinador y pueden omitirse en algunas aplicaciones. La resistencia variable VR1 es un control de balance para obtener la misma amplitud para cada tono. El nivel de salida lo establece VR2. Esto debe ajustarse para adaptarse al transmisor bajo prueba, a menudo tan solo 10 mV RMS es suficiente. Cuanto mayor sea el valor de RS, mejor será el aislamiento entre los dos osciladores y menor será la amplitud total de la señal de salida. El oscilador de cambio de fase que se muestra entregará al menos 1V de salida pico a pico, por lo que RS será un valor bastante alto. Las resistencias R9, R10 se pueden utilizar para establecer el nivel de atenuación variando RS. Si lo desea, una de las resistencias RS se puede reemplazar por una resistencia en serie variable para un control de nivel más fino de la salida. Solo uno necesita ser variable.

La conexión entre el generador y el transmisor debe realizarse con un cable bien blindado para garantizar que no pueda entrar RF en las etapas del amplificador del micrófono. Todo el generador también debe estar en un recinto bien protegido por la misma razón: ¡construir este generador de dos tonos en una caja de plástico generará problemas! La mejor opción es alimentar la unidad con una batería interna de 9 V en lugar de intentar suministrarla con una fuente de alimentación externa. La corriente consumida es baja y una pequeña batería de radio de transistores de batería de 9V debería durar mucho tiempo.

Los valores mostrados se adaptan a un corte de aproximadamente 1000 Hz. (Los valores de los componentes para el filtro de 1700Hz son R1, R2 = 27k Ω y C1 = 3.3nF, C2 = 1.5nF.) Al configurar el transmisor para realizar una prueba de dos tonos, es importante tener los niveles correctos. Si se dispone de un osciloscopio, los niveles de señal se pueden comprobar y ajustar con mayor precisión. La señal de salida compuesta cuando se ve en un osciloscopio debe verse como una señal AM modulada al 100%: los dos picos deben tener exactamente la misma amplitud y la señal debe caer a cero sin discontinuidad en el punto de amplitud cero. Sin embargo, ¡hay algunos problemas no revelados que podemos encontrar!

Ajuste de nivel correcto

Si el oscilador de dos tonos está configurado correctamente, cuando la salida se visualiza en un osciloscopio, debería aparecer exactamente igual que se muestra en la Fig . 7.2 . Si un tono tiene una mayor amplitud, la forma de onda se asemeja a una señal AM submodulada. Cuando los dos niveles de oscilador son idénticos, la forma de onda caerá a cero en el punto de cruce. Es simple lograr esto, pero a menudo encontramos que cuando conectamos esta señal al transmisor y miramos el espectro, produce un resultado pobre. La señal de RF en la salida del transmisor cuando se ve en un osciloscopio aparece como una señal de AM submodulada. ¡Mucho rascarse la cabeza!

La razón por la que la señal ya no parece tener la forma de onda correcta es que, a diferencia de un amplificador de alta fidelidad de audio de banda ancha, el transmisor SSB tiene algunas etapas que son de banda muy estrecha. Las dos partes del transmisor SSB generalmente responsables son el amplificador de micrófono y el filtro de cristal. En muchos casos, ninguna de estas etapas es realmente plana en respuesta de frecuencia.

Las etapas del amplificador de micrófono pueden tener un pronunciado roll-up o roll-off para ciertas frecuencias. De manera similar, los filtros de cristal SSB rara vez son totalmente planos en toda la banda de paso. Dependiendo del tipo, pueden tener una respuesta de pendiente ascendente o una respuesta de pendiente descendente. Ambas respuestas inclinadas colorean la señal de voz y mejoran o degradan la calidad del audio. Además, si el oscilador de inserción de portadora no está colocado correctamente en la pendiente de la respuesta del filtro, también altera la respuesta del ancho de banda para los componentes de voz de alta o baja frecuencia.

La única forma segura de asegurarnos de que tenemos dos portadoras iguales es ajustar los niveles de tono usando el transmisor.

Para configurar el transmisor, primero necesitamos algunos equipos de prueba. Como mínimo, necesitamos un vatímetro de RF de lectura promedio con una carga ficticia y un osciloscopio adecuado para la frecuencia para la que está diseñado el transmisor. Si se trata de un transmisor de HF, un osciloscopio de 10 MHz funcionará bien hasta unos 30 MHz. La respuesta de amplitud de un osciloscopio disminuye a medida que aumenta la frecuencia, pero la amplitud relativa de la señal normalmente no lo hace. El mejor método es comenzar en una banda de frecuencia más baja, como 80 m, donde el alcance dará resultados perfectos. A 10 m, para la misma potencia de salida, la altura vertical de la traza será un poco menor, eso es todo.

Configure el transmisor en LSB o USB y asegúrese de que el compresor de voz (o como se llame en el transmisor) esté apagado. Solo necesitamos las etapas normales de micrófono operativas. Conecte la caja de prueba de dos tonos al conector del micrófono.

Para empezar, apagamos uno de los osciladores de audio para que tengamos un tono de onda sinusoidal pura en las etapas del micrófono. Cargamos el transmisor de la manera normal para producir una señal CW (onda continua), simplemente una portadora pura. Al observar el medidor de potencia de RF, ahora ajustamos el control de ganancia del micrófono para establecer la portadora exactamente al 25% de la potencia de salida máxima nominal. (Debe configurar el potenciómetro de nivel de salida de dos tonos para que se adapte a la sensibilidad del micrófono del transmisor. La mayoría de los transmisores de HF necesitan aproximadamente 10 mV RMS para lograr una salida completa con el control de ganancia del micrófono a mitad de camino). Si el transmisor es del tipo PEP de 100 W, ajuste la salida exactamente a 25 W indicados en el vatímetro.

Habiendo configurado este nivel correctamente, no debemos ajustar el control de ganancia del micrófono en la configuración posterior.

El primer oscilador se apaga y el segundo oscilador se enciende. Si el segundo oscilador tiene exactamente la misma amplitud, el transmisor mostrará exactamente el 25% en el medidor de potencia. A menudo no es así. Esto se debe a que las etapas del amplificador del micrófono del transmisor o el filtro de cristal tienen una respuesta con pendiente de ganancia. Por lo tanto, ajustamos cuidadosamente el control de balance de salida de los osciladores para obtener exactamente el 25% del nivel de portadora. Si cambiamos entre los dos osciladores de audio cuando están configurados correctamente, ambos deberían dar exactamente el 25% del nivel de portadora como se indica en el vatímetro. Ahora la configuración de audio es correcta y ahora podemos, si es necesario, ajustar el control de ganancia del micrófono para mediciones posteriores.

Teclee el transmisor y encienda ambos tonos y observe el vatímetro. Ahora debería indicar el 50% de la potencia nominal del transmisor. El vatímetro lee el promedio de la potencia que genera la señal de dos tonos. Observe la envolvente de RF en el osciloscopio y busque cualquier falta de linealidad en los picos de la forma de onda. Si se ve exactamente como la forma de onda de AM de la figura 7.2 , significa que el transmisor está funcionando dentro de su región lineal.

Lo que es más probable es que se observe una pequeña capa plana. Si es así, ajuste con cuidado el control de ganancia del micrófono del transmisor para que desaparezcan. Lea la potencia promedio en el vatímetro. El verdadero PEP es el doble de la lectura promedio, así que no se sorprenda si su supuestamente transmisor PEP de 100W solo parece dar 80W PEP bajo esta condición; eso es bastante común con la mayoría de los transmisores. Si se trata de un transmisor de estado sólido, espere ver tan solo 60 W PEP, porque los transistores no son muy lineales en el extremo superior de su rango de potencia.

La prueba de envolvente de RF con un humilde alcance no es tan precisa como un analizador de espectro, que puede medir el nivel exacto de los productos IMD, pero es mucho mejor que confiar solo en los medidores. Cuando el transmisor esté configurado en la linealidad óptima, anote la lectura actual para referencia futura. También tenga en cuenta la potencia de RF promedio, esto es lo máximo que debe usar para garantizar la linealidad.

Una vez que haya aprendido cómo configurar una prueba de distorsión, puede intentar ajustar el control de ganancia del micrófono para ver el efecto que tiene en la forma de onda. A medida que aumenta la ganancia del micrófono, la envolvente comenzará a aplanarse y la amplitud aumentará en consecuencia. Si la ganancia avanza demasiado, la forma de onda se distorsiona mucho. Tenga en cuenta la potencia media indicada y duplique mentalmente la cifra para obtener una aproximación de la lectura de PEP. La lectura de PEP no será correcta ahora porque la señal está distorsionada y la relación promedio a PEP de 2: 1 ya no se aplica.

Supongamos que cuando avanza la ganancia del micrófono para obtener una cantidad notable de tope plano, la potencia promedio indicada es de 70W. Teóricamente, si el transmisor fuera perfectamente lineal, indicaría un valor de PEP de 140W. Pero esto nunca sucederá con un transmisor PEP nominal de 100 W, ya que se ha quedado sin energía disponible.

Prueba de un amplificador lineal externo

Para probar un amplificador lineal externo, configuramos el transmisor como se describió anteriormente para obtener una forma de onda perfecta y luego lo usamos para impulsar el amplificador. Seguimos el mismo procedimiento de configuración con cada tono individualmente para encontrar la configuración correcta del nivel de potencia del 25%. Cambie entre los dos tonos de audio diferentes para asegurarse de que la potencia media de cada uno sea idéntica. Luego encienda ambos tonos y observe la forma de onda de la envolvente de RF. Si muestra signos de que la parte superior está plana, retire la unidad de RF del transmisor bajando el control de ganancia del micrófono. Cuando haya establecido el ajuste óptimo, lea la potencia media y duplique el valor. Esta es la potencia lineal máxima absoluta en PEP que el amplificador puede producir con baja distorsión. Una vez más, ¡se sorprenderá de lo bajo que es en realidad!

Mientras tiene el equipo de prueba conectado, ahora es el momento de conectar y configurar los ajustes del ALC externo para que no pueda exceder este nivel de potencia. ¡Todos los demás usuarios de la banda te lo agradecerán!

Cuando los agentes configuran uno para un cliente, normalmente lo configuran en aproximadamente 80W, por lo que cumplen con la ley y también tienen menos problemas de confiabilidad y, por lo tanto, menores problemas de garantía a menor potencia, y estoy totalmente de acuerdo con este enfoque. El problema es que los transistores de salida de RF pueden entregar hasta 180W cuando están completamente saturados. El fabricante eligió este dispositivo en particular por varias razones. En primer lugar, al usar un dispositivo más grande y menos de un 50%, obtienen una mejor linealidad y, en segundo lugar, la confiabilidad es mucho mayor.

El problema es que los aficionados se han aferrado a esta potencia extra "gratuita" y muchos transceptores han sido ajustados a más de 150W PEP. Al girar el control ALC completamente hacia arriba, el transistor funciona en bucle abierto, por lo que es muy sencillo encender la alimentación de este transceptor. La señal no solo suena muy distorsionada al hacer esto, sino que los tweakers también tienen una oportunidad en el potenciómetro de ganancia del micrófono. ¡La salpicadura que generan estos transmisores debe ser escuchada para creerla! Puedes ver uno a una milla de distancia: cuando escuchas salpicaduras en la banda, simplemente sabes que 'Walter' está al aire y está usando uno de ellos. También parecen necesitar nuevos transistores de PA que se ajusten con bastante regularidad: me pregunto por qué es así. ¡Con suerte, todos explotarán y serán enviados a la basura!

Cómo Eimac prueba la linealidad

Cuando un fabricante como Eimac necesita probar el IMD en válvulas experimentales, tiene un dispositivo de prueba especial que opera a 2MHz. Por lo tanto, se muestran todas las hojas de datos de válvulas Eimac comunes para mediciones de IMD de 2MHz. En un boletín de ingeniería de Eimac , Bill Orr, W6SAI, divulgó los detalles de su sistema de prueba interno. ¡La escala de esto es alucinante!

Tienen dos osciladores de cristal de alta estabilidad que funcionan a 2MHz con una diferencia de 2kHz en las frecuencias. Estos osciladores son, por supuesto, válvulas (¿qué más?). Cada oscilador tiene un filtro de paso bajo severo para rechazar cualquier armónico, por lo que cada señal es perfectamente sinusoidal. Luego combinan estas dos señales de alta potencia en un combinador híbrido especial para obtener una señal compuesta de dos tonos. Finalmente, esta señal se alimenta a una válvula amplificadora, pero no a cualquier válvula. Es, espere, un tetrodo 4CX5000, con un suministro variable de alto voltaje, que funciona en Clase A. La carga del ánodo es una resistencia no inductiva de 50 Ω y la salida está acoplada a través de un condensador. Esto presenta un perfecto 50 Ωfuente al amplificador bajo prueba. La pureza espectral de esta configuración es tal que es prácticamente imposible ver los IMD; están muy bien suprimidos. Esto puede, si es necesario, generar una señal casi perfecta de varios cientos de vatios para aplicar a cualquier amplificador que necesite datos de prueba.




8: Diseño de un circuito de tanque de ánodo

Este tema se trata con todo detalle en uno de mis otros libros, RF Design Basics (RSGB, 2007), así que solo cubriré las partes básicas aquí. Si desea tener una explicación más detallada, el libro mencionado es una buena fuente de referencia. También contiene tablas completas para diferentes redes Q cargadas, lo que ahorra mucho esfuerzo para derivar los valores correctos para los componentes.

La red de adaptación de ánodos más comúnmente utilizada en los amplificadores lineales de HF se conoce como red pi y en la figura 8.1 se muestra un tipo típico .

Fig 8.1: La red pi.

La red consta de solo tres componentes que realizan la transformación de impedancia del ánodo de la válvula a la carga. Estos son dos condensadores de sintonización mostrados como CT y CO y el inductor en serie L1. El resto de componentes son necesarios para asegurar el correcto funcionamiento de la válvula y para otros fines. La alimentación de CC del ánodo se realiza a través del inductor RFC1 y se desacopla a tierra de RF mediante CD. El condensador CB es simplemente un bloque de CC para evitar que el alto voltaje aparezca a través del terminal de la antena y no tiene ninguna función en la red correspondiente. Generalmente, este condensador se elige para que tenga una reactancia baja a la frecuencia de funcionamiento. De manera similar, RFC2 es un dispositivo de seguridad para cortocircuitar el suministro de alto voltaje del ánodo en caso de que el CB falle en cortocircuito. El APC parásito de ánodo normal no está incluido ya que no tiene ningún efecto en la red de adaptación.

Para diseñar una red de ánodo, necesitamos tener ciertos datos a mano. Los detalles más importantes son la resistencia de carga del ánodo requerida, el valor Q de carga requerido en el que debe trabajar la red y la capacidad del ánodo a tierra de la válvula. Sin estos valores no podemos empezar.

Tenemos cierta libertad de elección con la Q cargada, pero puede influir en si una solución es posible o no con la válvula seleccionada y las condiciones de funcionamiento elegidas. A menudo, el fabricante indicará la resistencia de carga del ánodo, pero en algunos casos no se proporciona. Seleccionaremos la válvula 811 y optaremos por tener tres válvulas en paralelo para lograr la potencia de salida deseada.

En las hojas de datos más antiguas para la operación de la válvula 811 en Clase B no se da para aplicaciones de RF, pero se da una cifra para dos válvulas en contrafase para amplificadores de audio. La carga de ánodo citada para esta aplicación es 12,400 Ω ánodo-ánodo. Por tanto, cada ánodo representa la mitad de este valor, 6200 Ω . Esto es para un suministro de ánodo de 1500 V y una corriente de ánodo promedio de 313 mA para dos válvulas. La corriente anódica media por válvula es, por tanto, la mitad de esta cifra.

Como comprobación para ver si estos números se acumulan, podemos hacer una comprobación rápida. El suministro de ánodo de 1,5 kV, cuando se multiplica por 313 mA, da una potencia de entrada de 470 W y la potencia de salida cotizada es de 340 W. Por lo tanto, la eficiencia es (340/470) = 72%, lo que para un amplificador de audio de Clase B es razonable.

Entonces sabemos que los valores son correctos y ahora podemos continuar.

Poner en paralelo tres válvulas, cada una con una carga de ánodo de 6200 Ω, da una carga de ánodo equivalente de 2066 Ω , que podemos redondear a 2k Ω . La corriente del ánodo será (313/2) x 3 = 470mA. La corriente inactiva de cada válvula se expresa como 32 mA. Esto, sin embargo, necesita una polarización de red de -4,5 V a este voltaje de ánodo más alto. (Si usamos un suministro de ánodo más bajo de 1250 V, la polarización cero normal cubriría la configuración de corriente inactiva). Para convertir el voltaje negativo de la red en el voltaje del cátodo, simplemente invertimos el signo, por lo que debemos levantar el cátodo por encima del suelo en + 4.5V para obtener la corriente inactiva correcta. Por lo tanto, la corriente inactiva con tres válvulas será de aproximadamente 100 mA.

La Q cargada que elegiremos es 12, ya que es un buen compromiso.

Ahora necesitamos conocer la capacitancia de salida del ánodo y la hoja de datos da un valor de 5.6pF para una sola válvula. Teniendo en cuenta cierta capacidad parásita, asumiremos que cada válvula se ve como una capacitancia de derivación de 7pF a la red de ánodo. Entonces, con tres válvulas, el valor es 21pF.

El valor requerido del condensador de ajuste del ánodo está determinado por la Q cargada y el valor de la carga del ánodo. Esto nos da la reactancia de la capacitancia de ajuste del ánodo CT. En la fórmula estándar para la red pi, esto normalmente se denota C1. Las fórmulas estándar se muestran a continuación cuando R A es mucho mayor que R L .

XC 1 = Ra / Q L

Nota: Solo C1 determina la Q cargada y ningún otro componente tiene efecto. Esto da un valor de 172 Ω a la frecuencia de operación.

Ahora convertimos la reactancia al valor real del capacitor requerido a la frecuencia de operación. Diseñaremos para la banda de 80 m al principio, verá la sabiduría en esto más adelante.

A 3.5MHz requerimos un capacitor de sintonización de ánodo de 272pF. La capacitancia del ánodo ya es de 21pF, por lo que necesitamos un capacitor de sintonización de (272-21) = 251pF. Idealmente, este valor debería estar con el condensador aproximadamente a media malla. Entonces, en realidad, necesitamos un valor total de aproximadamente 500pF. Este es un condensador grande, por lo que elegimos usar algo más pequeño y compensar la diferencia con un condensador fijo.

Suponga que tenemos un condensador de ajuste máximo de 150pF con placas espaciadas anchas que manejarían de manera segura el cambio de voltaje del ánodo previsto. Podemos estimar que la oscilación máxima del voltaje del ánodo será aproximadamente el doble del suministro de CC, por lo que un mínimo de 3 kV es lo más bajo que podemos usar y un tipo de 5 kV sería una mejor opción. El condensador de sintonización del ánodo es, por tanto, un 150pF ajustado a aproximadamente la mitad de la malla y el resto es un condensador fijo cableado en paralelo. La mitad de 150pF es 75pF y el capacitor fijo debe ser de 175pF. El condensador estándar más cercano es 180pF y esto será suficiente.

Ahora calculamos el valor del condensador de carga C2 usando la fórmula para que coincida con la carga de salida de 50 Ω . Esto da un valor de 1474pF. El voltaje de RF en la salida es mucho menor que el voltaje del ánodo, ya que la transformación de 2k Ω a 50 Ω es una relación de 40: 1. Esto podría usar un tipo de transmisión de banda múltiple con aproximadamente 365pF por sección y una clasificación de voltaje de 350 VCA. (Si el voltaje de RF es de 150 V RMS, la potencia en 50 Ω es de 450 W.)

Un tipo de tres secciones sería un total de ~ 1100pF. Pero el valor requerido es el valor de carga completa y esto debería ocurrir con el capacitor casi completamente desacoplado. Entonces, el capacitor fijo adicional requerido es mayor. Estimamos que cuando se carga correctamente, el condensador variable está desacoplado 3/4 y tendrá un valor de aproximadamente 275pF. Por lo tanto, el capacitor fijo requerido es (1474 - 275) = 1199pF y un capacitor fijo de 1200pF será suficiente.

El componente final a calcular es la inductancia en serie y esto requiere un valor de 8.25 µ H.

Agregar bandas adicionales

Si también queremos atender a 40 m, una forma sencilla es dividir a la mitad todos los valores recién calculados para encontrar los valores correctos para 7MHz.

El valor de C1 requerido para 7MHz es 136pF en total. Restamos la capacitancia del ánodo de 21pF de las válvulas para obtener el valor del capacitor de sintonización de 115pF y nuestro máximo de 150pF encajará bien sin necesidad de ningún capacitor fijo adicional. El capacitor de carga necesario es (1474/2) = 737pF total y como el capacitor de tres gangs es 1100pF cuando está completamente mallado, solo estaría 67% mallado en esta configuración. Entonces necesitamos un capacitor fijo para acercar el capacitor de tres bandas a 3/4 sin engranar. Un valor fijo de aproximadamente 470pF funcionaría bien.

El inductor es simplemente (8.25 / 2) = 4.125 µ H.

Las otras bandas que están relacionadas armónicamente como 14, 21 y 28MHz se calculan de la misma manera dividiendo los valores por el factor correcto.

Las dos bandas más altas son donde podemos tener un problema ya que la capacitancia del ánodo es un valor fijo de 21pF y no podemos cambiar eso.

A 28MHz con un Q cargado de 12, la capacitancia total del ánodo debe ser de 34pF y, por lo tanto, el capacitor de sintonización solo debe ser de 13pF para hacer resonar el circuito. Con el condensador total de 150pF elegido, el condensador está casi totalmente desacoplado, por lo que es poco probable que funcione. Esto se debe a que su mínimo no es 0pF, sino que suele ser del 5 al 10% del valor máximo. Si el mínimo es 10%, no se puede reducir por debajo de 15pF en valor con las placas completamente desenredadas.

Hay dos soluciones para este problema. Una es conectar un condensador en serie de valor adecuado en el circuito entre el ánodo y el condensador de sintonización o aumentar el Q cargado. Cuanto mayor sea el Q cargado, menor debe ser la reactancia de C1. Una reactancia baja es un valor más alto de condensador.

Si la Q cargada se eleva a 16, el valor requerido para C1 aumenta a 45pF en total y cuando se resta la capacitancia del ánodo, deja un valor de 24pF, que es más práctico, pero aún es un valor demasiado bajo. Si la Q cargada aumenta a 18, C1 se convierte en 51pF en total y C1 debe ser (51 - 21) = 30pF y para una Q cargada de 20 es 56pF en total y un valor de capacitor de 35pF.

Este es un problema común cuando la capacitancia del ánodo es alta y la mayoría de los amplificadores de HF no usan un valor Q cargado constante para todas las bandas por esta razón. A medida que aumenta la frecuencia, la Q cargada normalmente debe aumentarse para mantener el condensador de sintonización del ánodo dentro de los valores prácticos.

En un caso como este, no podemos seguir empujando hacia arriba la Q cargada porque la pérdida de la red del ánodo aumenta con el aumento de la Q cargada.Una solución práctica sería usar un capacitor de sintonización de ánodo de menor valor y agregar capacitores de relleno de valor fijo para las bandas de frecuencia más bajas para llevar el condensador de valor más bajo a un valor medio. Alternativamente, podríamos hacer que el condensador C1 sea de un tipo de banda dual y cambiar la mitad por las bandas más altas. Ninguno de estos es ideal, ya que se requiere una conmutación adicional.

Sin embargo, una Q de alta carga no significa que la pérdida será muy alta y reducirá la potencia de salida en una cantidad considerable. A la frecuencia más alta, es más sencillo fabricar inductores de Q alto y la pérdida se define predominantemente por el inductor Q descargado.Una bobina de tanque de 80 m con muchas vueltas de alambre de diámetro pequeño tendrá un Q descargado mucho más bajo que un inductor de 10 m enrollado de calibre sustancial alambre o tubo de cobre. Este inductor es un pequeño número de vueltas y se pueden alcanzar valores Q descargados de 450 o más. La diferencia entre un Q descargado de 450 y uno de solo 100 es de aproximadamente un 5% en la pérdida de potencia en un amplificador práctico. Esto es 0.25dB y no es nada para emocionarse. La Q de un condensador de sintonización de ánodo de buena calidad es al menos 3000 y el condensador de carga suele ser aproximadamente la mitad de esta cifra. El inductor de la serie del tanque pi, si está enrollado correctamente,contribuye tan solo a una pérdida del 2 al 5%.

Estrangulador de alimentación de CC del ánodo

Hasta ahora no se ha hecho mención del inductor de RF del ánodo que suministra la corriente del ánodo. El elemento que normalmente aporta la mayor pérdida en una red pi es el estrangulador de alimentación de CC del ánodo, que representa aproximadamente el 30% de las pérdidas totales. Esto se debe a que es muy difícil hacer un estrangulador multibanda, enrollado con alambre de pequeño diámetro, con un buen Q.

Normalmente, al calcular los componentes de la red pi, se ignora el efecto del estrangulador del ánodo. Sin embargo, su inclusión en la red puede tener un efecto dramático en el funcionamiento de la red.

La regla general, cuando es necesario conectar un estrangulador en paralelo en un circuito, es evaluar cuánto perturba el funcionamiento de la red. Si la reactancia del estrangulador es infinita, tiene efecto cero. Pero en un circuito práctico no podemos hacer un estrangulador con una reactancia infinita. La suposición hecha es que si la reactancia del estrangulador es 10 veces mayor que la porción de la red que está conectada a la perturbación que produce, es insignificante y generalmente puede ignorarse. En nuestro ejemplo, está conectado a un punto de la red que en resonancia se comporta como una resistencia de 2k Ω . Por lo tanto, la reactancia del estrangulador debe ser de al menos 20 k Ω . A 3,5 MHz, este es un valor de 909 µ.H. Este valor no es práctico y sería muy difícil enrollar tal estrangulador y garantizar que no haya frecuencias resonantes espurias.

Dos estranguladores de ánodo típicos; izquierda: para un amplificador de HF multibanda, derecha: para un diseño de banda única.

La forma de evitar este problema es usar la capacitancia de derivación entre el ánodo y la tierra para hacer resonar un inductor de menor valor a la frecuencia de operación. Si el inductor de valor correcto está conectado a través del ánodo a tierra, se comportaría como una resistencia de valor muy alto debido a su resistencia dinámica (RD). Eimac cubre este aspecto en un boletín de ingeniería. En la figura 8.2 se muestra un gráfico que muestra la frecuencia de resonancia óptima .

Fig 8.2: Frecuencia óptima de resonancia del estrangulador de RF (de 'Cuidado y alimentación de los tubos de red eléctrica', Eimac).

Obtener la banda correcta de resonancia paralela puede estar plagado de problemas. Esto a menudo hace que se queme un estrangulador de ánodo aparentemente bien diseñado. La tensión en el estrangulador del ánodo es considerable. No solo tiene que soportar una carga de CC bastante alta para alimentar el ánodo, sino que, en la resonancia del circuito del tanque, el voltaje de RF impreso a través del estrangulador se acerca al doble del voltaje de suministro del ánodo. Si la reactancia o RD no es lo suficientemente alta, entonces una gran señal de CA puede fluir a tierra en el estrangulador. Está desacoplado en el extremo de suministro, por lo que tiene una conexión a tierra eficaz.

En nuestro amplificador de ejemplo que utiliza un suministro de ánodo de 1,5 kV, la oscilación del voltaje pico a pico del ánodo será de aproximadamente 2,5 kV a la salida de RF máxima. Si la reactancia de estrangulación o el R D es 10k Ω , el pico de corriente de RF que fluye a tierra es de 250 mA. Esto es poder efectivamente perdido . Si desajustar el capacitor del ánodo una pequeña cantidad hace que el estrangulador R D caiga a 5k Ω, la corriente de RF se duplica. Esta corriente adicional, más la corriente CC más alta del ánodo que se producirá cuando el ánodo se desajuste, puede ejercer demasiada tensión en el estrangulador y se opondrá al calentarse.

Un mejor método de estrangulamiento de alimentación de ánodo

La alimentación en derivación del suministro de ánodo siempre ha sido un problema difícil y debe haber una mejor manera de abordar esta parte de la red de adaptación del ánodo. Después de pensarlo mucho, se desarrolló un nuevo método.

El problema real es que las frecuencias de resonancia espúreas son muy difíciles de controlar o predecir si el valor del inductor es grande. Pero si el inductor no necesita tener un valor tan grande, existe una mejor posibilidad de mantener a raya a estos gremlins.

La solución es realmente muy sencilla. Si la red de tanques de pi se reorganiza ligeramente, es fácil incorporar una solución viable. El primer paso se muestra en la figura 8.3 .

En este método, el condensador de bloqueo del ánodo se mueve desde el ánodo al conector de salida. El estrangulador ahora alimenta la CC de alto voltaje al circuito del tanque en el punto de 50 Ω . En consecuencia, solo necesita tener una reactancia de ~ 500 Ω a la frecuencia de operación más baja. Esto da una respuesta de 22,7 µ H y un estrangulador mayor que esto sería suficiente. Como el voltaje de RF es mucho menor que en el ánodo, enrollar un estrangulador adecuado es mucho más simple. La única desventaja de este método es que los condensadores de sintonización y carga ahora tienen que flotar en el nivel de suministro de CC del ánodo. Estranguladores antiparasitarios

Este tema ya se ha abordado en una parte inicial del libro. Hay muchos mitos y leyendas urbanas sobre este tema. Mucho de esto se puede atribuir a una persona que parece tener una fijación con el tema. No mencionaré el nombre ni el indicativo, ¡pero los que han seguido esta saga saben a quién me refiero!

Las afirmaciones sobre los APC de que deben estar enrolladas con alambre de resistencia o algún otro material con pérdidas, como el acero inoxidable, son completamente incorrectas. El APC se inserta para formar una trampa en paralelo a la frecuencia ofensiva de modo que se convierta en resonante en paralelo con la capacitancia de la válvula o del circuito. La mayoría de las oscilaciones parásitas ocurren en la región de VHF, por lo que el valor del inductor debe adaptarse a este rango operativo. En la mayoría de los casos, se utilizan algunas vueltas de alambre de cobre formadas en una resistencia de composición de carbono de 2 W y conectadas a los cables de salida de la resistencia. El valor de la resistencia es muy bajo, típicamente alrededor de 22 a 56 Ω parece funcionar en el 99% de los casos.

El inductor tendrá un Q razonable y, por lo tanto, tendrá un rango estrecho en el que es efectivo con la capacitancia del ánodo. El propósito de la resistencia de bajo valor es aplanar la respuesta de frecuencia para que cubra un ancho de banda mucho mayor. También sirve como carga ficticia para cualquier poder parásito.

El efecto que tiene el APC en el circuito normal del tanque del ánodo es muy pequeño y como el ánodo y los capacitores de carga tienen un amplio rango, pueden compensar fácilmente la adición de este pequeño inductor adicional en serie con el terminal del ánodo.

El requisito de usar solo un cable con pérdidas es una tontería.

El alambre de cobre cuando se forma en un inductor y resuena por la capacidad parásita tendrá un Q de quizás 100 aproximadamente. Exhibirá una resistencia dinámica de aproximadamente 5k Ω o posiblemente más. Pero la conexión de la resistencia de bajo valor a través del inductor destruye completamente cualquier Q que tuviera. La Q cae a 1 o incluso por debajo, que es exactamente lo que necesitamos para formar un APC efectivo. La persona que persiste con este argumento afirma que aunque usa alambre de nicromo o una tira de acero inoxidable para enrollar el inductor, todavía necesita la resistencia. Esto muestra que el cable de resistencia no está siendo tan efectivo como ellos creen. Si el cable de resistencia estuviera haciendo lo que dicen que hace, la resistencia no sería necesaria.

En muchos casos de oscilación parásita, la inclusión de una resistencia de composición de carbono de 1/2 W en serie con un pin de rejilla en la válvula resuelve el problema. No tiene inductancia significativa pero detiene la oscilación. En la mayoría de los casos, no es necesario que la resistencia sea de un valor muy alto, de 10 Ω a 47 Ω suele ser más que suficiente.



9: Amplificadores lineales de cátodo con conexión a tierra

El tipo más común de amplificador de RF, antes de que el amplificador de rejilla con conexión a tierra
se pusiera de moda, era la etapa de cátodo con conexión a tierra . Esto también se conoce comúnmente como amplificador impulsado por rejilla . Este tipo de amplificador generalmente se reserva para válvulas de tetrodo y pentodo, ya que tienen una capacitancia de retroalimentación de rejilla de ánodo mucho menor. Por lo tanto, a veces se pueden usar sin neutralizar. Una ventaja del amplificador accionado por red es la baja potencia necesaria para impulsarlos. A menudo, tan solo 1 vatio es suficiente para impulsar un par de válvulas a una potencia máxima de aproximadamente 100 W PEP. Por lo tanto, este tipo de amplificador tiene una alta ganancia y se adapta bien a la clase de transmisor SSB de 100W. Muchos transmisores de ondas decamétricas utilizaban válvulas de televisión en color comúnmente disponibles, ya que estaban disponibles gratuitamente y eran baratas.

Una válvula popular fue la estadounidense 6JS6C y muchos fabricantes japoneses de válvulas, así como empresas estadounidenses, las fabricaron en grandes cantidades. Estos ganaron el nombre de tubos de barrido ya que en el servicio de televisión en color se usaban para desviar o "barrer" el punto de luz a través de la pantalla de fósforo. Sin embargo, no están diseñadas como válvulas verdaderamente lineales y tienen ciertas limitaciones.

Para resolver algunos de estos problemas, RCA intervino e hizo algunos trabajos en sus tubos de barrido y trajo una nueva válvula al mercado, la 6146. Esto fue a principios de 1952. En su forma original, era un tubo de barrido glorificado con algunas características alteradas. para adaptarse al servicio de amplificador de RF. También está el 6159, que es un verdadero tubo de barrido y modificado de un diseño anterior. El 6146 en el anuncio original no especificaba la operación usando SSB; en su lugar, estaba dirigido al mercado de radio móvil VHF que utiliza AM y FM.

Solo más tarde, RCA comenzó a especificar las necesidades operativas de SSB, ya que el nuevo modo ganó tracción entre los usuarios. La válvula pasó por numerosas mejoras, cada una con un pequeño ajuste a su construcción interna, pero solo más tarde sacó a la luz la primera revisión importante, la 6146A, a mediados de 1963. Indican en la hoja de datos del 6146A que “El 6146A es intercambiable unilateralmente con el 6146”.

A los militares les gustaban las nuevas válvulas, pero necesitaban una versión más resistente para resistir fuertes golpes y vibraciones. Válvulas de mayor potencia

Cuando se necesita más potencia, hay varias válvulas disponibles que pueden satisfacer el requisito. Por supuesto, la ganancia probablemente será un poco menor, pero la potencia de accionamiento necesaria no es excesiva. Muchos tetrodos con una disipación de ánodo de 200 W a 400 W todavía están disponibles y ofrecen mucha más potencia que un par de 6146.

El Eimac 4-250 y el 4-400 todavía están disponibles si busca lo suficiente y son verdaderos caballos de batalla para grandes amplificadores lineales de HF. También tienen diferentes números de pieza de otros fabricantes. Estas son válvulas de envolvente de vidrio tradicionales y bastante grandes. Una alternativa son las versiones compactas de metal-vidrio o metal-cerámica enfriadas por aire forzado, como el Eimac 4X150 y las series posteriores de 4X250, 4CX250 y 4CX350. Estos funcionarán hasta 432MHz, excepto el 4CX350 que realmente no funciona bien incluso en 144MHz, ya que cae por encima de 120MHz. Para los amplificadores de HF, un solo 4CX350 proporcionará mucha potencia y solo necesita un olfato de unidad para que funcionen.

Una de las válvulas más antiguas es la RCA 813. Esta válvula se introdujo por primera vez en 1938, justo a tiempo para el estallido de la guerra, y durante la guerra se fabricaron en grandes cantidades, principalmente por Westinghouse en Canadá. Después de la guerra, se convirtieron rápidamente en la válvula elegida para los transmisores de AM que funcionaban con el límite de entrada de CC de 1 kW en los EE. UU. Las ventas de excedentes de guerra vieron grandes cantidades vertidas en el mercado abierto a precios de ganga. El ánodo es una construcción de grafito y puede soportar una buena cantidad de abuso, pero la disipación oficial del ánodo es de solo 100W en las clasificaciones CCS y 125W en la clasificación ICAS. Como tales, son de menor potencia que el 572B posterior, pero al ser una válvula de alta ganancia se requiere muy poco impulso para que funcionen. A 30MHz, tan solo 5W era suficiente para un transmisor Clase C AM / CW.

Como la mayoría de las válvulas de transmisión de alta potencia, son válvulas de alta tensión y baja corriente. Muchos aficionados en el Reino Unido los usaron, pero debido a la limitación de la licencia de solo 150 W de entrada de CC para AM y CW, usaron un voltaje de ánodo anormalmente bajo. Cuando la operación SSB se hizo popular, el uso del suministro de ánodo inferior pareció persistir. El 813 realmente solo comienza a despertarse por encima de 1500V y 2.5kV es aún mejor. En la práctica, todavía no los destellarán con 5kV o incluso un poco más. Este era Richard (Dick) Thornley, G2DAF, cuyo indicativo de llamada se ha hecho famoso (o infame, según su punto de vista) por su nuevo método radical de suministrar la cuadrícula de la pantalla. Amplificador lineal G2DAF

El diseño G2DAF no requiere un suministro de polarización de red o un suministro de red de pantalla normal, pero proporciona una señal muy lineal, si se construye y configura correctamente. Muchas
de las modificaciones hechas por otros al diseño original solo sirvieron para convertirlo en un amplificador de 'caja de salpicaduras'. Esto se debió a un malentendido del concepto de diseño.

El concepto es utilizar un tetrodo normal en una condición de polarización cero con la señal de entrada alimentada a la red normal. La sutil diferencia es que el voltaje de la red de la pantalla no se suministra con un voltaje de CC constante. Se alimenta con una señal DC de media onda sinusoidal obtenida rectificando la señal de entrada. En consecuencia, cuando no se aplica ningún impulso, la rejilla de la pantalla y la rejilla de control están a cero voltios y la válvula está prácticamente cortada. Cuando se aplica una señal de entrada, el rectificador de rejilla de pantalla suministra un voltaje de CC que sigue con precisión el voltaje de la rejilla de entrada. Esto significa que cuando la válvula se acciona a la entrada máxima de la red, el voltaje de la red de la pantalla se eleva a un pico. Es un tipo de método de modulación de eficiencia que ahora se usa ampliamente en la industria de RF para proporcionar una amplificación muy lineal.

El amplificador G2DAF es bastante diferente a un amplificador normal que usa un tetrodo de haz. En un lineal normal, para garantizar una buena linealidad, la corriente del ánodo inactivo debe ser bastante alta; esto disipa mucho poder. El G2DAF se siente más como un tipo de amplificador de Clase C y, la primera vez que lo experimentas, lleva un tiempo acostumbrarse al comportamiento del medidor de corriente del ánodo. Debido a que la válvula no disipa energía cuando no se acciona, la temperatura del ánodo es mucho más baja y se puede hablar mucho más alto en los picos que un amplificador normal. La eficiencia es alta y la potencia de salida también es mayor que la de un amplificador normal que utiliza la misma válvula. De hecho, un solo 813 que funcione con un suministro de ánodo de 2,5 kV desarrollará fácilmente más de 400 W de PEP con buena linealidad en HF en un buen circuito.

Sin embargo, el sistema de rectificación de la pantalla y el condensador del depósito son elementos críticos para que el amplificador funcione correctamente. La mayoría de las modificaciones utilizan un condensador de depósito que es demasiado grande. Esto ralentiza el aumento de voltaje de la pantalla y también lo mantiene durante demasiado tiempo después de que el pico ha disminuido.

G2DAF no usó originalmente la 813, usó otras válvulas para probar el concepto y solo más tarde la 813 se convirtió en la válvula de elección porque eran muy económicas y estaban disponibles en grandes volúmenes. Más tarde, G2DAF proporcionó algunos datos experimentales sobre un amplificador que usa válvulas 4X150 y estos también funcionan muy bien. Prácticamente cualquier tetrodo de haz adecuado funcionará, aunque algunos son mejores que otros, y darán una gran potencia con buena linealidad, pero solo si sigue el circuito como lo describe G2DAF.

El rectificador de suministro de pantalla puede ser una fuente de problemas si se selecciona un diodo desfavorable. El mejor tipo es un rectificador de válvula. Aunque hoy tenemos muchos rectificadores de semiconductores capaces de manejar los altos voltajes requeridos, no todos funcionan tan bien como los tipos de válvulas anticuadas.

Otros métodos de modulación de la eficiencia

El amplificador G2DAF animó a otros experimentadores a realizar más investigaciones sobre este tipo de amplificador. Se experimentó alguna dificultad para obtener la relación correcta entre el accionamiento de entrada y el voltaje de la red de pantalla para adaptarse a diferentes válvulas. Una solución fue utilizar un método ligeramente diferente. Este utilizó una válvula reguladora en serie para alimentar la rejilla de la pantalla desde una fuente de CC, como en el diseño original de G2MA. Esta fue una modificación del sistema de válvula de compuerta o abrazadera en serie utilizado para amplificadores de Clase C para CW.

En este método, la unidad de entrada se rectifica y se utiliza para proporcionar un voltaje de CC negativo. Se utiliza una válvula de abrazadera que utiliza un tetrodo de viga, a menudo atada como triodo de alta potencia, para mantener la rejilla de la pantalla a un voltaje bajo a través de una resistencia de cuentagotas alimentada desde el suministro de ánodo normal o un riel de suministro inferior. En condiciones de activación, el voltaje de la pantalla se fija a un nivel bajo, de modo que la válvula de salida consume muy poca corriente de ánodo. Cuando la llave está hacia abajo y la portadora completa se aplica a la red, el voltaje rectificado negativo ahora desvía la válvula de sujeción y el voltaje de la pantalla se eleva al voltaje normal. 10: El amplificador de red con conexión a tierra

El uso del amplificador de rejilla conectado a tierra con válvulas de triodo ( Fig. 10.1 ) es una topología popular
para muchas frecuencias de funcionamiento diferentes. Una válvula de triodo cuando se configura como un amplificador de cátodo conectado a tierra adolece de un gran inconveniente. La gran capacitancia inherente (Cag) que existe entre el ánodo y la rejilla provoca una retroalimentación de señal no deseada. Si es de la fase correcta esto hará que la válvula oscile. Para evitar esto, la conexión a tierra de la red y el uso del cátodo como terminal de entrada resuelve el problema en el 90% de los casos. Siempre hay algún otro mecanismo que puede causar retroalimentación, por lo que generalmente la construcción física del amplificador está diseñada de tal manera que se minimiza.

Ganancia potencial

La válvula accionada por cátodo tiene una ganancia potencial mucho menor que un amplificador accionado por rejilla de cátodo conectado a tierra. Esto se debe a que las corrientes de entrada y salida están en fase, mientras que en el amplificador accionado por red están desfasadas 180º. La corriente del cátodo que fluye a tierra a través de la impedancia de la red de entrada resiste la señal de excitación de entrada. El resultado es que la impedancia vista por la fuente de excitación es ahora muy baja. La corriente en fase forma un tipo de señal de retroalimentación negativa y esto, en cierta medida, mejora la linealidad.

En el amplificador accionado por red, el controlador se presenta con una impedancia relativamente alta y, por lo tanto, puede desarrollar una gran oscilación de voltaje de la red-cátodo que requiere poca potencia de accionamiento. El amplificador accionado por cátodo requiere una potencia de excitación mucho mayor para obtener la misma oscilación de voltaje del cátodo de rejilla. El criterio de ganancia es la potencia de salida de RF frente a la potencia de entrada de RF, por lo que la ganancia es menor cuando se acciona por cátodo, ¡a menudo mucho menor!

La transconductancia, sin embargo, es la misma si la válvula está accionada por rejilla o cátodo. Cuando se requiere una alta ganancia, el amplificador impulsado por rejilla es la mejor opción. Esto a menudo necesita componentes de neutralización externos para contrarrestar la señal de retroalimentación interna causada por la capacitancia del ánodo a la red. Muy a menudo, la ganancia más baja puede ser útil.

Un amplificador de rejilla conectado a tierra es intrínsecamente más estable, y el cátodo como terminal de entrada está más cerca de 50 Ω que un amplificador de rejilla. Para los amplificadores lineales de HF, el controlador es normalmente un transceptor que tiene una potencia de salida de alrededor de 100W PEP. Si se eligiera el amplificador accionado por red, la potencia de excitación necesaria sería una pequeña fracción de esta cantidad. Teóricamente, un amplificador de triodo de cátodo conectado a tierra (impulsado por red), cuando no está conectado a la corriente de la red, no requiere potencia de accionamiento; solo tiene que desarrollar la oscilación de voltaje en alta impedancia.

Por lo tanto, un amplificador de red con conexión a tierra se adapta mejor a los transceptores PEP de 100 W, ya que la potencia requerida para impulsarlo está más cerca de los 100 W PEP.

Sesgo

En la topología de red con conexión a tierra, el cátodo vuelve a la tierra a través de una red resonante de entrada que puede conectarse a tierra por RF en el extremo inferior. La red está a 0 V en relación con la fuente de alimentación y el voltaje de polarización, si es necesario, significa que el cátodo debe elevarse por encima del suelo con el voltaje requerido para obtener el punto de polarización correcto. Si el triodo requiere, por ejemplo, un cátodo de red de –20 V para establecer la condición de funcionamiento correcta, significa que el cátodo debe elevarse por encima del suelo para que ahora esté a + 20 V. En otras palabras, es lo mismo que un amplificador polarizado en cátodo con una resistencia en el cátodo a tierra.

En amplificadores Clase A de señal pequeña que consumen una corriente relativamente constante, este método puede funcionar, pero para un triodo de potencia donde la corriente inactiva es bastante baja pero la corriente máxima del ánodo puede ser alta, un esquema de polarización de resistencia simple no funcionará. Si la polarización es de + 20 V para obtener la corriente inactiva requerida de, digamos, 50 mA, pero la corriente del ánodo alcanza un máximo de 400 mA, la caída de voltaje a través de la resistencia será ocho veces mayor cuando la corriente del ánodo se impulse por completo. Si el punto inactivo es + 20 V, la corriente del ánodo completo generará un voltaje de polarización de + 160 V. Es lo mismo que el cátodo de rejilla de –160V y cortará totalmente la conducción. Por lo tanto, se debe utilizar un método diferente y esto se muestra en la figura 10.2 .

Fig 10.2: Polarización de la resistencia de cátodo.

La corriente del cátodo fluye a tierra a través de RB y VR1. La señal de RF se aísla de los componentes de polarización mediante un estrangulador de RF y se desacopla a tierra mediante un condensador de baja reactancia. El ajuste de VR1 establece el voltaje del cátodo en reposo y, por lo tanto, el voltaje del cátodo de la red. Para cortar la válvula durante los períodos de recepción, se utilizan dos contactos de relé que son controlados por la línea PTT. En la transmisión, RY2B está cerrado y RY2A está abierto y la corriente del cátodo fluye a tierra a través del medidor M2. Al recibir, los contactos del relé se invierten y el voltaje de corte de + 50V tira del cátodo hasta este valor, por lo tanto, desvía la válvula. La resistencia R1 es un dispositivo de seguridad en caso de que falle el suministro de + 50 V o RY2A; el voltaje del cátodo flotará hasta un voltaje alto y cortará la válvula.

Triodos de sesgo cero

Se han desarrollado varios triodos que eliminan la necesidad de circuitos de polarización. Estos incluyen el 811, 572B y el 3-500Z y son atractivos ya que simplifican el diseño del amplificador. El único problema es que la válvula todavía necesita ser cortada durante la recepción, lo que requiere un alto voltaje de cátodo positivo o un alto voltaje negativo en la red. Si la red se utiliza para alimentar la polarización de corte, debe levantarse de la tierra de CC y desacoplarse con condensadores de baja inductancia. Si el condensador de desacoplamiento tiene una inductancia significativa, puede reintroducir inestabilidad ya que la red ya no está correctamente conectada a tierra. El otro problema potencial es que si se produce una descarga disruptiva, puede acabar con los condensadores de desacoplamiento y dejar la red flotando. Luego oscilará, destruyéndose violentamente.

Aunque los triodos populares de polarización cero son más fáciles de usar para polarizar, generalmente tienen cátodos calentados directamente. Esto significa que la corriente del calentador y la señal de entrada de RF deben manejarse correctamente (en un cátodo calentado indirectamente, el diseño es más fácil ya que ocurre poca o ninguna interacción). Esto significa que la corriente del filamento debe introducirse en el cátodo mediante reactancias de RF de alta corriente, que son una fuente de problemas. También significa que los dos extremos del calentador deben estar conectados sólidamente con un capacitor de baja reactancia para garantizar que ambos tengan el mismo potencial de RF.

Este condensador (y C4) está compuesto por dos condensadores de 10 nF en paralelo para manejar el flujo de alta corriente de RF. La corriente del calentador se alimenta a través de dos bobinas de ferrita de alta corriente RFC1 y RFC2. El devanado secundario del calentador es alimentado por el transformador T1 y el devanado tiene una toma central. El circuito del tanque del ánodo pi es idéntico a cualquier otro tipo de amplificador, por lo que no se ha incluido. Si el amplificador tiene que cubrir varias bandas, es necesario proporcionar redes de entrada para cada banda, que deben seleccionarse mediante una disposición de conmutador de banda.

Red de entrada de cátodos

En algunos amplificadores de rejilla simples con conexión a tierra, no se intenta producir una coincidencia de ROE baja con la fuente de excitación. Esto suele ser una fuente de distorsión indebida. El método preferido es utilizar una red de adaptación de impedancia para transformar de la fuente de excitación al cátodo. Sin embargo, esto no es tan simple como parece. Ha habido mucha confusión y declaraciones erróneas sobre este tema y tal vez sea necesario ver qué dijeron los expertos al respecto.

William (Bill) Orr, W6SAI, era un ingeniero de diseño senior en Eitel-McCullough Valve Company, que también conocemos como Eimac. W6SAI fue el editor de un popular boletín regular llamado Amateur Service ( AS ) publicado por Eimac para aficionados. En el boletín AS-3 se dan algunos detalles pertinentes sobre los amplificadores de rejilla conectados a tierra. También se publicó una versión ampliada en QST en agosto de 1961 con los coautores Ray Rinaudo, W6KEV y Robert Sutherland, W6UOV, también ingenieros de Eimac. Esto también aparece en el libro ARRL Single Side Band for Radio Amateurs.. En esta era no había transmisores de estado sólido, por lo que la información refleja puramente la de los transmisores / excitadores de válvulas. Como veremos más adelante, los transmisores de estado sólido son un problema importante para este tipo de amplificador.

Bill Orr dice lo siguiente acerca de los amplificadores de rejilla conectados a tierra: “ Cualquier etapa de Clase B de un solo extremo (independientemente del tubo utilizado) extrae corriente de la rejilla y la placa sólo en una parte del ciclo de funcionamiento (aproximadamente 180º). La impedancia de entrada de dicha etapa, por lo tanto, no representa una carga constante. La forma de onda entregada por el excitador a la etapa de rejilla conectada a tierra está muy distorsionada en la parte del ciclo en la que el amplificador extrae corriente de rejilla y placa. Aunque los valores de 'impedancia de entrada' publicados pueden parecer atractivos, en realidad representan solo el componente fundamental de la impedancia de entrada (útil para los cálculos del circuito del tanque 'Q'). Dado que la impedancia de carga de entrada de la etapa de red con conexión a tierra de Clase B no es un valor constante, es necesario transformarla a una impedancia constante que se asemejará a 50 Ωdurante todo el ciclo operativo. Esto se hace mejor con un circuito sintonizado de alta C colocado directamente en el cátodo de la etapa de rejilla conectada a tierra ".

La implicación de esta afirmación es que no se considera una buena práctica simplemente conectar la salida de una etapa de activación de 50 Ω al cátodo a través de un condensador de bloqueo de CC simple. El resultado será una gran distorsión de la señal de entrada y esto simplemente se amplificará para convertirse en una señal distorsionada aún más grande en la salida. La otra cosa, que quizás no sea obvia a partir de la declaración de Bill Orr, es que se estaba refiriendo a un nivel de potencia constante en el amplificador y observando la distorsión de la forma de onda de RF sinusoidal en ese nivel fijo de excitación. Y ahí es donde han surgido la mayoría de los hechos erróneos. No es posible diseñar una red de coincidencia simple que siempre coincidirá con los 50 Ωimpedancia del excitador cuando la impedancia del cátodo varía en un rango tan amplio. Por alguna razón, se convirtió en una leyenda urbana que una simple red puede solucionar todos los problemas del amplificador de red con conexión a tierra. Habiendo medido la oscilación de impedancia en varias válvulas diferentes a diferentes niveles de accionamiento, puedo decirles ahora que es imposible hacer una red como esta.

    Un ejemplo servirá para ilustrar el problema de la oscilación de la impedancia de la válvula. Hace algunos años construí un lineal de 6 m usando los triodos rusos GI7-BT y necesitaba establecer la impedancia del cátodo en el rango de funcionamiento de trabajo. Estos datos no se incluyeron en la hoja de datos, así que tuve que determinarlos experimentalmente. Se construyó y accionó un amplificador de prueba con una fuente de excitación de muy buena potencia fabricada por Hewlett Packard y se utilizó un acoplador direccional dual para muestrear la amplitud y la fase de la señal de entrada. A partir de esto, pude medir la impedancia a medida que variaba la unidad. Con un impulso de 1 W, el cátodo exhibía una impedancia de más de 600 Ω y la corriente de la red era muy baja. A 5 W, la impedancia era de 150 Ω.y la corriente de la red era el 50% del máximo seguro según la hoja de datos. A 10 W, la corriente de red había alcanzado el valor máximo, con una impedancia inferior a 30 Ω .

El uso de un circuito de tanque resonante en el cátodo proporciona el efecto de volante y esto restaura la forma de onda portadora a una verdadera forma sinusoidal. A partir de las mediciones realizadas en el laboratorio de Eimac en varios tipos de válvulas, demostraron que, independientemente del tipo de válvula que se usara, todas funcionaron mejor con la red de cátodos que sin ella. En general, se logró una mejora en la potencia de salida de aproximadamente un 5% con menos potencia motriz y valores de IMD más bajos. También mostró que la coincidencia con el excitador era menos crítica y no necesitaba longitudes específicas de cable coaxial entre el excitador y el amplificador para obtener una buena coincidencia de SWR baja. Sin embargo, todas las pruebas se realizaron con un controlador de alta potencia Clase A de muy alta calidad, que estaba muy amortiguado con carga resistiva y era en gran parte inmune a los efectos de desajuste. Un transmisor SSB promedio,que no se mencionó en el informe, cuando se sustituyó porque el excitador se notó como inferior a la etapa especial del conductor. Esto fue para equipos de válvulas como excitador o fuente de impulsión.

Controladores de estado sólido y amplificadores de red con conexión a tierra

Si avanzamos unos 50 años, todos los transceptores de HF modernos son de estado sólido y son completamente diferentes en la forma en que manejan cargas que no coinciden. Pueden ser extremadamente sensibles a los desajustes y cargas particularmente reactivas. Los excitadores de válvulas anteriores, con un tanque pi y un par de válvulas 6146, no sentían nada cuando se conectaba una carga deficiente. El operador podía ajustar la sintonización del ánodo y cargar los condensadores para compensar la carga reactiva de la antena, siempre que la ROE no fuera demasiado alta. A menudo, una ROE tan alta como 3: 1 podría igualarse felizmente, y en algunos casos hasta 5: 1. Si la ROE de la antena es más alta que eso, una ATU normal resolvería el problema sin problemas.

Los transmisores de estado sólido de hoy en día se encuentran en una gran desventaja cuando se trata de impulsar amplificadores de rejilla con conexión a tierra, porque el amplificador presenta una carga que varía en un amplio rango a medida que varía el nivel de potencia. Aunque a menudo tienen una ATU incorporada, esto no ayuda.

Si el modo utilizado es AM, hasta cierto punto el problema se alivia ya que la impedancia no varía en una cantidad tan grande, porque el nivel de portadora constante hace que la impedancia del cátodo sea más constante. Pero SSB varía desde cero potencia de accionamiento hasta el valor máximo.

La ATU puede hacer que una antena que no coincida se vea como una buena carga de 50 Ω , por lo que el transceptor está contento. Pero la impedancia de entrada del amplificador de rejilla con conexión a tierra no se comporta como una antena. La ROE de la antena es constante, sin importar el nivel de potencia. La impedancia del amplificador de rejilla con conexión a tierra varía desde casi un circuito abierto a niveles de excitación muy bajos y, a medida que aumenta la excitación, comienza a caer a un nivel más bajo. Cuando la válvula comienza a consumir una gran corriente de red, la impedancia cae a un valor muy bajo. En un nivel de unidad muy específico, se acercará a una coincidencia casi perfecta a 50 Ω , si la red de entrada se ha configurado correctamente, pero solo ocurre en ese nivel de unidad en particular. Como comentó una vez un diseñador de amplificadores sobre el cambio de coincidencia con la unidad,“Es como intentar golpear una paloma de arcilla en vuelo. ¡La maldita cosa siempre se está moviendo! "

Y ese es el problema con los transceptores de estado sólido que tenemos hoy. La ATU puede compensar la condición de desajuste de la antena estática, pero no puede variar su emparejamiento para seguir el nivel de potencia cuando cambia constantemente.

Cura para la variación de impedancia del cátodo

Muchos y variados han sido los intentos de estabilizar la impedancia de entrada del amplificador de rejilla conectado a tierra. Si comparamos la topología del amplificador accionado por rejilla y por cátodo, podemos ver una posible solución al problema.

Un método popular de proporcionar impedancia de entrada constante para la topología impulsada por la red (cátodo a tierra) es prescindir del circuito sintonizado y sustituir una resistencia de carga de baja inductancia. Si se elige esta resistencia para que sea ~ 50 Ω y el excitador se conecta directamente a ella a través de un condensador de bloqueo de CC para que no altere la condición de polarización, el excitador está suministrando efectivamente una carga ficticia. La oscilación del voltaje de la red es puramente una función de la potencia de activación de RF. Esta técnica se denomina comúnmente red de cuadrícula pasiva .

El método de entrada de red pasiva es atractivo ya que elimina los componentes de sintonización necesarios para cada banda y permite el funcionamiento en cualquier frecuencia necesaria. Por supuesto, la red de ánodos no puede tratarse de la misma manera y la cobertura está puramente limitada por la red de ánodos. La desventaja de la red de red pasiva es que necesita una potencia de accionamiento mucho mayor para obtener la oscilación de voltaje de red-cátodo requerida. Sin embargo, con un controlador PEP de 100 W, esto normalmente no es un problema. Necesita la mayor parte de la potencia de salida disponible del excitador para impulsar la válvula a la salida máxima y, como tal, significa que no necesitamos encontrar una forma de reducir mucho la potencia de accionamiento. La resistencia de la red, por supuesto, disipa una alta potencia, por lo que debe elegirse correctamente. La red con conexión a tierra y el amplificador de red pasivo necesitan aproximadamente la misma potencia de accionamiento,por tanto, también tiene una ganancia potencial baja.

    Hace algunos años estuve involucrado en un problema de diseño para un sistema de radio militar con salto de frecuencia. Al saltar sobre una banda ancha de frecuencias, es imposible hacer una antena que presente una ROE baja en todas las frecuencias y aún así irradie de manera eficiente. (Algunas antenas de banda ancha que parecen tener una ROE baja en un ancho de banda considerable se deben a que las resistencias de carga absorben la potencia no coincidente y, como tales, solo irradian una parte de la potencia que reciben, en algunos casos tan solo el 30%). ¡El único artículo que tiene una ROE constantemente baja es una carga ficticia! Intentamos hacer una ATU adaptativa que siempre presentara una ROE razonablemente baja para el transmisor y, de hecho, logramos acercarnos a un sistema viable después de aproximadamente dos años de esfuerzo. Pero la velocidad de salto fue de solo cuatro saltos por segundo.El usuario requería una velocidad de 50 saltos por segundo y eso no era ni de lejos posible.

En el caso del amplificador accionado por red, el nivel de excitación se puede reducir significativamente utilizando un transformador de banda ancha para aumentar la impedancia desde el excitador a la red. Si se utiliza un transformador 4: 1, los 50 Ω del excitador se transforman hasta 200 Ω y la tensión de excitación se duplica. Esta es una buena solución si solo tiene un transmisor de 10 o 20 W, pero desafortunadamente no funcionará para el amplificador de red con conexión a tierra.

Esta solución se utilizó en un amplificador lineal fabricado por Heathkit. En el SB-230 lineal
( Fig. 10.6 ), Heathkit utiliza el triodo de rejilla a tierra refrigerado por conducción Eimac 8873. El circuito de entrada del cátodo es una red pasiva pura con un banco de resistencias en paralelo para proporcionar una carga ficticia de 100 Ω conectada en derivación a través del cátodo a tierra. La impedancia de entrada del 8873 comienza casi como un circuito abierto a niveles de excitación bajos y cae gradualmente a aproximadamente 100 Ω en la clasificación de corriente máxima de la red. Por lo tanto, las dos cargas equivalen a una carga de 50 Ω a plena potencia y una carga de 100 Ωcarga a niveles mínimos de conducción. Tiene una ROE de entrada que es aproximadamente 2: 1 en el peor de los casos y cercana a 1: 1 a plena potencia. Esto se adapta mejor al excitador de estado sólido moderno. La desventaja es que solo necesita alrededor de 50W PEP para impulsar el amplificador a la salida completa de alrededor de 600W. (Eimac da una cifra de 26 W para lograr una salida de 587 W a 30 MHz. Dado que la carga ficticia absorbe aproximadamente la mitad de la potencia de transmisión, se duplica el nivel de transmisión requerido). En consecuencia, es susceptible de sobreexcitación por parte de un operador sin experiencia.

Fig 10.6: Red de entrada de cátodos Heathkit SB-230.

La resistencia de derivación del cátodo tiene una potencia nominal de 30 W y consta de muchas resistencias de composición de carbono de 2 W en paralelo montadas en una placa de tira de etiquetas. La resistencia está hecha de 15 x 1,5k Ω cableados en paralelo. Estos tienen la reputación de tener una potencia nominal demasiado baja y, a menudo, su resistencia aumenta debido al sobrecalentamiento. Muchos aficionados han reemplazado las resistencias de composición de carbono de 2 W originales con resistencias de película metálica de 3 W más apropiadas en un esfuerzo por ganar algo de disipación adicional. El problema básico, sin embargo, es que el operador no está teniendo suficiente cuidado con el nivel de la unidad y, a menudo, no tienen el ALC conectado al transceptor. La válvula 8873 necesita como máximo ~ 40 W para saturarla en las bandas de HF inferiores y el típico transceptor PEP de 100 W simplemente tiene demasiada potencia.

Un mejor método hoy en día sería una resistencia disipable de calor encapsulada de película delgada, como los tipos hechos para resistencias amortiguadoras en fuentes de alimentación conmutadas. Están disponibles en una disipación de hasta 500 W y son de baja inductancia. Un tipo empaquetado TO-220 tiene una disipación de 50 W y 100 Ω es un valor muy común para las resistencias amortiguadoras. Algunas de estas resistencias funcionarán felizmente hasta al menos 1 GHz y exhibirán una inductancia cercana a cero en esa frecuencia.

Una solución integral para el SB-230 sería insertar un atenuador de potencia entre la entrada y el cátodo para disipar el exceso de potencia de excitación y estabilizar la entrada Z. Sería apropiado un atenuador de 3dB / 50W, con el ALC conectado al transceptor para controlar el nivel de conducción absoluto.

Problemas de inestabilidad

Aunque conectar la válvula en una rejilla conectada a tierra reduce radicalmente el valor efectivo del mecanismo de retroalimentación del ánodo a la rejilla, no lo elimina por completo. La capacitancia de salida de una válvula de rejilla conectada a tierra es de hecho el capacitor que se muestra como Cag en la figura 10.1 . Esto también es válido para una válvula de rejilla de pantalla; la capacitancia de salida está entre el ánodo y la rejilla de la pantalla. Por lo tanto, si la conexión a tierra de RF de la red no es 100% efectiva, la gran corriente de salida circulante que fluye en esta interfaz hará que aparezca un voltaje de RF correspondiente a través de la interfaz de red-cátodo. Si la red está conectada a tierra en CC, esto normalmente proporciona una impedancia suficientemente baja para evitar la retroalimentación.

En algunos triodos de UHF y microondas, la ganancia potencial puede ser muy alta a frecuencias más bajas. Aunque el valor efectivo de Cag es bajo, no es cero. Si calculamos la reactancia del valor de Cag a diferentes frecuencias, podemos ver el problema potencial.

    El triodo Eimac 8873 es una válvula bastante delicada (y cara). La disipación máxima de la red es de solo 5 W, la disipación máxima del ánodo es de solo 200 W. El 8874 y el 8875, que son eléctricamente idénticos y solo se diferencian en el método de enfriamiento del ánodo, tienen una disipación de ánodo de 400 W y 300 W respectivamente. Por tanto, el 8873 no es mucho mejor que un solo 572B. Una válvula de reemplazo, si puede encontrar una, ya que no se han fabricado durante años, no será barata. Los precios actuales (2017) oscilan entre $ 400 y $ 600 para las válvulas Eimac NOS genuinas. ¡Manténgase alejado de las válvulas usadas!

Efecto de la capacitancia de retroalimentación de la red del ánodo

Suponga que la válvula funciona a una frecuencia baja, como 80 m. El valor de Cag para un triodo como el 3-500Z cuando se opera en una red con conexión a tierra se da como 0.07pF en la hoja de datos de Eimac. La capacitancia del ánodo de la red es 4.7pF. En el funcionamiento del cátodo conectado a tierra, el valor de Cag es, por tanto, 4,7 pF y la capacitancia de salida de la válvula sola es, por tanto, 0,07 pF. Los dos condensadores intercambian posiciones cuando se cambia la topología. La capacitancia del cátodo de rejilla permanece igual en 8.3pF sin importar qué topología se use.

A 3.5MHz, la reactancia de un capacitor de 0.07pF es muy alta, es de alrededor de 650k Ω , lo cual es bastante insignificante. Sin embargo, no debemos perder de vista el hecho de que la capacitancia interna de la válvula no es la única fuente posible de retroalimentación alrededor de la válvula. La válvula se montará en un enchufe y se encuentra muy cerca de la estructura metálica que la aloja. Existen altas corrientes de tierra en algunas partes del chasis de metal y si esto tiene una alta resistencia al efecto de piel, las grandes corrientes inducen voltajes en otras partes del circuito del amplificador.

Un mecanismo de retroalimentación clásico es a través de la bobina del tanque del ánodo. Muy a menudo, se monta cerca de las paredes verticales del compartimento del ánodo. Entran en juego dos factores. Si la bobina se monta a menos de aproximadamente dos diámetros de bobina de una superficie metálica conductora, fluyen corrientes parásitas considerables en este metal. Esto se comporta como una bobina de giro en cortocircuito y "des-Qs" severamente la bobina, lo que lleva a una mayor pérdida y una menor potencia de salida. Las corrientes parásitas son corrientes grandes que fluyen dentro del metal e inducen voltajes en otras partes del chasis del amplificador por acoplamiento magnético, similar a los devanados primario y secundario del transformador. Si no se previene o controla bien, esto puede ocasionar problemas.

Algunos amplificadores de aficionados están construidos con chasis de acero enchapado, debido a los pellizcos de los fabricantes, y exhiben extraños efectos de bucle de tierra. Un modelo en particular tiene la mala reputación de ser casi imposible de domesticar debido a esta elección de material. Necesita grandes correas de unión de cobre para domesticar completamente el amplificador. Un amplificador correspondiente fabricado por otro fabricante, que utiliza las mismas válvulas, tiene un diseño físico casi idéntico y este amplificador no sufre los mismos problemas. Este amplificador utiliza un chasis de aluminio.

Estos dos factores adicionales pueden aumentar radicalmente la realimentación efectiva de la red del ánodo. Si el terminal de la rejilla de válvulas varía en voltaje debido a la alta corriente de ánodo aplicada que fluye, puede causar inestabilidad. Muy a menudo, las correas de conexión a tierra para los pines de la rejilla introducen cierta inductancia y también son capaces de proporcionar la señal de retroalimentación necesaria para iniciar la oscilación.

Se comprenden bien las reglas básicas para garantizar la oscilación. Deben cumplirse dos criterios principales: primero, la ganancia del dispositivo debe ser mayor que 1 y, segundo, la señal de retroalimentación debe estar en fase con la señal de entrada. Si se cumplen ambos criterios, el dispositivo oscilará a alguna frecuencia, determinada por la inductancia y la capacitancia en el circuito.

Muy a menudo, la válvula no oscilará a la frecuencia de operación, sino a alguna otra frecuencia donde las correas de conexión a tierra y la capacitancia propia y parásita satisfacen un efecto resonante. En la mayoría de los casos, la frecuencia en la que esto ocurre es mucho más alta que la frecuencia para la que está diseñado el amplificador y, a menudo, en la región de VHF.

Oscilación parasitaria

Estas oscilaciones falsas se conocen comúnmente como oscilaciones parásitas y pueden ser un problema grave para algunos tipos de válvulas. Si examinamos las características de una válvula típica, la Eimac 3-500Z, encontramos que Eimac ofrece condiciones de funcionamiento de hasta 110 MHz a una salida casi completa. Todos los dispositivos tienden a perder ganancia a medida que aumenta la frecuencia. En semiconductores tenemos un factor llamado f max , que es la frecuencia donde la ganancia ha caído a 1. Cualquier frecuencia por encima de f maxes incapaz de sostener la oscilación. Las válvulas generalmente no se clasifican de esta manera, pero se puede inferir en la mayoría de los casos, si no se indica en la hoja de datos. Por ejemplo, el triodo ruso GI-7BT en la hoja de datos proporciona clasificaciones de salida de potencia para CW y operación de pulso autoexcitado. Con una longitud de onda de 18,5 cm (1,6 GHz) tiene una potencia nominal de salida mínima de 40 W y una potencia de 10 cm (3 GHz) en servicio pulsado tiene una potencia nominal de 12 kW para una longitud de pulso de 3 µ s. (Por cierto, en la operación de pulso autoexcitado se pulsa con un suministro de ánodo de 9kV y consume una corriente de cátodo de 7.5A, por lo que es una válvula bastante luchadora). Por lo tanto, podemos estimar que todavía tendrá alguna ganancia útil a 6GHz como un amplificador normal y probablemente superior. El f max del triodo 3-500Z es probablemente de unos 500 MHz, a juzgar por las características, por lo que en cualquier lugar hasta esa frecuencia podría experimentar problemas de oscilación parasitaria.

Las oscilaciones parasitarias no son un problema exclusivo de los circuitos de rejilla conectados a tierra. Una de las principales razones para elegir una red con conexión a tierra es porque normalmente no necesita neutralización, pero siempre hay excepciones a la regla. Algunos de los peores amplificadores con efectos parásitos son los tipos de cinchas con conexión a tierra, simplemente porque no hay un ajuste neutralizador para corregir el efecto.

Cualquier amplificador, si las condiciones del circuito lo permiten, puede sufrir este problema. Generalmente, las oscilaciones parásitas se pueden reducir o eliminar por completo mediante el uso de estranguladores antiparasitarios (APC) que se instalan en el ánodo y otros terminales para desalentar los efectos resonantes no deseados. En muchos casos, esto es simplemente un inductor de bajo valor derivado por una resistencia de bajo valor para desactivar el circuito. Se instalan lo más cerca posible del terminal con una longitud de cable cero. Arriba se muestra una imagen de un APC típico para un ánodo 6146. La resistencia de composición de carbono de 2 W normalmente utilizada sufre la alta temperatura a la que está expuesta y, durante un período, aumenta su valor. En aplicaciones de baja potencia, un APC adecuado es una resistencia bobinada de 5 W de aproximadamente 22 Ω a 47 Ω. Esto elimina el devanado de un inductor ya que el cable de resistencia forma el inductor. La resistencia de alambre enrollado es superior a la técnica normal, ya que puede soportar mejor las altas temperaturas y el costo es menor.

El otro factor a tener en cuenta es que aunque el fabricante en la hoja de datos proporciona los valores de capacitancia entre electrodos, estas mediciones se realizan a una frecuencia baja para facilitar la obtención de una medición precisa. Por lo general, se realizan en una válvula de frío en un accesorio blindado especial para minimizar los efectos de la capacidad de dispersión. No se puede suponer que estos valores de capacitancia sean válidos para frecuencias más altas. Los cables internos que sostienen las distintas partes de la válvula se utilizan para transportar las corrientes dentro de la estructura y salir de la base para las conexiones. Cualquier cable, si es relativamente largo y delgado, tiene una inductancia significativa. A alguna alta frecuencia, con la capacitancia entre electrodos, resonará.

Problemas de estrangulamiento de RF

Cuando se diseña un amplificador, es imperativo examinar todos y cada uno de los componentes seleccionados con mucho cuidado para determinar si tiene propiedades resonantes espurias no deseadas. Esto es particularmente importante para cualquier inductancia de RF que se utilice para alimentar las corrientes de suministro. En el clásico amplificador de rejilla con conexión a tierra que usa triodos de polarización cero, el área que a menudo causa problemas son los choques de filamento, ya que estos transportan alta corriente y están conectados en derivación con la señal de entrada. Estos también pueden ser una fuente de pérdida y absorber una buena cantidad de la potencia de accionamiento si se eligen incorrectamente. Si la temperatura del estrangulador excede la temperatura de Curie del material de ferrita, la inductancia cae a un valor muy bajo.

El tipo de estrangulador que sufre más problemas es la variedad de herida bifilar ya que ambos devanados están en una varilla de ferrita común. No debe pasarse por alto que los inductores, cuando se conectan en paralelo, tienen una inductancia neta menor: se comportan como resistencias en paralelo. Por lo tanto, cada estrangulador debe ser el doble de la inductancia requerida cuando está en circuito y en paralelo. Con el estrangulador bifilar también tiene el doble de efecto de calentamiento en comparación con un par de estranguladores individuales porque la misma corriente del calentador fluye en cada devanado. La saturación magnética de la corriente del calentador también es dos veces mayor. Esto puede empujar el estrangulador hacia el acantilado con menos aumento de temperatura.

El otro estrangulador de RF más importante es el estrangulador de alimentación del ánodo en la red de tanques pi normal. Esto es común a todos los amplificadores de este tipo, ya sean de red con conexión a tierra u otras topologías de circuitos. El estrangulador de RF del ánodo tiene que soportar un voltaje de RF muy alto y también transportar una cantidad sustancial de CC para alimentar el ánodo. La inductancia de este estrangulador debe ser lo suficientemente alta para evitar, o al menos reducir a un valor bajo, la corriente CA que fluye a tierra inducida por el alto voltaje de RF a través del ánodo a tierra.

Cuando se acopla en paralelo con el ánodo a la capacitancia de tierra, la inductancia del estrangulador se puede llevar a resonancia en paralelo para que parezca comportarse como una resistencia a tierra de muy alto valor. En un amplificador de banda única, esto es fácil de organizar. Sin embargo, en un amplificador de HF multibanda esto es casi imposible de lograr para cada banda ajustada. En consecuencia, a menudo se encuentra que se produce una gran pérdida en este componente en las bandas de frecuencia más baja y el calor generado causa problemas. En muchos casos, al inspeccionar un amplificador, el cable esmaltado utilizado para enrollar el estrangulador se decolora y, a veces, se quema por completo en algunos lugares. Si el adyacente se vuelve corto, la pérdida aumenta drásticamente y, por lo general, se produce una quemadura, lo que daña irreparablemente el estrangulador.

    Como experimento, se instaló una válvula Eimac 4CX250 en un enchufe SK-600 y se conectó a un analizador de red vectorial para medir el efecto. Se colocó un conector SMA RF en la base y una correa de cobre corta y ancha conectó el pin central del conector RF al terminal de la red. El valor de la hoja de datos de la capacitancia del cátodo de la rejilla es de aproximadamente 16 pF, medido a 1 MHz. La reactancia de un capacitor de 16pF a 144MHz es ~ 68 Ωy el VNA debe indicar este valor. Sin embargo, el valor medido fue radicalmente diferente. El valor efectivo se calculó a partir de los valores R y X mostrados. A 144MHz, la capacitancia efectiva había caído a aproximadamente 8pF, lo que muestra que la inductancia del cable interno estaba teniendo efecto. Cuando se midió a 432 MHz, el cátodo de rejilla ya no se comportaba como un condensador, había pasado por la frecuencia resonante y era inductivo. La medición se repitió con el calentador encendido y se produjo algún cambio, pero en realidad no fue muy diferente. Esto muestra que nunca se puede suponer que los valores de la hoja de datos son constantes con la frecuencia.

¿Puedo utilizar cualquier válvula en una red con conexión a tierra?

Esa es una buena pregunta y nuevamente ha habido mucha controversia sobre este tema. Bill Orr, W6SAI y sus colegas de Eimac publicaron un boletín de ingeniería sobre el tema. En él indicaron que algunos tubos no eran aptos para operar en rejilla puesta a tierra. Estos tubos tienen estructuras de rejilla de alta permeabilidad y debido a su construcción pueden dañarse por la alta corriente de rejilla cuando se operan en circuitos accionados por cátodos. A continuación, enumeraron los distintos tubos que no deben utilizarse en la rejilla con toma de tierra: se incluyen en ellos el 4X150 y sus variantes. La gente se dio cuenta y aceptó esto como el verdadero estado de cosas.

Pero luego se pone interesante. No mucho después, Eimac emitió otro boletín sobre el tema en el que parecían dar un giro completo. Ahora declararon que era aceptable operar algunos de los tubos de alta permeabilidad en una red con conexión a tierra, y mencionaron específicamente el 4X150, siempre que el circuito estuviera organizado de una manera particular. Aparentemente, la solución es conectar a tierra por RF la rejilla de control y la rejilla de la pantalla con condensadores de baja inductancia y alimentar los suministros de CC a estas rejillas. Esto, afirmaron, alivió el problema de la alta corriente de la red. El método anterior consistía en conectar a tierra CC la rejilla de control y aumentar la tensión CC del cátodo para obtener el punto de polarización correcto. Para el 4X150, eso requirió aproximadamente + 50V de cátodo a tierra para operación de Clase B.Para compensar que el cátodo se elevó 50V positivo significaba que la rejilla de la pantalla también necesitaba subir + 50V, para mantener el potencial correcto de la rejilla-cátodo de la pantalla para restaurar la ganancia perdida.

Pero no se detuvo ahí. Uno o dos años más tarde, emitieron otro boletín de ingeniería en el que volvieron a la opinión número uno, en el que declararon que la corriente de la red de cualquier magnitud significativa no es algo bueno y no se recomienda una red conectada a tierra. Quizás tuvieron muchos reclamos de garantía debido a tubos defectuosos, nunca lo sabremos.

¿Confuso?

Al poder reconstruir la historia casi 50 años después, parece claro que Eimac o la nueva empresa que surgió aún no ha cambiado de opinión al respecto. Por lo tanto, según los expertos, los tubos enumerados todavía no se recomiendan para su uso en una red con conexión a tierra.

¡En ese momento el estado de confusión era asombroso! Las peleas casi estallaron sobre el tema y 'Walter' estaba totalmente desconcertado y aturdido: ¿puede o no puede? Nadie pudo darle una respuesta definitiva. Un grupo agitó el boletín número 1, solo para ser derribado por otro grupo que presentó el boletín número 2 como evidencia, y otro grupo había logrado hacerse con una copia del boletín número 3.

Una pelea similar surgió sobre el método de enfriamiento que Dick Knadle, K2RIW, mostró para su amplificador de 70 cm usando el 4CX250B. En lugar de soplar el aire de enfriamiento a través de la toma de flujo de aire del compartimiento de la rejilla, que antes de su amplificador era el método aceptado de enfriamiento, presurizó el compartimiento del ánodo, tiró las chimeneas de cerámica SK-606 y colocó tubos de PTFE en el conducto. el aire de la caja del ánodo.

Inicialmente, Eimac tenía serias reservas sobre esta técnica y le dio un gran pulgar hacia abajo. Sin embargo, luego parecieron haber tenido un cambio de opinión y declararon que después de haber realizado algunas pruebas adicionales con este método, estaban felices de aceptarlo como una forma de enfriar la serie 4CX250B. No llegaron a recomendarlo, simplemente dijeron que no había evidencia convincente para decir que no debería usarse. ¡Más confusión! Cuando los expertos no pueden ponerse de acuerdo entre ellos, no es de extrañar que los humildes aficionados estén confundidos; ¿A quién crees realmente?

El tiempo ha demostrado que ciertas cosas funcionan y otras no, y ahora existen suficientes datos para ver la madera de los árboles. El sistema de enfriamiento K2RIW funciona bien y al 4X150 y su descendencia no les gusta la red conectada a tierra, ¡punto!

El 4X150 y las variantes posteriores, en la práctica, no se prestan realmente a una rejilla con conexión a tierra debido al diseño de los enchufes de flujo de aire Eimac. El enchufe que debe usar es el SK-600, que tiene los cuatro pines del cátodo flotando, por lo que puede atarlos todos y conducirlo de esa manera. Pero el pin de rejilla sale del zócalo en el centro y no es tan fácil organizar un método de puesta a tierra de inductancia realmente baja. En general, se vuelve muy complicado y funcionan bien en un cátodo conectado a tierra con suficiente amortiguación de rejilla. Los transmisores SSB aerotransportados de Collins Radio utilizados por la RAF en la década de 1970 usaban el 4CX250F / G con un controlador de estado sólido de 1/2 W y daban un buen PEP de 400 W en HF y nunca dieron ningún problema. Como dice el refrán, "si no está roto, no lo arregles" .

Triodos comunes utilizados en la red con conexión a tierra

Se utilizaron comúnmente varios triodos diferentes para los amplificadores de HF "tradicionales". Hoy en día, todas estas válvulas pueden considerarse tecnología antigua, ya que todas se originaron antes de la guerra o en los años no mucho después de la guerra. A pesar de esto, ofrecen un buen servicio, aunque hoy en día las válvulas más recientes a menudo son definitivamente mejores, aunque más caras. Dos de los tipos originales son compatibles pin por pin y pueden intercambiarse si surge la necesidad. Estas son la serie 811 y la serie 572B, que también tiene el número de pieza británico T160. Ambos tipos tienen un calentador de 6.3V a 4A y tienen una base de cuatro pines, que se originó en la era de la válvula rectificadora. De los dos, el 811 es el de menor potencia.

811 / 811A

Las clasificaciones oficiales para esta válvula son una disipación de ánodo de 45 W y un voltaje de ánodo máximo de 1250 V. Sin embargo, esta es la calificación CCS. La clasificación ICAS es un poco más alta, donde la vida útil máxima de la válvula no es tan importante. En esta clasificación, la disipación del ánodo es de 65 W y el voltaje máximo del ánodo es de 1500 V. En ambos sistemas de clasificación, la corriente máxima de ánodo de un solo tono es de 175 mA. La corriente de la red es de 50 mA como máximo absoluto para cualquier sistema de clasificación. La cuadrícula de la serie 811 es particularmente frágil si está sobrecargada.

La potencia de salida típica en el sistema ICAS es de aproximadamente 160 W, o 120 W para la clasificación CCS. Desafortunadamente, todas las hojas de datos más antiguas solo brindan operación de Clase C, pero la eficiencia bajo la Clase B parece ser aproximadamente del 60% en un buen circuito. Las hojas de datos más recientes del 811A muestran el funcionamiento de Clase AB 2 a 30 MHz, pero los números no cuadran : cuando se calcula, la eficiencia es increíblemente alta.

    Varios proveedores estadounidenses de amplificadores utilizan el 811A con dos, tres e incluso cuatro válvulas en paralelo. He usado uno de estos, pero no me sentiría tentado a comprar uno. En un concurso local tuvimos uno durante unas dos horas, antes de que llorara lo suficiente. ¡Herviría una taza de café en la parte superior del gabinete! El viejo Collins Radio 30S-1, traído como respaldo, funcionó sin fallas durante el período restante del concurso, sin apenas sudar.

572B

Esta válvula es un hermano mayor del 811 y tiene más disipación de ánodo y puede soportar un voltaje de ánodo mucho más alto. Las versiones modernas son fabricadas por Svetlana y varios otros fabricantes del Lejano Oriente, aunque algunas de las versiones chinas parecen tener un historial de confiabilidad irregular y algunas son notablemente inestables cuando son nuevas.

La disipación máxima del ánodo es 160 W y el voltaje máximo del ánodo es 2750 V. La corriente máxima de la red es de 50 mA, la misma que la de la serie 811, pero no parecen ser tan frágiles. La eficiencia del ánodo es un poco más alta que la de la serie 811, pero no mucho. El voltaje de ánodo máximo recomendado a plena carga es de 2,4 kV, pero resistirán más sin carga. 3kV parece estar perfectamente bien para la mayoría de las válvulas. La corriente inactiva en polarización cero es de aproximadamente 45 mA con voltaje de ánodo de 2,4 kV, pero es bastante variable de una válvula a otra.

La potencia de salida típica para la versión Svetlana es de alrededor de 300W PEP a 30MHz y la unidad requerida es de 50W, por lo que tiene una ganancia de 6 o 7.75dB. La corriente máxima de ánodo para una salida de 300 W se calcula como 275 mA con un suministro de ánodo de 2,4 kV. Con esta potencia de entrada de 660 W, la eficiencia del circuito es del 45%. A 80 m, la eficiencia es de aproximadamente el 50% en un buen circuito de tanques. No espere ver mucho más que esto si la válvula está operando de manera lineal.

'Plug & Play'

Un problema importante entre el 811 y el 572B es la resistencia de carga del ánodo requerida. Aunque son plug and play en los enchufes, las constantes del circuito del tanque son radicalmente diferentes. Para actualizar un amplificador que actualmente usa la válvula 811 al 572B, se requieren cambios importantes no solo en los valores del circuito del tanque pi, sino también en la fuente de alimentación. El suministro de ánodo de 1250 V para el 811 es demasiado bajo para obtener lo mejor de un 572B.

Si esta es la forma en que desea modificar un amplificador, la mejor opción es usar un 572B para reemplazar dos 811, pero necesitará casi duplicar el voltaje de suministro. Como muchos amplificadores comerciales utilizan el método de suministro de alto voltaje atroz duplicador de voltaje, el único recurso es reemplazar el transformador de red o rebobinar el existente. Dos válvulas 572B consumirán una entrada de CC máxima de aproximadamente 1400 W, mientras que las válvulas 811 solo consumen la mitad de esta potencia, por lo que el transformador de red realmente necesita ser reemplazado por algo más pesado. Una solución mucho más sensata sería encontrar un comprador para el amplificador 811 y luego comprar un amplificador con válvulas 572B.

    Una comparación interesante entre el 811 y el 572B es considerar la potencia del filamento. Ambos tienen el mismo voltaje y corriente de filamento. La disipación del ánodo del 811 es 65W ICAS y el 572B es 160W. Generalmente, la potencia del filamento para válvulas de mayor potencia es mayor ya que necesita hervir más electrones. La razón por la que son iguales es que el 811 todavía usa el diseño de cátodo ineficiente original, pero el 572B posterior usa un cátodo mucho más eficiente para que pueda hervir más electrones para la misma potencia. El 3-500Z, mucho más tardío, utiliza una potencia de filamento de 75 W para una disipación del ánodo de 500 W. Si esta última tecnología de cátodo se aplicara a la válvula 811, podría usar tan solo 2.5A.

3-500Z

Esta válvula es una propuesta mucho mejor cuando necesita o quiere mucha potencia. Una sola válvula es buena para al menos 700W de potencia razonablemente lineal y durarán más que el 811 o el 572B por un amplio margen si trata el 3-500Z correctamente. Va a holgazanear al nivel de 500W sin ningún esfuerzo en un buen circuito. Es diferente, ya que cuando el 3-500Z salió al mercado, el 811 y el 572B ya eran de mediana edad. Eimac, con la experiencia adquirida durante el largo período que habían estado diseñando válvulas, lo hizo bien con la 3-500Z. Tenían varias válvulas similares antes, en particular la 3-400Z, que era el banco de pruebas, y la versión 500Z es en realidad solo una 400Z modificada un poco a la luz de los cambios de producción. El gran padre fue el 3-1000Z, con una disipación de ánodo de 1kW.

Desafortunadamente, el 3-500Z no es plug & play para el 811 o 572B, pero eso no importa.

Como sugiere el número de pieza, el 3 designa el número de electrodo de triodo y el 500 es la disipación del ánodo en vatios. La Z indica que es un calentador de filamento de calentamiento rápido y es un cátodo calentado directamente. El enchufe también es diferente, siendo un tipo gigante de cinco pines. Debido a que la válvula tiene una corriente de ánodo potencial mucho más alta, el cátodo es mucho más grande y esto se refleja en la corriente del calentador. Consume 14,5 A a 5 V, por lo que aproximadamente 75 W solo para alimentar el calentador.

El rendimiento de alta frecuencia también es muy superior en comparación con el 811 o el 572B, que ya están disminuyendo en ganancia a 30MHz. El 3-500Z tiene una clasificación de hasta 110MHz con una clasificación ligeramente reducida, y se muestra en su figura de ganancia. Eimac afirma hasta 12dB, pero a menudo no podemos explotar esto en una red con conexión a tierra y una ganancia de 8dB a 10dB es más realista.

El 3-500Z necesita el enchufe y la chimenea correctos para un funcionamiento óptimo. El enchufe es el SK-400 o el SK-410. La última parte es necesaria cuando se utiliza la chimenea de vidrio opcional SK-406. Además, la tapa superior del ánodo necesita un enfriador con aletas para ayudar a mantener fría esta pieza.

El otro requisito previo para que un 3-500Z funcione correctamente es un buen suministro de alto voltaje. Realmente no les gusta mucho menos de 2.5kV y 3kV es una apuesta mucho mejor. Al igual que con todos los amplificadores de válvula, los voltajes utilizados son aterradores y potencialmente letales, en caso de que se conecte cuando el suministro de ánodo está encendido.

    Sobre el tema de la sustitución de válvulas, hace unos años un aficionado me trajo un antiguo amplificador comercial que compró recientemente pero que ahora había dejado de funcionar. Al interrogar al propietario, me dijo que había comprado dos válvulas 572B nuevas, sacó las viejas y colocó las nuevas. Habiendo vuelto a armar todo, descubrió que no funcionaba en absoluto. Sin corriente de ánodo y sin salida. No pasó mucho tiempo para resolver el problema. Los enchufes de cuatro clavijas utilizados en este amplificador eran del tipo antiguo de baquelita y se vuelven quebradizos con el calor. Cuando se insertaron las nuevas válvulas, el propietario no tuvo suficiente cuidado para alinear los pasadores, por lo que se deslizaron en los enchufes. En cambio, los forzó y en el proceso los enchufes se rompieron. Los amplificadores posteriores utilizaron bases de válvulas de cerámica y no sufrieron el calor,pero si intenta forzar un 811 o 572B en el receptáculo sin alinear correctamente los pasadores, dañará la válvula.

Válvulas rusas

En los últimos años se ha puesto a disposición una gran cantidad de válvulas rusas. A pesar de lo que algunos piensan, en muchos casos la industria de válvulas rusa estaba muy avanzada a la de Occidente pero, con las frías relaciones entre Oriente y Occidente, no estábamos al tanto de ellas. (Svetlana era la empresa más conocida, pero había cientos que fabricaban todo tipo de válvulas diferentes. Hasta hace poco, una empresa con sede en California distribuía válvulas Svetlana en los EE. UU. La fábrica todavía está en San Petersburgo, Rusia. Svetlana es un nombre de niña común y significa luz o brillante, ya que originalmente hacían lámparas eléctricas).

Aparte de los tipos de microondas GI-7BT, no tengo mucha experiencia con estas válvulas rusas. El GI-7BT tiene un precio razonable si busca y, en general, tiene un buen rendimiento de hasta 2,3 GHz.

He visto varios amplificadores 572B con ellos instalados y eliminan puntos de las válvulas más antiguas. Una modificación popular para la serie de amplificadores Yaesu FL-2100 es adaptarse a los GI-7BT. Pero es una empresa importante y no debe ser emprendida por un constructor sin experiencia. El mayor problema es siempre el enfriamiento: la serie Gi7-bt son válvulas de pulso de radar y son capaces de generar una potencia increíble cuando se pulsan. La hoja de datos oficial, si puede encontrar la correcta en ruso, proporciona los detalles correctos (las traducidas a menudo tienen errores). Tuve la suerte de adquirir una hoja de datos rusa genuina y una mujer rusa casada con un amigo me la tradujo. ¡Entonces tuvo sentido!

En servicio de radar a 3GHz, el GI-7BT tiene una potencia de pulso mínima de 12kW para 3 µs longitud del pulso. Para esta función, se pulsa con 9 kV y consume una corriente de cátodo de 7,5 A. Para el funcionamiento en CW a 1,6 GHz, tiene una potencia de salida mínima de 40 W con un voltaje de ánodo de 1,05 kV y una corriente de cátodo de 600 mA. En la práctica, con 2,5 kV en el ánodo y consumiendo hasta 450 mA, entregarán una potencia considerable en 2 my 70 cm en un buen circuito, si puede mantenerlos frescos. En el servicio de pulso corto de radar, el enfriador de aire con aletas pequeñas es adecuado, pero para el funcionamiento en CW realmente necesitan algo mejor. Mi adaptación fue hacer refrigeradores líquidos y luego realmente se pueden presionar con fuerza en CW. Como la mayoría de las válvulas, deben mantenerse por debajo de unos 200ºC y preferiblemente no más de 100ºC. Con refrigeración líquida esto es fácil con suficiente flujo a través del enfriador. Mi amplificador de 6 m ejecuta dos válvulas Gi7-bt en paralelo y entrega más de 1 kW con facilidad,y el refrigerante nunca supera los 60ºC. Por el precio de una válvula 572B, puedo comprar unas cinco válvulas GI-7BT.

Triodo ruso GI-7BT con el adaptador de refrigeración líquida y el radiador refrigerado por aire original.

Los enchufes para estas válvulas utilizan una técnica coaxial y realmente deberían tener anillos de conexión a tierra de rejilla de dedo. Pero pueden montarse sólidamente siempre que tenga cuidado de no estresar la parte del ánodo. El enfriador de aire se quita fácilmente y esto deja un perno de cobre roscado M6 al que puede conectar otro tipo de enfriador. El perno de cobre no debe apretarse demasiado, ya que dañará las roscas.

El tema de la refrigeración líquida se trata en el Capítulo 13.

El filamento / calentador tiene una potencia nominal de 12,6 V CA o CC y consume aproximadamente 2 A. Es un cátodo calentado indirectamente, pero en común con la mayoría de los triodos de microondas, el cátodo y un extremo del calentador comparten el pasador del cátodo. El pasador más pequeño es el otro extremo del calentador y el siguiente más grande es el cátodo. El cilindro más grande es el terminal de la red.

Para el funcionamiento de VHF / UHF, la activación del cátodo no es difícil con la red de adaptación correcta. Los detalles del amplificador de 6 m que construí están contenidos en Conceptos básicos de diseño de RF [1] , que cubre todos los aspectos constructivos necesarios. También hay diseños de amplificadores de microondas que escribí para Microwave Projects 2 [2] , con métodos de enfriamiento adicionales.

    Habiendo cantado las alabanzas del 3-500Z, creo que es necesario advertirles que estas no son válvulas baratas. El precio más reciente que pude encontrar fue de alrededor de los $ 200 para una nueva versión china, y algunos de estos tienen una reputación accidentada. Pero también estoy al tanto de al menos dos amplificadores que tienen los Eimac 3-500Z originales con más de 30 años en el reloj sin un reemplazo de válvula. Tienen un poco de poder, pero están lejos de extinguirse.
