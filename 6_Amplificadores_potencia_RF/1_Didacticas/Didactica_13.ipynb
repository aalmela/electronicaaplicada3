# Amplificadores de potencia de RF 

En los amplificadores de potencia (AP) la clase da una amplia indicación de las características y el rendimiento de un amplificador. 
Las clases están relacionadas con el período de tiempo en que el dispositivo amplificador activo está pasando corriente, expresado como una fracción del período de una forma de onda de señal aplicada a la entrada. 

Un amplificador de clase A está conduciendo a través de todo el período de la señal; la clase B sólo durante la mitad del período de entrada, la clase C durante mucho menos de la mitad del período de entrada. 
Un amplificador de clase D hace funcionar su dispositivo de salida de forma conmutada; la fracción del tiempo que el dispositivo está conduciendo se ajusta de forma que se obtiene una salida de modulación de anchura de pulso de la etapa.
Se definen clases de letras adicionales para los amplificadores con fines especiales, con elementos activos adicionales o mejoras particulares de la fuente de alimentación; a veces un fabricante utiliza un nuevo símbolo de letras para promover su diseño patentado. 

Los AP de RF lineales, es decir fundamentalmente aquellos denominados clase A y B (push-pull), son poco eficientes.
En estos se incluyen a menudo circuitos sintonizados o filtros pasa bajo para asegurar la supresión adecuada de armónicosos, pero no son componentes integrales de estos amplificadores.
Como el ancho de banda de un señal de RF suele ser una fracción pequeña de la frecuencia de portadora y del ancho de banda del AP, una parte de las señales no deseadas a la salida del amplificador de potencia pueden ser originadas por alinealidades en los circuitos. 

La falta de linealidad del amplificador produce dos tipos de señales no deseadas, Productos Armónicos y intermodulación (en inglés intermodulation, IMD abreviado). 

La IMD es la modulación de amplitud no deseada de señales, con dos o más frecuencias diferentes, producida en sistemas no lineales. 
La intermodulación entre cada componente de frecuencia formará señales adicionales en frecuencias que no son, en general, armónicos (múltiplos enteros) de cualquiera de ellas, sino a menudo las frecuencias suma y diferencia de las frecuencias originales.
Los IMD son identificados con el orden que los conforman (tercero, quinto, etc). Cerca de la frecuencia de portadora causan distorsión en la señal recibida e interferencias de canal adyacente o ambas. 

A otras señales no deseadas tales como oscilaciones subarmónicas, señales parásitas o productos indeseados de mezcladores, se los llama productos espurios. 

En un amplificador de potencia de RF, las armónicas y alguno de los productos espurios se pueden eliminar con filtros, sin embargo los IMD generados deben ser de un nivel aceptablemente bajo debido a que algunos de ellos se encuentran muy próximos a la señal deseada.

Las armónicas producidas por la no linealidad del dispositivo se filtran a través del circuito sintonizado paralelo. 
Cualquier componente de frecuencia de portadora de $i(\theta)$ fluye a través de la resistencia de carga R (suponiendo una sintonización adecuada del circuito paralelo sintonizado y efectos insignificantes provenientes del  choque y del Cb ).

Los amplificadores de potencia se utilizan del lado de transmisión de los circuitos de RF, normalmente para controlar trasmisores.
Los amplificadores de potencia normalmente compensan entre eficiencia y linealidad.
Una mayor eficiencia significa menor consumo, lo que podría resultar en una mayor duración de la batería, y esto es especialmente importante en la realización de productos pequeños y portátiles.
Hay algunos desafíos adicionales específicamente relacionados con la integración total. Los circuitos integrados suelen tener un voltaje de alimentación limitado para evitar averías, así como un límite de migración de metal para la corriente. Por lo tanto, simplemente lograr la potencia de salida deseada puede ser un desafío.

La aplicación más común del amplificador de Clase C son los circuitos de RF (radiofrecuencia) como el oscilador de RF, el amplificador de RF, etc., donde hay circuitos sintonizados adicionales para recuperar la señal de entrada original de la salida pulsada del amplificador de Clase C y, por lo tanto, la distorsión causada por el amplificador tiene poco efecto en la salida final. 



Los amplificadores de potencia disipan energía y generan calor, que debe eliminarse. Debido al pequeño tamaño de los circuitos integrados, este es un ejercicio desafiante de diseño y empaquetado.
Varias presentaciones generales recientes han destacado los problemas especiales que surgen al lograr alta eficiencia y linealidad en amplificadores de potencia totalmente integrados.

Los amplificadores de potencia se encuentran entre los últimos circuitos en estar completamente integrados.
En muchos casos, no queda más remedio que diseñarlos con transistores de potencia discretos, o al menos separados del resto de los circuitos frontales de radiofrecuencia.

Hay mucho interés en el diseño de amplificadores de potencia discretos y semiintegrados, y durante años la principal referencia fue el libro clásico de Krauss, Bostian y Raab.


##  MOSFETs de potencia versus transistores bipolares
(AN-860)

Varios fabricantes han introducido recientemente FETs de potencia para aplicaciones de amplificación de RF. 
Se dispone de dispositivos con capacidad de salida de $100 W$ para frecuencias de VHF y se fabrican unidades más pequeñas para el funcionamiento en UHF. 
Todos son dispositivos de modo de mejora, lo que significa que la puerta debe estar sesgada con voltaje positivo (canal N) con respecto a la fuente para "encenderla". 

Los primeros diseños fueron llamados V-MOSFETs, donde el canal está en una ranura en V. 
La ranura en V debe ser grabada con un proceso especial, y el material de silicio debe tener una orientación de cristal diferente del material normalmente utilizado para los transistores bipolares. La dificultad del proceso de grabado en la producción ha llevado al desarrollo de otros tipos de estructuras de canal como HEX y T, que siguen siendo estructuras de canal vertical, pero la ranura en V se elimina, y la puerta está en una superficie recta. 

Así pues, para una periferia de puerta igual, se requiere más espacio en la superficie. 
Los fabricantes japoneses parecen favorecer las geometrías con canales horizontales. 
Son similares a los pequeños MOSFETs de señal con varios de ellos paralelos en un chip. 
Esta técnica representa un uso aún más derrochador de la superficie del dado que el HEX o el TMOS. 

Típicamente un FET de potencia requiere de 50 a 100 por ciento más área del dado que un transistor bipolar para un rendimiento de salida de potencia igual. 


En el caso del TMOS, el número es de aproximadamente el 50 por ciento. 
Esto se debe principalmente al mayor voltaje de saturación, pero la geometría también da un 30 por ciento menos de periferia de la puerta que el área base disponible en bipolar. 
Dado que el precio de un dispositivo de estado sólido es una función del tamaño del dado, obtenemos menos vatios por dólar.


Para simplificar el analisis de estos circuitos en este apunte se propone emplear una trasferencia lineal como se muetra a continuación.


### Capacidad de Potencia

Uno de los principales objetivos del diseño de PA es entregar una potencia determinada a una carga.
Esto está determinado en gran medida por la resistencia de carga y la fuente de alimentación.
Dado un voltaje de fuente de alimentación particular $V_{CC}$, como 3 V, y una resistencia de carga $R_L =50 \Omega$, es posible determinar la potencia máxima que se debe usar.

$P = \frac{V^2_{CC}}{2R}$

Esto supone que tenemos un amplificador sintonizado y un punto de funcionamiento de por ejemplo $3 V$, una oscilación máxima negativa de $0 V$ y una oscilación máxima positiva de hasta $6 V$.

El voltaje minimo del colector sinusoidal debe garantizar mantener al dispositivo en la zona activa.
La suposición de que el voltaje del colector puede descender cerca de 0 V se justifica porque simplifica el análisis y normalmente produce sólo un error insignificante. 
Si bien se supone que el voltaje del colector es sinusoidal debido a la acción de filtrado del circuito sintonizado, la corriente del colector puede ser sinusoidal, como en la operación de clase A, o puede ser no sinusoidal, como en la operación de clase B o C, que está determinada principalmente por cómo está polarizado el transistor. 

La clasificación como A, AB, B o C describe la fracción del ciclo completo durante la cual fluye corriente en el transistor controlador. Esta fracción se puede describir como un ángulo de conducción, que es el número de grados (de 360\°) durante los cuales fluye corriente. Si siempre fluye corriente, el ángulo de conducción es de 360\° y el funcionamiento es de clase A.
Si la corriente fluye exactamente la mitad del tiempo, el ángulo de conducción es de 180\° y el funcionamiento es de clase B.

Para ángulos de conducción entre 180° y 360° el funcionamiento es clase AB. Si la corriente fluye entre 0\° y 180\°, se dice que el transistor funciona como clase C.
Más adelante se demostrará que un ángulo de conducción mayor dará como resultado una mejor linealidad pero una menor eficiencia.
En la Tabla se muestra un resumen.

|B     | 180             |78.5                             | 1  |
|C     |180–0            |78.5–100                         | 1  |


Para obtener una alta eficiencia, se debe minimizar la pérdida de potencia en el transistor, y esto significa que la corriente debe ser mínima mientras el voltaje sea alto, y el voltaje debe ser mínimo mientras la corriente sea alta.


Esto conduce a una mayor eficiencia; sin embargo, todavía habrá alguna pérdida, ya que hay una superposición de voltaje y corriente distintos de cero.
Otras clases de amplificadores, que se describirán en secciones posteriores, a saber, las clases D, E, F y S, están diseñadas de manera que el voltaje a través del transistor también sea no lineal, lo que conduce a eficiencias más altas, en algunos casos hasta el 100%.


Una forma diferente de mejorar la eficiencia, manteniendo potencialmente la linealidad, es alimentar un amplificador lineal desde una fuente de alimentación variable o conmutada.
Esta es la base para los diseños de clase G y H.

## Análisis de clase A, B y C


https://www.semanticscholar.org/paper/A-Nonlinear-Theory-of-Class-C-Transistor-Amplifiers-Harrison/a0f8575377636dca289978cd66c97a76a3526898

El amplificador de potencia de clase C es un tipo de amplificador en el que el elemento activo conduce durante menos de medio ciclo de la señal de entrada. Menos de un medio ciclo significa que el ángulo de conducción es inferior a $\theta_c \le 180$ y su valor típico es de $\theta_c = 80$ a $\theta_c = 120$. 
Un ángulo de conducción reducido mejora la eficiencia en gran medida pero causa mucha distorsión. 
La eficiencia máxima teórica de un amplificador de clase C es de alrededor del $90\%$.


Excepto en el caso de la clase A, la corriente que circula por el transistor no es sinusoidal, pero puede modelarse como una sinusoide polarizada.

![5-cuatrocientos](fig/waveform_ic.png)


Suponiendo una trasferencia lineal en la zona de conducción, la corriente del colector se puede expresar como

$i_C = I_{CC} cos(\omega t) − I_{CQ}$

Esto es válido desde $−\theta_c < \omega t < \theta_c$ (zona de conducción o activa).

Aquí,  $I_{CQ}$ es dado por

$I_{CQ} = I_{CC} cos(\theta)$

Podemos encontrar el componente de continua y el componente fundamental mediante determinar la serie de Fourier de esta forma de onda. 

Tenga en cuenta que el circuito sintonizado nos dará el componente fundamental (si está sintonizado en $f_o$ ).


$I_{dc} = \frac{1}{2 \pi} \int^{\theta}_{-\theta} (I_{CC} cos(\omega t) - I_{CQ}) d(\omega t)$

$I_{dc} = \frac{I_{CC}}{\pi}[sin(\theta) - \theta cos(\theta)]$

La potencia suministrada está dada por


$P_{CC} = V_{CC} I_{dc} = \frac{V_{CC} I_{CC}}{\pi}  (sin(\theta) - \theta cos(\theta))$

La corriente fundamental $i_1$ dada por

$i_1 = \frac{4}{2\pi}  \int^{\theta}_{-\theta}  (I_{CC} cos(\theta t) − I_{CQ} ) cos(\theta t) d (\omega t )$ 

$i_1 = \frac{I_{CC}}{2 \pi}  (2\theta − sin 2\theta )$


La potencia de salida está dada por

$P_{out} = \frac{i^2 R_L}{2} = \frac{v_{peak}}{\sqrt{2}} \frac{i_{peak}}{\sqrt{2}}$

El pico v máximo posible es cuando la salida oscila de aproximadamente 0 V a $2 V_{CC}$ o pico $v = V_{CC}$. 

De este modo,


$P_{salida, máx} = \frac{V_{CC}}{\sqrt{2}} \frac{i_{1}}{\sqrt{2}}$

La eficiencia para esta oscilación de voltaje máxima posible está dada por
 
$n_{max} = \frac{P_{out,max}}{P_{dc}}$

$n_{max} = \frac{2 \theta - sin(2\theta)}{4 (sin \theta - \theta cos(\theta)}$ 
 
La eficiencia se representa en la siguiente Figura.

```python
def nn(th):
    return (2 *th - np.sin(2 * th))/(4 *(np.sin(th)- th * np.cos(th)))

th = np.arange(0,np.pi,np.pi/100)

n    = nn(th)

plt.plot(th/np.pi, n*100, 'b')
plt.ylabel('$\eta$ %')
plt.xlabel('$\Theta_{c}/\pi$')
plt.grid()
plt.show()
```

La potencia de salida real para un voltaje pico de salida de $V_{op}$ se puede encontrar en función de ␪:

$P_{out} = \frac{V_{op} I_{CC}}{(2 \theta − sin 2\theta)}$
 
teniendo en cuenta que para obtener la potencia máxima, la resistencia de la carga debe ajustarse de modo que el voltaje máximo $v_{o,max}$ sea aproximadamente 2 V CC y el voltaje mínimo v o, min sea aproximadamente cero, por lo tanto, V op es igual a V CC .

La ecuación anterior será más conveniente de trazar si eliminamos I CC.
Recuerde que I CC es una medida de la corriente máxima, no con respecto a cero, sino con respecto al centro de la onda sinusoidal donde el centro para la clase C es menor que cero, como se muestra en la figura 10.12.

La corriente máxima es

$i_{pico} = I_{CC} (1 − cos \theta)$


Para la misma corriente máxima, la potencia de salida máxima distorsionada, o clase A, se puede determinar observando que el voltaje va de 0 a 2 V CC y la corriente va de 0 a i pico. Luego, convirtiendo pico a pico a rms, obtenemos la potencia de salida normalizada Po, norma como


$P_{o,norm} = \frac{(I_{max} − I{min}) (V_{max} − V_{min} )}{8} = \frac{I_{CC} V_{CC} (1 − cos(\theta)}{4}$


Entonces podríamos trazar la potencia de salida máxima normalizada como

$P_{o,max,norm} = \frac{P_{o,max}}{P_{o,norm}}$ 

$P_{o,max,norm} =\frac{1}{\pi} \frac{(2 \theta − sin 2\theta )}{1 - cos(\theta)}$




```python
def pp(th):
    return (2 *th - np.sin(2 * th))/(1-np.cos(th))*(1/np.pi)

th = np.arange(0,np.pi,np.pi/100)
p    = pp(th)


plt.plot(th/np.pi, p*100, 'b')
plt.ylabel('$P_{o_{max}}$')
plt.xlabel('$\Theta_{c}/\pi$')
plt.grid()
plt.show()
```

Se puede ver que en un ángulo de conducción de 180\° y 360\° (o $\theta = 90$\° y 180\°), la potencia de salida máxima normalizada es 1. 
En el medio hay un pico con un valor de aproximadamente 1,15 en un ángulo de conducción de alrededor de 240\°.

Para una potencia de salida máxima, este podría parecer el ángulo de conducción óptimo; sin embargo, se puede observar que en la vida real, o en simulaciones con otros modelos de la corriente (en lugar de la punta de una sinusoide), este pico no ocurre. 


## Sintetizando Amplificadores de Potencia de RF

Las técnicas más populares de diseño de pequeñas señales de RF son:
1. el uso de parámetros de cuadripolos 
2. el uso de algún tipo de circuito equivalente 

Los primeros intentos de adaptar estas técnicas al diseño del amplificador de potencia dieron lugar a malos resultados y a la frustración.

### IMPORTANCIA DE LOS PARÁMETROS DE LA GRAN SEÑAL 

La teoría de la red para el diseño de amplificadores de potencia es bien conocida pero no es útil, a menos que el diseñador tenga datos válidos de impedancia de entrada y salida para el transistor. 
El método de diseño descrito en este desarrollo depende principalmente de la medición directa de estos parámetros para su uso en ecuaciones de síntesis de redes. 
Los datos de impedancia de gran señal, junto con los datos de potencia de salida y de ganancia, proporcionan al diseñador la información necesaria para diseñar sus redes de amplificadores y para predecir el rendimiento que debe alcanzarse cuando se complete el diseño. 
Es importante comprender claramente las condiciones de prueba y el método de presentación de los datos de la impedancia de la gran señal. 

### Condiciones 

Los términos "impedancia de entrada de gran señal" e "impedancia de salida de gran señal" se refieren a la impedancia terminal real del transistor cuando funciona en un amplificador adaptado al nivel de salida de potencia de RF y al voltaje de suministro de continua deseados.

Adaptado se define como la condición en la que las redes de entrada y salida del amplificador de prueba proporcionan una coincidencia conjugada con el transistor, de tal manera que la impedancia de entrada y salida del amplificador es de $50 \Omega$.

Las impedancias de gran señal no debe confundirse con la de pequeña señal, los parámetros del cuadripolo normalmente se miden a niveles de bajas señal con sesgo de clase A y el transistor (o IC) conectado directamente con cortocircuito, circuito abierto o  $50 \Omega$.

La mayoría de los datos que aparecen en las hojas de datos de los transistores de potencia de RF de Motorola se miden como amplificadores de Clase C en configuración de emisor comun, ya que esta condición cubre la mayoría de las aplicaciones de los dispositivos.

### DISEÑO DEL AMPLIFICADOR 
Después de seleccionar un transistor con las capacidades de rendimiento requeridas, el siguiente paso en el diseño de un amplificador de potencia es determinar las impedancias de entrada y salida de grandes señales del transistor. 

Cuando se utilizan dispositivos para los que se dispone de datos, este paso no implica más que la lectura de los complejos valores de impedancia de la hoja de datos.

Una vez determinadas las impedancias de gran señal, el diseñador selecciona una configuración de red adecuada y procede a su síntesis de red.

### RESISTENCIA DE LA CARGA DEL COLECTOR 

Los datos de impedancia de gran señal en HF y VHF han sido publicados en su mayor parte por Motorola sin información de resistencia de carga del colector. 
La razón es que la resistencia de carga puede ser calculada fácilmente. 
Las condiciones necesarias para obtener esta derivación de la resistencia de carga se discutirán ahora. 

Si se hacen ciertas suposiciones para simplificar el analisis, el voltaje teórico del colector de un amplificador de potencia con una red de salida sintonizada es una onda sinusoidal que oscila de cero a $2 V_{CC}$. 
donde $V_{CC}$ es el voltaje de suministro del colector de la fuente DC. 

Estas suposiciones incluyen:
1.  $V_{sat}$ es igual a cero.
2.  La red de salida tiene suficiente $Q_C$ para producir un voltaje de onda sinusoidal independientemente del ángulo de conducción del transistor.
3.  La caída de voltaje en el sistema de alimentación del colector de CC es cero.
4.  La impedancia de carga del colector en todos los armónicos de la frecuencia de operación es cero.

Obviamente, ninguno de los supuestos anteriores es cierto, y las discrepancias más graves probablemente se derivan de los supuestos 1 y 4. 

Supongamos por un momento que esta condición teórica existe. 
La resistencia de carga del colector equivalente en paralelo, $R_L'$, se convierte entonces en una función de la potencia de salida de RF deseada y de la tensión $V_{CC}$ únicamente.  
La expresión para $R_L'$ dada en la ecuación se deduce fácilmente

$$R_L' =\frac{V_{CC}^2}{2 P_o}$$

Por lo tanto, la  impedancia compleja de carga del colector para el diseño de amplificador sería el conjugado de la capacitancia de salida equivalente en paralelo y la resistencia de carga del colector calculada con la Ecuación.


####  IMPEDANCIA DE SALIDA

La impedancia de salida de los transistores de potencia de RF, tal como se indica en las hojas de datos de todos los fabricantes, generalmente consiste sólo en un $C_{OUT}$ de capacitancia. 
Se supone que la resistencia interna del transistor es mucho mayor que la carga y normalmente se descuida. 
En el caso de una resistencia interna relativamente baja, la eficiencia del dispositivo disminuiría por el factor

$$1+ \frac{R_L}{R_T}$$

donde $R_L$ es la resistencia de carga, vista en los terminales del colector-emisor, y $R_T$ la resistencia del transistor interno igual a

$$\frac{1}{\omega_T (C_{CT}+C_{DC})}$$


definido como un pequeño parámetro de señal, donde: $\omega T=$ frecuencia angular de tránsito $C_{TC} + C_{DC}$ = capacitancias de transición y difusión en la unión del colector

La capacitancia de salida $C_{OUT}$, que es un parámetro de señal grande, está relacionada con el parámetro de señal pequeña $C_{CB}$, la capacitancia de transición de la base del colector.
Dado que la capacitancia de unión varía con el voltaje aplicado, $C_{OUT}$ difiere de $C_{CB}$ en que tiene que ser promediada sobre la oscilación total del voltaje.  
Para una unión abrupta y asumiendo ciertas simplificaciones, $C_{OUT} = 2 C_{CB}$. 

La figura muestra la variación del $C_{OUT}$ con la frecuencia. $C_{OUT}$ disminuye en parte debido a la presencia de la inductancia del colector de plomo, pero principalmente por el hecho de que el diodo emisor de base ya no se apaga cuando la frecuencia de funcionamiento se acerca a la frecuencia de tránsito $f_T$.


Suponiendo despresiable la resistencia de salida del dispositivo,  la carga de salida $R_L$ se toma igual a

$$R_L = \frac{(V_{CC}-V_{SAT})^2}{2 P_o}$$

La ecuación anterior sólo expresa una relación bien conocida, pero también muestra que la carga, en una primera aproximación, no está relacionada con el dispositivo, excepto para el $V_{SAT}$.  
El valor de la carga viene dictado principalmente por la potencia de salida requerida y el pico de tensión; no se corresponde con la impedancia de salida del dispositivo. 
A frecuencias más altas esta aproximación se hace menos exacta y para los dispositivos de microondas la carga que debe presentarse al dispositivo se indica en la hoja de datos. 

En sentido estricto, la adaptación de la impedancia se realiza sólo en la entrada.  
La coincidencia entre etapas y la coincidencia de carga son más transformaciones de la impedancia de entrada del dispositivo y de la carga en un valor $R_L$ (a veces con un componente reactivo adicional) que depende esencialmente de la potencia demandada y del voltaje de suministro.

## Consideraciones Practicas

La operación de un amplificador real no es tan buena como la pronosticada para el modelo ideal, pues los dispositivos reales tienen tensiones o resistencias de saturación, y cargas reales que incluyen reactancia. 
En altas frecuencias, los efectos de almacenamiento de carga pueden causar distorsión, de la forma de onda de corriente de colector, reduciendo en mas la eficiencia. 
En general, estos efectos se pueden despreciar cuando $f \le \frac{f_T}{10}$, aunque adquieren mucha importancia cuando $f < \frac{f_T}{2}$.

### Tensión de Saturación

Los transistores bipolares se caracterizan por una tensión colector-emisor bastante constante durante la saturación. 
Un BJT en lo circuitos amplificadores de potencia, entrará a la saturación siempre que el amplificador de potencia se excite lo suficiente, de tal suerte que el tensión de colector instantáneo mínimo sea igual al tensión $V_{CE(SAT)}$ de saturación. 


Para evitarla, la señal excitación se debe conservar lo suficientemente pequeña para mantener la oscilación de tensión de colector debajo de $V_{eff} = V_{CC} - V_{CE(SAT)}$.


La tensión de salida máxima, la potencia de salida y las corrientes de salida y entrada se calculan usando el tensión de alimentación efectivo $V_{eff}$ en lugar del $V_{CC}$.

Sin embargo, la potencia de entrada se calcula usando la $V_{CC}$. 
Esto reduce la eficiencia máxima en un factor de $\frac{V_{eff}}{V_{CC}}$, por ejemplo, la eficiencia máxima de un amplificador de potencia en clase A y B se reduce a  $\pi$ ($\frac{V_{eff}}{V_{CC}}$) y $\frac{\pi}{4}$ ($\frac{V_{eff}}{V_{CC}}$), respectivamente.


En bajas frecuencias, ($f < \frac{f_T}{10}$), $V_{sat}$ es simplemente la $V_{CE}(sat)$ especificado para el 
dispositivo (por lo común $V_{CE}(sat)=0.3V$ para transistores de silicio en corrientes débiles). 
Si se usa una configuración Darlington, $V_{sat} \sim V_{CE}(sat) + V_{p}$ donde $V_{p}$ es la caída de tensión entre base y emisor con polarización directa (por lo general, $V_{sat} \sim 0.7V$ para transistores de silicio). 


En frecuencias mas elevadas ($f > \frac{f_T}{10}$), $V_{sat}$ se hace mayor que $V_{CE}(sat)$ para operación en régimen permanente. 
El incremento en $V_{sat}$ está relacionado con el tiempo que se requiere para que el transistor completo llegue a saturarse. Esta tensión de RF de saturación puede ser tan alta como 1V o 2V y algunas veces se especifica en los manuales de operación.
Resistencia de Saturación 


La resistencia de saturación Ron (o resistencia de encendido) de los FET’s, como las tensiones de saturación de los BJTs, limita la salida máxima pero no afecta la operación en la región activa. 
La saturación se presenta cuando la tensión de drenaje instantánea mínima es igual al producto de la corriente de drenaje por la resistencia de saturación; es decir, para la clase B, $V_{DD} - I_{dm}R = I_{dm}R_{on}$.
Consecuentemente, para evitar saturación, $V_{dm}$ debe ser menor que la tensión de alimentación efectiva: $V_{eff} =R + Ron V_{DD}$

Para operación en clase A o para un amplificador de potencia clase B sintonizado con terminación única, se duplican los efectos de $R_{on}$; es decir, $V_{eff} = \frac{V_{DD}R}{R+2R_{on}}$. 

En cualquier caso, la tensión, la potencia de salida máximos y la corriente de entrada se calculan usando el $V_{eff}$ y la potencia de entrada se calcula entonces usando $V_{DD}$, como para los efectos de tensión de saturación.


### Distorsión    de  Intermodulación    y  Polarización  

La distorsión de intermodulación (IMD) consiste en aquellos productos de distorsión que ocurren cerca de la frecuencia portadora y proviene de la incapacidad de un amplificador de potencia en RF para reproducir exactamente la envolvente y la fase de la señal de entrada. 

La reproducción exacta de la portadora sinusoidal es innecesaria, pues las armónicas se remueven por el circuito de salida sintonizada o por el filtro. Dos causas importantes de IMD son los efectos de cruce por cero y la reducción de ganancia con corriente elevada. Otras dos de son la saturación del dispositivo y la variación de la capacitancia del colector con el tensión de colector. 

La saturación origina tanto un "aplanamiento de cima" como modulación de fase indeseada (debida a efectos de tiempo de almacenamiento) y se evita no sobreexcitando al amplificador (es decir, no excediendo la salida máxima para la clase de operación que se pretende). La capacidad variable con la tensión es inherente a las uniones de semiconductores, aunque su efecto es por lo común pequeño en la mayoría de las aplicaciones de amplificadores de potencia. 

La distorsión de intermodulación se mide generalmente con una señal de prueba 
compuesta de dos sinusoides de amplitudes iguales pero separadas en frecuencia por una frecuencia de 1 a 2 KHz. Los productos resultantes se denominan de tercer orden, de quinto orden, y así sucesivamente, de la misma manera que los productos de mezcladores. La medida que mas comúnmente se emplea para la IMD representa la razón que existe entre el producto mas grande de intermodulación con la amplitud de uno de los dos tonos; -30 dB es usualmente la razón IMD menos aceptable, pero aun tolerable, en un transmisor. Esta razón se mide fácilmente con un analizador de espectro, pero es demasiado pequeña para medirse con exactitud en un osciloscopio (o aun detectarse). Es interesante observar que si los productos de intermodulación en las bandas laterales superior e inferior son idénticas, son el resultado de una distorsión ya sea de la envolvente pura o de la fase pura; en cambio los productos de intermodulacion desiguales del mismo orden indican distorsión simultánea de fase y envolvente.

### Distorsión cruzada 

Los dispositivos reales (BJTs, FETs o tubos de vacío) operados en clase B, no 
cambian bruscamente de un modo de corte a uno activo (fuente de corriente lineal). El cambio real es gradual y no lineal. En consecuencia, el efecto denominado "distorsión cruzada", altera la forma de onda cuando un dispositivo entra en corte y el otro se torna activo. Estos efectos son más pronunciados con señales de amplitud relativamente pequeñas (respecto a la salida máxima del amplificador de potencia).

La distorsión cruzada se reduce al mínimo polarizando las bases o compuertas del dispositivo para producir una pequeña corriente estable en el colector o drenaje. Una corriente estable demasiado pequeña, no da lugar a salida cuando la entrada no alcanza la amplitud mínima requerida. Demasiada polarización se traduce en ganancia excesiva para señales de pequeña amplitud. 

La cantidad de corriente estable para tener distorsión mínima se determina experimentalmente en forma más directa, pues la predicción teórica es sumamente complicada. Por lo común, la razón IMD mínima ocurre con una corriente estable que tiene entre $1 y 10 \%$ de la corriente pico del colector o de drenaje; el valor exacto rara vez es crítico. 
En virtud de esta corriente estable, cada dispositivo lleva corriente durante un poco más de un medio ciclo de radio frecuencia. Esto explica el uso del término clase AB para los amplificadores de potencia con tubos de vacío y para algunos de los primeros amplificadores de potencia de estado sólido; sin embargo, muchos autores usan simplemente clase B, sobreentendiéndose que se usará una corriente estable.

### Fuente de Polarización

La polarización adecuada en los amplificadores de potencia con FET, se obtiene 
fácilmente de una tensión de alimentación regulada, mediante un divisor de tensión resistivo ajustable y una red de “puenteo” adecuada. Sin embargo, la polarización de tensión simple no se puede usar con los amplificadores de potencia con BJT. El efecto de incrementar la temperatura de una unión PN equivale más o menos en desplazar las características del dispositivo en aproximadamente 2 mV/C; es decir, un aumento de 1 C en la temperatura produce aproximadamente el mismo incremento en la corriente que produciría un incremento de 2 mV en el tensión de juntura. Supóngase que se aplica una tensión de polarización VBE de un BJT y que se ajusta la tensión rápidamente para producir la corriente de colector estable deseada. Esta corriente hace que se disipe potencia en el transistor, elevando le temperatura de la unión. Esto a su vez incrementa la corriente de colector, lo cual da lugar a un aumento en la disipación. Como hay poca resistencia de corriente contínua en el circuito de colector; casi no hay límite para la corriente de colector que se puede obtener. Este proceso se denomina desbordamiento térmico y conduce a la destrucción del dispositivo.

### Medicion de los parametros

Aunque los ingenieros de diseño encontrarán la caracterización de la impedancia de la gran señal en las hojas de datos para los transistores de potencia de RF, puede ayudar a saber cómo se obtienen estos datos.

El transistor se coloca en un circuito de prueba diseñado para proporcionar una amplia capacidad de sintonización. El diseño del primer amplificador de prueba para un nuevo tipo de transistor se basa en estimaciones de la impedancia de entrada y de salida.

Dado que las impedancias de entrada y salida son necesarias para diseñar un amplificador que luego se utiliza para medir las impedancias del dispositivo, tenemos un problema del tipo "el huevo o la gallina".  
Las redes de amplio rango de sintonía ayudan a compensar los errores en las estimaciones de impedancia y también permiten utilizar el mismo amplificador de caracterización en múltiples niveles de potencia de salida. 

El amplificador se sintoniza para una condicion luego de iteraciones de impedancia tanto en la entrada como en la salida.  
Se deben tomar varias precauciones para asegurar que esto se logre. 
La sintonización para la salida de máxima potencia es válida sólo si las impedancias de la fuente y de la carga son de $50 \Omega$.

Por lo general, en el laboratorio se dispone de una buena carga de $50 \Omega$. 
Dicha carga debe utilizarse, ya que la sintonización para la máxima potencia de salida para una potencia de entrada determinada es el mejor método a utilizar en la red de salida del amplificador. 

La red de entrada plantea algunos problemas adicionales. 
En primer lugar, muchas fuentes de energía de RF de laboratorio no son generadores de $50 \Omega$. 
Una impedancia del generador que no sea de $50 \Omega$ puede introducir errores en la medición de la ganancia así como en las impedancias de entrada. 
Además, una fuente con un alto nivel de exploración armónica causa dificultades en las redes de entrada de baja Q.
Una buena solución a este problema es utilizar un acoplador direccional doble o un medidor de potencia direccional en la línea coaxial entre el generador y el amplificador de prueba.  

A continuación, el amplificador se sintoniza para que la potencia reflejada sea cero, lo que indica que la red de entrada está realmente adaptada a la impedancia de entrada del transistor a $50 \Omega$.
En la práctica, la potencia reflejada no suele ser nula hasta el cero, por lo que hay que asegurarse de que el nulo sea al menos tan profundo como el que se obtiene con una buena pasiva terminación de 50 ohmios.

En algunos casos, el amplificador reflejará suficiente potencia armónica, para evitar que se obtenga nula potencia reflejada, una buena solución a este problema es colocar un filtro pasabanda de frecuencia fundamental en el puerto de potencia reflejada del acoplador direccional doble.
