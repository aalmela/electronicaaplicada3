### Tecnologías de transformadores de RF

Los dos tipos principales de transformadores de RF discretos son los de núcleo y cable y los de línea de transmisión. Además, dos tipos comunes de diseños de transformadores compactos y de bajo perfil son los LTCC y los MMIC.

#### Transformadores de RF de núcleo y alambre.

Los transformadores de núcleo y cable se fabrican envolviendo hilos conductores, normalmente hilos de cobre aislados, alrededor de un núcleo magnético como un toroide. 
Puede haber uno o más devanados secundarios, que también pueden tener una toma central para permitir funciones adicionales.

Sin embargo, el menor tamaño de los transformadores compactos aumenta las pérdidas resistivas de los devanados y del núcleo, lo que da lugar a una mayor pérdida de inserción a frecuencias más bajas.

#### Transformadores de RF de línea de transmisión

Las topologías de los transformadores de línea de transmisión pueden incluir líneas de transmisión diseñadas con precisión y colocadas entre dos cargas desajustadas o una disposición compleja de varias líneas de transmisión. Por ejemplo, se puede utilizar una longitud de línea de transmisión para implementar una transformación de impedancia entre dos cargas desajustadas. Algunos transformadores de líneas de transmisión utilizan hilos aislados envueltos en núcleos de ferrita, lo que los asemeja a los típicos transformadores de núcleo y cable, a menudo considerados transformadores de núcleo y cable. No obstante, el siguiente análisis se ofrece más para describir el comportamiento de los transformadores que para mejorar su comprensión.

### Transformadores LTCC

Los transformadores LTCC son componentes multicapa fabricados con un sustrato de base cerámica. Los transformadores LTCC utilizan líneas acopladas que actúan como líneas de transmisión para lograr la transformación de la impedancia y la conversión de la señal de monopolar a equilibrada. Los transformadores LTCC se basan en el acoplamiento capacitivo, lo que permite a los transformadores LTCC funcionar a frecuencias más altas en comparación con los transformadores ferromagnéticos. Sin embargo, esto puede provocar una degradación del rendimiento a bajas frecuencias. Una de las ventajas de la tecnología LTCC es la posibilidad de fabricar transformadores pequeños y resistentes, ideales para aplicaciones de alta fiabilidad (véase la figura 7).

Al igual que los transformadores LTCC, los transformadores MMIC se fabrican utilizando sustratos 2D con una metalización plana de capas de precisión. Normalmente, los transformadores MMIC se fabrican utilizando inductores en espiral impresos en un sustrato en una configuración de dos líneas de transmisión, con las líneas en paralelo. ## Balun y Unun


Las topologías de transformadores de RF más complejas, como los transformadores con varios devanados, derivaciones y elementos adicionales, presentan una dinámica de rendimiento variable basada en la topología y la construcción del transformador. 
Por ejemplo, un dispositivo de RF conocido como balun se utiliza para interconectar eficazmente circuitos equilibrados (es decir, de señal diferencial) con circuitos no equilibrados (es decir, de señal de un solo extremo) mediante la transformación de la impedancia; puede realizarse con un transformador de RF. 

Otro dispositivo similar a un balun, conocido como unun, se utiliza para interconectar circuitos de RF no equilibrados con otros no equilibrados, y puede realizarse con un transformador de RF. 
Un balun común fabricado a partir de un transformador es un transformador balun acoplado por flujo, que se construye enrollando cables separados alrededor de un núcleo magnético y conectando a tierra un lado del devanado primario. 
Las señales de RF de un solo extremo que entran en el devanado primario desequilibrado se someten a una transformación de impedancia a una salida diferencial (es decir, equilibrada) a través del devanado secundario.


Los baluns y ununs son esenciales en las cadenas de señales de RF para muchas aplicaciones. Los diseños de balun de RF se asocian más comúnmente con transformadores de núcleo y cable, pero también se pueden realizar a través de tecnologías de líneas de banda acopladas y coaxiales. 
La principal diferencia entre los ununs y los baluns están diseñados para igualar las impedancias entre los circuitos balanceados y no balanceados, mientras que los no balanceados proporcionan un emparejamiento de impedancias entre dos circuitos no balanceados.

### Aplicaciones comunes de balun

El uso más común de los baluns es cuando se usa un amplificador de potencia de un solo extremo para impulsar una carga balanceada. 
Los ejemplos incluyen antenas dipolo o antenas de un solo extremo, como látigos, que se necesitan para alimentar un amplificador frontal adicional (consulte la figura 1). 
En el pasado, los baluns también se usaban ampliamente en la industria de la televisión por cable, por ejemplo, al combinar una antena dipolo de $300 \Omega$ para transmisión de televisión con un cable coaxial de $75 \Omega$. 
Con el desarrollo de circuitos integrados de RF (RFIC), los baluns ahora también se utilizan ampliamente para mejorar la inmunidad al ruido y el rechazo de modo común. 
El crecimiento de las aplicaciones 5G también ha generado una enorme demanda de baluns pequeños de banda ancha para interactuar con transceptores de radio altamente integrados que utilizan entradas y salidas diferenciales.

Los Ununs se utilizan a menudo si una línea de alimentación desequilibrada está impulsando una antena desequilibrada y hay una falta de coincidencia de impedancia entre la línea de alimentación y la antena. Una antena de látigo con una impedancia de entrada baja se beneficiaría de una impedancia transformadora para acoplar eficientemente una línea de alimentación de 50 Ω con la antena.



### Introducción a la teoría balun y a los sistemas equilibrados y desequilibrados

Antes de abordar la teoría del balun, es importante comprender las diferencias entre fuentes y cargas de dos terminales equilibradas y no equilibradas. 
En un circuito balanceado, las señales viajan a lo largo de dos caminos, cada uno con la misma impedancia a tierra. La impedancia de un sistema balanceado se define por la impedancia entre las dos rutas, mientras que en un sistema no balanceado, un terminal está conectado a tierra.


![32-cuatrocientos](fig/balan_un.png)
 Circuito desbalanceado y circuito balanceado (derecha).

La Figura ilustra la diferencia entre la respuesta de la señal en un circuito balanceado y no balanceado. 
El circuito desbalanceado muestra el voltaje establecido entre una sola línea y tierra. 
La cantidad de corriente que fluye entre tierra y la fuente es igual a la corriente en el circuito. 
El circuito balanceado muestra una señal diferencial que fluye donde el voltaje es la diferencia de potencial entre las dos líneas. 
En este caso, la corriente que fluye a tierra en una línea es igual a la corriente que fluye desde tierra en la otra línea.

Comprender la diferencia entre las señales de modo común y las señales diferenciales es crucial para comprender la importancia de los balunes. 
En un circuito balanceado, las señales de modo común son aquellas que son iguales en magnitud y polaridad entre las dos líneas. 
Las señales diferenciales, por otro lado, son iguales en magnitud, pero opuestas en polaridad. En general, las señales diferenciales son mucho más robustas. 
Esto se debe al hecho de que la señalización diferencial tiene inmunidad al ruido inherente. 
En un sistema típico, el ruido externo está igualmente presente en ambas líneas de una configuración balanceada y aparece como una señal de modo común. Una señal diferencial está representada por la diferencia de voltaje entre las dos líneas. Dado que la señal de modo común es igual en ambas líneas, se cancela.

### Modos y tecnologías Balun

Hay dos modos principales de baluns: baluns de corriente y baluns de voltaje (ver Figura 3). Los balunes de corriente operan forzando corrientes iguales en ambas líneas balanceadas, eliminando efectivamente las corrientes de modo común. Los balunes de voltaje fuerzan el mismo voltaje en cada línea balanceada; esto es, en última instancia, un mejor ajuste para aplicaciones de adaptación de impedancia.

Como la mayoría de los transformadores, los baluns se pueden fabricar utilizando líneas de transmisión de núcleo y cable (por ejemplo, coaxial), tecnologías de cerámica cocida a baja temperatura (LTCC) y circuito integrado de microondas monolítico (MMIC). 
Hay dos versiones principales de balunes de núcleo y cable: transformadores de aislamiento y autotransformadores, ambos son balunes de tensión (Figura 4). 
Por encima de unos pocos gigahercios, a menudo es necesario utilizar balunes de línea de transmisión para lograr un rendimiento deseable. Una de las variedades de balun de línea de transmisión de mejor rendimiento es el balun de Marchand (Figura 5). 
Muchos de los baluns LTCC y MMIC de Mini-Circuits utilizan la topología balun de Marchand.

Baluns de núcleo y alambre (baluns de aislamiento y autotransformador)

Los transformadores balun de aislamiento son transformadores de RF con el lado desequilibrado conectado a tierra y el lado equilibrado conectado a la carga. De manera similar a los transformadores de RF estándar, la relación de devanados también se puede utilizar para crear una transformación de impedancia. La figura 4 ilustra un balun autotransformador con entradas balanceadas en cada extremo del devanado, una toma central a tierra y un extremo del devanado que crea el puerto desequilibrado. Un beneficio importante de este tipo de transformador es que la entrada y la salida están separadas eléctricamente, lo que permite un grado de protección para los sistemas que son propensos a bucles de tierra en sus voltajes a nivel del suelo.

Los autotransformadores tienen una configuración diferente a la de los transformadores de RF típicos, ya que esta topología tiene una única ruta conductora. 
Los balunes de autotransformador se pueden fabricar enrollando un solo cable alrededor de un núcleo de ferrita o mediante un cableado cruzado de los devanados primario y secundario. 
Se utiliza un punto de derivación entre los dos extremos del devanado para acceder a diferentes potenciales de voltaje respectivos al voltaje de entrada del transformador. Esta configuración incorpora una ruta de corriente CC a tierra para cada terminal, disipando cualquier acumulación de estática.



### Baluns y Ununs de la línea de transmisión

Los transformadores balun de línea de transmisión generalmente se construyen a partir de una línea de transmisión, como un cable coaxial enrollado alrededor de un núcleo de ferrita y, en algunos casos, simplemente aire. 
Este tipo de transformador balun 1: 1 crea una alta reactancia de estrangulamiento en el conductor exterior del cable coaxial, reduciendo efectivamente las señales de modo común mientras permite que las corrientes internas de la línea de transmisión coaxial pasen sin obstáculos (es decir, un balun Guanella). Además, las configuraciones incluyen un balun bifilar acoplado capacitivamente que usa dos cables enrollados entre sí, así como líneas de transmisión acopladas capacitivamente y magnéticamente enrolladas entre sí y luego enrolladas alrededor de un núcleo común. El propósito de utilizar un núcleo magnético con acopladores de línea de transmisión de banda ancha es permitir el funcionamiento de baja frecuencia.

Los transformadores balun de banda ancha también se construyen con varias transformaciones de impedancia utilizando varias líneas de transmisión en serie y arreglos en paralelo. En este caso, la transformación de impedancia es 1: n 2 , donde n es el número de líneas de transmisión en serie-paralelas. También son posibles los transformadores balun de línea de transmisión de cuarto de onda y media onda, aunque estos tipos de balunes son los más adecuados para aplicaciones con un rango de frecuencia de funcionamiento estrecho.
Baluns LTCC y MMIC

La producción de balun también puede implicar el uso de estructuras metálicas planas con tecnologías como LTCC y MMIC. Esta elección de diseño y fabricación suele ser patentada. Generalmente, el principal beneficio de estos diseños es el tamaño reducido que se puede integrar fácilmente en un conjunto de microondas. 
A diferencia de otros balunes, los baluns basados en LTCC y MMIC se producen utilizando máquinas de ensamblaje de alta precisión y métodos de fabricación de semiconductores que producen una repetibilidad mucho mayor.
Parámetros clave de rendimiento de balun

Los baluns comparten la mayoría de los mismos parámetros de rendimiento que los transformadores de RF, pero la construcción y el uso únicos de los baluns presentan consideraciones adicionales. Estos parámetros incluyen:

    Equilibrio de fase
    Equilibrio de amplitud
    Relación de rechazo de modo común
    Aislamiento de puerto equilibrado
    Aislamiento DC / tierra
    Planitud de retardo de grupo

### Equilibrio de fase y amplitud

El equilibrio (o desequilibrio) de fase y amplitud es una medida de cuán iguales y opuestas son la tensión y la corriente de salida equilibradas de un balun. En esencia, cualquier desequilibrio en la fase o amplitud de voltaje / corriente de una línea balanceada produce pérdidas adicionales y corrientes parásitas. El balance de fase mide qué tan cerca está la diferencia de fase entre la entrada invertida y la entrada no invertida a un desfase ideal de 180. El balance de amplitud es el valor absoluto de la potencia de salida en cada línea balanceada. Para estos parámetros, cada línea balanceada debe ser lo más parecida posible. Los factores determinantes para el equilibrio de fase y amplitud en baluns reales incluyen las propiedades del material, los métodos de fabricación y la coincidencia entre las líneas de salida.Los balunes modernos de alto rendimiento generalmente se especifican para no más de unos pocos grados de desequilibrio de fase y algunos dB de desequilibrio de amplitud.
Relación de rechazo de modo común

El índice de rechazo de modo común (CMRR) es una medida de qué tan bien se atenúan las señales de modo común desde el puerto balanceado al puerto no balanceado. Esta relación depende del desequilibrio de amplitud y fase. Un balun con mejor amplitud y equilibrio de fase también exhibirá CMRR mejorado. Una pauta generalmente aceptada es que una mejora de 0,1 dB en el equilibrio de amplitud, o una mejora de 1 grado en el equilibrio de fase, mejorará la CMRR de un balun en aproximadamente 1 dB.
Aislamiento de puerto balanceado y aislamiento de CC / tierra

El aislamiento de puerto balanceado es una medida de la relación entre la intensidad de la señal de entrada y la intensidad de la señal de salida de un puerto balanceado a otro (pérdida de inserción). La mayoría de las veces, este parámetro no es muy alto en la mayoría de los diseños de balun. El aislamiento de CC es una medida de la conductividad de CC entre el puerto no balanceado y los puertos balanceados, mientras que el aislamiento de tierra es una medida del aislamiento entre la tierra del puerto no balanceado y la tierra (o pseudo tierra) de los puertos balanceados. Estas cifras de aislamiento son importantes para determinar la inmunidad de un dispositivo al ruido y la interferencia. El potencial de aislamiento de un balun a menudo está limitado por la topología del balun.
Planitud de retardo de grupo

El retardo de grupo es el tiempo que tardan los componentes de frecuencia de una señal en pasar a través de un dispositivo. La uniformidad del retardo de grupo es una forma de medir la cantidad de distorsión que encontrarán las señales de banda ancha o de alta velocidad al pasar por un dispositivo. En general, es deseable que todos los componentes de frecuencia de una señal pasen simultáneamente. En el caso de un balun, una buena correspondencia de banda ancha está relacionada con un buen rendimiento de retardo de grupo. Los baluns con una mejor pérdida de retorno normalmente demostrarán una mejor planitud de retraso de grupo.
Aplicaciones importantes de Balun y Unun

Los baluns se utilizaron originalmente para proporcionar adaptación de impedancia y para presentar salidas balanceadas a antenas balanceadas alimentadas por líneas de transmisión no balanceadas. Con la llegada de la electrónica de estado sólido, las salidas y entradas de muchos dispositivos de estado sólido (por ejemplo, amplificadores, mezcladores, DAC / ADC) presentan puertos desequilibrados. Aunque las salidas no balanceadas se pueden conectar fácilmente a líneas de transmisión comunes como cables coaxiales, las líneas de señalización diferencial son generalmente menos susceptibles al ruido y la diafonía. Muchas de estas líneas también son menos costosas que los cables coaxiales para transmisión a larga distancia.

Por ejemplo, equilibrar amplificadores y mezcladores push-pull permite diseños con contenido espurio reducido y CMRR mejorado, es decir, mezcladores de doble equilibrio (ver Figuras 6 y 7). Tenga en cuenta que las derivaciones centrales de los baluns en la Figura 7 se utilizan para polarizar el amplificador balanceado. Otra aplicación común es el equilibrio y la adaptación de impedancias de convertidores de banda ancha a fuentes no balanceadas, que generalmente presentan impedancias de puerto que son muy diferentes de las de los puertos del convertidor (consulte la Figura 8).
