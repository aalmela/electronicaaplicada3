```python
import myplot as myp
import matplotlib.pyplot as plt
import numpy as np

fo = 10.7e6
B  = 200e3
R  = 10e3
Ro = 1e3
rg = 10e3
Qo = 80
wo = 2*np.pi*fo

Qc = fo/B
print('$$Q_c= {:1.2f}$$'.format(Qc))

rext = rg*R/(rg+R)
XL = rext*(1/Qc-1/Qo)
L  = XL/wo
print('$$L= {:1.2e} Hy$$'.format(L))

XC = XL
C  = 1/(wo*XC)
print('$$C= {:1.2e} C$$'.format(C))



### Calculo del divisor

N = (R/Ro)**0.5
print('$$N = {:1.2f}$$'.format(N))

Qm2 = wo * C * R
print('$$Q_m2 = {:1.2f}$$'.format(Qm2))

Qm1 = Qm2/N
print('$$Q_m1 = {:1.2f}$$'.format(Qm1))


C2  = C*N
print('$$C_2= {:1.2e} F$$'.format(C2))


C1  = C2/(N-1)
print('$$C_1= {:1.2e} F$$'.format(C1))



#### Plot!

f  =  np.arange(10e6,11.4e6,100)                   # Defino el array de frecuencia

w  = 2*np.pi*f

ZC1 = 1/(1j*w*C1) 
ZC2 = 1/(1j*w*C2)

rp  = Qo*w*L
ZL  = rp*(1j*w*L)/(rp+(1j*w*L))

Z2 = ZC2*Ro/(ZC2+Ro)

Zs = ZC1 + Z2
Zi = Zs*ZL/(Zs+ZL)
        
print(Zi[5000])    

# Plot Za
myp.PlotCpx(f,Zi , labelx = 'Frecuencia/Hz', labely = '$Z_{in}$')

plt.savefig('figures/ejemplodivCb.png')

plt.show() 
```
