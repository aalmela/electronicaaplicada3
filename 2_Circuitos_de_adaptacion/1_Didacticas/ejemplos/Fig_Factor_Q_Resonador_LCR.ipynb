```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# Ejemplo de respuesta en frecuencia 
R  = 1
fo = 1

# respuesta vg/ig
def resp(f, fo, Q, R):
    return  R/( 1+ 1j * Q * (fo**2 - f**2)/(fo*f) )

#Defino el vector Frecuencia
freq = np.arange(0.1, 10, 0.001)

for Q in [1,10,50]:
    plt.semilogx(freq, abs(resp(freq,fo,Q,R)), label=Q )
    plt.xlim([0.9,1.2])
    plt.legend()
plt.ylabel('vo/ig')
plt.xlabel('frecuencia[Hz]')
plt.grid()
#plt.savefig('figures/Qcon.png')
```
