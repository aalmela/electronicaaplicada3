### Ejemplo adaptador tipo  L, adaptador para antena de 11m.

Supóngase querer adaptar una antena látigo de cuarto de onda diseñada para operar a $28.5MHz$, cuya impedacia es de $37.5+j21.25 ohms$.

Se desea emplear la misma antena a la frecuencia de $f_{RF} = 3 MHz$, donde presenta una impedancia de  $Z_{ant}= 2 \Omega - j \frac{1}{2 \pi 3MHz 30pF}$.

![30-cuatrocientos](fig/eje_a_antena.png)

Se desea que el equipo transmisor, la fuente de corriente, tenga una carga de $Z_{in} = 50\Omega+j0\Omega$ @ $28.5 MHz$ para que esté adaptada, se recurre así al circuito de adaptación mostrado en la figura compuesto por L y C. 

Suponga que la resistencia de perdida de L es despresiable. 

La potencia disponible del transmisor es $P_{disp} = 100 W$. 

Realcular el adaptador para la frecuencia de $f_{RF} = 3 MHz$.

```python
#import myplot as myp
import numpy as np
import matplotlib.pyplot as plt
import csv

#f = np.linspace(2e6,4e6,100)

#w    = 2*np.pi*f


Pdisp = 100

# Datos para los calculos 
fo    = 28.5e6
wo    = 2*np.pi*fo
Za    = 37.5+ 21.25j
rg    = 50


print('.param ra =  {:1.2f} '.format(Za.real) )
print('.param rg =  {:1.2f} '.format(rg) )
if (Za.imag > 0) : print('.param La  =  {:1.4e} '.format(Za.imag/(wo)) )
else             : print('.param Ca  =  {:1.4e} '.format(1/(Za.imag*wo)) )

## Etapa A

# Conversion serie-paralelo
Qm = ((rg/(Za.real)) -1)**0.5    # Calculo Qm necesario
#print(Qm)
XL  = Qm*(Za.real)           
L   = (XL-Za.imag)/wo
print('.param L  =  {:1.4e} '.format(L) )

#mean diameter of the coil    D = 200.0 mm
#number of turns              N = 26.0
#length of the coil           ℓ = 180.0 mm
#wire or tubing diameter      d = 3.0 mm
#design frequency             f = 3.0 MHz
#effective series inductance @ design frequency          L_eff_s = 9.433e-05 H
#effective series AC resistance @ design frequency       R_eff_s = 1.095e+00 Ω


XLp = XL*(1+1/Qm**2)

C   = 1/(XLp*wo)        # Capacitor sintonia 
print('.param C  =  {:1.4e} '.format(C) )
#https://www.knowlescapacitors.com/getattachment/ac79327c-78b2-4790-b841-4541dcd9a291/High-Power-RF
#http://datasheets.avx.com/HQ-Series.pdf
#effective series AC resistance @ design frequency       R_eff_s = 0.001 Ω


vg = ((Pdisp*8)*rg)**0.5

### Para simular con LTspice
print('.param vg =  {:1.2f} '.format(vg) )
print('.param fo=  {:1.2e} '.format(fo) )

```
