# Ejercicio auto-transformador

El siguiente circuito corresponde a un adaptador de impedancias realizado con un auto-trasformador, cuyo  $Q_{oL} = \infty$ a la frecuencia de trabajo, y un capacitor de $ C_1$ conectado en paralelo con el primario, cuyas pérdidas $Q_{oC} = 100$. 

![11-cuatrocientos](figg/eje3.png)

La impedancia interna de la fuente es de $z_g(100MHz)= (74-j7) \Omega$ a y la corriente que entrega es $i_{g}(100MHz) = 1 \mu A$. 

La resistencia de carga es de $R_L = 1500 \Omega$.


Nota: en SPICE la sentencia "k1 L1 L2 1" indica que el transformador tiene $K = 1$ (acople máximo).


Para máxima transferencia de energía a un $Q_c = 20$, se busca que $R_L' = r_g$.  


**2. Tensión en el primario, en el secundario y la potencia sobre $R_L$.**   


**1. Para $Q_c= 20$ a $100MHz$: $C$,$L_1$, $L_{1a}$, $Ls$ y la relación de vueltas del transformador.**


![11-cuatrocientos](./figg/eje3a.png)


El autotransformador tiene conectado en su primeario $R_L$ y debe refleja en el secundario $R_L'= r_g$.


La relación de espiras entre primario y secundario  

N = (N1 + N2)/ N1 (pri/sec) 
N  = (RL/RLs)**0.5


El calculo de L se hace en el primario (podria hacerse en el secundario también):
Rext  = rgp * RL / (rgp + RL )

    
L en el primario
L =  (1/Qc - 1/Qo)  * Rext / wo


Calculo de L1 y L2

N = (N1 + N2)/ N1 = (L/L1)**(1/2) (**(1/2) es la raiz ) 

$N^2 = \frac{L}{L_1}$

L1 es la inductancia del secundario
L1  = L / N**2


N = (N1 + N2)/ N1 = (L/L1)**(1/2) (**(1/2) es la raiz ) 
# N = (1 +  N2/N1)
# N = (1 +  N21) 
N21 = N - 1

# (L/L2) = [(N1 + N2)/ N2]**2
# (L/L2) = [(N1/N2 + 1)]**2
# (L/L2) = [(N12 + 1)]**2
#  N12 = 1/N21
N12 = 1/N21

# L2 = L/ [(N12 + 1)]**2
L2 = L / ( N12 + 1)**2

# La capacidad que resuena con L
Ct  =  1/ (L * (wo)**2)  

# La capacidad de la fuente en el primario del autotrafo
cgp = (cg/N**2)

# C1 entonces
C1  = Ct - cgp 


```python
import numpy as np
from IPython.display import Image
# Datos
Qc = 20
Qo = 100
zg = 74-7j  
ig  = 1e-6 # A pico
fo  =  100e6
RL  = 1500

wo = 2*np.pi*fo

# Calculo la impedancia de la fuente para el modelo paralelo
rg = 1/((1/zg).real )
cg  = ( (1/zg).imag )/wo 
print('rg =  {:1.1f} ohms'.format(rg) )
print('cg =  {:1.3f} pF'.format(cg/1e-12) )

Image(filename= "fig/eje3a.png",width=350)
```

```python
# a)


# El autotransformador refleja en el primario rg como RL ya que se calcula para MTEQcte.
rgp  = RL

# El autotransformador refleja en el secundario RL como rg ya que se calcula para MTEQcte.
RLs  = rg

# Relación de espiras N = (N1 + N2)/ N1 (pri/sec) 
N  = (RL/RLs)**0.5


# El calculo de L se hace en el primario (podria hacerse en el secundario también):
Rext  = rgp * RL / (rgp + RL )

    
# L en el primario
L =  (1/Qc - 1/Qo)  * Rext / wo


# Calculo de L1 y L2
# N = (N1 + N2)/ N1 = (L/L1)**(1/2) (**(1/2) es la raiz ) 
# N**2 = L/L1
# L1 es la inductancia del secundario
L1  = L / N**2


# N = (N1 + N2)/ N1 = (L/L1)**(1/2) (**(1/2) es la raiz ) 
# N = (1 +  N2/N1)
# N = (1 +  N21) 
N21 = N - 1

# (L/L2) = [(N1 + N2)/ N2]**2
# (L/L2) = [(N1/N2 + 1)]**2
# (L/L2) = [(N12 + 1)]**2
#  N12 = 1/N21
N12 = 1/N21

# L2 = L/ [(N12 + 1)]**2
L2 = L / ( N12 + 1)**2

# La capacidad que resuena con L
Ct  =  1/ (L * (wo)**2)  

# La capacidad de la fuente en el primario del autotrafo
cgp = (cg/N**2)

# C1 entonces
C1  = Ct - cgp 


print('a1) L1  = {:2.2e} Hy'.format(L1) ) 
print('a2) L2  = {:2.2e} Hy'.format(L2) ) 
print('a3) C1  = {:2.3e} F'.format(C1) ) 
print('a4) N   = {:2.2f} '.format(N) ) 
```

```python
#b)
# resistencia total que ve la fuente 
rt  = Qc*wo*L1

# tensión en el secundario
vs = ig * rt # V pico

# Tensión en el primario
vp = vs * N

# Potencia en RL
PRL = vp**2/RL/2 # W (Eficaz)

print('b1) vs = {:2.2e} V (pico)'.format(vs) ) 
print('b2) vp = {:2.2e} V (pico)'.format(vp) ) 
print('b3) PRL= {:2.2e} W (eficaz)'.format(PRL) ) 
```
