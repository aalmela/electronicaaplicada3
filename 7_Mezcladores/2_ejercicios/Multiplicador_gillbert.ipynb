## Mezclador basado en la celda de Gilbert

El mezclador basado en la celda de Gilbert aparece muy a menudo tanto en equipos comerciales como de radioaficionados. 

Es popular el integrado NE602 (ahora conocido como SA602 ó SA612) usado en muchos circuitos de radioaficionados es en realidad un mezclador Gilbert en empaque IC.

# Multiplicador: Celda de Gilbert.

La configuración conocida como celda de Gilbert basa su funcionamiento en circuitos pares diferenciales los cuales puede utilizarse como circuito multiplicador a partir del hecho que la corriente diferencial de salida es proporcional al producto entre la corriente de los emisores acoplados. 


![5-cuatrocientos](fig/diferencial.png)



La figura muestra un circuito diferencial formado por un par acoplado por emisor. 
El circuito de polarización conectado a los emisores de Q1 y Q2 pueden ser una fuente de corriente de transistor, que se denomina $ I_{TAIL}$. Esta fuente tiene una resistencia $R_{TAIL}$.

El comportamiento para gran señal del par acoplado por el emisor es importante en parte porque ilustra el rango limitado de voltajes de entrada en el que el circuito se comporta de forma casi lineal.

Además, el comportamiento de la señal grande muestra que la amplitud de las señales analógicas en los circuitos bipolares se puede limitar sin llevar a los transistores a la saturación, donde el tiempo de respuesta aumentaría debido al exceso de almacenamiento de carga en la región base.

Para simplificar el análisis, asumimos que la resistencia de salida de la fuente de corriente $R_{TAIL} \rightarrow \infty$, que la resistencia de salida de cada transistor $r_o \rightarrow \infty$, y que la resistencia de base de cada transistor $R_{b} \rightarrow \infty$.
Estas suposiciones no afectan en gran medida el comportamiento de baja señal y gran señal del circuito. 

Si se analiza por la ley de mallas de Kirchhoff el circuito de entrada, podemos plantear:

\begin{equation}
    V_{i1} - V_{be1} + V_{be2} - V_{i2} = 0 
\end{equation}

Suponga que las resistencias de  colector son lo suficientemente pequeñas como para que los transistores no operen en saturación si $V_{i1} \le V_{CC}$ y $V_{i2} \le V_{CC}$.
Si $V_{be1} \gg V_{T}$ y $V_{be2} \gg V_{T}$, las ecuaciones de Ebers-Moll muestran que:

\begin{equation}
    I_{c1} = I_{s1} e^{\frac{V_{be1}}{V_T}}
\end{equation}

\begin{equation}
    I_{c2} = I_{s2} e^{\frac{V_{be2}}{V_T}}
\end{equation}

Asumiendo que los transistores son idénticos para que $I_{s1} = I_{s2}$.

\begin{equation}
    V_{be1} = V_T ln(\frac{I_{c1}}{I_s})
\end{equation}

\begin{equation}
    V_{be2} = V_T ln(\frac{I_{c2}}{I_s})
\end{equation}

Combinando ambas ecuaciones y desarrollando. 

\begin{equation}
    V_{i1} - V_T ln(\frac{I_{c1}}{I_s}) + V_T ln(\frac{I_{c2}}{I_s}) - V_{i2} = 0 
\end{equation}

\begin{equation}
    \frac{V_{i1} - V_{i2} }{V_T} = ln(\frac{I_{c1}}{I_s}) - ln(\frac{I_{c2}}{I_s}) 
\end{equation}

\begin{equation}
    \frac{V_{i1} - V_{i2} }{V_T} = ln(\frac{I_{c1}}{I_{c2}})  
\end{equation}

\begin{equation}
    \frac{I_{c1}}{I_{c2}} = e^(\frac{V_{i1} - V_{i2} }{V_T}) = e^{\frac{V_{id}}{V_T}}   
\end{equation}

donde $V_{id} = V_{i1} - V_{i2}$.
Entonces, empleando nodos en los emisores de los transistores:

\begin{equation}
  -(I_{e1} - I_{e2} ) = I_{TAIL} = \frac{I_{c1}-I_{c2}}{\alpha_F}  
\end{equation}

\begin{equation}
  I_{TAIL} \alpha_F = I_{c1} + I_{c1} e^{-\frac{V_{id}}{V_T}} 
\end{equation}

\begin{equation}
  I_{c1} =    \frac{I_{TAIL} \alpha_F}{1 +  e^{-\frac{V_{id}}{V_T}} } 
\end{equation}


\begin{equation}
  I_{c2} =    \frac{I_{TAIL} \alpha_F}{1 +  e^{\frac{V_{id}}{V_T}} } 
\end{equation}

Estas dos corrientes se muestran como una función de $V_{id}$ en la figura.
Cuando la magnitud de $V_{id}$ es mayor que aproximadamente $3V_{T}$, que es aproximadamente 78 mV a temperatura ambiente, las corrientes del colector son casi independientes de Vid porque uno de los transistores se apaga y el otro conduce toda la corriente que fluye.
Además, el circuito se comporta de forma aproximadamente lineal solo cuando la magnitud de $V_{id}$ es menor que aproximadamente $V_{T}$.

![5-cuatrocientos](fig/diferencial_plot.png)

Respuesta de la corriente $I_{c1}$ e $I_{c2}$ en función de la tensión $V_{id}$  del circuito diferencial.}


Si se asume que la tensión diferencial es $ V_{id} \ll 2 V_T $

\begin{equation}
 \frac{d I_{c1}}{d V_{id}} =    \frac{I_{TAIL} \alpha_F e^{-\frac{V_{id}}{V_T}}}{V_T (1 +  e^{-\frac{V_{id}}{V_T}})^2 } 
\end{equation}

\begin{equation}
 (\frac{d I_{c1}}{d V_{id}}) |_{V_{id}=0} =    \frac{I_{TAIL} \alpha_F }{4 V_T} 
\end{equation}




\begin{equation}
  I_{c1} |_{V_{id}=0} =    \frac{I_{TAIL} \alpha_F}{2 } 
\end{equation}

\begin{equation}
  I_{c1} \approx  I_{TAIL} \alpha_F ( \frac{1}{2 } +  \frac{V_{id}}{4 V_T}) 
\end{equation}
