# Mezcladores a diodo balanceados

Como los circuitos a diodo que se van a describir producen sumas y diferencias de las dos frecuencias de entrada, pueden usarse como moduladores y demoduladores de amplitud, así como mezcladores; es por esta razón que
los términos "modulador balanceado" y "mezclador balanceado" son sinónimos. 

En la discusión que sigue, las frecuencias de entrada serán $f_{RF}$ y $f{LO}$, dando lugar a la salida $f_{IF}$.
En el estudio de moduladores DSB/SC, (doble banda lateral, portadora única)por otra parte, las frecuencias de entrada serán $f_c$(portadora) y $f_m$(señal moduladora).
y la salida deseada estará en $f_c \pm f_m$.

Un mezclador balanceado es aquel en el que las frecuencias de entrada no aparecen en la salida. Idealmente, las únicas frecuencias que se producen son la suma y la diferencia de las frecuencias de entrada.
Un circuito multiplicador, donde la amplitud de salida es proporcional al producto de dos señales de entrada, puede utilizarse como un mezclador balanceado. Es fácil la demostración matemática.

Analicemos el circuito para el ciclo positivo del oscilador local. El oscilador local enciende los diodos D2 y D4, y dado que tiene una resistencia de serie, el valor de la resistencia dinamica se aproximará a este valor para valores de $V_D$ mayores a los de encendido. 

El secundario $L_3$ tiene una resistencia de aproximadamente $25 \Omega$ a masa dado por el pararelo de las impedancias reflejadas por la impedancia del generador de $OL$.

La impedancia de la malla $L_3$ reflejada a la malla de entrada ($L_1$), se mantiene sin varaciones por la relación $1:1$. 

La tensión sobre $L_1$ se calcula a partir del divisor.

$V_{R1} =  V_{L_3} \frac{50}{50+25}$

$V_{R1} =  V_{L_3} 0.66$

Dado que el doble balanceado produce un inversión en la señal de  RF para el siguiente ciclo del oscilador local (señal $\pm 1$, de amplitud 2. 


$a_1 = 2 \frac{2}{\pi}$

$V_{R1} = V_{RF} \frac{2}{\pi} 0.66 0.6$

Siendo $V_{RF} = 50 \times 10^{-3} V$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy                  # http://scipy.org/
from scipy import signal

from IPython.display import Image  
%matplotlib inline
params = {'legend.fontsize': 24,
          'figure.figsize': (8, 8),
         'axes.labelsize': 24,
         'axes.titlesize': 24,
         'xtick.labelsize':24,
         'ytick.labelsize':24}
plt.rcParams.update(params) 

from scipy.optimize  import fsolve, root

## Ejemplo, transferencia de diodo
mydiode            = {} 
mydiode['1N4148'] =  [2.52e-9,1.72,0.56,25e-3] #Is,n,Rs
mydiode['1N5817'] =  (  31e-6,1.37,0.50,25e-3) #Is,n,Rs


def IDT(x, *data):
    VD,diode = data
    IS = diode[0]
    n  = diode[1]
    Rs = diode[2]
    vt = diode[3]
    return IS* np.exp((VD - x*Rs)/(n*vt))- x

vgo = np.arange(0,1.2,0.01)

ido = [ fsolve( IDT, [0.1], args=(vg,mydiode['1N5817']))[0] for vg in vgo]


plt.plot(vgo, ido)

plt.ylabel('$i_{o}$ [A]')
plt.xlabel('$v_g$[V]')
#plt.ylim(0, Ic*1.5)
plt.grid()
plt.show()

```
