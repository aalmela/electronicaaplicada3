# 3. Amplificadores de bajo ruido.

El siguiente proyecto presenta el diseño de un amplificador de bajo ruido LNA (Low Noise Amplifier). 
Los LNA son muy usados en receptores de RF como en los teléfonos celulares, receptores de GPS, redes LANs (WiFi) y comunicación satelital. 
Este tipo de amplificadores, por lo general, se encuentra inmediatamente despues de las antenas en los sistemas de RF, aunque a veces puede haber un filtro pasa banda de por medio. 
Al ser los primeros amplificadores del sistema, su función es amplificar toda la señal, incluyendo el ruido, recibida por la antena, sin agregar mas ruido propio del amplificador. 
De esta manera se amplifica la señal y no se presentan cambios muy significativos en la relación señal a ruido a la salida. 
Son un componente clave en el front-end de los receptores de RF ya que como agregan poco ruido a la señal, se caracterizan por tener una cifra de ruido baja, y una ganancia buena, y como sabemos por la formula de Friis la cifra de ruido total del sistema es dominada por las primeras etapas del sistema. 

La cifra de ruido del LNA se minimiza eligiendo componentes de bajo ruido, el punto de operación y la topología del circuito. 
Los dispositivos activos mas usados en los LNA son fabrica dos usando GaAs pHEMTS o transistores bipolares usando SiGe, pero también se usan transistores de silicio para aplicaciones menos criticas o de bajo costo. 
En general los dispositivos de GaAs tienen menor cifra de ruido y mayor linealidad en comparación con los transistores de SiGe, pero los transistores fabricados usando SiGe trabajan de manera mas eficiente y son de mas fácil integración. 

La elección de la topología depende de la aplicación especifica que se le de al LNA, por ejemplo si queremos un amplificador con muy baja cifra de ruido podemos usar un emisor común, pero tenemos considerar que posible que necesite una estabilización externa, y no tendremos gran ancho de banda, a diferencia de la topología de base común, la cual es estable y además tiene mayor linealidad, pero menor ganancia.

### Adaptadores de Impedancia

En el curso vimos que se debe adaptar impedancias para asegurar la máxima trasferencia de energía entre generador y carga, pero observando la expresión de la cifra de ruido, vemos que depende de la adaptación. 
En teoría es posible adaptar para máxima transferencia de energía y mínima cifra de ruido simultáneamente, pero no siempre es posible. 

Muchas veces se debe optar por degradar la trasferencia de energía para obtener la cifra de ruido óptima. 
La cifra de ruido depende de la impedancia de fuente, por lo tanto se busca la impedancia de generador que logra la menor cifra de ruido. Derivando la expresión del la cifra de ruido $F$ con respecto a $B_{s}$ y $G_{s}$, igualando a cero y resolviendo el sistema de ecuaciones se obtiene el valor óptimo de admitancia de fuente $Y_{opt}=G_{opt}+jB_{opt}$. 
Las componentes de conductancia y susceptancia expresada en términos de $R_{n}$ , $G_{u}$ , $G_{\gamma}$  y $B_{\gamma}$ quedan:

$$G_{opt}=[\frac{G_{u}+R_{u}G_{\gamma}^2}{R_{u}}]^{\frac{1}{2}}$$

$$B_{opt}=-B_{\gamma}$$

Y el valor mínimo de cifra de ruido es:

$$ F_{opt}=1 + 2R_{n}(G_{\gamma}+G_{opt})$$

Por lo tanto la cifra de ruido puede ser descrita mediante cuatro parámetros fundamentales $F_{opt}$, $G_{opt}$ , $B_{opt}$ y $R_{n}$. La cifra de ruido queda:

$$ F=F_{opt} + \frac{R_{n}}{G_{s}}[(G_{s}-G_{opt})^2+(B_{s}+B_{opt})^2]$$

Estos parámetros se pueden medir variando la admitancia de fuente y midiendo el ruido de salida. Cuando el ruido es mínimo, se obtiene $Y_{opt}$, el valor de potencia de ruido permite e $Y_{opt}$ permite obtener los demás parámetros. El trabajo del diseñador será obtener un valor de $Y_{s}$ tal que cumpla los requerimientos de ruido y potencia.

Ahora veamos el caso del amplificador sintonizado.
El circuito de adaptación transforma la impedancia de generador $R_{g}$ a la necesaria a la entrada del transistor $R_{in}$. A la frecuencia de sintonía $f_{0}$, se anulan las componentes reactivas por lo tanto, la reactancia vista en la base del transistor $B'_{s}$ es igual a cero. Solo se ve una resistencia $R'_{s}$, o de forma equivalente una conductancia $G'_{s}$. La expresión para la cifra de ruido de un amplificador sintonizado seria la siguiente:

$$F=F_{opt} + \frac{R_{n}}{G'_{s}}[(G'_{s}-G_{opt})^2+B_{opt}^2]$$

Si seleccionamos $G'_{s}=G_{opt}$, se obtiene la figura de ruido mínima y es:

$$ F=F_{opt} + \frac{R_{n}B_{opt}^2}{G_{opt}}$$

Cada diseño será una relación de compromiso entre mínima cifra de ruido, la máxima trasferencia de energía y el $Q_{c}$ del circuito.


Los sistemas receptores casi siempre se especifican en términos de niveles de potencia, en lugar de voltajes. 
Estas son mediciones logarítmicas para que la ganancia se pueda calcular sumando niveles en lugar de multiplicación. 
La unidad mas familiar es ''dBm'', que no es un voltaje, sino una potencia en relación con un nivel de potencia de referencia de $1 mW$

$$P [dBm] = 10\cdot log_{10}( \frac{P[W]}{1mW} )$$

Si la potencia de ruido "kTB" se considera con más detalle, se puede introducir una simplificación interesante:

$$kTB = (kT) \cdot B  = (Densidad de potencia de ruido) x (Ancho de banda)$$

La densidad de potencia de ruido "kT" representa la potencia en cada Hz; esto debe ser multiplicado por el ancho de banda para calcular la potencia de ruido producida. Convirtiendo esto a un cálculo del nivel de potencia, podemos decir que:

Cada resistencia produce a temperatura ambiente ($T_o = 290 K$) un nivel de ruido disponible y, por lo tanto, una densidad de potencia de ruido disponible de -174dBm por Hz.

Para un ancho de banda B mayor a 1 Hz el nivel de potencia de ruido disponible en [dBm]:

$$N_{disp} = -174 dBm + 10 \cdot log(B [Hz])$$

Como ejemplo, se calculará el voltaje de ruido sin carga vn en los terminales de una resistencia de 50 $\Omega$ para un ancho de banda de 100 kHz:

$$N_{disp} = -174dBm + 10log(10K) = -174dBm + 50dB = -124dBm$$

Eso da como resultado una potencia de ruido disponible de

$$N_{disp} = 1 mW \cdot 10^{\frac{-124}{10}} = 1 mW \cdot 10^{-12.4}= 4\cdot 10^{-16} W$$

Eso producirá un voltaje en la resistencia de $50\,\Omega$ de:

$$v_n = \sqrt[]{ N_{disp} \cdot R} = \sqrt[]{ 4\cdot 10^{-16} W \cdot 50\Omega } = 141 nV$$

Entonces el voltaje de circuito abierto es el doble de este valor, por lo tanto 282nV.

El ruido de disparo se produce en diodos de válvula y transiciones PN por distorsiones del flujo de corriente cuando cruzando las diferencias de potencial. Produce banda ancha, ruido blanco.

Ruido Flicker o $1/f$: como resultado de defectos en la estructura cristalina causados por impurezas conducen a fluctuaciones de tipo de pulso corto en el flujo de corriente que producen un espectro cuya la densidad de potencia disminuye con la frecuencia creciente.
