Diseño de amplificador de potencia y bajo ruido


![fig1](fig/PA_schematic.png)

El esquema de la polarización del amplificador se muestra en la Fig. 2(a). Usamos el mismo transistor, el BFU550, para PA y LNA para simplificar la construcción. La diferencia entre los dos diseños será la condición de polarización y las impedancias de red coincidentes de entrada y salida. La parte gris de la red de polarización es la polarización de CC. R2 establece la corriente base para Q1. R1 y C3 forman un circuito de retroalimentación simple que ayuda a regular el voltaje de operación en C3. El voltaje de entrada, aproximadamente 9V de una batería de 9V, se reduce por la caída de voltaje en R1. A medida que aumenta la corriente del colector (DC), esta caída aumentará, reduciendo el voltaje en la resistencia de polarización R2, provocando que la polarización se reduzca y con ello una reducción en la corriente del colector del colector. Esta red de retroalimentación no es crítica para el sesgo; sin embargo, ayuda en la regulación del sesgo. La ruta de RF es a través de Q1, hasta la carga de salida (C1 y C2 son capacitores de bloqueo de CC simples). Los amplificadores funcionan en Clase A (lineal) y Clase AB (no lineal) según la señal de entrada. Las señales pequeñas serán Clase-A [5].

El PA está diseñado optimizando la impedancia de salida, a través de L3 y L4, para maximizar la potencia de salida. El transistor está polarizado a una corriente de CC de 30 mA, que está cerca de la potencia de salida máxima en la hoja de datos del transistor. Para diseñar el PA, realizamos un análisis Load-Pull, en el que hacemos un barrido de la impedancia de salida en la simulación en un rango de la carta de Smith. Luego medimos la potencia de salida y encontramos la impedancia de carga con la potencia de salida más alta. La figura 1(b) muestra los resultados de carga y tracción. Las cruces negras son la impedancia de carga simulada y los círculos son de potencia de salida constante (azul) y eficiencia de potencia añadida (PAE) (rojo). Seleccionamos una impedancia de carga de 50+58j que estaba cerca del punto de máxima potencia (círculo más central en azul). Transformamos la impedancia de una carga de 50 a esta impedancia objetivo en nuestro PA para que vea la impedancia de carga óptima para la máxima potencia de salida. Se supone que la red de adaptación de impedancia de entrada opera en el rango de señal pequeño, por lo que simplemente buscamos una red de adaptación de impedancia que iguale la impedancia del cable de entrada 50 a la entrada del PA. La entrada PA es 40-22j, Fig. 2(b), lo que significa que nuestra red debe transformar 50 a 40+22j, es decir, el conjugado complejo para garantizar la coincidencia de impedancia. Es importante tener en cuenta que para la carga de PA, barrimos las cargas reales que queríamos, por lo que no hay una conjugación compleja. De hecho, normalmente no se piensa en "hacer coincidir" la salida del PA con una red, sino en encontrar la carga que produce la máxima potencia [6]. Esto se debe a que la mayoría de los PA operan en un régimen no lineal donde no se aplica la coincidencia de señal pequeña (o no es el mejor enfoque).

El LNA está polarizado a 3 mA, que está más cerca de la operación de ruido más bajo del transistor como se muestra en la hoja de datos. Luego se simula el amplificador para encontrar la impedancia de salida que logra la ganancia máxima (Gmax). Se encontró que esto, Fig. 2(c), era 60+100j. Tenga en cuenta que los círculos de ganancia máxima (azul) se extienden fuera del gráfico de Smith. Elegimos un punto en el círculo más alto, pero más cercano al origen, es decir, más cercano a 50 , para minimizar la transformación de impedancia necesaria. La red fuente es donde el LNA se diferencia de un simple amplificador de CA. El ruido de entrada referido del amplificador depende de la impedancia vista en la entrada. Como resultado, es posible que una entrada de impedancia adaptada no proporcione el funcionamiento con el ruido más bajo. Muchos transistores se han medido y modelado para la impedancia de la fuente que produce el ruido más bajo, medido por la figura de ruido (NF). El factor de ruido, F, es la relación entre la relación señal/ruido de la entrada y la relación señal/ruido de la salida, NF es 10log10F. La figura de ruido (NF) siempre es mayor que cero. La Figura 2(c) muestra los círculos de ganancia para la impedancia de entrada que alcanza la máxima ganancia (verde), se supone que la carga es la carga que alcanza la máxima ganancia, es decir, 60+100j. Trazadas en marrón están las impedancias de fuente para NF mínimo. Tenga en cuenta que los dos se superponen pero no son idénticos. Seleccionamos una impedancia de fuente de 11+10j como compromiso entre alta ganancia y ruido mínimo.

Para todas las redes de adaptación de impedancia, se eligió una red simple de dos líneas de transmisión (líneas t), que consta de una línea t en serie con un ramal de línea t abierto, Fig. 2(a). Si bien se pueden calcular las longitudes de cada línea t [L1 a L4 Fig. 2(a)], la mayoría de las herramientas de simulación comerciales pueden "afinar" los parámetros de los componentes. La impedancia característica de las líneas t se fijó en 50 para facilitar la fabricación; solo se modificaron las longitudes. Nuestro radar fue diseñado en la oficina de microondas AWR de National Instruments, lo que nos permitió ajustar las longitudes en un control deslizante. En menos de un minuto se pueden ajustar las longitudes por ensayo y error para obtener las impedancias deseadas. Una característica de optimización en el software también puede encontrar valores adecuados. Luego, la red se conecta a una línea de 50 t para extender la red hasta el borde de la placa para conectar un conector SMA de montaje en PCB.

La Figura 2 (c) muestra los resultados del parámetro S de PA y LNA. El rendimiento general es bueno, sin embargo, hay discrepancias en las simulaciones. Estas discrepancias se deben a que las redes de adaptación de impedancia no son precisas debido a la falta de consideración de los bordes de la placa y las variaciones en la fabricación utilizando un simple cortador de láminas 2-D.

```python
def idi(v):
    i =  0.             if (v < 3.8) else 0
    i = (2/1.2)*(v-3.8) if (v >= 3.8  and v <  5) else i
    i = 2+(0.5/3)*(v-5) if (v >= 5  and v <  8.8) else i
    i = (2.64) if (v >= 8.8  ) else i
    return i



vgsa = np.arange(0,10,0.01)
idsa = np.array([idi(v) for v in vgsa])

plt.plot(vgsa,idsa,'b.')
#plt.axis([0, 4, 0, 4.5])
plt.grid()
plt.ylabel('Id(A)')
plt.xlabel('VGS(V)')
plt.show()
```

```python
# Transformada Discreta de Fourier     
def DFT(x):
    N = len(x)
    X = list()
    for k in range(int(N/2)):
        a = 2*sum([s*np.cos(2*np.pi*n*k/N) for n,s in enumerate(x)])/N
        b = 2*sum([s*np.sin(2*np.pi*n*k/N) for n,s in enumerate(x)])/N
        if k == 0: X.append(a/2)
        else     : X.append(a+1j*b)
    return np.array(X) 

Idsf = DFT(ids)

plt.stem(abs(Idsf[:5]) )
plt.show()
```
