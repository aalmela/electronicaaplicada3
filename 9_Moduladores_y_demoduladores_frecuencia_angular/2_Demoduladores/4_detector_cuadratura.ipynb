
El detector de cudratura es un demodulador de FM en cuadratura.  
Un detector de cuadratura utiliza un desplazador de fase de 90° y un detector de producto para demodular las señales de FM. 
El desplazador de fase de 90° produce una señal que está en cuadratura con las señales recibidas. 
El circuito sintonizado convierte las variaciones de frecuencia a variaciones de fase y el detector de producto multiplica las señales de IF recibidas por la señal de IF desplazadas en
fase.

El circuito tanque se sintoniza a la frecuencia central de IF y produce un desplazamiento de fase adicional $\varphi$ que es proporcional a la desviación de frecuencia.
